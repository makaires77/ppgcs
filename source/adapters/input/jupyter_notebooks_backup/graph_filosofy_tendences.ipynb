# Grafos de tendências Filósofas durante a história

### Polaridades em relação aos temas

Para analisar e classificar os 50 filósofos em diferentes temas e subtemas, utilizamos dois critérios principais: a posição filosófica de cada um em relação a determinados temas e a relevância que esses temas tiveram dentro de suas próprias obras. Abaixo estão as descrições das classificações efetuadas nos diferentes níveis e subníveis:

1. Classificação de polaridade temática em uma área filosófica

    Positiva (+1): Indica que o filósofo apoia, defende ou tem uma visão favorável sobre o tema. Isso é evidente em suas obras e argumentações.
    
    Neutra (0): O filósofo é considerado neutro em relação ao tema, não demonstrando apoio ou oposição significativos, ou o tema não é central em sua obra.
    
    Negativa (-1): O filósofo se opõe ou critica o tema, rejeitando-o explicitamente em suas obras.

2. Relevância em cada tema na obra de cada Filósofo

    Pontuação de 0 a 5: Uma escala Likert é usada para indicar a importância do tema na obra do filósofo, onde 0 significa que o tema não é tratado e 5 indica que o tema é um ponto central ou extensivamente abordado em seus trabalhos.

### Categorias e Subcategorias

Para cada uma das três categorias - Transcendente, Antropológica e Imanente - temos três temas relevantes, organizando-os de cima para baixo, do mais transcendente para o mais imanente dentro de cada categoria. Esses temas representam áreas de interesse filosófico significativas dentro de cada categoria e podem ser usados para entender melhor as contribuições e as áreas de foco de diversos filósofos ao longo da história. 

As categorias foram atribuídas com base nas principais áreas de interesse de cada filósofo:

Transcendente: Para filósofos que lidaram significativamente com questões metafísicas ou teológicas.

Antropológica: Para aqueles que se concentraram em questões humanas, éticas e sociais.

Imanente: Para pensadores com um foco no mundo material, físico ou na crítica da metafísica.

2.1 Interesse Transcendente

    Metafísica: Aborda questões da natureza da realidade, existência e o mundo além do físico.
    Teologia: Envolve discussões sobre divindade, religião e espiritualidade.
    Ética Normativa: Trata de princípios morais e a base do certo e errado.

2.2 Interesse Antropológico

    Filosofia Política: Enfoca na organização social, governo, justiça e direitos.
    Filosofia da Mente: Explora a natureza da mente, consciência e pensamento.
    Ética Aplicada: Lida com a aplicação prática de princípios éticos a situações específicas.

2.3 Interesse Imanente

    Lógica e Raciocínio: Foco na validade dos argumentos, estruturas e o processo de raciocínio.
    Epistemologia Empírica: Enfatiza a experiência sensorial como fonte de conhecimento.
    Estudos Sociais e Culturais: Explora aspectos da sociedade, cultura e relações humanas.

Cada filósofo foi avaliado nessas categorias e subcategorias, resultando em um perfil detalhado que reflete sua posição filosófica e a importância que deram a cada tema em suas obras. Esta classificação proporciona uma visão abrangente e multifacetada de cada pensador, destacando tanto suas ideias centrais quanto as nuances de seu pensamento.

```python
def cadastrar_filosofo(filosofos_info, nome_filosofo, nascimento, morte, obras=None):
    """
    Função para cadastrar um novo filósofo no dicionário filosofos_info com informações básicas.

    Parâmetros:
    filosofos_info (dict): Dicionário existente com informações dos filósofos.
    nome_filosofo (str): Nome do novo filósofo a ser cadastrado.
    nascimento (int): Ano de nascimento do filósofo.
    morte (int): Ano de morte do filósofo.
    obras (list): Lista de obras principais do filósofo.

    Retorna:
    dict: Dicionário atualizado com o novo filósofo.
    """
    filosofos_info[nome_filosofo] = {
        'nascimento': nascimento,
        'morte': morte,
        'obras': obras
    }
    return filosofos_info

def atualizar_relevancia_temas(filosofos_info, nome_filosofo, relevancia_temas):
    """
    Função para atualizar a relevância dos temas de um filósofo no dicionário filosofos_info.

    Parâmetros:
    filosofos_info (dict): Dicionário existente com informações dos filósofos.
    nome_filosofo (str): Nome do filósofo a ser atualizado.
    relevancia_temas (dict): Dicionário com as pontuações de relevância dos temas para o filósofo.

    Retorna:
    dict: Dicionário atualizado com as pontuações de relevância dos temas para o filósofo.
    """
    if nome_filosofo in filosofos_info:
        filosofos_info[nome_filosofo]['relevancia_temas'] = relevancia_temas
    else:
        print(f"Filósofo {nome_filosofo} não encontrado no dicionário.")
    return filosofos_info

def atualizar_polaridades_temas(polaridades_info, nome_filosofo, polaridades):
    for tema, polaridade in polaridades.items():
        if tema in polaridades_info:
            polaridades_info[tema][nome_filosofo] = polaridade
```

```python
def plot_themes(polaridades_info, filosofos_info, ax, classificacoes_indices, filosofos_ocultos):
    temas_por_categoria = {
        'Transcendente': ['Metafísica', 'Teologia', 'Ética Normativa'],
        'Antropológica': ['Filosofia Política', 'Filosofia da Mente', 'Ética Aplicada'],
        'Imanente': ['Lógica e Raciocínio', 'Epistemologia Empírica', 'Estudos Sociais e Culturais']
    }
    posicoes_bolhas = {}

    # Ajuste na divisão de cada subdivisão (altura)
    subdivisoes_altura = 1 / len(temas_por_categoria['Transcendente'])  # assumindo que cada categoria tem a mesma quantidade de temas

    for filosofo, info in filosofos_info.items():
        if filosofo not in filosofos_ocultos:
            pos_x_filosofo = (info['nascimento'] + info['morte']) / 2 / 100
            posicoes_bolhas[filosofo] = []

            for categoria, temas in temas_por_categoria.items():
                pos_y_base = classificacoes_indices[categoria]
                temas_ordenados = sorted(temas, key=lambda t: -abs(info['relevancia_temas'].get(t, 0)))[:3]

                for i, tema in enumerate(temas_ordenados):
                    relevancia = info['relevancia_temas'].get(tema, 0)
                    if relevancia != 0:
                        # Ajuste para centralizar cada bolha em sua subdivisão
                        pos_y_tema = pos_y_base + subdivisoes_altura * (i + 0.5) - 0.5
                        posicoes_bolhas[filosofo].append(pos_y_tema)
                        tamanho_bolha = 5 * abs(relevancia)
                        cor_bolha = 'green' if polaridades_info[tema].get(filosofo, 0) > 0 else 'red'
                        ax.plot(pos_x_filosofo, pos_y_tema, 'o', markersize=tamanho_bolha, color=cor_bolha, alpha=0.5)

    return posicoes_bolhas

def plot_thinkers(filosofos_info, polaridades_info, filosofos_ocultos=[]):
    import matplotlib.pyplot as plt
    import matplotlib.patches as mpatches
    import datetime

    # Cores para cada categoria
    cores = {
        'Imanente': 'gray',
        'Antropológica': 'blue',
        'Transcendente': 'yellow',
    }

    # Obtemos o ano corrente
    ano_corrente = datetime.datetime.now().year

    # Encontramos o menor ano de nascimento entre os filósofos
    menor_nascimento = min(filosofo['nascimento'] for filosofo in filosofos_info.values() if filosofo not in filosofos_ocultos)
    limite_inferior = (menor_nascimento // 100)
    limite_superior = ((ano_corrente // 100))

    # Criação da figura e do eixo
    fig, ax = plt.subplots(figsize=(30, 10))

    # Desenha as camadas horizontais de fundo para cada classificação
    classificacoes = ['Imanente', 'Antropológica', 'Transcendente']
    classificacoes_indices = {c: i for i, c in enumerate(classificacoes)}

    for i, classificacao in enumerate(classificacoes):
        ax.add_patch(mpatches.Rectangle((limite_inferior, i - 0.5), limite_superior - limite_inferior, 1, color=cores[classificacao], alpha=0.1))

    # Desenha as camadas verticais de fundo para cada século
    for i in range(limite_inferior, limite_superior):
        ax.axvline(x=i, color='grey', linestyle=':', alpha=0.5)

    posicoes_bolhas = plot_themes(polaridades_info, filosofos_info, ax, classificacoes_indices, filosofos_ocultos)

    # Adicionar linhas horizontais tênues para subdivisões
    subdivisoes_por_categoria = 3
    for i in range(subdivisoes_por_categoria * len(classificacoes)):
        ax.axhline(y=i / subdivisoes_por_categoria - 0.5, color='grey', linestyle=':', alpha=0.5)

    for filosofo, info in filosofos_info.items():
        if filosofo in filosofos_ocultos:
            continue  # Pula os filósofos na lista de ocultos
        
        pos_x_filosofo = (info['nascimento'] + info['morte']) / 2 / 100
        posicoes_filosofo = posicoes_bolhas.get(filosofo, [])

        if posicoes_filosofo:
            min_y, max_y = min(posicoes_filosofo), max(posicoes_filosofo)
            altura = max_y - min_y
            centro_y = (min_y + max_y) / 2
            elipse = mpatches.Ellipse((pos_x_filosofo, centro_y), 
                                      width=(info['morte'] - info['nascimento']) / 100, 
                                      height=altura + 0.2, edgecolor='blue', 
                                      facecolor='none', linestyle='dashed', alpha=0.5)
            ax.add_patch(elipse)
            plt.text(pos_x_filosofo, centro_y, filosofo, ha='center', va='center', rotation=90, color='blue')

    ax.set_xlim(limite_inferior, limite_superior)
    ax.set_ylim(-0.5, len(classificacoes) - 0.5)
    plt.xticks([x for x in range(limite_inferior, limite_superior)], [f"{x*100}" for x in range(limite_inferior, limite_superior)], rotation=90)
    plt.yticks(range(len(classificacoes)), classificacoes)

    plt.show()

# Lista de filósofos a serem ocultados
filosofos_ocultos = ['Diógenes de Sinope', 'Marco Aurélio', 'Clemente de Alexandria', 'Sexto Empírico', 
                     'Alberto Magno', 'Guilherme de Ockham', 'Dante Alighieri', 'Beda, o Venerável', 'Marsílio de Pádua', 'George Berkeley','Étienne Gilson']
plot_thinkers(filosofos_info, polaridades_info, filosofos_ocultos)

```
