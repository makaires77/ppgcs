```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Function to draw a single block with text
def draw_block(ax, center, text, width=1.5, height=0.6, arrow=None):
    # Create a rectangle
    rect = patches.Rectangle(
        (center[0] - width / 2, center[1] - height / 2), width, height, 
        linewidth=1, edgecolor='black', facecolor='none'
    )
    # Add the rectangle to the plot
    ax.add_patch(rect)
    
    # Add text to the rectangle
    plt.text(center[0], center[1], text, ha='center', va='center', fontsize=8)
    
    # Draw an arrow if specified
    if arrow:
        ax.arrow(*arrow['start'], *arrow['length'],
                 head_width=0.1, head_length=0.1, fc='black', ec='black')

# Define positions for each block (x, y)
positions = {
    'dfd': (1, 6),
    'pca': (3, 6),
    'ccl': (5, 6),
    'ploa': (3, 4),
    'procurement_unit': (3, 2),
    'technical_unit': (1, 2),
    'purchase_unit': (5, 2)
}

# Define arrows
arrows = {
    'arrow_1': {'start': (positions['dfd'][0], positions['dfd'][1] - 0.3), 'length': (0, -1.4)},
    'arrow_2': {'start': (positions['pca'][0], positions['pca'][1] - 0.3), 'length': (0, -1.4)},
    'arrow_3': {'start': (positions['ccl'][0], positions['ccl'][1] - 0.3), 'length': (0, -1.4)},
    'arrow_4': {'start': (positions['ploa'][0], positions['ploa'][1] - 0.3), 'length': (0, -1.4)},
    'arrow_5': {'start': (positions['procurement_unit'][0] + 0.75, positions['procurement_unit'][1]), 'length': (1.5, 0)},
    'arrow_6': {'start': (positions['technical_unit'][0] + 0.75, positions['technical_unit'][1]), 'length': (1.5, 0)},
    'arrow_7': {'start': (positions['purchase_unit'][0] - 0.75, positions['purchase_unit'][1]), 'length': (-1.5, 0)}
}

# Create figure and axis
fig, ax = plt.subplots(figsize=(18, 12))

# Set limits and remove axes
ax.set_xlim(0, 6)
ax.set_ylim(0, 8)
plt.axis('off')

# Draw blocks with placeholder text
for key, position in positions.items():
    draw_block(ax, position, key.upper())

# Draw arrows
for arrow in arrows.values():
    draw_block(ax, position, '', arrow=arrow)

plt.show()
```

```python
# Function to draw a single block with text
def draw_block(ax, x, y, width, height, text='', edgecolor='black'):
    # Create a rectangle
    rect = patches.Rectangle((x, y), width, height, linewidth=1, edgecolor=edgecolor, facecolor='none')
    # Add the rectangle to the plot
    ax.add_patch(rect)
    # Add text to the rectangle
    if text:
        plt.text(x + width / 2, y + height / 2, text, ha='center', va='center', fontsize=8)

# Function to draw an arrow between two points
def draw_arrow(ax, from_rect, to_rect):
    # Calculate arrow properties
    start_x = from_rect[0] + from_rect[2] / 2
    start_y = from_rect[1]
    end_x = to_rect[0] + to_rect[2] / 2
    end_y = to_rect[1] + to_rect[3]
    ax.annotate('', xy=(start_x, start_y), xytext=(end_x, end_y),
                arrowprops=dict(facecolor='black', arrowstyle='->', lw=1))

# Create figure and axis
fig, ax = plt.subplots(figsize=(19, 12))

# Set the aspect of the plot to equal
ax.set_aspect('equal')

# Set limits and remove axes
plt.axis('off')

# Define the size of each block
block_width = 1
block_height = 0.5

# Define the positions and sizes for the 5 columns
# The y-positions are inverted because the coordinate system starts from top-left
columns = {
    1: [(0, 3.5, 0.5, 2)],  # One vertical rectangle
    2: [(1, 2, block_width, block_height), (1, 3, block_width, block_height),
        (1, 4, block_width, block_height), (1, 5, block_width, block_height)],  # Four horizontal rectangles
    3: [(2, 3.5, 0.5, 2), (2, 2, block_width, block_height)],  # One vertical and one horizontal rectangle
    4: [(3, 2, block_width, block_height), (3, 3, block_width, block_height), 
        (3, 4, block_width, block_height)],  # Three horizontal rectangles
    5: [(4, 2, block_width, block_height), (4, 3, block_width, block_height), 
        (4, 4, block_width, block_height)]   # Three horizontal rectangles
}

# Draw the blocks
for col in columns.values():
    for rect in col:
        draw_block(ax, *rect)

# Draw the arrows between blocks
draw_arrow(ax, columns[1][0], columns[2][0])
draw_arrow(ax, columns[2][1], columns[3][0])
draw_arrow(ax, columns[3][1], columns[4][0])
draw_arrow(ax, columns[4][2], columns[5][2])

# Set the plot limits
plt.xlim(-0.5, 5.5)
plt.ylim(1.5, 7.5)

plt.show()
```

```python
# Create figure and axis
fig, ax = plt.subplots(figsize=(19, 12))

# Set the aspect of the plot to equal and remove axes
ax.set_aspect('equal')
plt.axis('off')

# Define the size of each block and spacing
block_width = 1
block_height = 0.5
vertical_spacing = 0.8   # Vertical space between blocks
horizontal_spacing = 0.5 # Horizontal space between columns
vertical_gap = 0.2       # Vertical space within a column between blocks

# Positions of the columns, starting x positions
column_positions = [1, 3 + horizontal_spacing, 5 + 2 * horizontal_spacing, 
                    7 + 3 * horizontal_spacing, 9 + 4 * horizontal_spacing]

# Draw the blocks for each column
for i, x_pos in enumerate(column_positions):
    # First column, one vertical block
    if i == 0:
        draw_block(ax, x_pos, 4, block_width, 2 * block_height)
    # Other columns
    else:
        for j in range(4 - i % 2):  # 4 blocks for even columns, 3 for odd ones after the first
            y_pos = 1 + j * (block_height + vertical_gap)
            draw_block(ax, x_pos, y_pos, block_width, block_height)

# Draw the arrows between columns
for i in range(len(column_positions) - 1):
    from_x = column_positions[i] + block_width / 2
    to_x = column_positions[i + 1] - block_width / 2
    # For the first column with vertical block, the arrow starts from the middle of the block
    if i == 0:
        from_y = 4 + block_height
    # For other columns, the arrow starts from the bottom of the first block
    else:
        from_y = 1 + block_height
    
    # Arrow to the top block of the next column
    to_y = 1 + 3 * (block_height + vertical_gap) - block_height / 2
    ax.annotate('', xy=(to_x, to_y), xytext=(from_x, from_y),
                arrowprops=dict(facecolor='black', arrowstyle='->', lw=1))

# Set the plot limits
plt.xlim(0, 11)
plt.ylim(0, 7)

plt.show()
```
