# Ilustração Modelo espiral

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

def plotar_espiral():
    # Configurações iniciais
    num_iters = 5  # Quantidade de fases
    num_points = 1000  # Número de pontos na espiral

    theta = np.linspace(0, 2 * np.pi * num_iters, num_points)
    z = np.linspace(0, 100, 1000)  # Height of the spiral
    radius = theta**1.35
    x = radius * np.cos(theta)
    y = radius * np.sin(theta)

    fig = plt.figure(figsize=(14, 14))
    ax = fig.add_subplot(111, projection='3d', facecolor='none')

    # Remove linhas de grade e eixos
    ax.grid(False)
    # ax.axis('off')

    # Segmentar a linha e colorir
    colors = ['cyan', 'blue', 'green', 'yellow', 'orange']
    segment_length = num_points // num_iters
    for i in range(num_iters):
        start_idx = i * segment_length
        end_idx = (i + 1) * segment_length if i < num_iters - 1 else num_points
        ax.plot(x[start_idx:end_idx], y[start_idx:end_idx], z[start_idx:end_idx],
                linewidth=16, color=colors[i % len(colors)])

    # Plotar a linha interna com estilo tracejado
    ax.plot(x, y, z, linestyle='--', linewidth=2, color='w')

    # Distância para mover os rótulos para a direita
    label_offset = 10

    # Criação dos rótulos de cada fase
    labels = ['Prototipar', 'Socializar', 'Externalizar', 'Combinar', 'Internalizar']
    for i, label in enumerate(labels):
        idx = i * segment_length
        # Calcular deslocamento baseado na direção do segmento
        label_x = x[idx] + label_offset * np.cos(theta[idx])
        label_y = y[idx] + label_offset * np.sin(theta[idx])
        ax.text(label_x, label_y, z[idx], label, fontsize=22, color='black')    

    # Adicionar rótulo 'Redefinição' no final da espiral
    label_x = x[-1] + label_offset * np.cos(theta[-1])
    label_y = y[-1] + label_offset * np.sin(theta[-1])
    ax.text(label_x, label_y, z[-1], 'Adaptar...', fontsize=24, color='black')

    # Ajustar a inclinação da visualização
    ax.view_init(elev=15, azim=300)  # azim ajusta a rotação em torno do eixo z

    # Adicionar rótulos aos eixos
    axis_fontsize=18
    labelpad=4
    ax.set_xlabel('Proposta de Valor', fontsize=axis_fontsize, labelpad=labelpad)
    ax.set_ylabel('Público-alvo', fontsize=axis_fontsize, labelpad=labelpad)
    ax.set_zlabel('Recursos', fontsize=axis_fontsize, labelpad=labelpad)

    plt.show()
```

```python
plotar_espiral()
```

## Prospecção Tecnológica

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def plot_tech_roadmap(tech_data, start_year, end_year):
    # Cores para cada camada
    cores = ['lightblue', 'lightgreen', 'lightcoral']

    # Criação da figura e do eixo
    fig, ax = plt.subplots(figsize=(20, 10))

    # Desenha as camadas horizontais de fundo
    for i, cor in enumerate(cores):
        ax.add_patch(mpatches.Rectangle((start_year, i), end_year - start_year, 1, color=cor, alpha=0.5))

    # Desenha as linhas das tecnologias
    for tech, info in tech_data.items():
        camada = info['camada']
        inicio = info['inicio']
        fim = info['fim']
        ax.plot([inicio, fim], [camada + 0.5, camada + 0.5], color='black', linewidth=2)
        ax.text(inicio, camada + 0.5, f"{tech} ({inicio})", va='bottom', ha='right', rotation=45)

    # Configurações do gráfico
    ax.set_xlim(start_year, end_year)
    ax.set_ylim(0, len(cores))
    ax.set_yticks([0.5, 1.5, 2.5])
    ax.set_yticklabels(['Conhecimentos', 'Competências', 'Tecnologias'])
    plt.grid(True, which='both', axis='x', color='gray', linestyle='--', linewidth=0.5)
    plt.title('Roadmap Tecnológico')

    plt.show()

# Exemplo de dados
tech_data = {
    'Tecnologia A': {'camada': 0, 'inicio': 2022, 'fim': 2025},
    'Tecnologia B': {'camada': 1, 'inicio': 2023, 'fim': 2026},
    'Tecnologia C': {'camada': 2, 'inicio': 2024, 'fim': 2027}
}

plot_tech_roadmap(tech_data, 2020, 2030)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def plot_tech_trl_roadmap(tech_data):
    # Definição das cores para cada TRL
    trl_colors = ['lightgray', 'silver', 'darkgray', 'rosybrown', 'sandybrown', 'peachpuff', 
                  'darkorange', 'orangered', 'red']
    
    # Preparação da figura e do eixo
    fig, ax = plt.subplots(figsize=(20, 10))

    # Desenha as camadas horizontais de fundo para cada TRL
    for i in range(9):
        ax.add_patch(mpatches.Rectangle((2020, i), 30, 1, color=trl_colors[i], alpha=0.3))
    
    # Adiciona linhas representando a evolução das TRLs para cada tecnologia
    for tech, info in tech_data.items():
        for stage in info:
            start_year, end_year, trl = stage['start'], stage['end'], stage['trl']
            ax.plot([start_year, end_year], [trl-1, trl-1], label=tech, linewidth=4)

    # Configurações do eixo
    ax.set_xlim(2020, 2050)
    ax.set_ylim(0, 8)
    plt.xticks([year for year in range(2020, 2051, 5)])
    plt.yticks(range(9), [f"TRL {i+1}" for i in range(9)])
    plt.xlabel("Ano")
    plt.ylabel("Nível de Maturidade Tecnológica (TRL)")
    plt.title("Roadmap de Evolução Tecnológica por TRL")
    ax.legend()

    plt.show()

# Dados de exemplo
tech_data_example = {
    'Tecnologia 1': [{'start': 2020, 'end': 2025, 'trl': 3}, {'start': 2025, 'end': 2030, 'trl': 5}],
    'Tecnologia 2': [{'start': 2022, 'end': 2026, 'trl': 2}, {'start': 2026, 'end': 2031, 'trl': 4}],
    'Tecnologia 3': [{'start': 2023, 'end': 2028, 'trl': 1}, {'start': 2028, 'end': 2032, 'trl': 3}]
}

plot_tech_trl_roadmap(tech_data_example)
```

```python
import plotly.graph_objects as go
from plotly.offline import plot

def plot_tech_trl_roadmap_plotly(tech_data):
    fig = go.Figure()

    # Definindo as cores de fundo para cada nível TRL
    colors = ['lightgreen'] * 3 + ['yellow'] * 3 + ['orange'] * 3

    # Adicionando retângulos coloridos para o fundo
    for i, color in enumerate(colors, start=1):
        fig.add_shape(type="rect",
                      xref="paper", yref="y",
                      x0=0, y0=i, x1=1, y1=i+1,
                      fillcolor=color,
                      opacity=0.2,
                      layer="below",
                      line_width=0)

    # Adicionando as linhas de dados
    for tech, data in tech_data.items():
        years = [year for year, _ in data]
        trls = [trl for _, trl in data]
        fig.add_trace(go.Scatter(x=years, y=trls, mode='lines+markers', name=tech))

    # Ajustando o layout
    fig.update_layout(
        title='Roadmap Tecnológico com TRLs',
        xaxis_title='Ano',
        yaxis_title='Nível de TRL',
        yaxis=dict(tickmode='array', tickvals=list(range(1, 10)), ticktext=[f'TRL {i}' for i in range(1, 10)]),
        showlegend=True
    )

    return fig

# Dados de exemplo para tecnologias
tech_data = {
    'Tech A': [(2020, 1), (2021, 2), (2022, 3), (2023, 4)],
    'Tech B': [(2020, 2), (2022, 3), (2024, 5), (2026, 7)],
    'Tech C': [(2020, 3), (2021, 4), (2022, 5), (2023, 6), (2024, 7)],
}

# Criar e mostrar o gráfico
fig = plot_tech_trl_roadmap_plotly(tech_data)
plot(fig, filename='tech_roadmap.html', auto_open=True, include_plotlyjs='cdn')
```

```python
import plotly.graph_objects as go
from plotly.offline import plot

def plot_tech_roadmap_graph(tech_data, start_year, end_year):
    fig = go.Figure()

    # Cores de fundo para cada camada
    colors = ['lightblue', 'lightgreen', 'lightcoral']

    # Altura de cada retângulo
    retangulo_altura = 0.3

    for tech, info in tech_data.items():
        inicio = info['inicio']
        fim = info['fim']
        camada = info['camada']
        nome = info['nome']

        # Calculando a posição central do retângulo
        y_center = camada + 0.5

        # Adiciona o retângulo
        fig.add_shape(type="rect",
                      xref="x", yref="y",
                      x0=inicio, y0=y_center - retangulo_altura/2, 
                      x1=fim, y1=y_center + retangulo_altura/2,
                      fillcolor=colors[camada],
                      opacity=0.6,
                      line=dict(color="black", width=2))

        # Adiciona o texto
        fig.add_annotation(x=(inicio+fim)/2, y=y_center,
                           text=nome,
                           showarrow=False,
                           yshift=10)

    # Adicionando as setas
    for tech, info in tech_data.items():
        if 'liga_com' in info:
            for destino in info['liga_com']:
                destino_info = tech_data[destino]
                fig.add_trace(go.Scatter(x=[info['fim'], destino_info['inicio']], 
                                         y=[info['camada']+0.5, destino_info['camada']+0.5],
                                         mode='lines',
                                         line=dict(color='black', width=1, dash='dot'),
                                         showlegend=False))

    # Ajustando o layout do gráfico
    fig.update_layout(
        title='Roadmap Tecnológico',
        xaxis=dict(range=[start_year, end_year], tickmode='linear'),
        yaxis=dict(range=[0, 3], tickvals=[0.5, 1.5, 2.5],
                   ticktext=['Conhecimentos', 'Competências', 'Tecnologias']),
        showlegend=False
    )

    return fig

# Exemplo de dados
tech_data = {
    'Conhecimento A': {'camada': 0, 'inicio': 2024, 'fim': 2026, 'nome': 'Know A', 'liga_com': ['Competência B']},
    'Competência B': {'camada': 1, 'inicio': 2026, 'fim': 2028, 'nome': 'Comp B', 'liga_com': ['Tecnologia C']},
    'Tecnologia C': {'camada': 2, 'inicio': 2028, 'fim': 2030, 'nome': 'Tech C'},
}

# Criar e mostrar o gráfico
fig = plot_tech_roadmap_graph(tech_data, 2024, 2030)
plot(fig, filename='tech_roadmap_graph.html', auto_open=True, include_plotlyjs='cdn')
```

```python
import subprocess
import io
from IPython.display import display, SVG

def render_plantuml_inline(uml_code: str, jar_path: str = "plantuml-gplv2-1.2023.12.jar"):
    # Salve o código PlantUML temporariamente
    with open("temp.uml", "w") as f:
        f.write(uml_code)
    
    # Execute o PlantUML e capture a saída SVG
    result = subprocess.run(["java", "-jar", jar_path, "-pipe", "-tsvg"], 
                            input=uml_code.encode('utf-8'), 
                            stdout=subprocess.PIPE, 
                            stderr=subprocess.PIPE)
    
    if result.returncode != 0:
        raise Exception("Erro ao executar PlantUML: " + result.stderr.decode('utf-8'))
    
    # Converta a saída SVG em BytesIO para exibição
    svg_content = io.BytesIO(result.stdout).getvalue().decode('utf-8')
    return SVG(svg_content)

# Seu código PlantUML
uml_code = """
@startuml

skinparam package {
    borderColor black
    backgroundColor white
    shadowing 0
    fontColor black
    borderThickness 1
    fontSize 12
}

skinparam rectangle {
    borderColor black
    backgroundColor #EDEDED
    shadowing 0
    fontColor black
    borderThickness 1
    fontSize 12
}

package "Feature extraction module" {
    [Input] -down-> rectangle "Dilated convolution D=2" as DC1
    DC1 -down-> rectangle "Channel attentional" as CA1
    DC1 -right-> rectangle "W1" as W1
    W1 -down-> rectangle "1x1 Conv" as Conv1
    Conv1 -right-> rectangle "Z1" as Z1
    CA1 -down-> rectangle "Spatial attentional" as SA1

    rectangle "Dilated convolution D=3" as DC2
    DC1 -right-> DC2
    DC2 -down-> rectangle "Channel attentional" as CA2
    DC2 -right-> rectangle "W2" as W2
    W2 -down-> rectangle "1x1 Conv" as Conv2
    Conv2 -right-> rectangle "Z2" as Z2
    CA2 -down-> rectangle "Spatial attentional" as SA2

    rectangle "Dilated convolution D=4" as DC3
    DC2 -right-> DC3
    DC3 -down-> rectangle "Channel attentional" as CA3
    DC3 -right-> rectangle "W3" as W3
    W3 -down-> rectangle "1x1 Conv" as Conv3
    Conv3 -right-> rectangle "Z3" as Z3
    CA3 -down-> rectangle "Spatial attentional" as SA3
}

package "Category relationship discrimination module" {
    [Wear safety helmet] as helmet
    [Saw the workpiece] as saw
    [..] as dot
    [Pick up parts] as pick
    [Clean up] as clean

    rectangle "Fully Connect Layer" as FCL

    helmet --> FCL
    saw --> FCL
    dot --> FCL
    pick --> FCL
    clean --> FCL

    FCL -right-> rectangle "Dynamic graph convolution" as DGC
    FCL -right-> rectangle "Static Graph convolution" as SGC
}

@enduml
"""

display(render_plantuml_inline(uml_code))
```
