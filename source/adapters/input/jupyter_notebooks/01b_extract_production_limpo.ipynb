```python
lista_restante = lista_busca[:]
print(f'{len(lista_busca)} currículos buscados')
print(f'{len(dom_dict_list)} currículos extraídos com sucesso')
total_extraidos = 0
total_nao_extraidos = 0

# Função para normalizar os nomes
def normalizar_nome(nome):
    # Normaliza o nome para comparar de forma mais flexível
    nome_normalizado = nome.lower().replace(' ', '').replace('á', 'a').replace('é', 'e').replace('í', 'i').replace('ó', 'o').replace('ú', 'u').replace('â', 'a').replace('ê', 'e').replace('ô', 'o').replace('ã', 'a').replace('õ', 'o').replace('ç', 'c').replace('ü', 'u')
    return nome_normalizado

# Lista para armazenar nomes extraídos com sucesso
nomes_extraidos = []

for i in dom_dict_list:
    nome = i.get('Identificação').get('Nome')
    nome_normalizado = normalizar_nome(nome)
    encontrado = False

    # Verifica se o nome ou uma forma similar já foi extraído
    for nome_original in lista_restante:
        nome_original_normalizado = normalizar_nome(nome_original)
        if nome_original_normalizado == nome_normalizado:
            lista_restante.remove(nome_original)
            encontrado = True
            break

    # Incrementa o contador correspondente
    if encontrado:
        total_extraidos += 1
    else:
        total_nao_extraidos += 1

    # Imprime o status da extração
    if encontrado:
        print(f'({total_extraidos:>02}) {nome}')
    else:
        print(f'({total_nao_extraidos}) Não extraído: {nome}')
    # Adiciona à lista de nomes extraídos apenas se não for exatamente igual ao buscado
    if encontrado and nome_original_normalizado != nome_normalizado:
        nomes_extraidos.append(nome)

print(f'\n{len(lista_restante)} currículos não extraídos')

for i in lista_restante:
    print(f'   {i}')
```

```python
import logging
from neo4j import GraphDatabase
from neo4j.exceptions import Neo4jError

logging.basicConfig(level=logging.INFO, format='%(asctime)s %(levelname)s: %(message)s', filename='logs/persister.log')
logger = logging.getLogger(__name__)

class Neo4jPersister:
    def __init__(self, uri, user, password):
        self._uri = uri
        self._user = user
        self._password = password
        self._driver = GraphDatabase.driver(self._uri, auth=(self._user, self._password))
        self.logger = logger

        # Devem ser persistidos como 
        self.tipos = ['Identificação',
                      'Idiomas',
                      'Formação',
                      'Atuação Profissional',
                      'Linhas de Pesquisa',
                      'Áreas',
                      'Produções',
                      'ProjetosPesquisa',
                      'ProjetosExtensão',
                      'ProjetosDesenvolvimento',
                      'ProjetosOutros',
                      'Bancas',
                      'Orientações',
                      ]

        self.subtipos= ['Acadêmica',
                        'Pos-Doc',
                        'Complementar',
                        'Artigos completos publicados em periódicos',
                        'Resumos publicados em anais de congressos',
                        'Apresentações de Trabalho',
                        'Outras produções bibliográficas',
                        'Entrevistas, mesas redondas, programas e comentários na mídia',
                        'Concurso público',
                        'Outras participações',
                        'Livros publicados/organizados ou edições',
                        'Capítulos de livros publicados',
                        'Resumos expandidos publicados em anais de congressos',
                        'Resumos publicados em anais de congressos (artigos)',
                        'Trabalhos técnicos',
                        'Demais trabalhos',
                        'Mestrado',
                        'Teses de doutorado',
                        'Qualificações de Doutorado',
                        'Qualificações de Mestrado',
                        'Monografias de cursos de aperfeiçoamento/especialização',
                        'Trabalhos de conclusão de curso de graduação',
                        'Orientações e supervisões concluídas',
                        'Citações',
                        'Trabalhos completos publicados em anais de congressos',
                        'Produtos tecnológicos',
                        'Artigos  aceitos para publicação',
                        'Assessoria e consultoria',
                        'Programas de computador sem registro',
                        'Professor titular',
                        'Avaliação de cursos',
                        'Orientações e supervisões em andamento',
                        'Processos ou técnicas',
                        'Outras produções artísticas/culturais',
                        'Textos em jornais de notícias/revistas',
                        'Redes sociais, websites e blogs',
                        'Artes Visuais'            
                        ]

        self.propriedades = ['Nome',
                             'ID Lattes',
                             'Última atualização',
                             ]
    # def run(self, query, **parameters):
    #     with self.driver.session() as session:
    #         return session.run(query, **parameters)
        
    def close(self):
        self._driver.close()
    
    def persistir_revistas_da_planilha(self):
        """
        Persiste dados de revistas a partir da planilha 'classificações_publicadas_todas_as_areas_avaliacao1672761192111.xlsx' no Neo4j.

        Args:
            session: Objeto de sessão do Neo4j.
        """
        # Leitura da planilha
        dados_qualis = pd.read_excel(os.path.join(LattesScraper.find_repo_root(),'data','classificações_publicadas_todas_as_areas_avaliacao1672761192111.xlsx'))

        # Extração e persistência de dados de revista
        with self._driver.session() as session:
            for index, row in dados_qualis.iterrows():
                issn = row['ISSN'].replace('-','')
                nome_revista = row['Título']
                area_avaliacao = row['Área de Avaliação']
                estrato = row['Estrato']

                # Verificação de existência da revista
                revista_node = session.run("""
                    MATCH (j:Revista {issn: $issn})
                    RETURN j
                """, issn=issn).single()

                if not revista_node:
                    # Criação da revista se não existir
                    session.run("""
                        CREATE (j:Revista {issn: $issn, nome_revista: $nome_revista, area_avaliacao: $area_avaliacao, estrato: $estrato})
                    """, nome_revista=nome_revista, issn=issn, area_avaliacao=area_avaliacao,  estrato=estrato)

    # Testes Ok! 
    def persist_pessoa_nodes(self, dict_list):
        query_pessoa = """
        MERGE (p:Pesquisador {id_lattes: $id_lattes})
        ON CREATE SET p.nome = $nome, p.ultima_atualizacao = $ultima_atualizacao
        ON MATCH SET p.nome = $nome, p.ultima_atualizacao = $ultima_atualizacao
        """
        try:
            with self._driver.session() as session:
                for item in dict_list:
                    identificacao = item.get('Identificação')
                    nome = identificacao.get('Nome')
                    id_lattes = identificacao.get('ID Lattes')
                    ultima_atualizacao = identificacao.get('Última atualização')
                    if nome:
                        session.run(query_pessoa, id_lattes=id_lattes, nome=nome, ultima_atualizacao=ultima_atualizacao)
        except Exception as e:
            self.logger.error('Erro ao criar node "Pesquisador": {}'.format(e))

    # Testes Ok!         
    def persist_pesquisador_grande_area_relationships(self, dict_list):
        query_rel_pessoa_grande_area = """
        MATCH (p:Pesquisador {id_lattes: $id_lattes})
        MATCH (ga:GrandeArea {nome: $grande_area_nome})
        MERGE (p)-[:ATUA_EM]->(ga)
        """

        with self._driver.session() as session:
            for item in dict_list:
                identificacao = item.get('Identificação')
                id_lattes = identificacao.get('ID Lattes')
                areas = item.get('Áreas').values()
                for area_string in areas:
                    grande_area_nome, _, _ = self.extract_area_info(area_string)
                    if grande_area_nome:
                        session.run(query_rel_pessoa_grande_area, id_lattes=id_lattes, grande_area_nome=grande_area_nome)

    # Testes Ok! 
    def persist_areas_nodes(self, dict_list):
        query_grande_area = """
        MERGE (ga:GrandeArea {nome: $nome})
        """
        query_area = """
        MATCH (ga:GrandeArea {nome: $grande_area_nome})
        MERGE (a:Area {nome: $nome}) ON CREATE SET a:Area
        MERGE (ga)-[:CONTEM]->(a)
        """
        query_subarea = """
        MATCH (a:Area {nome: $area_nome})
        MERGE (sa:Subarea {nome: $nome}) ON CREATE SET sa:Subarea
        MERGE (a)-[:CONTEM]->(sa)
        """
        query_rel_pessoa_grande_area = """
        MATCH (p:Pesquisador {id_lattes: $id_lattes})
        MATCH (ga:GrandeArea {nome: $grande_area_nome})
        MERGE (p)-[:ATUA_EM]->(ga)    
        """

        with self._driver.session() as session:
            for item in dict_list:
                areas = item.get('Áreas').values()
                for area_string in areas:
                    grande_area_nome, area_nome, subarea_nome = self.extract_area_info(area_string)
                    
                    # Verificar se o nome não está vazio
                    if grande_area_nome:
                        session.run(query_grande_area, nome=grande_area_nome)
                    if area_nome:
                        session.run(query_area, grande_area_nome=grande_area_nome, nome=area_nome)
                    if subarea_nome:
                        session.run(query_subarea, area_nome=area_nome, nome=subarea_nome)
                    
                    # Adicionar relacionamento Pesquisador - GrandeÁrea
                    id_lattes = item['Identificação']['ID Lattes']
                    if grande_area_nome:
                        session.run(query_rel_pessoa_grande_area, id_lattes=id_lattes, grande_area_nome=grande_area_nome)

    # Testes Ok! 
    @staticmethod
    def extract_area_info(area_string):
        # Extraindo os nomes de GrandeÁrea, Área e Subárea da string
        try:
            grande_area_nome = area_string.split('/')[0].strip().split(': ')[1]
        except:
            grande_area_nome = ''
        try:
            area_nome = area_string.split('/')[1].strip().split(': ')[1]
        except:
            area_nome = ''
        try:
            subarea_nome = area_string.split('/')[2].strip().split(': ')[1]
        except:
            subarea_nome = ''
        return grande_area_nome, area_nome, subarea_nome

    ## PRODUÇÕES
    def persist_producoes_pesquisador(self, dict_list):
        with self._driver.session() as session:
            for pesq in dict_list:
                identificacao = pesq.get('Identificação')
                id_lattes = identificacao.get('ID Lattes')
                producoes = pesq.get('Produções')

                if not isinstance(producoes, dict):
                    print(f"Erro!! Dicionário da seção 'Produções' não encontrado para {id_lattes}")
                    continue

                for chave_producao, valores_producao in producoes.items():
                    print(f'{chave_producao} | {valores_producao}')
                    if chave_producao == 'Artigos completos publicados em periódicos':
                        # self.persistir_artigos_completos(session, id_lattes, valores_producao)
                        self.persistir_artigos_revistas(session, id_lattes, valores_producao)

    def _get_or_create_node(self, session, label, properties):
        properties = [x.rstrip('.') for x in properties]
        node = session.run("MATCH (n: {label}) WHERE {properties} RETURN n", {"label": label, "properties": properties}).single()

        if not node:
            node = session.run("CREATE (n: {label} {properties}) RETURN n", {"label": label, "properties": properties}).single()["n"]
            self._node_created_count += 1

        return node

    def persist_tipo_producao(self, session, id_lattes, tipo_producao):
        query_create_node = """
        MERGE (p:Pesquisador {id_lattes: $id_lattes})
        MERGE (t:TipoProducao {nome: $tipo_producao})
        MERGE (p)-[:PRODUZ]->(t)
        """
        result = session.run(query_create_node, id_lattes=id_lattes, tipo_producao=tipo_producao)
        summary = result.consume()
        return summary.counters.nodes_created, summary.counters.nodes_deleted, summary.counters.relationships_created, summary.counters.relationships_deleted

    def persist_subtipo_producao(self, session, id_lattes, tipo_producao, subtipo_producao, dados_producao):
        def checar_e_serializar(dados):
            """ Verifica e serializa dicionários recursivamente """
            if isinstance(dados, dict):
                for chave, valor in dados.items():
                    if isinstance(valor, dict):
                        dados[chave] = json.dumps(valor)
                    # Checagem adicional para outros tipos inválidos, se necessário
            return dados
        # Serialização recursiva do dicionário
        dados_producao = checar_e_serializar(dados_producao)

        query_create_node = """
        MERGE (p:Pesquisador {id_lattes: $id_lattes})
        MERGE (t:TipoProducao {nome: $tipo_producao})
        MERGE (s:SubtipoProducao {nome: $subtipo_producao})
        MERGE (p)-[PRODUZ:]->(t)-[:DO_TIPO]->(s)
        """

        if subtipo_producao in ["ArtigoCompleto", "ResumoCongresso", "ApresentacaoTrabalho", "OutrasProducoesBibliograficas"]:
            query_create_node += """
            MERGE (o:Ocorrencia {tipo: $subtipo_producao, dados: $dados})
            MERGE (s)-[:OCORRENCIA]->(o)
            """

        result = session.run(query_create_node, id_lattes=id_lattes, tipo_producao=tipo_producao, 
                            subtipo_producao=subtipo_producao, dados=dados_producao)

        # Obtendo as informações de contadores
        summary = result.consume()

        return summary.counters.nodes_created, summary.counters.nodes_deleted, summary.counters.relationships_created, summary.counters.relationships_deleted

    def persistir_artigos_completos(self, session, id_lattes, dados):
        created_nodes = 0
        updated_nodes = 0
        created_relations = 0
        updated_relations = 0
        
        for dados_artigo in dados:
            dados_artigo['dados'] = json.dumps(dados_artigo)  # Conversão para JSON da estrutura completa
            query_create_node_artigo = """
                MERGE (p:Pesquisador {id_lattes: $id_lattes})
                CREATE (a:ArtigoPublicado {
                    ano: $ano,
                    fator_impacto_jcr: $fator_impacto_jcr,
                    ISSN: $ISSN,
                    titulo: $titulo,
                    revista: $revista,
                    autores: $autores,
                    Qualis: $Qualis,
                    DOI: $DOI,
                    dados: $dados_artigo
                })
                CREATE (p)-[:PRODUZ]->(a)

                MERGE (j:Revista {nome: $revista, issn: $ISSN})
                CREATE (a)-[:PUBLICADO_EM]->(j)
            """

            result_artigo = session.run(query_create_node_artigo, 
                                id_lattes=id_lattes, 
                                ano=dados_artigo['ano'],
                                fator_impacto_jcr=dados_artigo['fator_impacto_jcr'],
                                ISSN=dados_artigo['ISSN'],
                                titulo=dados_artigo['titulo'],
                                revista=dados_artigo['revista'],
                                autores=dados_artigo['autores'],
                                Qualis=dados_artigo['Qualis'],
                                DOI=dados_artigo['DOI'],
                                dados_artigo=dados_artigo
                                )
            summary_artigo = result_artigo.consume()
            created_nodes += summary_artigo.counters.nodes_created
            updated_nodes += summary_artigo.counters.nodes_deleted
            created_relations += summary_artigo.counters.relationships_created
            updated_relations += summary_artigo.counters.relationships_deleted

        return created_nodes, updated_nodes, created_relations, updated_relations


    def persistir_artigos_completos(self, session, id_lattes, dados):
        created_nodes = 0
        updated_nodes = 0
        created_relations = 0
        updated_relations = 0
                
        for dados_artigo in dados:
            ano = dados_artigo['ano']
            impact_jcr = dados_artigo['fator_impacto_jcr']
            issn = dados_artigo['ISSN']
            titulo = dados_artigo['titulo']
            revista = dados_artigo['revista']
            autores = dados_artigo['autores']
            qualis = dados_artigo['Qualis']
            doi = dados_artigo['DOI']

            query_create_node_artigo = """
                MERGE (p:Pesquisador {id_lattes: $id_lattes})
                CREATE (a:ArtigoPublicado {ano: $ano, impact_jcr: $impact_jcr, issn: $issn, titulo: $titulo, revista: $revista, autores: $autores, qualis: $qualis, doi: $doi})
                CREATE (p)-[:PRODUZ]->(a)
                MERGE (j:Revista {nome: $revista, issn: $issn})
                CREATE (a)-[:PUBLICADO_EM]->(j)
            """
            print(query_create_node_artigo)
            result_artigo = session.run(query_create_node_artigo, 
                                id_lattes=id_lattes, 
                                ano=ano,
                                impact_jcr=impact_jcr,
                                issn=issn,
                                titulo=titulo,
                                revista=revista,
                                autores=autores,
                                qualis=qualis,
                                doi=doi,
                                )
            summary_artigo = result_artigo.consume()
            created_nodes += summary_artigo.counters.nodes_created
            updated_nodes += summary_artigo.counters.nodes_deleted
            created_relations += summary_artigo.counters.relationships_created
            updated_relations += summary_artigo.counters.relationships_deleted

        return created_nodes, updated_nodes, created_relations, updated_relations

    ## Agrupando nós de artigos por revista
    # def persistir_artigos_revistas(self, session, id_lattes, dados):
    #     created_nodes = 0
    #     updated_nodes = 0
    #     created_relations = 0
    #     updated_relations = 0

    #     for dados_artigo in dados:
    #         ano = dados_artigo['ano']
    #         impact_jcr = dados_artigo['fator_impacto_jcr']
    #         issn = dados_artigo['ISSN']
    #         titulo = dados_artigo['titulo']
    #         revista = dados_artigo['revista']
    #         autores = dados_artigo['autores']
    #         qualis = dados_artigo['Qualis']
    #         doi = dados_artigo['DOI']
            
    #         query_create_node_artigo = """
    #             MERGE (p:Pesquisador {id_lattes: $id_lattes})
    #             CREATE (a:ArtigoPublicado {ano: $ano, impact_jcr: $impact_jcr, issn: $issn, titulo: $titulo, revista: $revista, autores: $autores, qualis: $qualis, doi: $doi})
    #             CREATE (p)-[:PRODUZ]->(a)
    #             MERGE (j:Revista {nome: $revista, issn: $issn})
    #             CREATE (a)-[:PUBLICADO_EM]->(j)
    #         """

    #         with session.begin_transaction() as tx:
    #             print("DEBUG: ISSN da revista:", issn)
    #             # Verificação de existência da revista
    #             revista_node = tx.run("""
    #                 MATCH (j:Revista {issn: $issn})
    #                 RETURN j
    #             """, issn=issn).single()

    #             if not revista_node:
    #                 # Revista não encontrada, crie-a
    #                 tx.run("""
    #                     CREATE (j:Revista {nome: $revista, issn: $issn})
    #                 """, revista=revista, issn=issn)

    #                 revista_node = tx.run("""
    #                     MATCH (j:Revista {issn: $issn})
    #                     RETURN j
    #                 """, issn=issn).single()

    #             if revista_node is not None:
    #                 print("DEBUG: Propriedades do nó Revista:", revista_node.items())  
    #                 node_revista = revista_node[0]  # Extrair o objeto Node
    #                 print(f"revista_node[0] {node_revista}")
    #                 issn = node_revista['issn']  # Acessar a propriedade 'issn'
    #                 print(f"node_revista['issn'] {issn}")
    #                 revista_nome = node_revista['nome_revista']
    #                 revista_area_avaliacao = node_revista['area_avaliacao']
    #                 revista_estrato = node_revista['estrato']
    #             else:
    #                 print("Erro: O retorno para a revista com ISSN", issn, "é None.")

    #             # Criação do nó do artigo
    #             tx.run(query_create_node_artigo, 
    #                 id_lattes=id_lattes, 
    #                 ano=ano,
    #                 impact_jcr=impact_jcr,
    #                 issn=issn,
    #                 titulo=titulo,
    #                 revista=revista,
    #                 autores=autores,
    #                 qualis=qualis,
    #                 doi=doi)

    #             # Criação do relacionamento PUBLICADO_EM
    #             tx.run("""
    #                 MATCH (a:ArtigoPublicado {doi: $doi}), (j:Revista {nome_revista: $revista_nome, issn: $issn, area_avaliacao: $revista_area_avaliacao, estrato: $revista_estrato})
    #                 CREATE (a)-[:PUBLICADO_EM]->(j)
    #             """, doi=doi, revista_nome=revista_nome, issn=issn, revista_area_avaliacao=revista_area_avaliacao, revista_estrato=revista_estrato)

    #             tx.commit()

    #         # Atualização dos contadores
    #         created_nodes += tx.run("MATCH (n) WHERE n:ArtigoPublicado RETURN count(n)").single()[0]
    #         updated_nodes += tx.run("MATCH (n) WHERE n:ArtigoPublicado SET n.updated_at = datetime() RETURN count(n)").single()[0]
    #         created_relations += tx.run("MATCH (r) WHERE r:PUBLICADO_EM RETURN count(r)").single()[0]
    #         updated_relations += tx.run("MATCH (r) WHERE r:PUBLICADO_EM SET r.updated_at = datetime() RETURN count(r)").single()[0]

    #     return created_nodes, updated_nodes, created_relations, updated_relations

    def buscar_revista_por_issn(session, issn):
        query = """
            MATCH (revista:Revista {issn: $issn})
            RETURN revista
        """

        try:
            result = session.run(query, issn=issn)
            return result.single()
        except Neo4jError as e:
            print(f"Erro Neo4j ao buscar a revista por ISSN: {e}")
            return None
        
    def persistir_artigos_revistas(self, session, id_lattes, dados):
        """
        Função para persistir os dados de artigos completos publicados em periódicos.

        Args:
            session (neo4j.Session): Sessão Neo4j.
            id_lattes (str): ID do Lattes do pesquisador.
            dados (dict): Dicionário contendo os dados dos artigos.

        Returns:
            None
        """

        for artigo in dados:
            # Extraindo informações do artigo
            ano = artigo['ano']
            impact_jcr = artigo['fator_impacto_jcr']
            issn = artigo['ISSN']
            titulo = artigo['titulo']
            revista = artigo['revista']
            autores = artigo['autores']
            data_issn = artigo['data_issn']
            doi = artigo['DOI']
            qualis = artigo['Qualis']

            query_create_node_artigo = """
                MERGE (p:Pesquisador {id_lattes: $id_lattes})
                CREATE (a:ArtigoPublicado {ano: $ano, impact_jcr: $impact_jcr, issn: $issn, titulo: $titulo, revista: $revista, autores: $autores, qualis: $qualis, doi: $doi})
                CREATE (p)-[:PRODUZ]->(a)
                MERGE (j:Revista {nome: $revista, issn: $issn})
                CREATE (a)-[:PUBLICADO_EM]->(j)
            """
            
            # Buscando o nó da revista
            revista_node = self.buscar_revista_por_issn(session, issn)

            # Criando o nó do artigo
            with session.begin_transaction() as tx:
                tx.run(query_create_node_artigo,
                    id_lattes=id_lattes,
                    ano=ano,
                    impact_jcr=impact_jcr,
                    issn=issn,
                    titulo=titulo,
                    revista=revista,
                    autores=autores,
                    data_issn=data_issn,
                    doi=doi,
                    qualis=qualis
                    )

                # Criando o relacionamento PUBLICADO_EM
                if revista_node is not None:
                    node_revista = revista_node[0][1]
                    if node_revista is not None:
                        revista_nome = node_revista['nome_revista']
                        revista_issn = node_revista['issn']
                        revista_area_avaliacao = node_revista['area_avaliacao']
                        revista_estrato = node_revista['estrato']
                    else:
                        print("Erro: O nó da revista não foi encontrado para o ISSN", issn)
                        # Lógica de tratamento de erro (opcional)
                else:
                    print("Erro: O retorno para a revista com ISSN", issn, "é None.")

                if revista_nome:
                    tx.run("""
                        MATCH (a:ArtigoPublicado {doi: $doi}), (j:Revista {nome_revista: $revista_nome, issn: $revista_issn, area_avaliacao: $revista_area_avaliacao, estrato: $revista_estrato})
                        CREATE (a)-[:PUBLICADO_EM]->(j)
                    """, doi=doi, revista_nome=revista_nome, revista_issn=revista_issn, revista_area_avaliacao=revista_area_avaliacao, revista_estrato=revista_estrato)

                tx.commit()

        # Atualização dos contadores
        with session.begin_transaction() as tx:
            created_nodes += tx.run("MATCH (n) WHERE n:ArtigoPublicado RETURN count(n)").single()[0]
            updated_nodes += tx.run("MATCH (n) WHERE n:ArtigoPublicado SET n.updated_at = datetime() RETURN count(n)").single()[0]
            created_relations += tx.run("MATCH (r) WHERE r:PUBLICADO_EM RETURN count(r)").single()[0]



    def persistir_resumos_congressos(self, session, id_lattes, dados):
        query_create_node = """
        MERGE (p:Pesquisador {id_lattes: $id_lattes})
        MERGE (r:ResumoCongresso {titulo: $titulo, ano: $ano, evento: $evento, autores: $autores, data_issn: $data_issn, doi: $doi})
        MERGE (p)-[:PRODUZ]->(r)
        """
        result = session.run(query_create_node, id_lattes=id_lattes, **dados)
        # Obtendo as informações de contadores
        summary = result.consume()
        created_nodes = summary.counters.nodes_created
        updated_nodes = summary.counters.nodes_deleted  
        created_relations = summary.counters.relationships_created
        updated_relations = summary.counters.relationships_deleted 

        return created_nodes, updated_nodes, created_relations, updated_relations

    def persistir_apresentacoes_trabalho(self, session, id_lattes, dados):
        query_create_node = """
        MERGE (p:Pesquisador {id_lattes: $id_lattes})
        MERGE (a:ApresentacaoTrabalho {
            titulo: $titulo,
            ano: $ano,
            evento: $evento,
            autores: $autores
        })
        MERGE (p)-[:PRODUZ]->(a)
        """
        result = session.run(query_create_node, id_lattes=id_lattes, **dados)

        # Obtendo as informações de contadores
        summary = result.consume()
        created_nodes = summary.counters.nodes_created
        updated_nodes = summary.counters.nodes_deleted  
        created_relations = summary.counters.relationships_created
        updated_relations = summary.counters.relationships_deleted 

        return created_nodes, updated_nodes, created_relations, updated_relations

    def persistir_outras_producoes_bibliograficas(self, session, id_lattes, dados):
        query_create_node = """
        MERGE (p:Pesquisador {id_lattes: $id_lattes})
        MERGE (o:OutrasProducoesBibliograficas {
            titulo: $titulo,
            ano: $ano,
            autores: $autores,
            doi: $doi
        })
        MERGE (p)-[:PRODUZ]->(o)
        """
        result = session.run(query_create_node, id_lattes=id_lattes, **dados)

        # Obtendo as informações de contadores
        summary = result.consume()
        created_nodes = summary.counters.nodes_created
        updated_nodes = summary.counters.nodes_deleted  
        created_relations = summary.counters.relationships_created
        updated_relations = summary.counters.relationships_deleted 

        return created_nodes, updated_nodes, created_relations, updated_relations

    def persistir_orientacoes_concluidas(self, session, id_lattes, dados):
        query_create_node = """
        MERGE (p:Pesquisador {id_lattes: $id_lattes})
        MERGE (o:OrientacaoConcluida {
            tipo: $tipo,
            titulo: $titulo,
            ano: $ano,
            autor: $autor,
            instituicao: $instituicao
        })
        MERGE (p)-[:ORIENTA]->(o)
        """
        result = session.run(query_create_node, id_lattes=id_lattes, **dados)

        # Obtendo as informações de contadores
        summary = result.consume()
        created_nodes = summary.counters.nodes_created
        updated_nodes = summary.counters.nodes_deleted  
        created_relations = summary.counters.relationships_created
        updated_relations = summary.counters.relationships_deleted 

        return created_nodes, updated_nodes, created_relations, updated_relations

    def persistir_participacoes_bancas(self, session, id_lattes, dados):
        query_create_node = """
        MERGE (p:Pesquisador {id_lattes: $id_lattes})
        MERGE (b:Banca {
            tipo: $tipo,
            titulo: $titulo,
            ano: $ano,
            instituicao: $instituicao
        })
        MERGE (p)-[:PARTICIPA_BANCA]->(b)
        """
        result = session.run(query_create_node, id_lattes=id_lattes, **dados)

        # Obtendo as informações de contadores
        summary = result.consume()
        created_nodes = summary.counters.nodes_created
        updated_nodes = summary.counters.nodes_deleted  
        created_relations = summary.counters.relationships_created
        updated_relations = summary.counters.relationships_deleted 

        return created_nodes, updated_nodes, created_relations, updated_relations

    def persistir_projetos_pesquisa(self, session, id_lattes, dados):
        query_create_node = """
        MERGE (p:Pesquisador {id_lattes: $id_lattes})
        MERGE (pr:ProjetoPesquisa {
            titulo: $titulo,
            ano_inicio: $ano_inicio,
            ano_fim: $ano_fim,
            agencia_financiadora: $agencia_financiadora,
            valor_financiamento: $valor_financiamento
        })
        MERGE (p)-[:COORDENA]->(pr)
        """
        result = session.run(query_create_node, id_lattes=id_lattes, dados=dados)

        # Obtendo as informações de contadores
        summary = result.consume()
        created_nodes = summary.counters.nodes_created
        updated_nodes = summary.counters.nodes_deleted  
        created_relations = summary.counters.relationships_created
        updated_relations = summary.counters.relationships_deleted 

        return created_nodes, updated_nodes, created_relations, updated_relations

    def persistir_premios_distincoes(self, session, id_lattes, dados):
        query_create_node = """
        MERGE (p:Pesquisador {id_lattes: $id_lattes})
        MERGE (pd:PremioDistincao {
            titulo: $titulo,
            ano: $ano,
            instituicao: $instituicao,
        })
        MERGE (p)-[:RECEBE]->(pd)
        """
        result = session.run(query_create_node, id_lattes=id_lattes, **dados)

        # Obtendo as informações de contadores
        summary = result.consume()
        created_nodes = summary.counters.nodes_created
        updated_nodes = summary.counters.nodes_deleted  
        created_relations = summary.counters.relationships_created
        updated_relations = summary.counters.relationships_deleted 

        return created_nodes, updated_nodes, created_relations, updated_relations

uri = "bolt://localhost:7687"
user = "neo4j"
password = "password"
persister = Neo4jPersister(uri, user, password)
```

```python
import json
import logging

from neo4j import GraphDatabase

class Neo4jDataPersister:
    def __init__(self, uri, user, password):
        self._uri = uri
        self._user = user
        self._password = password
        self._driver = GraphDatabase.driver(self._uri, auth=(self._user, self._password))
        self.logger = logging.getLogger(__name__)

        self._node_created_count = 0
        self._node_updated_count = 0
        self._node_deleted_count = 0

        self._relationship_created_count = 0
        self._relationship_updated_count = 0
        self._relationship_deleted_count = 0

    def close(self):
        self._driver.close()

    def persist_data_from_json(self, json_data):
        with self._driver.session() as session:
            self._persist_data(session, json_data)

        self.logger.info("Total nodes created: %d", self._node_created_count)
        self.logger.info("Total nodes updated: %d", self._node_updated_count)
        self.logger.info("Total nodes deleted: %d", self._node_deleted_count)

        self.logger.info("Total relationships created: %d", self._relationship_created_count)
        self.logger.info("Total relationships updated: %d", self._relationship_updated_count)
        self.logger.info("Total relationships deleted: %d", self._relationship_deleted_count)

    def _persist_data(self, session, data):
        if isinstance(data, dict):  # Check if it's a dictionary
            for key, value in data.items():
                if key == 'Identificação':
                    self._handle_identificacao(session, value)
                elif key == 'Atuação Profissional':
                    self._handle_atuacao_profissional(session, value)
                elif key == 'Atuação Profissional':
                    self._handle_atuacao_profissional(session, value)                    
                elif key == 'Produções':
                    self._handle_producoes(session, value)
                elif isinstance(value, list) and not value:
                    # Ignore empty lists
                    pass
                elif isinstance(value, dict):
                    if 'JCR2' in value:
                        # Ignore the 'JCR2' subdictionary
                        del value['JCR2']

                    self._persist_data(session, value)

        elif isinstance(data, list):  # Check if it's a list
            for item in data:
                self._persist_data(session, item)  # Recurse on list items

        else:
            # Handle other data types (strings, numbers, etc.) or raise an error
            self.logger.warning("Unexpected data type: %s", type(data))


    def _handle_identificacao(self, session, data):
        for item in data:
            if item['campo'] == 'ID Lattes':
                node_id = item['valor']
                node = self._get_or_create_node(session, 'Pesquisador', {'ID Lattes': node_id})
                self._persist_other_properties(session, node, data)
            else:
                node = self._get_or_create_node(session, 'Pesquisador', {data.items()})

    def _handle_atuacao_profissional(self, session, data):
        for item in data:
            node = self._create_node(session, 'AtuaçãoProfissional', item)

    def _handle_producoes(self, session, data):
        for production_type, production_data in data.items():
            if production_type == 'Artigos completos publicados em periódicos':
                for article in production_data:
                    node = self._create_node(session, production_type, article)
            elif production_type.startswith('1.'):
                for index, article in enumerate(production_data):
                    node = self._create_node(session, production_type, article)
                    node['OrdemCronológica'] = index + 1

    def _get_or_create_node(self, session, label, properties):
        node = session.run("MATCH (n: {label}) WHERE {properties} RETURN n", {"label": label, "properties": properties}).single()

        if not node:
            node = session.run("CREATE (n: {label} {properties}) RETURN n", {"label": label, "properties": properties}).single()["n"]
            self._node_created_count += 1

        return node

    def _create_node(self, session, label, properties):
        node = session.run("CREATE (n: {label} {properties}) RETURN n", {"label": label, "properties": properties}).single()["n"]
        self._node_created_count += 1

        return node

    def _persist_other_properties(self, session, node, data):
        for key, value in data.items():
            if key not in ['campo', 'valor']:
                node[key] = value  

    def _create_relationship(self, session, start_node, relationship_type, end_node, properties={}):
        session.run("MATCH (a), (b) WHERE ID(a) = {start_node_id} AND ID(b) = {end_node_id} CREATE (a)-[r:{type} {props}]->(b) RETURN r", 
                    {"start_node_id": id(start_node), "end_node_id": id(end_node), "type": relationship_type, "props": properties})
        self._relationship_created_count += 1
```

```python
from datetime import datetime
from neo4j import GraphDatabase
import ast

class GraphPreparer:
    def __init__(self, uri, user, password):
        self._uri = uri
        self._user = user
        self._password = password
        self._driver = None

    def connect(self):
        self._driver = GraphDatabase.driver(self._uri, auth=(self._user, self._password))

    def prepare_data_from_file(self, file_path):
        nodes = []
        relationships = []

        with open(file_path, 'r', encoding='utf-8') as file:
            data = json.load(file)

        for person_data in data:
            print(person_data.keys())
            person_id = person_data['Identificação'].get('ID Lattes')
            nodes.append(('Pesquisador', {'id_lattes': person_id}))
            
            for language in person_data.get('Idiomas', []):
                language_name = language['Idioma']
                proficiency = language['Proficiência']
                nodes.append(('Idioma', {'nome_idioma': language_name, 'proficiencia': proficiency}))
                relationships.append(('Pesquisador', {'id_lattes': person_id}, 'FALA', {'nome_idioma': language_name}))
            
            for academic_info in person_data.get('Formação', {}).get('Acadêmica', []):
                description = academic_info['Descrição']
                nodes.append(('Curso', {'descricao': description}))
                relationships.append(('Pesquisador', {'id_lattes': person_id}, 'REALIZOU', {'descricao': description}))
                
            for professional_info in person_data.get('Atuação Profissional', []):
                institution = professional_info['Instituição']
                description = professional_info['Descrição']
                relationships.append(('Pesquisador', {'id_lattes': person_id}, 'ATUA', {'instituicao': institution, 'descricao': description}))

        return nodes, relationships

    def persist_data(self, data_for_neo4j):
        try:
            with self._driver.session() as session:
                for data in data_for_neo4j:
                    if data[0] == 'Node':
                        label, properties = data[1:]
                        session.write_transaction(self._create_node, label, **properties)
                    elif data[0] == 'Relationship':
                        start_label, rel_type, end_label = data[1:]
                        session.write_transaction(self._create_relationship, start_label, rel_type, end_label)
        except Exception as e:
            print(f"Error persisting data: {e}")

    @staticmethod
    def _create_node(tx, label, **properties):
        query = f"CREATE (n:{label} $props)"
        tx.run(query, props=properties)

    @staticmethod
    def _create_relationship(tx, start_label, rel_type, end_label):
        query = "MATCH (a:{start_label}), (b:{end_label}) " \
                "CREATE (a)-[r:{rel_type}]->(b)"
        tx.run(query, start_label=start_label, rel_type=rel_type, end_label=end_label)

    def close(self):
        if self._driver is not None:
            self._driver.close()

# Exemplo de utilização:
graph_preparer = GraphPreparer(uri="bolt://localhost:7687", user="neo4j", password="password")
graph_preparer.connect()

filepath = os.path.join(folder_data_input,'dict_list_fioce.json')
```

```python
import json

class Neo4jGraphGenerator:
    def __init__(self, dict_list):
        self.dict_list = dict_list
        self.graph_data = {'nodes': [], 'relationships': []}

    def generate_graph_json(self):
        for item in self.dict_list:
            id_lattes = item.get('Identificação', {}).get('ID Lattes')
            pesquisador = item.get('Identificação', {}).get('Nome')
            
            # Adicionar nó de pessoa (Person)
            person_node = {'id': id_lattes, 'label': 'Person', 'name': pesquisador}
            self.graph_data['nodes'].append(person_node)

            areas = item.get('Áreas', {})
            for area in areas.values():
                # Adicionar nó de Grande Área (GrandeArea)
                grande_area_node = {'id': area['Grande Área'], 'label': 'GrandeArea', 'description': area['Grande Área']}
                self.graph_data['nodes'].append(grande_area_node)
                self.graph_data['relationships'].append({'source': id_lattes, 'target': area['Grande Área'], 'type': 'BELONGS_TO'})
                
                for subarea in area.get('Subáreas', []):
                    # Adicionar nó de Área (Area)
                    area_node = {'id': subarea['Área'], 'label': 'Area', 'description': subarea['Área']}
                    self.graph_data['nodes'].append(area_node)
                    self.graph_data['relationships'].append({'source': area['Grande Área'], 'target': subarea['Área'], 'type': 'BELONGS_TO'})
                    
                    # Adicionar nó de Subárea (Subarea)
                    subarea_node = {'id': subarea['Subárea'], 'label': 'Subarea', 'description': subarea['Subárea']}
                    self.graph_data['nodes'].append(subarea_node)
                    self.graph_data['relationships'].append({'source': subarea['Área'], 'target': subarea['Subárea'], 'type': 'BELONGS_TO'})

        return json.dumps(self.graph_data, indent=4)
```

```python
import json
import re
from neo4j import GraphDatabase

class AreasHandler:

    def __init__(self, uri, user, password):
        self._driver = GraphDatabase.driver(uri, auth=(user, password))

    def close(self):
        self._driver.close()
        
    def consult_areas_atuacao(self, name):
        with self._driver.session() as session:
            result = session.run("MATCH (p:Person {name: $name}) RETURN p.`Áreas de atuação` as areas_atuacao", name=name)
            record = result.single()
            if record:
                return record['areas_atuacao']
            return None

    def debug_and_convert(self, areas_str):
        try:
            return json.loads(areas_str)
        except json.JSONDecodeError:
            print(f"Failed to deserialize JSON string: {areas_str}")
            return None

    def extract_subarea(self, area_detail):
        # Extract the 'Subárea' content from the area detail
        match = re.search(r'Subárea: ([^/]+)', area_detail)
        if match:
            return match.group(1).strip()
        return None

    def extract_areas(self, area_detail):
        # Extract the 'Grande Área', 'Área', and 'Subárea' contents from the area detail
        grande_area_match = re.search(r'Grande área: ([^/]+)', area_detail)
        area_match = re.search(r'Área: ([^/]+)', area_detail)
        subarea_match = re.search(r'Subárea: ([^/]+)', area_detail)
        
        grande_area = grande_area_match.group(1).strip() if grande_area_match else None
        area = area_match.group(1).strip() if area_match else None
        subarea = subarea_match.group(1).strip() if subarea_match else None
        
        return grande_area, area, subarea

    def create_areas_relations(self, name):
        # Get 'Áreas de atuação' properties
        areas_properties = self.consult_areas_atuacao(name)

        # Convert the serialized JSON strings back into dictionaries
        try:
            deserialized_areas_properties = self.debug_and_convert(areas_properties)
        except Exception as e:
            print(f"Error deserializing 'Áreas de atuação' properties: {e}")
            return

        successful_areas_creations = 0

        with self._driver.session() as session:
            for _, area_detail in deserialized_areas_properties.items():
                try:
                    # Extracting Grande Área, Área, and Subárea from the details
                    match = re.match(r'Grande área: (.*?) / Área: (.*?) / Subárea: (.*?)(?:/Especialidade: (.*?))?\.?$', area_detail)
                    if not match:
                        print(f"Unexpected format for 'Áreas de atuação' detail: {area_detail}")
                        continue
                    grande_area, area, subarea = match.groups()[:3]

                    # Creating or merging nodes for Subárea, Área, and Grande Área
                    session.run("MERGE (s:Subárea {name: $subarea})", subarea=subarea)
                    session.run("MERGE (a:Área {name: $area})", area=area)
                    session.run("MERGE (ga:GrandeÁrea {name: $grande_area})", grande_area=grande_area)

                    # Creating or merging relationships. Using MERGE ensures no duplicate relationships are created.
                    session.run("MATCH (p:Person {name: $name}), (s:Subárea {name: $subarea}) MERGE (p)-[r:ATUA_EM]->(s)", name=name, subarea=subarea)
                    session.run("MATCH (ga:GrandeÁrea {name: $grande_area}), (a:Área {name: $area}) MERGE (ga)-[r:CONTÉM_ÁREA]->(a)", grande_area=grande_area, area=area)
                    session.run("MATCH (a:Área {name: $area}), (s:Subárea {name: $subarea}) MERGE (a)-[r:CONTEM_SUBÁREA]->(s)", area=area, subarea=subarea)

                    successful_areas_creations += 1
                except Exception as e:
                    print(f"Error processing 'Áreas de atuação' detail '{area_detail}': {e}")

            # Inform the user about areas
            print(f"{successful_areas_creations} 'Áreas de atuação' relations successfully created.")
```

```python
class ProjectsHandler:

    def __init__(self, uri, user, password):
        self._driver = GraphDatabase.driver(uri, auth=(user, password))

    def close(self):
        self._driver.close()

    def consult_data_by_property(self, name, property_name):
        with self._driver.session() as session:
            result = session.run(f"MATCH (p:Person {{name: $name}}) RETURN p.`{property_name}` as data", name=name)
            record = result.single()
            return record['data'] if record else None

    def create_projects_relations(self, name):
        successful_creations = 0

        with self._driver.session() as session:
            # Process 'Atuação Profissional' data
            professional_data = self.consult_data_by_property(name, 'Atuação Profissional')
            if professional_data:
                for institution_name, _ in json.loads(professional_data).items():
                    session.run("MERGE (i:Instituição {name: $institution_name})", institution_name=institution_name)
                    print(f"Institution node created/merged for: {institution_name}")

                    session.run("MATCH (p:Person {name: $name}), (i:Instituição {name: $institution_name}) MERGE (p)-[:TEM]->(i)", name=name, institution_name=institution_name)
                    print(f"Relationship established between {name} and {institution_name}.")

            # Process other dynamic nodes
            key_labels_to_check = ['Linhas de pesquisa', 'Projetos de pesquisa', 'Projetos de extensão', 'Projetos de desenvolvimento']
            for key in key_labels_to_check:
                formatted_key = f"`{key}`"  # Wrap the key with backticks
                project_data = self.consult_data_by_property(name, key)
                if project_data:
                    for project_time, project_name in json.loads(project_data).items():
                        if project_name:  # to avoid empty names
                            session.run(f"MERGE (p:{formatted_key} {{name: $project_name}})", project_name=project_name)
                            print(f"{key} node created/merged for: {project_name}")

                            session.run(f"MATCH (a:Person {{name: $name}}), (p:{formatted_key} {{name: $project_name}}) MERGE (a)-[:TEM]->(p)", name=name, project_name=project_name)
                            print(f"Relationship established between {name} and {project_name} ({key}).")
                            successful_creations += 1
                else:
                    print(f"'{key}' data not found for {name}")

        print(f"{successful_creations} projetos atualizados com sucesso.")
```

```python
import json
import re
from neo4j import GraphDatabase

class ArticleHandler:

    def __init__(self, uri, user, password):
        self._driver = GraphDatabase.driver(uri, auth=(user, password))

    def close(self):
        self._driver.close()

    def fetch_person_productions(self, name):
        with self._driver.session() as session:
            result = session.run("MATCH (p:Person {name: $name}) RETURN p.Produções as produções", name=name)
            record = result.single()
            return record['produções'] if record else None

    def extract_article_info(self, input_str):
        # Encontre todas as abreviaturas de iniciais em maiúsculas e seus índices
        abbreviations = [(match.group(), match.start()) for match in re.finditer(r'\b[A-Z]\.', input_str)]

        # Encontre a posição da maior ocorrência de abreviaturas de iniciais, se houver
        if abbreviations:
            max_abbr_position = max(abbreviations, key=lambda x: x[1])

            # Encontre a primeira ocorrência de '. ' ou ' . ' após a maior ocorrência de abreviaturas de iniciais
            first_separator_candidates = [
                input_str.find('. ', max_abbr_position[1] + 3),
                input_str.find(' . ', max_abbr_position[1] + 3),
                input_str.find('.. ')
            ]
            first_separator_candidates = [pos for pos in first_separator_candidates if pos != -1]

            if first_separator_candidates:
                first_separator = min(first_separator_candidates)

                # Encontre a primeira ocorrência de '. ' após o primeiro separador
                second_separator = input_str.find('. ', first_separator + 2)

                # Encontre a primeira ocorrência de ', ' após o segundo separador
                third_separator = input_str.find(', ', second_separator + 2)
            else:
                first_separator = second_separator = third_separator = -1
        else:
            first_separator = second_separator = third_separator = -1

        # Defina o padrão para encontrar "p." e o conteúdo até a próxima vírgula
        pages_match = re.search(r' p\.\s*(.*?),', input_str)
        pages = pages_match.group(1) if pages_match else ""

        # Defina o padrão para encontrar "v." e o conteúdo até a próxima vírgula
        volume_match = re.search(r' v\.\s*(.*?),', input_str)
        volume = volume_match.group(1) if volume_match else ""

        # Encontre a primeira ocorrência de um ano de quatro dígitos seguido de ponto final após o terceiro separador
        year_match = re.search(r' \d{4}\.', input_str[third_separator + 2:])
        year = year_match.group().strip('.').strip() if year_match else ""

        # Extraia os dados com base nas posições dos separadores
        authors = input_str[:first_separator].strip()
        title = input_str[first_separator + 2:second_separator].strip()
        journal = input_str[second_separator + 2:third_separator].strip()

        # Verifique se a lista de autores e o título não estão vazios
        if not authors or not title:
            return None  # Retorna None para indicar falha

        # Crie um dicionário com os dados extraídos
        article_info = {
            "authors": authors,
            "title": title,
            "original_title": journal,
            "pages": pages,
            "volume": volume,
            "year": year
        }

        return article_info
    
    def deserialize_and_create_nodes(self, name):
        print(f"Fetching 'Produções' data for {name}...")
        productions_data = self.fetch_person_productions(name)
        
        if not productions_data:
            print(f"'Produções' data not found or empty for {name}.")
            return

        print(f"Attempting to deserialize 'Produções' data for {name}...")
        try:
            productions_data = json.loads(productions_data)
        except json.JSONDecodeError as e:
            print(f"Failed to deserialize 'Produções' data for {name}: {e}")
            return

        successful_articles = 0
        unsuccessful_articles = []

        with self._driver.session() as session:
            print(f"Processing 'Produção bibliográfica' for {name}...")
            bibliographic_production = productions_data.get("Produção bibliográfica", {})
            
            if isinstance(bibliographic_production, str):
                print(f"Attempting to deserialize 'Produção bibliográfica' for {name}...")
                try:
                    bibliographic_production = json.loads(bibliographic_production)
                except json.JSONDecodeError as e:
                    print(f"Failed to deserialize 'Produção bibliográfica' for {name}: {e}")
                    return

            articles = json.loads(bibliographic_production.get("Artigos completos publicados em periódicos", "{}"))

            for _, article_str in articles.items():
                article_details = self.extract_article_info(article_str)

                # Vamos imprimir os detalhes de cada artigo e verificar se os autores estão presentes.
                print(f"Original Article: {article_str}")
                print(f"Extracted Details: {article_details}")

                if article_details:
                    article_details["title"] = article_details["title"].strip()
                    article_details["original_title"] = article_details["original_title"].strip()

                    session.run(f"MERGE (a:Artigo {{title: $title}}) SET a += $details", title=article_details["title"], details=article_details)
                    session.run(f"MATCH (p:Person {{name: $name}}), (a:Artigo {{title: $title}}) MERGE (p)-[:PUBLICOU]->(a)", name=name, title=article_details["title"])
                    successful_articles += 1
                else:
                    unsuccessful_articles.append(article_str)

        print(f"Processed {successful_articles} articles successfully for {name}.")

        if unsuccessful_articles:
            print("Failed to process the following articles:")
            for article in unsuccessful_articles:
                print(article)

    def process_articles(self, name):
        self.deserialize_and_create_nodes(name)
```

```python
from neo4j import GraphDatabase
import urllib.parse
import json
import re

class AdvisorHandler:

    def __init__(self, uri, user, password):
        self._driver = GraphDatabase.driver(uri, auth=(user, password))

    def close(self):
        self._driver.close()

    @staticmethod
    def convert_to_primitives(input_data):
        if input_data is None:
            return None
        if isinstance(input_data, dict):
            for key, value in input_data.items():
                if isinstance(value, dict):  
                    input_data[key] = json.dumps(AdvisorHandler.convert_to_primitives(value), ensure_ascii=False)
                else:
                    input_data[key] = AdvisorHandler.convert_to_primitives(value)
            return input_data
        elif isinstance(input_data, list):
            return [AdvisorHandler.convert_to_primitives(item) for item in input_data]
        elif isinstance(input_data, str):
            if 'http://' in input_data or 'https://' in input_data:
                parts = input_data.split(" ")
                new_parts = [urllib.parse.quote(part) if part.startswith(('http://', 'https://')) else part for part in parts]
                return " ".join(new_parts)
            return input_data
        elif isinstance(input_data, (int, float, bool)):
            return input_data
        else:
            return str(input_data)

    @staticmethod
    def debug_and_convert(input_data):
        try:
            return AdvisorHandler.convert_to_primitives(input_data)
        except:
            print("Conversion failed for:", input_data)
            raise

    def consult_orientacoes(self, name):
        with self._driver.session() as session:
            query = (
                "MATCH (p:Person {name: $name})"
                "RETURN p.Orientações AS orientacoes"
            )
            result = session.run(query, name=name)
            orient_data = result.single()["orientacoes"]
            if orient_data is None:
                raise ValueError(f"No data found for 'Orientações' attribute for Person '{name}'")
            orient_properties_list = json.loads(orient_data)
            return orient_properties_list

    def create_advisor_relations(self, name):
        # Get Orientações properties
        orient_properties = self.consult_orientacoes(name)

        # Convert the serialized JSON strings back into dictionaries
        try:
            deserialized_orient_properties = self.debug_and_convert(orient_properties)
        except Exception as e:
            print(f"Error deserializing Orientações properties: {e}")
            return

        # Advisory relationship mapping
        advisory_types = {
            "Dissertação de mestrado": "ORIENTOU_MESTRADO",
            "Tese de doutorado": "ORIENTOU_DOUTORADO",
            "Trabalho de conclusão de curso de graduação": "ORIENTOU_GRADUAÇÃO"
        }

        successful_advisory_creations = 0

        with self._driver.session() as session:
            for orientacao_category, advisories in deserialized_orient_properties.items():
                if isinstance(advisories, str):
                    try:
                        advisories = json.loads(advisories)
                    except json.JSONDecodeError:
                        print(f"Failed to deserialize JSON string in 'Orientações' for category '{orientacao_category}': {advisories}")
                        continue

                if not isinstance(advisories, dict):
                    print(f"Unexpected data type in 'Orientações' for category '{orientacao_category}': {advisories}")
                    continue

                for advisory_type, relationships in advisories.items():
                    relation_label = advisory_types.get(advisory_type)
                    if not relation_label:
                        continue  # skip if the advisory type is not one of the specified ones

                    for _, detail in json.loads(relationships).items():
                        try:
                            student_name = detail.split(".")[0]
                            title = detail.split(".")[1]
                            
                            # Extract the year from the detail string
                            year_match = re.search(r'(\d{4})', detail)
                            year = year_match.group(1) if year_match else None

                            # Create or merge the Orientações node
                            node_query = (
                                "MERGE (a:Orientações {Title: $title}) "
                                "ON CREATE SET a.StudentName = $student_name, a.Tipo = $advisory_type, a.Year = $year "
                                "ON MATCH SET a.Tipo = $advisory_type, a.Year = $year "  # Ensure the 'Tipo' and 'Year' are always updated
                                "RETURN a"
                            )
                            session.run(node_query, title=title, student_name=student_name, advisory_type=advisory_type, year=year)

                            # Create or update the advisory relationship
                            relation_query = (
                                f"MATCH (p:Person {{name: $name}}), (a:Orientações {{Title: $title}}) "
                                f"MERGE (p)-[r:{relation_label}]->(a) "
                            )
                            session.run(relation_query, name=name, title=title)

                            successful_advisory_creations += 1
                        except Exception as e:
                            print(f"Error processing advisory '{detail}': {e}")

        # Inform the user about advisories
        print(f"{successful_advisory_creations} orientações atualizadas com sucesso.")
```

```python
from neo4j import GraphDatabase

class DataRemovalHandler:

    def __init__(self, uri, user, password):
        """
        Inicializa a classe DataRemovalHandler com informações de conexão ao banco de dados Neo4j.

        Parâmetros:
        - uri (str): URI de conexão ao Neo4j.
        - user (str): Nome de usuário para autenticação.
        - password (str): Senha para autenticação.
        """
        self._driver = GraphDatabase.driver(uri, auth=(user, password))

    def close(self):
        """
        Fecha a conexão com o banco de dados Neo4j.
        """
        self._driver.close()

    def delete_nodes_by_label(self, label):
        """
        Deleta todos os nós associados a um label específico no Neo4j.

        Parâmetro:
        - label (str): O label dos nós a serem deletados.

        Retorna:
        - int: O número de nós deletados.
        """
        with self._driver.session() as session:
            # Esta consulta combina com todos os nós do label especificado e os deleta
            result = session.run(f"MATCH (n:{label}) DETACH DELETE n RETURN count(n) as deleted_count")
            deleted_count = result.single()["deleted_count"]
            return deleted_count
```

```python
from py2neo import Graph, Node, NodeMatcher, Relationship
from typing import Dict, List

def create_or_update_publications(graph: Graph, publications_dict: Dict[str, Dict[str, List[Dict[str, str]]]]) -> None:
    """
    Create or update a researcher node and its associated publications in the Neo4j database.

    Parameters:
        graph (Graph): The py2neo Graph object connected to the Neo4j database.
        publications_dict (Dict[str, Dict[str, List[Dict[str, str]]]]): Dictionary containing publication information.

    Returns:
        None
    """

    # Initialize counters
    publications_created = 0
    properties_updated = 0

    # Create a NodeMatcher object
    matcher = NodeMatcher(graph)

    # Search for existing researcher node by name
    existing_node = matcher.match("Researcher", name=publications_dict['Node Name']).first()

    if not existing_node:
        raise ValueError("Researcher node must exist before adding publications.")

    # Process publications
    for publication in publications_dict['Properties']['Produções']:
        # Check if a similar publication already exists
        existing_pub_node = matcher.match("Publication", doi=publication['doi']).first()

        if not existing_pub_node:
            pub_node = Node("Publication", **publication)
            graph.create(pub_node)
            publications_created += 1
        else:
            for key, value in publication.items():
                if key not in existing_pub_node or existing_pub_node[key] != value:
                    existing_pub_node[key] = value
                    properties_updated += 1
            pub_node = existing_pub_node
            graph.push(pub_node)

        # Create or update relationship between researcher and publication
        rel = Relationship(existing_node, "PUBLICOU", pub_node)
        graph.merge(rel)

    # Print statistics
    print(f"Publications created: {publications_created}")
    print(f"Properties updated: {properties_updated}")
```

```python
from bs4 import BeautifulSoup
from typing import Dict, Any

def generate_lattes_dict(soup: BeautifulSoup) -> Dict[str, Any]:
    """
    Generate a dictionary from a BeautifulSoup object to be persisted in Neo4j.

    Parameters:
        soup (BeautifulSoup): The BeautifulSoup object containing the parsed HTML/XML data.

    Returns:
        Dict[str, Any]: A dictionary containing the relevant information for Neo4j persistence.
    """
    # Initialize an empty dictionary to hold the extracted data
    lattes_data = {}
    
    # Extracting researcher's name as an example
    name_section = soup.find('div', {'id': 'name-section'})
    if name_section:
        lattes_data['researcher_name'] = name_section.text.strip()
    
    # Extracting list of publications as an example
    publications = []
    for pub in soup.find_all('div', {'class': 'publication'}):
        publication_data = {}
        title = pub.find('span', {'class': 'title'})
        authors = pub.find('span', {'class': 'authors'})
        
        if title:
            publication_data['title'] = title.text.strip()
        
        if authors:
            publication_data['authors'] = authors.text.strip().split(',')
        
        publications.append(publication_data)
    
    lattes_data['publications'] = publications
    
    # Additional extractions can be performed as per the requirements
    
    return lattes_data

# Example usage (Assuming 'some_html_content' contains the HTML content)
# soup = BeautifulSoup(some_html_content, 'html.parser')
# lattes_dict = generate_lattes_dict(soup)
```

```python
from py2neo import Graph, Node, NodeMatcher, Relationship
from typing import Dict

def create_or_update_professional_links(graph: Graph, researcher_dict: Dict[str, str]) -> None:
    """
    Create or update a researcher node and its professional links in the Neo4j database based on the given dictionary.

    Parameters:
        graph (Graph): The py2neo Graph object connected to the Neo4j database.
        researcher_dict (Dict[str, str]): A dictionary containing the researcher's professional information.

    Returns:
        None
    """

    # Initialize counters
    relationships_created = 0
    properties_updated = 0

    # Create a NodeMatcher object
    matcher = NodeMatcher(graph)

    # Look for existing nodes with the same name
    existing_node = matcher.match("Researcher", name=researcher_dict.get('Nome')).first()

    if not existing_node:
        raise ValueError("Researcher node must exist before adding professional links.")

    # Process professional affiliations and activities
    for key, value in researcher_dict.items():
        if key not in ['Nome', 'Endereço Profissional']:
            # Create or find the organization/affiliation node
            org_node = matcher.match("Organization", name=key).first()
            if not org_node:
                org_node = Node("Organization", name=key)
                graph.create(org_node)

            # Create or update the relationship
            rel_type = "AFFILIATED_WITH" if 'Atual' in value else "HAS_BEEN_AFFILIATED_WITH"
            rel = Relationship(existing_node, rel_type, org_node, details=value)

            # Check if a similar relationship already exists
            existing_rel = None
            for r in graph.match((existing_node, org_node), r_type=rel_type):
                if r['details'] == value:
                    existing_rel = r
                    break

            # Create new or update existing relationship
            if not existing_rel:
                graph.create(rel)
                relationships_created += 1
            else:
                for property_name, property_value in rel.items():
                    if property_name not in existing_rel or existing_rel[property_name] != property_value:
                        existing_rel[property_name] = property_value
                        properties_updated += 1
                graph.push(existing_rel)

    # Print statistics
    print(f"Relationships created: {relationships_created}")
    print(f"Properties updated: {properties_updated}")

# Example usage
# Initialize the Neo4j graph (Replace 'neo4j' and 'password' with your database credentials)
# graph = Graph("bolt://localhost:7687", auth=("neo4j", "password"))

# Create or update professional links in Neo4j based on the dictionary
# create_or_update_professional_links(graph, dict_vinculo)
```

```python
from py2neo import Graph, Node, NodeMatcher
import pandas as pd
from typing import Union

def persist_journals_from_xlsx(graph: Graph, xlsx_path: str) -> None:
    """
    Persist journal information into a Neo4j database from an Excel (.xlsx) file.

    Parameters:
        graph (Graph): The py2neo Graph object connected to the Neo4j database.
        xlsx_path (str): The path to the Excel (.xlsx) file containing the journal information.

    Returns:
        None
    """
    
    # Initialize counters for tracking
    nodes_created = 0
    properties_updated = 0
    
    # Create NodeMatcher object
    matcher = NodeMatcher(graph)
    
    # Read the Excel file into a pandas DataFrame
    df = pd.read_excel(xlsx_path)
    
    # Iterate over each row in the DataFrame
    for _, row in df.iterrows():
        # Extract journal properties
        properties = {'ISSN': row['ISSN'],
                      'Título': row['Título'],
                      'Área de Avaliação': row['Área de Avaliação'],
                      'Estrato': row['Estrato']}
        
        # Check if a node with the same ISSN already exists
        existing_node = matcher.match("original_title", ISSN=row['ISSN']).first()
        
        if existing_node:
            for key, value in properties.items():
                if key not in existing_node or existing_node[key] != value:
                    existing_node[key] = value
                    properties_updated += 1
            graph.push(existing_node)
        else:
            new_node = Node("original_title", **properties)
            graph.create(new_node)
            nodes_created += 1
    
    # Print statistical data
    print(f"Nodes created: {nodes_created}")
    print(f"Properties updated: {properties_updated}")

# Example usage
# Initialize the Neo4j graph (Replace 'neo4j' and 'password' with your database credentials)
# graph = Graph("bolt://localhost:7687", auth=("neo4j", "password"))

# Given the path to the Excel file
# xlsx_path = "path/to/excel/file.xlsx"

# Persist the journal information from the Excel file
# persist_journals_from_xlsx(graph, xlsx_path)
```

Para extrair informações precisamos que o método procure pelo tipo de tag, nome do id ou nome de classe corretas.

#### Formação Acadêmica/Titulação
Essas informações estão no HTML sob a tag do tipo div e classe "layout-cell layout-cell-12 data-cell"
Esta tag é seguida por diversos itens seguindo a divisão:
a tag do tipo div e classe "layout-cell layout-cell-3 text-align-right" para o ano e 
a tag do tipo div e classe "layout-cell layout-cell-9" para a descrição da formação.

#### Produção Bibliográfica
As produções bibliográficas, especificamente artigos completos publicados em periódicos, estão listadas sob a tag do tipo div com id = "artigos-completos" 
Cada artigo é marcado pela tag do tipo div e classe "artigo-completo", com detalhes dentro a tag do tipo div e classe "transform"

#### Achando os marcadores corretos no HTML

Na classe que traz os artigos os dados estão hierarquicamente como filhos da classe "artigo-completo". Dentro da classe "title-wrapper" há a classe filha "layout-cell layout-cell-12 data-cell" que contém todas as produções científicas, dentro dessa classe data-cell há as divisões de tipos pela classe "inst_back", dentro da qual em uma tag 'b' está o texto com nome de cada tipo de produção Produção bibliográfica

```python
# html3 = '''
# <html lang="pt-BR"><head>
# <meta http-equiv="Content-Type" content="text/html; charset=ISO-8859-1">
# <link href="images/v2/fav_ico_lattes.ico" rel="Shortcut Icon">
# <meta content="Sistema de Currículos Lattes 2.0 (CNPq)" http-equiv="generator">
# <meta content="currículo,curriculo,curriculum,cv,vitae,lattes,produção,producao,científica,cientifica,Brasil" http-equiv="keywords">
# <title>Currículo do Sistema de Currículos Lattes (Anya Pimentel Gomes Fernandes Vieira Meyer)</title>
# <link type="text/css" rel="stylesheet" href="css/import.css">
# <link type="text/css" href="css/tablet-mobile.css" rel="stylesheet">
# <link type="text/css" rel="stylesheet" href="css/jquery.megamenu.css">
# <link type="text/css" rel="stylesheet" href="css/style_curiculo.css">
# <link type="text/css" rel="stylesheet" href="css/v2/espelho.css">
# <link type="text/css" rel="stylesheet" href="css/impressao.css">
# <script type="text/javascript" async="" src="https://ssl.google-analytics.com/ga.js"></script><script type="text/javascript" src="js/v2/jquery.min.js"></script><script src="js/jquery.cookie.js" type="text/javascript"></script><script src="js/acessibilidade.js" type="text/javascript"></script><script type="text/javascript" src="js/v2/tooltip-position.js"></script><script type="text/javascript" src="js/v2/jquery.tipsy.js"></script><script type="text/javascript" src="js/v2/jquery.megamenu.js"></script><script type="text/javascript" src="js/v2/byOrder.js"></script><script type="text/javascript" src="js/v2/font-size.js"></script><script type="text/javascript" src="js/v2/cv_citacoes.js"></script><script type="text/javascript" src="js/v2/hint_ajax.js"></script><script type="text/javascript" src="js/v2/menuFlowScroll.js"></script><script type="text/javascript" src="js/v2/videoUtils.js"></script><script type="text/javascript">
			
# 			var menu = {
# 				iniciar: function(id) {
# 					jkmegamenu.definemenu("ancora-menu-" + id, "menu-" + id, "focus");
# 				}
# 			};
			
# 			function abreLinkHistoricoPublicacoes(codigo,nome){
#             	var j = window.open("historicoAtualizacao.jsp?id="+codigo+"&nome="+nome+"&fromBusca=1", "historico","resizable=no,menubar=no,toolbar=no,status=yes,scrollbars=yes,width=550px,height=470px,left=0,top=0");
#             	if(j){
#             		j.focus();
#             	} else {
#             		alert('Por favor, desabilite o bloqueador de popups');
#             	}
# 			}
			
# 			var autores = {
# 				exibir: function() {
# 					$(this).hide().parent().find(".autores-et-al").show();
# 				},
# 				ocultar: function() {
# 					$(this).parent().hide().parent().find(".autores-et-al-plus").show();
# 				}
# 			};
				
# 			$(document).ready(function(){
			
# 				$(".ajaxJCR").ajaxJCR();
				
# 				$(".ajaxCAPES").ajaxCAPES();
				
# 				$(".ajaxINPI").ajaxINPI();
				
# 				$("span.citacoes").verificarCitacoes();
				
# 				$(".tooltip").tipsy({gravity: 's'});
				
# 				$(".autores-et-al-plus").click(autores.exibir);
				
# 				$(".autores-et-al-minus").click(autores.ocultar);
        			
#       		});	
      				  
# <!--       		$( function() { -->
# <!-- 			    $( document ).tooltip(); -->
# <!-- 			 } ); -->
      		                                
		
# 	</script><script type="text/javascript">
					
# 					  var _gaq = _gaq || [];
# 					  _gaq.push(['_setAccount', 'UA-35652740-1']);
# 					  _gaq.push(['_trackPageview']);
					
# 					  (function() {
# 					    var ga = document.createElement('script'); ga.type = 'text/javascript'; ga.async = true;
# 					    ga.src = ('https:' == document.location.protocol ? 'https://ssl' : 'http://www') + '.google-analytics.com/ga.js';
# 					    var s = document.getElementsByTagName('script')[0]; s.parentNode.insertBefore(ga, s);
# 					  })();
					
# 					</script>
# <link rel="stylesheet" href="https://use.fontawesome.com/releases/v6.2.1/css/all.css"></head>
# <body>
# <div class="page min-width">
# <div class="header">
# <div class="header-content max-width">
# <div class="identity center">
# <div class="combo">
# <h1 class="escondido" tabindex="0"> Currículo Lattes - Busca Textual - Visualização do Currículo </h1>
# <a href="http://lattes.cnpq.br/" title="Currículo Lattes" target="_blank" tabindex="1"><img title="Currículo Lattes" alt="Currículo Lattes" src="images/titulo-sistema.png"></a>
# </div>
# <div class="menu-header" style="position: absolute; top: 15px;">
# <span class="tituloFlow"></span><span class="linksFlow"><a href="http://buscatextual.cnpq.br/buscatextual/visualizacv.do?id=K4761906A8&amp;idiomaExibicao=2" class="btn-br" tabindex="2">English</a><a title="Imprimir currículo" class="bt-menu-header titBottom  icons-top icons-top-print" href="javascript:window.print()" tabindex="3"></a><a title="Aumentar a fonte" class="bt-menu-header titBottom fontMais icons-top icons-top-fontMais" href="javascript:void(0)" tabindex="4"></a><a title="Fonte padrão" class="bt-menu-header titBottom fontMenos icons-top icons-top-fontMenos" href="javascript:void(0)" tabindex="5"></a><a title="Alto contraste" class="bt-menu-header titBottom contraste icons-top icons-top-contraste-en" href="javascript:void(0)" tabindex="6"></a><a href="download.do?metodo=apresentar&amp;idcnpq=7659758489387356" target="_blank" class="bt-menu-header titBottom fontMenos icons-top icons-top-xml" title="Baixar Currículo em XML" tabindex="7"></a><a title="Ajuda" class="bt-menu-header titBottom icons-top icons-top-help" target="_blank" href="http://ajuda.cnpq.br/index.php" tabindex="8"></a></span>
# </div>
# <div class="logo">
# <a target="_blank" href="http://cnpq.br/" tabindex="9">
# <h2 tabindex="0">CNPq - Conselho Nacional de Desenvolvimento Científico e Tecnológico</h2>
# </a>
# </div>
# </div>
# </div>
# </div>
# <div class="navigation-wrapper">
# <div class="center navigation-wrapper-line">
# <div class="menuPrincipal">
# <div class="menuCtl menuAcesso">
# <div class="menucent"><script type="text/javascript">menu.iniciar("dados-gerais");</script>
		
		
# 		<a id="ancora-menu-dados-gerais" href="javascript:void(0)" class="acessibilidade" tabindex="10" _megamenupos="0pos"><span></span>Dados gerais</a>
		
	
		
		
		
		
# 	<script type="text/javascript">menu.iniciar("formacao");</script>
		
		
# 		<a id="ancora-menu-formacao" href="javascript:void(0)" class="acessibilidade separador" tabindex="16" _megamenupos="1pos"><span></span>Formação</a>
		
	
		
		
		
		
# 	<script type="text/javascript">menu.iniciar("atuacao");</script>
		
		
# 		<a id="ancora-menu-atuacao" href="javascript:void(0)" class="acessibilidade separador" tabindex="19" _megamenupos="2pos"><span></span>Atuação</a>
		
	
		
		
		
		
	
		
		
# 		<a id="ancora-menu-projetos" href="#ProjetosPesquisa" class="acessibilidade separador" tabindex="25" _megamenupos="3pos"><span></span>Projetos</a>
		
# 	<script type="text/javascript">menu.iniciar("projetos");</script>
		
		
		
		
# 	<script type="text/javascript">menu.iniciar("producoes");</script>
		
		
# 		<a id="ancora-menu-producoes" href="#ProducoesCientificas" class="acessibilidade separador" tabindex="30" _megamenupos="4pos"><span></span>Produções</a>
		
	
		
		
		
		
# 	<script type="text/javascript">menu.iniciar("potencialInovacao");</script>
		
		
# 		<a id="ancora-menu-potencialInovacao" href="#PotencialInovacao" class="acessibilidade separador" tabindex="63" _megamenupos="5pos"><span></span>Inovação</a>
		
	
		
		
		
		
# 	<script type="text/javascript">menu.iniciar("educacaoPopularizacaoCTA");</script>
		
		
# 		<a id="ancora-menu-educacaoPopularizacaoCTA" href="#EducacaoPopularizacaoCTA" class="acessibilidade separador" tabindex="78" _megamenupos="6pos"><span></span>Educação e Popularização de C &amp; T</a>
		
	
		
		
		
		
# 	<script type="text/javascript">menu.iniciar("eventos");</script>
		
		
# 		<a id="ancora-menu-eventos" href="#Eventos" class="acessibilidade separador" tabindex="93" _megamenupos="7pos"><span></span>Eventos</a>
		
	
		
		
		
		
# 	<script type="text/javascript">menu.iniciar("orientacoes");</script>
		
		
# 		<a id="ancora-menu-orientacoes" href="#Orientacoes" class="acessibilidade separador" tabindex="96" _megamenupos="8pos"><span></span>Orientações</a>
		
	
		
		
		
		
# 	<script type="text/javascript">menu.iniciar("bancas");</script>
		
		
# 		<a id="ancora-menu-bancas" href="#Bancas" class="acessibilidade separador" tabindex="99" _megamenupos="9pos"><span></span>Bancas</a>
		
	
		
		
		
		
# 	<script type="text/javascript">menu.iniciar("citacoes");</script>
		
		
# 		<a id="ancora-menu-citacoes" href="#Citacoes" class="acessibilidade separador" tabindex="102" _megamenupos="10pos"><span></span>Citações</a>
		
	
		
		
		
		
# 	<script type="text/javascript">menu.iniciar("informacoes-complementares");</script>
		
		
# 		<a id="ancora-menu-informacoes-complementares" href="javascript:void(0)" class="acessibilidade separador" tabindex="107" _megamenupos="11pos"><span></span>+</a>
		
	
		
		
		
		
# 	</div>
# </div>
# </div>
# </div>
# </div>
# <div class="content-wrapper">
# <div class="main-content max-width min-width" id="annot-div">
#   <div class="layout-cell-pad-main" style="padding: 10px 10px">
#     <div class="rodape-cv" style="margin: 0px 5px; color: #666666; font-size: 13px;">
    
#     <a href="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/html/visualization.html" target="_blank" id="qlattes-logo" style="margin: 0px 0px;" title="Abrir QLattes">
#       <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/qlattes-logo.png" width="70" style="margin-left: -3px; margin-bottom: -4px;">
#     </a>anotou o Qualis de 101 artigos em periódicos  neste CV.
#     <br>
#     <a href="#artigos-completos">
#       <button style="color: #edf2f7; background-color: #3569a7; box-shadow: 1px 1px 1px #2e5469; padding: 4px 0.5em 3px; font-size: 12px; border-radius: 4px; cursor: pointer; width: 142px; text-align: left; margin-top: 15px; margin-bottom: 10px;  margin-right: 5px;">
#         <i class="fa-solid fa-note-sticky fa-flip-vertical" style="font-size: 1.1em; margin-top: 0px; margin-right: 2px;"></i> Ver anotações
#       </button>
#     </a>
#   <a href="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/html/visualization.html" target="_blank" id="qlattes-visualization">
#     <button style="color: #edf2f7; background-color: #3569a7; box-shadow: 1px 1px 1px #2e5469; padding: 4px 0.5em 3px; font-size: 12px; border-radius: 4px; cursor: pointer; width: 142px; text-align: left; margin-top: 10px; margin-bottom: 10px; margin-right: 5px;">
#       <i class="fa-solid fa-chart-simple" style="font-size: 1.1em; margin-top: 0px; margin-right: 2px;"></i>
#       Visualizar dados
#     </button>
#   </a>
#   <a href="https://github.com/nabormendonca/qlattes/releases" target="_blank" id="recent-updates">
#   <button style="color: #edf2f7; background-color: #3569a7; box-shadow: 1px 1px 1px #2e5469; padding: 4px 0.5em 3px; font-size: 12px; border-radius: 4px; cursor: pointer; width: 142px; text-align: left; margin-top: 10px; margin-bottom: 10px;">
#     <i class="fa-solid fa-bullhorn" style="font-size: 1.1em; margin-top: 0px; margin-right: 2px;"></i>
#     Últimas atualizações
#   </button>
#   </a>
     
#     </div>
#   </div></div><div class="main-content max-width min-width center">
# <div class="layout-cell layout-cell-12">
# <div class="layout-cell-pad-main">
# <img class="foto" src="http://servicosweb.cnpq.br/wspessoa/servletrecuperafoto?tipo=1&amp;id=K4761906A8"><div class="infpessoa">
# <h2 class="nome" tabindex="0">Anya Pimentel Gomes Fernandes Vieira Meyer</h2>
# <h2 class="nome" tabindex="0">
# <span align="center" style="font-weight: bold" class="texto">Bolsista de Produtividade em Pesquisa do CNPq - Nível 2 (***)</span>
# </h2>
# <ul class="informacoes-autor">
# <li>
# <span class="img_link icone-informacao-autor"></span>Endereço para acessar este CV: http://lattes.cnpq.br/7659758489387356</li>
# <li>
# <span class="img_identy icone-informacao-autor"></span>ID Lattes: <span style="font-weight: bold; color: #326C99;">7659758489387356</span>
# </li>
# <li>
# <span class="img_cert icone-informacao-autor"></span>Última atualização do currículo em 15/02/2024</li>
# </ul>
# </div><br class="clear"><div class="title-wrapper">
# <h1 class="ui-hidden" tabindex="0"></h1>
# <hr class="separator">
# <p class="resumo">Possui Graduação em Odontologia pela Universidade Federal do Ceará, Mestrado em Odontopediatria (MSc) - University of London-Inglaterra, Doutorado em Odontologia na University of Toronto-Canadá e pós-doutorado em Saúde Pública pela Universidade da Califórnia Berkeley. É Pesquisadora Especialista da Fundação Oswaldo Cruz/FIOCRUZ  Ce, onde também exerce o papel de SecretariaExecutiva da Rede Nordeste de Formação em Saúde daFamília (RENASF), além de docente do curso de odontologia da Unichristus. Atualmente está como Visiting Researcherna Universidade de Harvard-USA (2024). Suas linhas de pesquisa atuais referem-se à avaliação de serviços de saúde; acesso a serviços de saúde e violência sofrida por populações vulneráveis (HIV, crianças, idosos, pacientes com necessidades especiais, beneficiários do Bolsa Família etc.); Saúde da Família; o impacto do Programa Bolsa Família e da Estratégia Saúde da Família na saúde da população; além da análise da ressonância da violência urbana nos processos de trabalho e saúde mental dos profissionais da Atenção Primária, especialmente os Agentes Comunitários de Saúde. Estes enfocam a avaliação de conhecimentos, práticas, cuidados e condições de saúde em pessoas em grupos de risco; avaliar os serviços de saúde na área da atenção básica - na perspectiva do usuário, profissionais de saúde e gestores; bem como o impacto das políticas públicas na saúde da população.<span style="font-weight: bold" class="texto"> (Texto informado pelo autor)</span>
# </p>
# </div><br class="clear"><div class="title-wrapper">
# <a name="Identificacao" tabindex="109">
# <h1 tabindex="0">Identificação</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Nome</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Anya Pimentel Gomes Fernandes Vieira Meyer<img src="/buscatextual/images/icon-rfb.png" class="alinhamento-texto-img" title="Dados cadastrais validados junto a Receita Federal do Brasil"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Nome em citações bibliográficas</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vieira, A P G F;Vieira, A;Anya Vieira;Vieira, A P;Vieira, Anya Pimentel Gomes Fernandes;VIEIRA, ANYA P. G. F.;Vieira  Anya Pimentel Gomes Fernandes;Anya Pimentel Gomes Fernandes Vieira;VIEIRA, ANYA PIMENTEL;Anya Pimentel Gomes Fernandes Vieira-Meyer;PIMENTEL GOMES FERNANDES VIEIRA, ANYA;PIMENTEL GOMES FERNANDES VIEIRA-MEY, ANYA;VIEIRA-MEYER, ANYA PIMENTEL GOMES FERNANDES;MEYER, ANYA PIMENTEL G. F. VIEIRA;PIMENTEL G. F. VIEIRA-MEYER, ANYA;Vieira-Meyer, AP;Vieira-Meyer, Anya PGF;VIEIRA-MEYER, ANYA P G F;Anya P G F Vieira-Meyer;Anya P. G. F. Vieira-Meyer;Anya Vieira-Meyer;Anya P. G. F. Vieira - Meyer;Anya P G F Vieira - Meyer;Anya P G F Vieira Meyer;MEYER, ANYA PIMENTEL GOMES FERNANDES VIEIRA;VIEIRA-MEYER, Anya Pimentel Gomes Fernandes;VIEIRA-Meyer, Anya P G F;Vieira-Mey Anya Pimentel Gomes Fernandes;VIEIRA-MEY, ANYA PIMENTEL GOMES FERNANDES;VIEIRA'MEYER, ANYA PIMENTEL GOMES FERNANDES;Anya, Pimentel Gomes Fernandes Vieira-Meyer;Anya Pimentel, Gomes Fernandes Vieira-Meyer;GOMES FERNANDES VIEIRA-MEYER, ANYA PIMENTEL;Anya Meyer;Meyer, Anya;Vieira-Meyer,;Vieira-Meyer, Anya;VIEIRA-MEYER, ANYA P. G. F</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Lattes iD</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><img src="./images/v2/icone_lattes.png" style="width: 15px; margin-right: 10px;">http://lattes.cnpq.br/7659758489387356</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Orcid iD </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><a class="icons-aviso titBottom icons-aviso-help icon-cor" href="javascript:void(0)" rel="tooltip" title="O que é ORCID? ORCID é uma organização independente sem fins lucrativos que fornece um identificador persistente - um ORCID iD - que distingue de outros pesquisadores e um mecanismo para conjuntos de resultados e atividades de pesquisa ao seu ID. O ORCID está bem localizado em editores, financiadores, instituições e outros serviços relacionados à pesquisa. Saiba mais em orcid.org " style="font-weight:bold;font-size:+1;color:blue;position:relative;top:-3px;" tabindex="110">?</a> <img src="./images/v2/icone_orcid.png" style="width: 15px; margin-right: 10px;"><a href="https://orcid.org/0000-0003-4237-8995" tabindex="111">https://orcid.org/0000-0003-4237-8995</a></div>
# </div>
# </div>
# </div><br class="clear"><div class="title-wrapper">
# <a name="Endereco" tabindex="112">
# <h1 tabindex="0">Endereço</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Endereço Profissional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Fundação Oswaldo Cruz, Presidência da Fiocruz, Escritório Fiocruz Ceará. <br class="clear">Rua São José, S/N<br class="clear">Precabura<br class="clear">61760000 - Eusébio, CE - Brasil<br class="clear">Telefone: (85) 32156450<br class="clear">URL da Homepage: <a target="blank" href="http://www.fiocruz.br" tabindex="113">http://www.fiocruz.br</a></div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="FormacaoAcademicaTitulacao" tabindex="114">
# <h1 tabindex="0">Formação acadêmica/titulação</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2001 - 2005</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Doutorado em Odontologia. <br class="clear">University of Toronto, UTORONTO, Canadá.
		
# 	<br class="clear">Título: Fluoride, Dental Fluorosis and Tooth Quality, Ano de obtenção: 2005. <br class="clear">Orientador: Dr. Marc Daniel Grynpas. <br class="clear">Bolsista do(a): U of T - Harron and Connaugh Scholarships, U OF T, Canadá. <br class="clear">Palavras-chave: Fluoretacao; Fluoride; crystal size; dureza knoop; x-ray diffraction; Dental Fluorosis. <br class="clear">Grande área: Ciências da Saúde<br class="clear">Grande Área: Ciências da Saúde / Área: Odontologia / Subárea: Materiais Odontológicos. <br class="clear">Grande Área: Ciências da Saúde / Área: Odontologia / Subárea: Biomateriais. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1999 - 2000</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Mestrado em Odontopediatria (MSc). <br class="clear">University of London, UL, Inglaterra.
		
# 	<br class="clear">Título: Orthodontic Treatment and Bacteraemia, Ano de Obtenção: 2001.<br class="clear">Orientador: Professor Graham Jonh Roberts.<br class="clear">Palavras-chave: Orthodontia; Bacteremia; Endocardite Bacteriana; Lysis Filtration; Prevencao.<br class="clear">Grande área: Ciências da Saúde<br class="clear">Grande Área: Ciências da Saúde / Área: Odontologia. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1994 - 1998</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Graduação em Odontologia. <br class="clear">Universidade Federal do Ceará, UFC, Brasil.
		
# 	<br class="clear">Bolsista do(a): Iniciacao Cientifica do Cnpq e Iniciacao Cientifica Ufc, CNPQ-UFC, Brasil. </div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="FormacaoAcademicaPosDoutorado" tabindex="115">
# <h1 tabindex="0">Pós-doutorado</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2015 - 2016</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Pós-Doutorado. <br class="clear">University of California Berkeley, UC BERKELEY, Estados Unidos.
		
# 	<br class="clear">Bolsista do(a): Conselho Nacional de Desenvolvimento Científico e Tecnológico, CNPq, Brasil. <br class="clear">Grande área: Ciências da Saúde</div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="FormacaoComplementar" tabindex="116">
# <h1 tabindex="0">Formação Complementar</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2018 - 2019</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Tecnologias e Metodologias para a Docência na Saúde.  (Carga horária: 40h). <br class="clear">Fundação Oswaldo Cruz, FIOCRUZ, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2015 - 2015</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Intensif Travailler apprendre et collaborer dans le système de santé uniqué.  (Carga horária: 40h). <br class="clear">Université de Montreal, UdeM, Canadá.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2009 - 2009</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"> Extensão universitária em Epidemiologia e Estatística para Monitoramento.  (Carga horária: 40h). <br class="clear">Universidade de Fortaleza, UNIFOR, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2009 - 2009</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Operador de Informática - nível I.  (Carga horária: 48h). <br class="clear">Universidade de Fortaleza, UNIFOR, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2008 - 2008</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"> Extensão universitária em Curso Básico de EPI INFO.  (Carga horária: 20h). <br class="clear">Universidade de Fortaleza, UNIFOR, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2008 - 2008</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"> Extensão universitária em Ciências Sociais e Humanas em Saúde.  (Carga horária: 40h). <br class="clear">Universidade de Fortaleza, UNIFOR, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2008 - 2008</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"> Extensão universitária em Introdução ao Epi Info.  (Carga horária: 20h). <br class="clear">Universidade de Fortaleza, UNIFOR, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2008 - 2008</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Abordagem por Competências.  (Carga horária: 20h). <br class="clear">Consórcio Internacional de Desenvolvimento em Educação, CIDE, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2007 - 2007</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Metodologias Problematizadoras no Ensino Superior.  (Carga horária: 20h). <br class="clear">Universidade de Fortaleza, UNIFOR, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2007 - 2007</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Curso de desenvolvimento de ensino por competência.  (Carga horária: 20h). <br class="clear">Secretaria Municipal de Saúde de Fortaleza, SMS, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2007 - 2007</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Formação de Facilitadores de Educação Permanente.  (Carga horária: 266h). <br class="clear">Fundação Oswaldo Cruz, FIOCRUZ, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2007 - 2007</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Gestão por Resultados.  (Carga horária: 32h). <br class="clear">Programa de Cooperação para o Desenvolvimento da Embaixada do Canadá, PCDEC, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2007 - 2007</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Atualização em Formação Pedagógica em EAD.  (Carga horária: 40h). <br class="clear">Fundação Oswaldo Cruz, FIOCRUZ, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2007 - 2007</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Prevenção em Odontopediatria em 2007 e no futuro.  (Carga horária: 15h). <br class="clear">Associação Brasileira de Odontopediatria, ACEOP, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2007 - 2007</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Reabilitação bucal em Odontopediatria e Ortodontia.  (Carga horária: 16h). <br class="clear">Associação Brasileira de Odontopediatria, ACEOP, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2007 - 2007</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Estética em dentes anteriores.  (Carga horária: 16h). <br class="clear">Associação Brasileira de Odontopediatria, ACEOP, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2006 - 2006</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Seminário de Informãções para docentes.  (Carga horária: 40h). <br class="clear">Universidade de Fortaleza, UNIFOR, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2006 - 2006</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Seminário Internacional de Promoção da Saúde.  (Carga horária: 4h). <br class="clear">Universidade de Fortaleza, UNIFOR, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2005 - 2005</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Sedation in the Pediatric Dental Office. <br class="clear">University Of Toronto Facult of Dentistry, UTFD, Canadá.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2004 - 2004</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">A Modified Paediatric Life Support (PALS). <br class="clear">Hospital for Sick Children, HSC, Canadá.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2002 - 2002</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Prevenindo e Controlando a Doença Cárie.  (Carga horária: 4h). <br class="clear">Associação Paulista de Cirurgiões Dentistas, APCD, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2002 - 2002</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Odontologia Intra-uterina.  (Carga horária: 4h). <br class="clear">Associação Paulista de Cirurgiões Dentistas, APCD, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2000 - 2000</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Basic Course in Clinical Hypnosis. <br class="clear">British Society of Medical and Dental Hypnosis, BSMDH, Inglaterra.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2000 - 2000</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Facial Deformit. <br class="clear">Great Ormond Street Hospital for Children, GOSHC, Canadá.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1998 - 1998</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"> Extensão universitária em Endodontia.  (Carga horária: 20h). <br class="clear">Universidade Federal do Ceará, UFC, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1998 - 1998</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"> Extensão universitária em Programa Crutac.  (Carga horária: 240h). <br class="clear">Universidade Federal do Ceará, UFC, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1998 - 1998</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"> Extensão universitária em Curso Básico de Fotografia.  (Carga horária: 90h). <br class="clear">Universidade Federal do Ceará, UFC, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1998 - 1998</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Dentística para o clínico geral e especialista.  (Carga horária: 8h). <br class="clear">Universidade Federal do Ceará, UFC, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1998 - 1998</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Odontologia estética e adesiva - conceitos atuais.  (Carga horária: 8h). <br class="clear">Universidade Federal do Ceará, UFC, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1998 - 1998</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Como implantes e enxertos solucionam problemas est.  (Carga horária: 8h). <br class="clear">Universidade Federal do Ceará, UFC, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1997 - 1997</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"> Extensão universitária em Dentística.  (Carga horária: 12h). <br class="clear">Universidade Federal do Ceará, UFC, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1997 - 1997</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Libre Echange.  (Carga horária: 350h). <br class="clear">Alliance Française du Brésil, AFB, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1996 - 1996</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"> Extensão universitária em Odontologia de Promoção de Saúde.  (Carga horária: 12h). <br class="clear">Universidade Federal do Ceará, UFC, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1996 - 1996</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"> Extensão universitária em Tratamentos Clínicos Integrados.  (Carga horária: 12h). <br class="clear">Universidade Federal do Ceará, UFC, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1996 - 1996</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"> Extensão universitária em Prótese Clínica.  (Carga horária: 8h). <br class="clear">Universidade Federal do Ceará, UFC, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1996 - 1996</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"> Extensão universitária em Atualização em Dentística.  (Carga horária: 8h). <br class="clear">Universidade Federal do Ceará, UFC, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1996 - 1996</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"> Extensão universitária em Perio-implantes.  (Carga horária: 8h). <br class="clear">Universidade Federal do Ceará, UFC, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1996 - 1996</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"> Extensão universitária em Oclusão.  (Carga horária: 8h). <br class="clear">Universidade Federal do Ceará, UFC, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1996 - 1996</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"> Extensão universitária em Implantes.  (Carga horária: 8h). <br class="clear">Universidade Federal do Ceará, UFC, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1996 - 1996</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Mantenedor de Espaço.  (Carga horária: 4h). <br class="clear">Associação Brasileira de Odontologia, ABO, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1996 - 1996</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Odontopediatria.  (Carga horária: 8h). <br class="clear">Associação Brasileira de Odontologia, ABO, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1996 - 1996</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Atualização Clínica.  (Carga horária: 22h). <br class="clear">Associação Brasileira de Odontologia, ABO, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1996 - 1996</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Odontologia para babês.  (Carga horária: 4h). <br class="clear">Associação Brasileira de Odontologia, ABO, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1996 - 1996</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Periodontia.  (Carga horária: 8h). <br class="clear">Associação Brasileira de Odontologia, ABO, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1996 - 1996</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Odontopediatria (teórico).  (Carga horária: 8h). <br class="clear">Associação Brasileira de Odontologia, ABO, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1996 - 1996</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Curso com prof. John Burges/University of Texas.  (Carga horária: 8h). <br class="clear">Grupo Brasileiro de Materiais Dentários, GBMD, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1996 - 1996</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Estética.  (Carga horária: 4h). <br class="clear">Associação Brasileira de Odontologia, ABO, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1995 - 1995</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"> Extensão universitária em Prótese Clínica.  (Carga horária: 8h). <br class="clear">Universidade Federal do Ceará, UFC, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1995 - 1995</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Teacher´s Training Course.  (Carga horária: 50h). <br class="clear">Yázigi Internacional, YI, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1995 - 1995</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Dentística Estética.  (Carga horária: 12h). <br class="clear">Universidade Federal do Ceará, UFC, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1995 - 1995</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Atualização Clínica. <br class="clear">Associação Brasileira de Odontologia, ABO, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1995 - 1995</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Radio-diagnóstico oral.  (Carga horária: 8h). <br class="clear">Associação Brasileira de Odontologia, ABO, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1994 - 1994</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Diagnóstico Oral (teórico).  (Carga horária: 30h). <br class="clear">Associação Brasileira de Odontologia, ABO, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1994 - 1994</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Controle de Infecção nos Consultórios.  (Carga horária: 20h). <br class="clear">Associação Brasileira de Odontologia, ABO, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1993 - 1993</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Francês. <br class="clear">Arlington High School, AHS, Estados Unidos.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1991 - 1991</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Inglês.  (Carga horária: 600h). <br class="clear">Instituto Brasileiro - Estados Unidos no Ceará, IBEUC, Brasil.
		
# 	</div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="AtuacaoProfissional" tabindex="117">
# <h1 tabindex="0">Atuação Profissional</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell"><br class="clear"><div class="inst_back">
# <b>Universidade Estadual Vale do Acaraú, UVA-CE, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2022 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: , Enquadramento Funcional: </div>
# </div><br class="clear"><br class="clear"><div class="inst_back">
# <b>Harvard School of Public Health, HSPH, Estados Unidos.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2021 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: , Enquadramento Funcional: </div>
# </div><br class="clear"><br class="clear"><div class="inst_back">
# <b>Universidade Estadual do Ceará, UECE, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2019 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: , Enquadramento Funcional: </div>
# </div><br class="clear"><br class="clear"><div class="inst_back">
# <b>FIOCRUZ/CEARÁ, FIOCRUZ/CE, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2010 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Servidor Público, Enquadramento Funcional: Servidor Publico Federal, Carga horária: 40</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2009 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Servidor Público, Enquadramento Funcional: Coordenadora Geral do CMPSF/RENASF, Carga horária: 8</div>
# </div><br class="clear"><br class="clear"><div class="inst_back">
# <b>Fundação Oswaldo Cruz, FIOCRUZ, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2011 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Servidor Público, Enquadramento Funcional: Pesquisadora Especialista em Saúde da Familia, Carga horária: 40, Regime: Dedicação exclusiva. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Outras informações</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Coordenadora do Mestrado Profissional em Saúde da Família - MPSF 
# Docente dos Módulos Atenção e Integral a Saúde da Família e Atenção Gestão do Cuidado</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2010 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Servidor Público, Enquadramento Funcional: Professor titular, Carga horária: 40</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Outras informações</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Coordenadora Geral do Programa de Pós-Graduação Stricto Senso Profissional em Saúde da Família em Rede - RENASF (Fiocruz CE, UFMA, UFPI, UFC, UECE, UVA, URCA, UFRN, UFPB).</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Atividades</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2015 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Pesquisa e desenvolvimento, Fundação Oswaldo Cruz - Ceará. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Linhas de pesquisa <br class="clear"><a href="#LP_Impacto do Programa Bolsa Família e da Estratégia Saúde da Família na saúde da população" tabindex="118">Impacto do Programa Bolsa Família e da Estratégia Saúde da Família na saúde da população</a><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>02/2010 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Ensino, Mestrado Profissional em Saúde da Família, Nível: Pós-Graduação</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Disciplinas ministradas<br class="clear">Coordnadora e docente do Programa de mestrado profissional em saúde da família - RENASF<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>01/2010 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Direção e administração, Fundação Oswaldo Cruz - Ceará. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Cargo ou função<br class="clear">Coordenadora Geral do Programa de Pós-Graduação Stricto Sensu Profissional em Saúde da Família em Rede - RENASF (Fiocruz, UFMA, UFPI, UFC, UECE, UVA, URCA, UFRN, UFPB).</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>02/2014 - 12/2014</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Direção e administração, Fundação Oswaldo Cruz - Ceará. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Cargo ou função<br class="clear">Coordenadora interina do Escritório Fiocruz - Ceará.</div>
# </div><br class="clear"><div class="inst_back">
# <b>Centro Universitário Christus, UNICHRISTUS, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2014 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Celetista, Enquadramento Funcional: Professora do curso de odontologia, Carga horária: 20</div>
# </div><br class="clear"><br class="clear"><div class="inst_back">
# <b>Ministério da Saúde, MS, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2014 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: ,, Enquadramento Funcional: Coordenadora, Carga horária: 3</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Outras informações</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">PMAQ-AB: Avaliação Externa no Estado do Ceará "Saúde Mais Perto de Você" - 2 Ciclo, aprovado pela PROPESQ n PDV 10304/2013 / PROPLAN n 60/2013 / GESCON n 243665710001/13-002. Termo de cooperação n 25000.107681/2013-90</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2011 - 2012</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: ,, Enquadramento Funcional: Coordenadora, Carga horária: 3</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Outras informações</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">(PMAQ-AB - Saúde Bucal e Avaliação Externa no Estado do Ceará - 1 Ciclo), projeto PROPLAN: 606/2011 - FUNPEC: 86/2011 GESCON: 243657/0000 1120-29, Departamento de Saúde Coletiva da Universidade Federal do Rio Grande do Norte/UFRN</div>
# </div><br class="clear"><br class="clear"><div class="inst_back">
# <b>Universidade de Fortaleza, UNIFOR, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2006 - 2012</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Professora do Mestrado, Enquadramento Funcional: Professor, Carga horária: 20</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Outras informações</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Professora do Mestrado Saúde Coletiva (lesionando e orientando alunos) e Professora na Graduação (Curso de Farmácia e Odontologia).</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Atividades</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>02/2006 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Pesquisa e desenvolvimento, Mestrado de Saúde Coletiva. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Linhas de pesquisa <br class="clear"><a href="#LP_Acesso a serviços de Saúde em população de risco (HIV, crianças, idosos, trabalhadores, etc.)" tabindex="119">Acesso a serviços de Saúde em população de risco (HIV, crianças, idosos, trabalhadores, etc.)</a><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>02/2006 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Ensino, Saúde Coletiva, Nível: Pós-Graduação</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Disciplinas ministradas<br class="clear">Seminário de Pesquisa<br class="clear">Bioestatística<br class="clear">Epidemiologia<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>02/2006 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Ensino, Saúde Coletiva, Nível: Graduação</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Disciplinas ministradas<br class="clear">Odontopediatria<br class="clear">Saúde Bucal Coletiva<br class="clear">Saúde Coletiva<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>02/2006 - 12/2008</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Pesquisa e desenvolvimento, Mestrado de Educação em Saúde. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Linhas de pesquisa <br class="clear"><a href="#LP_Avaliação dos serviços de saúde" tabindex="120">Avaliação dos serviços de saúde</a><br class="clear"></div>
# </div><br class="clear"><div class="inst_back">
# <b>Prefeitura Municipal de Fortaleza, PMF, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2006 - 2011</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Servidor Público, Enquadramento Funcional: Odontologa do Programa Saúde da Família, Carga horária: 40</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2005 - 2005</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Revisora técnica, Enquadramento Funcional: Revisora Técnica das "Normas técnicas: Ações</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Outras informações</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Revisao técnica das "Normas técnicas: Ações coletivas em Saúde Bucal"</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Atividades</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>01/2007 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Outras atividades técnico-científicas , Secretaria Municipal de Saúde, Secretaria Municipal de Saúde. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Atividade realizada<br class="clear">Assessoria Técnica do Departamento de Educação Permanente da SMS. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>08/2006 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Serviços técnicos especializados , Secretaria Municipal de Saúde. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Serviço realizado<br class="clear">Dentista do Programa Saúde da Família. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>08/2005 - 10/2005</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Outras atividades técnico-científicas , Secretaria de Saúde, Secretaria de Saúde. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Atividade realizada<br class="clear">Revisora Técnica. </div>
# </div><br class="clear"><div class="inst_back">
# <b>Associacao Brasileira de Odontologia, ABO-CE, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2005 - 2012</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Professora, Enquadramento Funcional: Professora do Curso de Especialização, Carga horária: 4</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Outras informações</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Professora do Curso de Especialização em Odontopediatria (lesionando e orientando alunos)</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2005 - 2006</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Professora colaboradora, Enquadramento Funcional: Professora colaboradora, Carga horária: 4</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Outras informações</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Professora colaboradora dos cursos de aperfeiçoamento e atualização em Saúde da Família</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1998 - 1999</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Pesquisadora, Enquadramento Funcional: Administrativo - Pesquisa  Coodernadora do P, Carga horária: 20</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Outras informações</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"> Coodernadora do Projeto de pesquisa: Vigilância sanitária da fluoretação das águas de abastecimento público no Município de Fortaleza-CE</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Atividades</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2/2006 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Ensino, Odontopediatria, Nível: Especialização</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Disciplinas ministradas<br class="clear">Odontopediatria<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>10/2005 - 04/2006</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Ensino, Saúde da Família, Nível: Aperfeiçoamento</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Disciplinas ministradas<br class="clear">Curso Saúde da Família<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>08/1998 - 08/1999</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Pesquisa e desenvolvimento, ABO. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Linhas de pesquisa <br class="clear"><a href="#LP_Fluor e Fluorose Dental" tabindex="121">Fluor e Fluorose Dental</a><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>07/1998 - 07/1999</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Direção e administração, ABO-Ce. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Cargo ou função<br class="clear"> Coodernadora do Projeto de pesquisa: Vigilância sanitária da fluoretação das águas de abastecimento público no Município de Fortaleza-CE.</div>
# </div><br class="clear"><div class="inst_back">
# <b>Academia Cearense de Odontologia, ACO/CEC, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2005 - 2007</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Colaborador, Enquadramento Funcional: Professora colaboradora, Carga horária: 1</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Outras informações</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Professora colaboradora do curso de especialização em odontopediatria</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Atividades</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>07/2005 - 12/2007</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Ensino, Odontopediatria, Nível: Especialização</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Disciplinas ministradas<br class="clear">Odontopediatria<br class="clear"></div>
# </div><br class="clear"><div class="inst_back">
# <b>Conselho Nacional de Desenvolvimento Científico e Tecnológico, CNPq, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1998 - 1999</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Bolsista, Enquadramento Funcional: Pesquisadora, Carga horária: 20</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Outras informações</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Fui bolsista do CNPq - Bolsa de Aperfeicoamento Atividade de Pesquisa CNPq (para formados) durante o periodo de agosto de 1998 a julho de 1999. Trabalhei na Coordenação do Projeto ?Prevenção de Cáries em Crianças?, desenvolvido em escolas na periferia de Fortaleza - CE</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1996 - 1998</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Pesquisadora, Enquadramento Funcional: Bolsista, Carga horária: 20</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Outras informações</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Pesquisa na área de fluorterapia, materiais odontológicos e anatomia</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Atividades</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>04/2007 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Pesquisa e desenvolvimento, Mestrado de Saúde Coletiva. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Linhas de pesquisa <br class="clear"><a href="#LP_Saúde da Família" tabindex="122">Saúde da Família</a><br class="clear"></div>
# </div><br class="clear"><div class="inst_back">
# <b>University of London, UL, Inglaterra.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1999 - 2000</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Pesquisa, Enquadramento Funcional: Aluna Mestrado, Carga horária: 40, Regime: Dedicação exclusiva. </div>
# </div><br class="clear"><br class="clear"><div class="inst_back">
# <b>University of Toronto, UTORONTO, Canadá.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2001 - 2005</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Pesquisadora, Enquadramento Funcional: Aluna de Doutorado, Carga horária: 40, Regime: Dedicação exclusiva. </div>
# </div><br class="clear"><br class="clear"><div class="inst_back">
# <b>Rotary Club, ROTARY, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1998 - 1999</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Pesquisadora, Enquadramento Funcional: Administrativo - Pesquisa, Carga horária: 20</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Outras informações</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Coordenadora do Projeto ?Prevenção de Cáries em Crianças?, desenvolvido em escolas na periferia de Fortaleza - CE</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Atividades</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>07/1998 - 07/1999</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Direção e administração, Rotary Club - Programa de Prevenção. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Cargo ou função<br class="clear"> Coordenadora do Projeto ?Prevenção de Cáries em Crianças?, desenvolvido em escolas na periferia de Fortaleza - CE.</div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="LinhaPesquisa" tabindex="123">
# <h1 tabindex="0">Linhas de pesquisa</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell"><a name="LP_Saúde da Família" tabindex="124"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Saúde da Família</div>
# </div><a name="LP_Fluor e Fluorose Dental" tabindex="125"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Fluor e Fluorose Dental</div>
# </div><a name="LP_Acesso a serviços de Saúde em população de risco (HIV, crianças, idosos, trabalhadores, etc.)" tabindex="126"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Acesso a serviços de Saúde em população de risco (HIV, crianças, idosos, trabalhadores, etc.)</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Objetivo: Avaliar conhecimentos, práticas, cuidados e condições de saúde, em especial saúde bucal, de pessoas em grupos de risco.. </div>
# </div><a name="LP_Avaliação dos serviços de saúde" tabindex="127"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Avaliação dos serviços de saúde</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Objetivo: Avaliar serviços de saúde na área na atenção primária - sob a ótica do usuário, profissionais de saúde e gestores.. </div>
# </div><a name="LP_Impacto do Programa Bolsa Família e da Estratégia Saúde da Família na saúde da população" tabindex="128"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Impacto do Programa Bolsa Família e da Estratégia Saúde da Família na saúde da população</div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="ProjetosPesquisa" tabindex="129">
# <h1 tabindex="0">Projetos de pesquisa</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell"><a name="PP_PESQUISA DE AVALIABILIDADE DA POLÍTICA NACIONAL DE PROMOÇÃO DA SAÚDE" tabindex="130"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2022 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">PESQUISA DE AVALIABILIDADE DA POLÍTICA NACIONAL DE PROMOÇÃO DA SAÚDE<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: O projeto desenvolve um estudo de avaliabilidade da Política Nacional de Promoção da Saúde (PNPS). Constitui-se em um estudo multicêntrico, multimétodos, desenvolvido em duas etapas: Estudo de Avaliabilidade e Estudos de Casos Múltiplos. Realizado nos estados de Goiás, Ceará, Pernambuco, Maranhão, Tocantins, São Paulo,  Minas Gerais, Rio Grande do Sul, Paraná e Distrito Federal; e incluindo 50 municípios representativo destes. O estudo tem o foco em municípios que tenham a promoção da saúde em sua estrutura organizacional ou no cerne de suas ações, e que estejam em diferentes estratos de porte populacional. A magnitude e a processualidade deste Estudo de Avaliabilidade requer como unidades de análise documentos (nacionais, estaduais e municipais) que subsidiam e registram a implementação da Promoção da Saúde nos cenários; e informantes-chave (gestores, profissionais da saúde, usuários, e representantes da sociedade civil) nos locais onde são tencionados movimentos concernentes à Promoção da Saúde. Como itinerário empírico analítico, considerando o método que reveste a primeira  etapa da pesquisa, esta organiza-se na delimitação da política, desenho do modelo teórico-lógico, elaboração de perguntas avaliativas e elaboração de recomendações. Para segunda etapa serão consideradas duas técnicas de coleta a observação não-participante e a entrevista em profundidade a fim de  colaborar com a exploração do fenômeno. Para análise, serão resgatadas as proposições fundantes desta pesquisa e propor-se-á realizar uma síntese cruzada dos dados. O estudo é regido  pelos preceitos éticos descritos na Resolução Nº 466/2012 do CNS. Pesquisa de interesse nacional financiada pelo Ministério da Saúde, mediante carta acordo entre OPAS, Fiocruz-CE e Universidade Estadual Vale do Acaraú. A relevância desta pesquisa reside no tempo de implantação da política e a incipiência de estudos com foco na avaliação da PNPS em vigor, abordando os desafios de sua implementação nas distintas regiões brasileiras. Ademais, identifica-se dificuldade da gestão e da academia em produzir evidências sobre a implementação da PNPS, com delineamento de seus indicadores de monitoramento e avaliação de ações intra e intersetoriais relacionadas. Situação que pode contribuir para a invisibilidade do processo e dos resultados da PNPS para tomadores de decisão e sociedade, realçando a necessidade de reconhecimento dos fatores estruturais e processuais que possam estar contribuindo para esta invisibilidade ou para um processo de implementação distinto do desenho inicialmente proposto. Antecipa-se que os resultados deste estudo poderão colaborar na efetivação da PNPS nos diversos entes federados. 

# Equipe: Maria Socorro de Araújo Dias (Coordenação), Anya Pimentel Gomes Fernandes Vieira Meyer, Dais Gonçalves Rocha, Maria da Conceição Coelho Brito, Lielma Carla Chagas da Silva.
# Apoiadores Estaduais: Claudia Schneck, Cristianne Maria Famer Rocha, Kelva Karina Nogueira de Carvalho de Aquino, Luisa Azevedo Silveira, Márcia Maria Santos da Silva, Marco Antonio de Moraes, Rejane Christine de Sousa Queiroz, Roberta Corrêa de Araújo de Amorim, Rosane da Silva Santos, Simone Matias Gondim Silva.
# Instituições Executoras: Universidade Estadual Vale do Acaraú/Fundação Oswald Cruz-Ceará.. <br class="clear">Situação: Em andamento; Natureza: Pesquisa. <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Integrante / DE ARAÚJO DIAS, MARIA SOCORRO - Coordenador / Dais Gonçalves Rocha - Integrante / Maria da Conceição Coelho Brito - Integrante / Lielma Carla Chagas da Silva - Integrante.<br class="clear"></div>
# </div><a name="PP_EFFECT OF COVID-19 AND VIOLENCE ON THE WORK PROCESS AND MENTAL HEALTH OF COMMUNITY HEALTH AGENTS IN BRAZIL" tabindex="131"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2021 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">EFFECT OF COVID-19 AND VIOLENCE ON THE WORK PROCESS AND MENTAL HEALTH OF COMMUNITY HEALTH AGENTS IN BRAZIL<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: The COVID-19 pandemic has been a challenge for countries with profound social inequalities including Brazil. Vulnerable population groups, already exposed to adversities including violence, face the most serious impacts from the COVID-19 pandemic. Violence is recognized as a multifaceted, complex phenomenon, which is a significant public health problem that compromises health and impacts both access and demand for health services often compounded by poor governance. Moreover, the epidemiological picture of the social and health crisis of COVID-19, the increase in the number of people unemployed and social isolation can have repercussions on the levels of violence and crime, historically higher in contexts of social vulnerability, as well as on the health agent?s mental health and work process, thereby negatively affecting their ability to provide care. However, there are no studies that analyze the relationships between violence and countering COVID-19 in vulnerable territories. Countries that have been successful in tackling and controlling this pandemic have done so with actions to control the spread of the virus, identifying infected patients and preventing them from spreading the infection through isolation. Data recently indicated that combined isolation and screening strategies would reduce the transmission of COVID-19 more than mass testing alone or self-isolation alone. Well established Primary Health Care (PHC) systems can effectively contribute to the challenges posed by the COVID-19 pandemic.Brazil has a public-funded unified health system (SUS), established more than 20 years ago. The PHC system is organized by the Family Health Strategy (FHS) through Family Health Teams (FHT), which comprise a family doctor, a nurse, a dentist, nurse auxiliaries and Community Health Agents (CHAs). The CHAs are individuals from the community trained to provide a range of services in the territories, including home visits, health promotion activities, and serve as liaisons between health units and communities. In this context, the CHA has the potential to play an important role in fighting the pandemic by working on contact tracing, collecting information on infected people and providing guidance to them and the community in order to contain community transmission. However, contextual issues, such as violence, can influence their work process and their role in combating COVID-19 and other health problems, especially in groups living in vulnerable territories. Unfortunately, Brazil has a very high level of community violence, among the 30 most violent cities in the world, there are six Northeastern capitals, namely: Natal (4°), Fortaleza (7°), Maceió (14°) Aracaju (18°), Recife (22°) and João Pessoa (30°). Furthermore, this is one of the poorest regions in Brazil, with poor social indicators including high illiteracy, low human development index, and high utilization of the Brazilian conditional cash transfer program (Bolsa Família Program). It can be hypothesized that the violence in the territory affect the CHAs? mental health and work process, including absenteeism and their ability to deal with the COVID-19 pandemic and other health problems. CHA characteristics (e.g., age, gender, experience), self-efficacy and coping mechanisms, as well as their social and structural support may moderate the impact and quality of health care provided. Therefore, it is critical to investigate the interrelationships between violence, COVID-19 and the impact on CHAs? wellbeing and work using a robust research methodology to generate pertinent evidence for policy makers, which can inform strategies and public policies to tackle COVID-19 and to support the health workforce in the communities with high exposure to adversities. Thus, this study aim is to analyzethe associations between violence, COVID-19, CHA work process and CHA mental health. <br class="clear">Situação: Em andamento; Natureza: Pesquisa. <br class="clear">Alunos envolvidos: Graduação: (6)  / Mestrado profissional: (1) . <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Coordenador / Maria Socorro Araújo Dias - Integrante / Ana Patricia Pereira Morais - Integrante / Maristela Inês Osawa Chagas - Integrante / José Maria Ximenes Guimarães - Integrante / Alice Maria Correia Pequeno - Integrante / Franklin Delano Soares Forte - Integrante / Claudete Ferreira de Souza Monteiro - Integrante / Maria de Fátima. A. S. Machado - Integrante / sidney feitoza farias - Integrante / André Luiz Sá Oliveira - Integrante / Elaine F do Nascimento - Integrante / Fernando José G S Júnior - Integrante / Maria Rosilene Cândido Moreira - Integrante / Milena Silva Costa - Integrante / Paulo Goberlanio Barros Silva - Integrante / Aisha Yousafzai - Integrante / carlos Henrique Campos Meyer - Integrante.<br class="clear">Financiador(es): Harvard School of Public Health - Auxílio
# 										financeiro.</div>
# </div><a name="PP_Impacto da atenção primária à saúde na morbimortalidade materno-infantil no Brasil" tabindex="132"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2020 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Impacto da atenção primária à saúde na morbimortalidade materno-infantil no Brasil<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: O projeto tem dois objetivos: 1) Avaliar o impacto da Atenção Primária á Saúde (APS) nos indicadores de morbimortalidade materno-infantil. O estudo utilizará os dados da avaliação externa do Programa Nacional para Melhoria do Acesso e da Qualidade da Atenção Primária (PMAQ), que serão linkados com os dados individuados de registros administrativos do Cadastro Único (CADU) e dos Sistemas de Informação em Saúde (SIS). Espera-se que os resultados desta pesquisa forneçam subsídios para o aprimoramento das políticas de saúde no Brasil, com o intuito de melhorar as condições de saúde da população. 2) Promover a  expansão do Centro de Integração de Dados e Conhecimentos para a Saúde (CIDACS) e o uso das Coortes epidemiológicas, desenvolvendo um planejamento estratégico para o aprimoramento da gestão de dados relacionados a interoperabilidade, ontologias, processamento, vinculação, acurácia e compartilhamento de dados, além da ampliação da capacidade computacional e ampliação de parcerias nacionais e internacionais.

# Financiado pela Bill e Melinda Gates Foundation. <br class="clear">Situação: Em andamento; Natureza: Pesquisa. <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Integrante / Helena Paula Guerra dos Santos - Integrante / Mauricio Barreto - Coordenador / Elzo Pereira Pinto Junior - Integrante / Rosana aquino guimaraes pereira - Integrante / Maria Yury Ichihara - Integrante / Valentina Martufi - Integrante / Leila Denise Alves Ferreira Amorim - Integrante / Cristiana Paula Matos de Almeida - Integrante / Carlos Antônio de Souza Teles Santos - Integrante / Robespierre Dantas da Rocha Pita - Integrante / Roberto Pérez Carreiro - Integrante / Samila Oliveira Lima Sena - Integrante / Maira Lima de Souza - Integrante / Thomas Hone - Integrante / Christopher Millett - Integrante / Acácia Mayra Pereira de Lima - Integrante / José Carlos Vaz - Integrante.<br class="clear"></div>
# </div><a name="PP_Impacto da violência no processo de trabalho e saúde mental dos agentes comunitários desaúde" tabindex="133"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2020 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Impacto da violência no processo de trabalho e saúde mental dos agentes comunitários desaúde<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: Estudo multicêntrico, desenvolvido em quatro capitais do Nordeste brasileiro: Fortaleza, João Pessoa, Recife e Teresina. Tem por objetivo analisar as dimensões da violência vivenciada pelos ACS e seu impacto na saúde e no processo de trabalho da estratégia saúde da família no contexto da pandemia de COVID-19, assim como desenvolver ações de disseminação da informação e mitigação do impacto da violência nos ACS.  Trata-se de estudo de métodos mistos, na modalidade sequencial, combinando-se a abordagem quantitativa descritiva e a qualitativa hermenêutica. Os achados da presente pesquisa apoiarão a implementação de estratégias e políticas públicas para a melhoria da saúde do trabalhador, e a qualificação do processo de trabalho do ACS ante a convivência com a violência no território, e a necessidade de seu enfrentamento em territótios  vulneráveis no contexto da pandemia de COVID-19.
# Palavras-chaves: Violência. Saúde mental. Agentes Comunitários de Saúde. Atenção Primária à Saúde.  
# Projeto selecionado pelo Edital 2020, com foco na Atenção Primária de Saúde, do Programa de Políticas Públicas e Modelos de Atenção e Gestão à Saúde (Fiocruz/VPPCB/PMA). <br class="clear">Situação: Em andamento; Natureza: Pesquisa. <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Coordenador / Antônio Silva Lima Neto - Integrante / José Maria Ximenes Guimarães - Integrante / Alice Maria Correia Pequeno - Integrante / kilma Wanderley Lopes Gomes - Integrante / Franklin Delano Soares Forte - Integrante / Ana Patrícia Pereira Morais - Integrante / Cibelly Melo Ferreira - Integrante / Maria de Fátima Antero Sousa Machado - Integrante / CAMPELO, ISABELLA LIMA BARBOSA - Integrante / VIEIRA, NEIVA FRANCENELY CUNHA - Integrante / sidney feitoza farias - Integrante / Marcia Cladas Castro - Integrante / André Luiz Sá Oliveira - Integrante / Elaine F do Nascimento - Integrante / Fernando José G S Júnior - Integrante / Claudete Monteiro - Integrante / Altamira P S Reichert - Integrante / Maximiliano Loiola P de Souza - Integrante / Elenice Araújo Andrade - Integrante / Janaína Rocha de Sousa Almeida - Integrante / Felipe Proenço Oliveira - Integrante / Luana Rodrigues de Almeida - Integrante / carlos Henrique Campos Meyer - Integrante.<br class="clear">Financiador(es): Fundação Oswaldo Cruz - Auxílio
# 										financeiro.</div>
# </div><a name="PP_COVID19 e A VIOLÊNCIA URBANA: REPERCUSSÕES NO TRABALHO E NA SAÚDE MENTAL DOS AGENTES COMUNITÁRIOS DE SAÚDE NO ESTADO DO CEARÁ" tabindex="134"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2020 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">COVID19 e A VIOLÊNCIA URBANA: REPERCUSSÕES NO TRABALHO E NA SAÚDE MENTAL DOS AGENTES COMUNITÁRIOS DE SAÚDE NO ESTADO DO CEARÁ<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: Estudo multicêntrico, desenvolvido nos municípios cearenses: Fortaleza, Sobral, Crato, Juazeiro do Norte e Barbalha. Considera-se que o trabalho em saúde em territórios de violência urbana é desafiador, sobretudo no contexto da pandemia de Covid-19, podendo ter impacto negativo no trabalho e na saúde mental dos Agentes Comunitários de Saúde. Nesse contexto, é possível a ocorrência de limitações do acesso, assim como a redução da oferta de cuidado, nos territórios de vulnerabilidade e governança limitada. Tem-se por objetivo analisar as associações entre COVID-19, violência, condução dos cuidados de outros problemas de saúde na população, processo de trabalho e saúde mental de Agentes Comunitários de Saúde. Trata-se de estudo transversal, de natureza quantitativa, descritivo-analítico, que será realizado em três fases: aplicação de questionários aos ACS; aquisição de dados secundários e organização de banco de dados; e, construção de Mapas de Violência e COVID-19, assim como a disseminação do conhecimento gerado. Os achados da presente pesquisa apoiarão a implementação de estratégias e políticas públicas para o enfrentamento da COVID-19, a melhoria da saúde do trabalhador, e a qualificação do processo de trabalho do ACS ante a convivência com a violência no território.

# EDITAL 01/2020 PROGRAMA INOVA FIOCRUZ-CE/Funcap. <br class="clear">Situação: Em andamento; Natureza: Pesquisa. <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Coordenador / Neiva Francenely Cunha Vieira - Integrante / Ana Patricia Pereira Morais - Integrante / José Maria Ximenes Guimarães - Integrante / Maristela Inês Osawa Vasconcelos - Integrante / Alice Maria Correia Pequeno - Integrante / DIAS, MARIA DO SOCORRO DE ARAÚJO - Integrante / Cibelly Melo Ferreira - Integrante / MACHADO, MARIA DE FATIMA A S - Integrante / CAMPELO, ISABELLA LIMA BARBOSA - Integrante / André Luiz Sá Oliveira - Integrante / Maximiliano Loiola P de Souza - Integrante / Elenice Araújo Andrade - Integrante / Janaína Rocha de Sousa Almeida - Integrante / CANDEIRO, GEORGE T DE M - Integrante / Maria Rosilene Cândido Moreira - Integrante / Milena Silva Costa - Integrante / Erlemus Ponte Soares - Integrante / Paulo Goberlanio Barros Silva - Integrante / carlos Henrique Campos Meyer - Integrante.<br class="clear">Financiador(es): Fundação de Amparo a Pesquisa do Ceará - Auxílio
# 										financeiro.</div>
# </div><a name="PP_Produção do cuidado integral às pessoas com sífilis na atenção primária à saúde: saberes, práticas e fluxos na articulação com a rede de atenção à saúde" tabindex="135"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2019 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Produção do cuidado integral às pessoas com sífilis na atenção primária à saúde: saberes, práticas e fluxos na articulação com a rede de atenção à saúde<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: Aborda a produção do cuidado às pessoas com sífilis no âmbito da atenção primária à saúde, com ênfase nos saberes e práticas de cuidado. Nesse contexto, considera acesso, coordenação do cuidado e fluxos construídos para atenção aos usuários na articulação com a rede de atenção à saúde, cuja organização dos serviços de saúde par o cuidado integral requer a identificação das necessidades de saúde do território e a integração entre vigilância e atenção à saúde. Reconhece-se o papel estratégico das equipes da Estratégia Saúde da Família na prevenção, na detecção da sífilis adquirida e em gestantes, bem como no tratamento. Considerando-se a transmissão vertical da mãe para o feto, o diagnóstico oportuno e a oferta do tratamento adequado da sífilis na gravidez ainda são desafios para o controle da sífilis congênita e importante indicador de qualidade da APS. Tem-se por objetivo compreender os modos de produção do cuidado às pessoas com sífilis, com vistas ao desenvolvimento de processos e tecnologias de organização dos serviços e das práticas de saúde no âmbito da atenção primária à saúde em sua articulação com a redes de atenção à saúde, sob a ótica do acesso, integralidade, resolubilidade e qualidade. Trata-se de um estudo de caso, de natureza quanti-qualitativa. Tem como participantes os gestores, trabalhadores e usuários diagnósticos com sífilis atendidos na Estratégia Saúde da Família..
# Situação: Em andamento; Natureza: Pesquisa.
# Alunos envolvidos: Graduação: (1) / Especialização: (0) / Mestrado profissional: (3). <br class="clear">Situação: Em andamento; Natureza: Pesquisa. <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Integrante / Ana Patricia Pereira Morais - Integrante / José Maria Ximenes Guimarães - Coordenador / Carlos Garcia Filho - Integrante / Mirnis Aparecida Macedo - Integrante / Ednaiane Priscila de Andrade Amorim - Integrante / Mariana Vale Francelino Sampaio - Integrante.<br class="clear"></div>
# </div><a name="PP_Projeto de Pesquisa Violência ACS, Saúde mental e processo de trabalho" tabindex="136"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2018 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Projeto de Pesquisa Violência ACS, Saúde mental e processo de trabalho<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: Pesquisa para avaliar o impacto da violência comunitária no processo de trabalho e na saúde mental dos agentes comunitários de saúde de Fortaleza. <br class="clear">Situação: Em andamento; Natureza: Pesquisa. <br class="clear">Alunos envolvidos: Graduação: (1)  / Mestrado profissional: (1) . <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Coordenador / Ana Patricia Pereira Morais - Integrante / Anya Pimentel Gomes Fernandes Vieira-Meyer - Integrante / Helena Paula Guerra dos Santos - Integrante / carlos Henrique Campos Meyer - Integrante.<br class="clear"></div>
# </div><a name="PP_ESTUDO DE PREVALÊNCIA DE DOENÇA PERIODONTAL, MALOCLUSÕES, LESÕES CERVICAIS NÃO CARIOSAS E DE CÁRIE DENTÁRIA EM COMUNIDADES DE GUINÉ-BISSAU" tabindex="137"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2017 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">ESTUDO DE PREVALÊNCIA DE DOENÇA PERIODONTAL, MALOCLUSÕES, LESÕES CERVICAIS NÃO CARIOSAS E DE CÁRIE DENTÁRIA EM COMUNIDADES DE GUINÉ-BISSAU<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: O projeto visa estabelecer o primeiro levantamento de doenças e necessidades orais em indivíduos adultos e crianças em Guiné-Bissau. De janeiro a abril de 2023 realizamos o primeiro levantamento de dados (doença periodontal, cárie, lesão cervical não-cariosa e malocusão) em várias regiões do país. Esses resultados culminaram na defesa de 05 dissertações de mestrado e representam o primeiro estudo de doenças orais em subpopulações de Guiné-Bissau. Esses resultados preliminares contribuirão para o estabelecimento de políticas de saúde bucal junto ao governo guineense. <br class="clear">Situação: Em andamento; Natureza: Pesquisa. <br class="clear">Alunos envolvidos: Mestrado acadêmico: (5) . <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Integrante / Juliana Paiva Prado - Integrante / Fabrício Bitu Sousa - Coordenador / CANDEIRO, GEORGE T DE M - Integrante / DAMASCENO, JULIANA XIMENES - Integrante / Paulo Goberlanio Barros Silva - Integrante / Ramille Araujo Lima - Integrante / Thinali Sousa Dantas - Integrante.<br class="clear"></div>
# </div><a name="PP_Estratégia Saúde da Família e Programa Bolsa Família: seus avanços e impacto na saúde da populacão" tabindex="138"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2015 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Estratégia Saúde da Família e Programa Bolsa Família: seus avanços e impacto na saúde da populacão<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: Avaliação dos programas Bolsa Família e Saúde da Família na saúde da população. <br class="clear">Situação: Em andamento; Natureza: Pesquisa. <br class="clear">Alunos envolvidos:  / Mestrado profissional: (1) . <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Coordenador / Paola Gondim Calvasina - Integrante / SAINTRAIN, MARIA VIEIRA DE LIMA - Integrante / Helena Paula Guerra dos Santos - Integrante.<br class="clear"></div>
# </div><a name="PP_Pesquisa de avaliação Externa do Programa de Melhoria do Acesso e da Qualidade da Atenção Básica (PMAQ-AB)" tabindex="139"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2013 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Pesquisa de avaliação Externa do Programa de Melhoria do Acesso e da Qualidade da Atenção Básica (PMAQ-AB)<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: (PMAQ-AB: Avaliação Externa no Estado do Ceará "Saúde Mais Perto de Você" - 2° Ciclo), aprovado pela PROPESQ n° PDV 10304/2013 / PROPLAN n° 60/2013 / GESCON n° 243665710001/13-002. Termo de cooperação n° 25000.107681/2013-90. <br class="clear">Situação: Em andamento; Natureza: Pesquisa. <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Coordenador / Maria Socorro Araújo Dias - Integrante / Ana Patricia Pereira Morais - Integrante / Cibelly Aliny SiqueiraLima Freitas - Integrante / Fabiane do Amaral Gubert - Integrante / Eliany Nazaré Oliveira - Integrante / Maria de Fatima Antero Sousa Machado - Integrante / Ana Mattos Brito de Almeida - Integrante / Emilia Soares Chaves - Integrante / Maristela Inês Osawa Vasconcelos - Integrante / Yana Paula Coelho Correia Sampaio - Integrante / Evanira Rodrigues Maia - Integrante / Rogério Sampaio de Oliveira - Integrante / Alice Maria Correia Pequeno - Integrante / Caroline Antero Machado Mesquita - Integrante / Paola Colares de Borba - Integrante.<br class="clear">Financiador(es): Ministério da Saúde - Auxílio
# 										financeiro.</div>
# </div><a name="PP_Pesquisa de avaliação Externa do Programa de Melhoria do Acesso e da Qualidade da Atenção Básica (PMAQ-AB)" tabindex="140"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2011 - 2012</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Pesquisa de avaliação Externa do Programa de Melhoria do Acesso e da Qualidade da Atenção Básica (PMAQ-AB)<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: (PMAQ-AB - Saúde Bucal: Avaliação Externa no Estado do Ceará - 1° Ciclo), projeto PROPLAN: 606/2011 - FUNPEC: 86/2011 GESCON: 243657/0000 1120-29, Departamento de Saúde Coletiva da Universidade Federal do Rio Grande do Norte/UFRN.. <br class="clear">Situação: Concluído; Natureza: Pesquisa. <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Coordenador / Maria Socorro Araújo Dias - Integrante / Ana Patricia Pereira Morais - Integrante / Fabiane do Amaral Gubert - Integrante / Jocileide Sales Campos - Integrante / Maria de Fatima Antero Sousa Machado - Integrante / Ana Maria Cavalcante - Integrante / Ana Mattos Brito de Almeida - Integrante / Emilia Soares Chaves - Integrante / Maristela Inês Osawa Vasconcelos - Integrante.<br class="clear">Financiador(es): Ministério da Saúde - Auxílio
# 										financeiro.</div>
# </div><a name="PP_Pesquisa de avaliação Externa do Programa de Melhoria do Acesso e da Qualidade da Atenção Básica (PMAQ-AB)" tabindex="141"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2011 - 2012</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Pesquisa de avaliação Externa do Programa de Melhoria do Acesso e da Qualidade da Atenção Básica (PMAQ-AB)<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: PMAQ-AB: Avaliação Externa no Estado do Ceará - 1° Ciclo), projeto PROPLAN: 606/2011 - FUNPEC: 86/2011 GESCON: 243657/0000 1120-29, Departamento de Saúde Coletiva da Universidade Federal do Rio Grande do Norte/UFRN.. <br class="clear">Situação: Concluído; Natureza: Pesquisa. <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Coordenador / Maria Socorro Araújo Dias - Integrante / Ana Patricia Pereira Morais - Integrante / Fabiane do Amaral Gubert - Integrante / Jocileide Sales Campos - Integrante / Maria de Fatima Antero Sousa Machado - Integrante / Ana Maria Cavalcante - Integrante / Ana Mattos Brito de Almeida - Integrante / Emilia Soares Chaves - Integrante / Maristela Inês Osawa Vasconcelos - Integrante.<br class="clear">Financiador(es): Ministério da Saúde - Auxílio
# 										financeiro.</div>
# </div><a name="PP_Saúde Bucal e Negligência em Crianças" tabindex="142"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2010 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Saúde Bucal e Negligência em Crianças<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Situação: Em andamento; Natureza: Pesquisa. <br class="clear">Alunos envolvidos: Mestrado acadêmico: (1) . <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Coordenador.<br class="clear">Financiador(es): Conselho Nacional de Desenvolvimento Científico e Tecnológico - Auxílio
# 										financeiro.</div>
# </div><a name="PP_Validação do protocolo de Treinamento para Agentes Comunitário de Saúde na utilização do Índice Comunitário em Saúde Bucal" tabindex="143"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2007 - 2009</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Validação do protocolo de Treinamento para Agentes Comunitário de Saúde na utilização do Índice Comunitário em Saúde Bucal<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: Projeto de pesquisa (financiado pelo CNPq) para validação da utilização do ICSB por agentes comunitários em saúde em municpio do estado do Ceará. <br class="clear">Situação: Concluído; Natureza: Pesquisa. <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Integrante / Maria Vieira de Lima Saintrain - Coordenador.<br class="clear">Financiador(es): Conselho Nacional de Desenvolvimento Científico e Tecnológico - Auxílio
# 										financeiro.<br class="clear">Número de produções C, T &amp; A: 5</div>
# </div><a name="PP_Avaliação de políticas, serviços e práticas relacionadas a atenção básica" tabindex="144"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2007 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Avaliação de políticas, serviços e práticas relacionadas a atenção básica<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: Desenvolver avaliação de políticas, serviços e práticas relacionadas a atenção básica, através de dados primários e secundários, utilizando distintas abordagens metodológicas, com vistas a melhoria da qualidade da atenção a saúde nos distinto ciclos de vida da população.. <br class="clear">Situação: Em andamento; Natureza: Pesquisa. <br class="clear">Alunos envolvidos: Graduação: (0)  / Mestrado acadêmico: (0)  / Mestrado profissional: (7) . <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Coordenador / Sharmênia de Araújo Soares Nuto - Integrante / Helena Paula Guerra dos Santos - Integrante / Roberto Wagner Jr Freire de Freitas - Integrante / Antonio Leonel de Lima júnior - Integrante / Carla Kerr - Integrante / Diego Magalhães Siqueira - Integrante / Marcia Gomes Marinheiro Coelho - Integrante / Ana Paula do Nascimento Monteiro de Barros Rafael - Integrante / Cibelly Melo Ferreira - Integrante.<br class="clear"><br class="clear">Número de produções C, T &amp; A: 18</div>
# </div><a name="PP_Saúde em grupos de risco (crianças, idosos, trabalhadores, estudantes, HIV/AIDS, hipertensos, endocrinopatias, etc.)" tabindex="145"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2006 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Saúde em grupos de risco (crianças, idosos, trabalhadores, estudantes, HIV/AIDS, hipertensos, endocrinopatias, etc.)<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: Estudo das condções de saúde bucal e sistêmica em populações de risco, como crianças, idosos, trabalhadores, estudantes, HIV/AIDS, hipertensos e endocrinopatias.. <br class="clear">Situação: Em andamento; Natureza: Pesquisa. <br class="clear">Alunos envolvidos: Especialização: (3)  / Mestrado acadêmico: (6) . <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Coordenador.<br class="clear">Financiador(es): Fundação Cearense de Apoio ao Desenvolvimento Científico e Tecnológico - Bolsa / Cnpq - Bolsa.<br class="clear">Número de produções C, T &amp; A: 44</div>
# </div><a name="PP_Tooth fluoride concentration, Dental Fluorosis and Tooth Quality" tabindex="146"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2001 - 2005</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Tooth fluoride concentration, Dental Fluorosis and Tooth Quality<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: Avaliar a influencia do flúor e da fluorose dental nas propriedades dentais em pacientes que vivem em área endemica de fluorose dental e áreas com níveis otimos e subotimos de flúor na água de abastecimento.. <br class="clear">Situação: Concluído; Natureza: Pesquisa. <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Coordenador.<br class="clear">Financiador(es): U of T - Harron and Connaugh Scholarships - Bolsa / Canadian Institutes of Health Research - Auxílio
# 										financeiro.<br class="clear">Número de produções C, T &amp; A: 26</div>
# </div><a name="PP_The relationship between odontogenic bacteraemia and orthodontic treatment procedures" tabindex="147"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1999 - 2000</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">The relationship between odontogenic bacteraemia and orthodontic treatment procedures<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: Avalia o nível de bacteremia em pacientes que estão se submetendo a alguns procedimentos ortodonticos. <br class="clear">Situação: Concluído; Natureza: Pesquisa. <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Coordenador.<br class="clear"></div>
# </div><a name="PP_Pesquisa com fluorterapia - Nível do flúor na água de abastecimento do Município de Fortaleza" tabindex="148"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1995 - 1999</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Pesquisa com fluorterapia - Nível do flúor na água de abastecimento do Município de Fortaleza<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: Avaliação dos níveis de flúor nas águas de abastecimento de Fortaleza - heterocontrole do flúor na água consumida pela população de Fortaleza. <br class="clear">Situação: Concluído; Natureza: Pesquisa. <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Integrante / Sergio Luiz e Silva - Coordenador.<br class="clear">Financiador(es): Iniciacao Cientifica do Cnpq e Iniciacao Cientifica Ufc - Bolsa / Associacao Brasileira de Odontologia - Auxílio
# 										financeiro.<br class="clear">Número de produções C, T &amp; A: 5</div>
# </div><a name="PP_Materiais odontológicos/Biomateriais" tabindex="149"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1995 - 1999</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Materiais odontológicos/Biomateriais<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: Estudo de materiais odontológicos e biomateriais quanto a aplicabilidade, caracteísticas e utilização por profissionais da área odontológica. <br class="clear">Situação: Concluído; Natureza: Pesquisa. <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Integrante / Julio Jorge D'Albuquerque Lossio - Coordenador.<br class="clear">Financiador(es): Iniciacao Cientifica do Cnpq e Iniciacao Cientifica Ufc - Bolsa.<br class="clear">Número de produções C, T &amp; A: 8</div>
# </div><a name="PP_Anatomia das hemorragias faciais" tabindex="150"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1995 - 1997</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Anatomia das hemorragias faciais<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: Projeto de pesquisa com o departamento de anatomia financiado pelo Conselho Nacional de Desenvolvimento Científico e Tecnológico.. <br class="clear">Situação: Concluído; Natureza: Pesquisa. <br class="clear">Alunos envolvidos: Graduação: (2) . <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Integrante / Alan Marcus Neves da Silva - Coordenador.<br class="clear">Financiador(es): Iniciacao Cientifica do Cnpq e Iniciacao Cientifica Ufc - Bolsa.<br class="clear">Número de produções C, T &amp; A: 3</div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="ProjetosExtensao" tabindex="151">
# <h1 tabindex="0">Projetos de extensão</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell"><a name="PP_OBSERVATÓRIO DE PESQUISA EM ODONTOLOGIA BASEADA NA REDE DE ATENÇÃO A SAÚDE" tabindex="152"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2017 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">OBSERVATÓRIO DE PESQUISA EM ODONTOLOGIA BASEADA NA REDE DE ATENÇÃO A SAÚDE<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: Esse projeto tem como principal objetivo aproximar as demandas e necessidades odontológicas do serviço em saúde dos centros de análise e pesquisa. Visa contribuir para o direcionamento de melhorias das principais barreiras apresentadas pelas redes de atenção e permitir o acompanhamento de indicadores que possam transformar as rotinas de acesso, promoção, prevenção, diagnóstico e tratamento de enfermidades orais e ampliar a qualidade de atenção a saúde. Rotineiramente e de forma sistemática, o mestrado em Ciências Odontológicas da UNICHRISTUS e as Coordenações de Saúde Bucal do município de Fortaleza e do estado do Ceará reúnem-se para avaliar os processos de intervenções e de pesquisa a partir das demandas consideradas prioritárias pelo serviço. Atualmente, 03 grandes temas são focos de estudo e acompanhamento desse Observatório: Lesões potencialmente maligna/Câncer oral, Pacientes portadores de fissuras labio-palatinas e Complicações das infecções odontogênicas. <br class="clear">Situação: Em andamento; Natureza: Extensão. <br class="clear">Alunos envolvidos: Mestrado acadêmico: (5) . <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Integrante / Juliana Paiva Prado - Integrante / Fabrício Bitu Sousa - Coordenador / isabella fernandes Carvalho - Integrante / DAMASCENO, JULIANA XIMENES - Integrante / Paulo Goberlanio Barros Silva - Integrante / CANDEIRO, GEORGE TM - Integrante / NERI, JIOVANNE R - Integrante / Thinali Sousa Dantas - Integrante.<br class="clear"></div>
# </div><a name="PP_UNICHRISTUS pelo mundo: Missão Guiné Bissau" tabindex="153"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2017 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">UNICHRISTUS pelo mundo: Missão Guiné Bissau<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: O projeto visa o estimulo a politicas publicas em saúde bucal e a formação técnica em Odontologia na Republica de Guine Bissau. Em 2017, tivemos todo planejamento de atividades a serem realizadas em 2018, incluindo o 1 Curso de Técnico em Odontologia, com atividades programadas para abril de 2018. Nossas atividades de imersão e ensino em 2018 serão a base para planejarmos todas as intervenções em pesquisa a serem realizadas no país, com foco nas necessidades e realidades da população... <br class="clear">Situação: Em andamento; Natureza: Extensão. <br class="clear">Alunos envolvidos: Mestrado acadêmico: (6) . <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Integrante / Vicente de Paulo Aragão Sabóia - Integrante / Carlos Santos de Castro Filho - Integrante / Fabrício Bitu Sousa - Coordenador / Juliana Ximenes Damasceno - Integrante / lidiany karla azevedo rodrigues gerage - Integrante / Mario Rogerio Lima Mota - Integrante / Ana Paula negreiros Nunes Alves - Integrante / Danna Mota Moreira - Integrante / Ramille Araujo Lima - Integrante / Antonio Sergio Teixeira de Menezes - Integrante / Smyrna Luiza Ximenes de Sousa - Integrante.<br class="clear"></div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="ProjetosDesenvolvimento" tabindex="154">
# <h1 tabindex="0">Projetos de desenvolvimento</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell"><a name="PP_Odontopediatria" tabindex="155"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2000 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Odontopediatria<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: Pesquisa, projetos de extensão, grupo de estudo e participação em eventos relacionados ao desenvolvimento da área.. <br class="clear">Situação: Em andamento; Natureza: Desenvolvimento. <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Coordenador.<br class="clear"><br class="clear">Número de produções C, T &amp; A: 18</div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="OutrosProjetos" tabindex="156">
# <h1 tabindex="0">Outros Projetos</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell"><a name="PP_Nos APS Brasil" tabindex="157"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2019 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Nos APS Brasil<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: Participação no Grupo de pesquisa Nós APS (Atenção Primária à Saúde) Brasil, grupo colaborativo que tem como interesse o entendimento do processo de trabalho dos atores envolvidos na Estratégia de Saúde da Família (ESF), principalmente dos Agentes Comunitários de Saúde (ACS), e a influência de estressores, como a Violência, a COVID-19, na organização do serviço, na labuta e saúde dos trabalhadores. O grupo foi constituído em 2020, a partir do encontro de pesquisadores de duas redes pré-existentes: a Rede Nordeste de Formação em Saúde da Família (RENASF) e a Rede de unidades da Fiocruz no Nordeste. As investigações iniciaram-se com a busca pela compreensão dos efeitos da violência urbana no cotidiano dos ACS, entendendo-se que, por serem linha de frente dos serviços de saúde da atenção primária no país, esses profissionais, que vivem e trabalham no território, tem o potencial de melhor compreender o impacto da violência urbana nos serviços de saúde e na atenção à saúde da população. O grupo tem também atuado em estudo com outros profissionais de saúde e investigando a resiliência dos Sistemas de Saúde na atenção primária. Ao compreender a realidade dos trabalhadores da atenção primária e dos territórios por eles assistidos, o grupo objetiva colaborar para o enfrentamento dos desafios encontrados. Como lócus de trabalho tem se priorizado municípios do Nordeste brasileiro (capitais e cidades do interior), coletando e analisando, desde 2019, dados quantitativos e qualitativos nas cidades de Fortaleza/CE, Recife/PE, Teresina/Pi, João Pessoa/PB Sobral/CE, Juazeiro do Norte/CE, Barbalha/CE e Crato/CE.. <br class="clear">Situação: Em andamento; Natureza: Outra. <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Coordenador / Neiva Francenely Cunha Vieira - Integrante / Andrea Silvia Walter de Aguiar - Integrante / Ana Patricia Pereira Morais - Integrante / Maristela Inês Osawa Chagas - Integrante / José Maria Ximenes Guimarães - Integrante / Alice Maria Correia Pequeno - Integrante / CORREIA SAMPAIO, YANA PAULA COÊLHO - Integrante / FORTE, FRANKLIN DELANO SOARES - Integrante / Ardigleusa Alves Coelho - Integrante / DE ARAÚJO DIAS, MARIA SOCORRO - Integrante / MACHADO, MARIA DE FATIMA A S - Integrante / Isabella Lima Barbosa Campelo - Integrante / André Luiz Sá Oliveira - Integrante / SILVA JÚNIOR, FERNANDO JOSÉ GUEDES DA - Integrante / FERREIRA, REGINA GLAUCIA LUCENA AGUIAR - Integrante / ALBUQUERQUE, GRAYCE ALENCAR - Integrante / FARIAS, SIDNEY FEITOZA - Integrante / NASCIMENTO, ELAINE FERREIRA DO - Integrante.<br class="clear">Financiador(es): Fundação Oswaldo Cruz - Auxílio
# 										financeiro / Fundação Cearense de Apoio ao Desenvolvimento Científico e Tecnológico - Auxílio
# 										financeiro / Instituto Lemann - Auxílio
# 										financeiro.</div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="MembroCorpoEditorial" tabindex="158">
# <h1 tabindex="0">Membro de corpo editorial</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2014 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Periódico: Journal of Dental Applications</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2013 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Periódico: Sanare (Sobral)</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2012 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Periódico: Cadernos ESP (Online)</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2011 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Periódico: Revista Brasileira em Promoção da Saúde (UNIFOR) (1806-1222)</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2011 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Periódico: Revista Brasileira em Promoção da Saúde (Online)</div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="RevisorPeriodico" tabindex="159">
# <h1 tabindex="0">Revisor de periódico</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2006 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Periódico: Revista Brasileira em Promoção da Saúde (UNIFOR)</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2007 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Periódico: Journal of Dental Research</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2008 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Periódico: Revista Panamericana de Salud Pública / Pan American Journal of Public Heal</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2008 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Periódico: European Journal of Dentistry</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2013 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Periódico: Interfase - Comunicação Saúde Educação</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2014 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Periódico: Revista de Ciências Médicas (PUCCAMP)</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2014 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Periódico: Cadernos Saúde Coletiva (UFRJ)</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2015 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Periódico: Revista Trabalho Educação e saúde</div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="AreasAtuacao" tabindex="160">
# <h1 tabindex="0">Áreas de atuação</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Grande área: Ciências da Saúde / Área: Saúde Coletiva / Subárea: Saúde Pública. <br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Grande área: Ciências da Saúde / Área: Odontologia / Subárea: Odontopediatria. <br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Grande área: Ciências da Saúde / Área: Odontologia / Subárea: Biomateriais/Materiais Odontologicos. <br class="clear"></div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="Idiomas" tabindex="161">
# <h1 tabindex="0">Idiomas</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Inglês</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Compreende Bem, Fala Bem, Lê Bem, Escreve Bem. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Português</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Compreende Bem, Fala Bem, Lê Bem, Escreve Bem. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Francês</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Compreende Bem, Fala Razoavelmente, Lê Bem, Escreve Razoavelmente. </div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="PremiosTitulos" tabindex="162">
# <h1 tabindex="0">Prêmios e títulos</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2017</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Information for IEA Financial Support Participants, WORLD Congress of Epidemiology (WCE 2017). </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2013</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Prêmio de Incentivo em Ciências e Tecnologia para o SUS 2013, Ministério de Estado da Saúde. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2007</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">PRÊMIO DE INCENTIVO EM CIÊNCIAS E TECNOLOGIA PARA O SUS 2007, Ministério da Saúde. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2005</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">2nd Place - Poster Presentation - Research Day, Faculty of Dentistry - University of Toronto (February 22nd, 2005). </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2005</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Nomeação para Gold Medal Governor Award (University of Toronto) - pelo alto desempenho acadêmico durante seus estudos nesta instituição, Faculdade de Odontologia - Universidade de Toronto. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2004</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">AADR Mineralized tissue Award, IADR-AADR (82nd IADR Meeting - Honolulu). </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2004</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Best Poster Presentation, 10th Annual Canadian Connective Tissue Conference (Toronto-Canada). May 30th 2004. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2003</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">1st Place - 2003 CADR-IMHA Student Research Award (sendo selecionada para representar o Canadá na competição mundial do IADR Hatton Award), Canadian Association of Dental Research. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2003</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Hatton Awards Competition - Gotemburg (participation), International Association of Dental Research. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2002</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">2nd Award - Bone and Mineral Research Day, Toronto General Hospital - Canada. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2000</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">2nd Place Award - BSPD Annual Meeting in London, British Society of Paediatric Dentistry. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1999</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">1º Lugar no exame de seleção da Escola de Aperfeiçoamento Profissional, Associação Brasileira de Odontologia-CE. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1998</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Distincao Academica Magna cum Laude, Universidade Federal do Ceara. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1997</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">2º lugar no forum academico, ABO/XII Congresso Cearense de Odontologia. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1997</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Mensão honrosa  pelo trabalho apresentado no XII Congresso Cearense de odontologia: Fórum Acadêmico: Sistemática de vigilância sanitária no sistema de abastecimento d´àgua no Município de Fortaleza, Associação Brasileira de Odontologia. </div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="ProducoesCientificas" tabindex="163">
# <h1 tabindex="0">Produções</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <a name="ProducaoBibliografica" tabindex="164"></a><br class="clear"><div class="inst_back">
# <b>Produção bibliográfica</b>
# </div>
# <a name="Citacoes" tabindex="165"></a><div class="cita-artigos"> <b>Citações</b> </div><br class="clear"><div class="layout-cell layout-cell-12"><div class="layout-cell layout-cell-6"><div class="layout-cell-pad"> <div class="science_cont"><div class="web_s">Outras</div><div class="trab">Total de trabalhos:15</div><div class="cita">Total de citações:15</div><div class="detalhes">Vieira, A&nbsp;&nbsp;Data:&nbsp;28/08/2012</div></div></div></div></div><br class="clear"><div id="artigos-completos">
# <div class="cita-artigos">
# <b><a name="ArtigosCompletos" tabindex="166"></a>Artigos completos publicados em periódicos</b>
# </div>
# <br class="clear"><br class="clear"><div class="layout-cell-6">
# <div class="sub_tit_form">
# <label for="combo-ordenacao-citacao">Ordenar por</label><select class="input-text input-login-select" id="combo-ordenacao-citacao"><option value="1"> Ordem Cronológica </option><option value="2"> Número de citações Web of science </option><option value="3"> Número de citações Scopus </option><option value="4"> Numero de citações Scielo </option><option value="5"> Primeiro autor </option><option value="6"> Impacto JCR </option><option value="7"> Ordem de Importância </option></select>
# </div>
# </div>
# <script type="text/javascript">
			
# 				var ordenacao = {
# 					engine: {
# 						textual: function(seletor) {
						
# 							$("#artigos-completos").find(".artigo-completo").sortElements(function(a, b){
# 								var v1 = $(a).find(seletor).text();
# 								var v2 = $(b).find(seletor).text();
# 								return (v1 > v2) ? 1 : -1;
# 							});
# 						},
# 						numerico: function(seletor) {
						
# 							$("#artigos-completos").find(".artigo-completo").sortElements(function(a, b){
# 								var v1 = parseInt($(a).find(seletor).text(), 10);
# 								var v2 = parseInt($(b).find(seletor).text(), 10);
# 								if (isNaN(v1)) v1 = 0;
# 								if (isNaN(v1)) v2 = 0;
# 								return (v1 < v2) ? 1 : -1;
# 							});
# 						},
# 						numerico_crescente: function(seletor) {
						
# 							$("#artigos-completos").find(".artigo-completo").sortElements(function(a, b){
# 								var v1 = parseInt($(a).find(seletor).text(), 10);
# 								var v2 = parseInt($(b).find(seletor).text(), 10);
# 								if (isNaN(v1)) v1 = 0;
# 								if (isNaN(v1)) v2 = 0;
# 								return (v1 > v2) ? 1 : -1;
# 							});
# 						}
# 					},
# 					change: function(e) {
					
# 						var combo = $(e.target),
# 							valor = combo.val();
							
# 						if (valor == 1) ordenacao.engine.numerico("span[data-tipo-ordenacao='ano']");
# 						if (valor == 2) ordenacao.engine.numerico("span[data-tipo-ordenacao='1']");
# 						if (valor == 3) ordenacao.engine.numerico("span[data-tipo-ordenacao='3']");						
# 						if (valor == 4) ordenacao.engine.numerico("span[data-tipo-ordenacao='2']");
# 						if (valor == 5) ordenacao.engine.textual("span[data-tipo-ordenacao='autor']");
# 						if (valor == 6) ordenacao.engine.numerico("span[data-tipo-ordenacao='jcr']");
# 						if (valor == 7) ordenacao.engine.numerico_crescente("span[data-tipo-ordenacao='importancia']");
							
# 					}
# 				};
			
# 				$("#combo-ordenacao-citacao").bind("change", ordenacao.change);
			
# 			</script>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">VIEIRA-Meyer, Anya P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2023</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1590/0102-311XEN022122" target="_blank" tabindex="167"></a><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="168">VIEIRA-Meyer, Anya P G F</a></b>; MORAIS, A. P. P. ; SANTOS, H. P. G. ; YOUSAFZAI, A. ; CAMPELO, ISABELLA LIMA BARBOSA ; GUIMARÃES, JOSÉ MARIA XIMENES . Violence in the neighborhood and mental health of community health workers in a Brazilian metropolis. Cadernos de Saúde Pública (Online)<sup></sup>, v. 38, p. e00022122, 2023. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1590/0102-311XEN022122&amp;issn=16784463&amp;volume=38&amp;issue=&amp;paginaInicial=e00022122&amp;titulo=Violence in the neighborhood and mental health of community health workers in a Brazilian metropolis&amp;sequencial=1&amp;nomePeriodico=Cadernos de Saúde Pública (Online)" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> Não classificado, ISSN 1678-4463 <a href="https://scholar.google.com/scholar?q=intitle%3A%22Violence+in+the+neighborhood+and+mental+health+of+community+health+workers+in+a+Brazilian+metropolis%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2023</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1007/s10900-023-01221-9" target="_blank" tabindex="169"></a><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="170">VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</a></b>; FERREIRA, REGINA GLAUCIA LUCENA AGUIAR ; ALBUQUERQUE, GRAYCE ALENCAR ; GUIMARÃES, JOSÉ MARIA XIMENES ; MORAIS, Ana Patrícia Pereira ; MEYER, CARLOS HENRIQUE CAMPOS ; DE LIMA SAINTRAIN, MARIA VIEIRA ; CASTRO, MARCIA C. ; YOUSAFZAI, AISHA K. . Gender and Violence in the Daily Routine of Community Health Workers in Fortaleza, Brazil. JOURNAL OF COMMUNITY HEALTH<sup><img id="00945145_2" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="00945145" original-title="Journal Of Community Health (0094-5145)<br />Fator de impacto (JCR 2022): 5.9"></sup>, v. 2023, p. 1221-9, 2023. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1007/s10900-023-01221-9&amp;issn=00945145&amp;volume=2023&amp;issue=&amp;paginaInicial=1221-9&amp;titulo=Gender and Violence in the Daily Routine of Community Health Workers in Fortaleza, Brazil&amp;sequencial=2&amp;nomePeriodico=JOURNAL OF COMMUNITY HEALTH" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A2, ISSN 0094-5145, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Gender+and+Violence+in+the+Daily+Routine+of+Community+Health+Workers+in+Fortaleza%2C+Brazil%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/15877" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">DIAS, Maria Socorro de Araújo</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2023</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1590/1413-81232023288.07272023" target="_blank" tabindex="171"></a>DIAS, Maria Socorro de Araújo ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5035858383519296" target="_blank" tabindex="172">GUBERT, FABIANE DO AMARAL</a> ; PESSOA, TALITHA RODRIGUES RIBEIRO FERNANDES ; <b>VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</b> ; VASCONCELOS, MARISTELA INÊS OSAWA ; NUTO, SHARMÊNIA DE ARAÚJO SOARES ; FREITAS, ROBERTO WAGNER JÚNIOR FREIRE DE . Trilhas da Promoção e da Atenção à Saúde: contribuições da Rede Nordeste de Formação em Saúde da Família. Ciência &amp; Saúde Coletiva<sup><img id="16784561_3" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="16784561" original-title="Ciencia &amp; Saude Coletiva (1678-4561)<br />Fator de impacto (JCR 2022): 1.7"></sup>, v. 28, p. 2170-2170, 2023. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1590/1413-81232023288.07272023&amp;issn=16784561&amp;volume=28&amp;issue=&amp;paginaInicial=2170&amp;titulo=Trilhas da Promoção e da Atenção à Saúde: contribuições da Rede Nordeste de Formação em Saúde da Família&amp;sequencial=3&amp;nomePeriodico=Ciência &amp; Saúde Coletiva" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 1678-4561, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Trilhas+da+Promo%C3%A7%C3%A3o+e+da+Aten%C3%A7%C3%A3o+%C3%A0+Sa%C3%BAde%3A+contribui%C3%A7%C3%B5es+da+Rede+Nordeste+de+Forma%C3%A7%C3%A3o+em+Sa%C3%BAde+da+Fam%C3%ADlia%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2023</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1590/1413-81232023288.06462023" target="_blank" tabindex="173"></a><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="174">VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</a></b>; FARIAS, SIDNEY FEITOZA ; FORTE, FRANKLIN DELANO SOARES ; COSTA, MILENA SILVA ; GUIMARÃES, JOSÉ MARIA XIMENES ; MORAIS, Ana Patrícia Pereira ; OLIVEIRA, ANDRÉ LUIZ SÁ DE ; SILVA JÚNIOR, FERNANDO JOSÉ GUEDES DA ; NASCIMENTO, ELAINE FERREIRA DO ; VASCONCELOS, MARISTELA INÊS OSAWA ; DIAS, Maria Socorro de Araújo ; OLIVEIRA, FELIPE PROENÇO DE ; MACHADO, MARIA DE FÁTIMA ANTERO SOUSA ; CASTRO, MARCIA C. ; YOUSAFZAI, AISHA KHIZAR . Saúde mental de agentes comunitários de saúde no contexto da COVID-19. Ciência &amp; Saúde Coletiva<sup><img id="16784561_4" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="16784561" original-title="Ciencia &amp; Saude Coletiva (1678-4561)<br />Fator de impacto (JCR 2022): 1.7"></sup>, v. 28, p. 2363-2376, 2023. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1590/1413-81232023288.06462023&amp;issn=16784561&amp;volume=28&amp;issue=&amp;paginaInicial=2363&amp;titulo=Saúde mental de agentes comunitários de saúde no contexto da COVID-19&amp;sequencial=4&amp;nomePeriodico=Ciência &amp; Saúde Coletiva" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 1678-4561, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Sa%C3%BAde+mental+de+agentes+comunit%C3%A1rios+de+sa%C3%BAde+no+contexto+da+COVID-19%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">PRAXEDES, RAQUEL CRISTINA SANTANA</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2023</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1590/1413-81232023288.07042023" target="_blank" tabindex="175"></a>PRAXEDES, RAQUEL CRISTINA SANTANA ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5035858383519296" target="_blank" tabindex="176">GUBERT, FABIANE DO AMARAL</a> ; SOUSA, GYZELDA DE BARROS ; CASTRO JÚNIOR, ANDRÉ RIBEIRO DE ; MARTINS, MARIANA CAVALCANTE ; ALVES, RENATA DE SOUSA ; BESERRA, EVELINE PINHEIRO ; <b>VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</b> . Saúde bucal na infância: construção e validação de instrumento sobre conhecimento, atitude e prática de cuidadores. Ciência &amp; Saúde Coletiva<sup><img id="16784561_5" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="16784561" original-title="Ciencia &amp; Saude Coletiva (1678-4561)<br />Fator de impacto (JCR 2022): 1.7"></sup>, v. 28, p. 2203-2214, 2023. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1590/1413-81232023288.07042023&amp;issn=16784561&amp;volume=28&amp;issue=&amp;paginaInicial=2203&amp;titulo=Saúde bucal na infância: construção e validação de instrumento sobre conhecimento, atitude e prática de cuidadores&amp;sequencial=5&amp;nomePeriodico=Ciência &amp; Saúde Coletiva" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 1678-4561, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Sa%C3%BAde+bucal+na+inf%C3%A2ncia%3A+constru%C3%A7%C3%A3o+e+valida%C3%A7%C3%A3o+de+instrumento+sobre+conhecimento%2C+atitude+e+pr%C3%A1tica+de+cuidadores%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Vieira-Meyer, Anya PGF</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2023</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1590/0102-311XEN007223" target="_blank" tabindex="177"></a><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="178">Vieira-Meyer, Anya PGF</a></b>; FORTE, FRANKLIN DELANO SOARES ; GUIMARÃES, JOSÉ MARIA XIMENES ; FARIAS, SIDNEY FEITOZA ; OLIVEIRA, A. L. S. ; DIAS, MARIA DO SOCORRO DE ARAÚJO ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4276567857553527" target="_blank" tabindex="179">MONTEIRO, C. F. S.</a> ; SILVA JÚNIOR, FERNANDO JOSÉ GUEDES DA ; MORAIS, A. P. P. ; MOREIRA, M. R. C. ; CASTRO, MARCIA C. ; YOUSAFZAI, A. . Community health workers perspective on the COVID-19 impact on primary health care in Northeastern Brazil. CADERNOS DE SAÚDE PÚBLICA<sup><img id="16784464_6" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="16784464" original-title="Cadernos De Saude Publica (1678-4464)<br />Fator de impacto (JCR 2022): 2.8"></sup>, v. 39, p. e00007223, 2023. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1590/0102-311XEN007223&amp;issn=16784464&amp;volume=39&amp;issue=&amp;paginaInicial=e00007223&amp;titulo=Community health workers perspective on the COVID-19 impact on primary health care in Northeastern Brazil&amp;sequencial=6&amp;nomePeriodico=CADERNOS DE SAÚDE PÚBLICA" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 1678-4464, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Community+health+workers+perspective+on+the+COVID-19+impact+on+primary+health+care+in+Northeastern+Brazil%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/20135" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">ORTELAN, NAIÁ</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2023</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1093/ijcoms/lyad014" target="_blank" tabindex="180"></a>ORTELAN, NAIÁ ; FLORES-QUISPE, MARIA DEL PILAR ; LUZ, LEANDRO A. DA ; MARTUFI, VALENTINA ; ANJOS, EDUARDA F. DOS ; LIMA, ACÁCIA M. P. DE ; ALMEIDA, JOSEMIR R. DE ; PASSOS, MICHELLE P. V. DOS ; SANTOS, YTHALO H. S ; AMORIM, LEILA D.A.F ; AQUINO, ROSANA ; <b>VIEIRA-MEYER, ANYA P. G. F</b> ; ICHIHARA, MARIA YURY T ; BARRETO, MAURICIO L ; PINTO-JUNIOR, ELZO P . Quality of&nbsp;antenatal care in&nbsp;primary health care in&nbsp;Brazil: a latent class analysis. IJQHC Communications<sup></sup>, v. 3, p. 1-9, 2023. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1093/ijcoms/lyad014&amp;issn=26345293&amp;volume=3&amp;issue=&amp;paginaInicial=1&amp;titulo=Quality of&nbsp;antenatal care in&nbsp;primary health care in&nbsp;Brazil: a latent class analysis&amp;sequencial=7&amp;nomePeriodico=IJQHC Communications" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> Não classificado, ISSN 2634-5293 <a href="https://scholar.google.com/scholar?q=intitle%3A%22Quality+of%C2%A0antenatal+care+in%C2%A0primary+health+care+in%C2%A0Brazil%3A+a+latent+class+analysis%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2023</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.54620/cadesp.v17i1.1712" target="_blank" tabindex="181"></a><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="182">VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</a></b>. RAIZES E PONTES NO FORTALECIMENTO DO SUS. Revista da ESP<sup></sup>, v. 17, p. e1712, 2023. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.54620/cadesp.v17i1.1712&amp;issn=18087329&amp;volume=17&amp;issue=&amp;paginaInicial=e1712&amp;titulo=RAIZES E PONTES NO FORTALECIMENTO DO SUS&amp;sequencial=8&amp;nomePeriodico=Revista da ESP" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B4, ISSN 1808-7329, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22RAIZES+E+PONTES+NO+FORTALECIMENTO+DO+SUS%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">LOURENÇO, CAROLINE BARBOSA</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2022</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.22161/ijaers.91.33" target="_blank" tabindex="183"></a><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9886955951397288" target="_blank" tabindex="184">LOURENÇO, CAROLINE BARBOSA</a> ; SAINTRAIN, MARIA VIEIRA DE LIMA ; Almeida, Rosa Lívia Freitas de ; BIZERRIL, DAVI OLIVEIRA ; Braga, Débora Rosana Alves ; DE BRITO, EDLA HELENA SALLES ; MARTINS, CAROLINE FERREIRA ; FERNANDES, MARIA ISABEL DAMASCENO MARTINS ; <b>VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</b> . Hospital admissions of Brazilian older adults due to oral health problems: A trend analysis. INTERNATIONAL JOURNAL OF ADVANCED ENGINEERING RESEARCH AND SCIENCE<sup></sup>, v. 9, p. 300-308, 2022. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.22161/ijaers.91.33&amp;issn=23496495&amp;volume=9&amp;issue=&amp;paginaInicial=300&amp;titulo=Hospital admissions of Brazilian older adults due to oral health problems: A trend analysis&amp;sequencial=9&amp;nomePeriodico=INTERNATIONAL JOURNAL OF ADVANCED ENGINEERING RESEARCH AND SCIENCE" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> C, ISSN 2349-6495, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Hospital+admissions+of+Brazilian+older+adults+due+to+oral+health+problems%3A+A+trend+analysis%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>10. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">PEQUENO, L. L.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2022</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.22161/ijaers.94.27" target="_blank" tabindex="185"></a>PEQUENO, L. L. ; MARQUES, P. L. P. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="186">SAINTRAIN, M. V. L.</a> ; BEZERRA, I. N. ; <b>VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</b> ; FREITAS, S. K. S. ; PEIXOTO CALDAS, JOSÉ MANUEL . Factors associated with self-rated health in older adults receiving oral prosthetic rehabilitation. International Journal of Advanced Engineering Research and Science (IJAERS)<sup></sup>, v. 9, p. 232-244, 2022. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.22161/ijaers.94.27&amp;issn=24561908&amp;volume=9&amp;issue=&amp;paginaInicial=232&amp;titulo=Factors associated with self-rated health in older adults receiving oral prosthetic rehabilitation&amp;sequencial=10&amp;nomePeriodico=International Journal of Advanced Engineering Research and Science (IJAERS)" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> C, ISSN 2456-1908, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Factors+associated+with+self-rated+health+in+older+adults+receiving+oral+prosthetic+rehabilitation%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>11. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Oliveira, Janayne de Sousa</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2022</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.37118/ijdr.24443.05.2022" target="_blank" tabindex="187"></a>Oliveira, Janayne de Sousa ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="188">SAINTRAIN, M. V. L.</a> ; Vilar, Nathalie Barreto Saraiva ; ALBUQUERQUE, F. H. A. ; HOLANDA, I. M. M. ; CALDAS, J. M. P. ; ARAGAO, J. A. ; POS, L. S. F. ; Bezerra, Carina Bandeira ; <b>VIEIRA-Meyer, Anya P G F</b> . PREVALENCE AND HEALTH PROMOTION ACTIONS RELATED TO HYPOSALIVATION AND MOUTH DISCOMFORT IN A NURSING HOMESFOR ELDERS. INTERNATIONAL JOURNAL OF DEVELOPMENT RESEARCH<sup></sup>, v. 12, p. 56022-56027, 2022. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.37118/ijdr.24443.05.2022&amp;issn=22309926&amp;volume=12&amp;issue=&amp;paginaInicial=56022&amp;titulo=PREVALENCE AND HEALTH PROMOTION ACTIONS RELATED TO HYPOSALIVATION AND MOUTH DISCOMFORT IN A NURSING HOMESFOR ELDERS&amp;sequencial=11&amp;nomePeriodico=INTERNATIONAL JOURNAL OF DEVELOPMENT RESEARCH" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> C, ISSN 2230-9926, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22PREVALENCE+AND+HEALTH+PROMOTION+ACTIONS+RELATED+TO+HYPOSALIVATION+AND+MOUTH+DISCOMFORT+IN+A+NURSING+HOMESFOR+ELDERS%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>12. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2022</span><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="189">VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</a></b>; VIEIRA-MEYER, A. P. G. F. ; GUIMARÃES, JOSÉ MARIA XIMENES ; MORAIS, ANA PATRICIA P ; DIAS, MARIA DO SOCORRO DE ARAÚJO ; VIEIRA, NEIVA FRANCENELY CUNHA ; FARIAS, S. F. ; OLIVEIRA, A. L. S. ; MEYER, C. H. C. ; VASCONCELOS, MARISTELA INES OSAWA ; YOUSAFZAI, A. . Readiness of Community Health Agents for COVID19. EUROPEAN JOURNAL OF PUBLIC HEALTH<sup><img id="11011262_12" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="11011262" original-title="European Journal Of Public Health (1101-1262)<br />Fator de impacto (JCR 2022): 4.4"></sup>, v. 32, p. iii221, 2022. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=11011262&amp;volume=32&amp;issue=&amp;paginaInicial=iii221&amp;titulo=Readiness of Community Health Agents for COVID19&amp;sequencial=12&amp;nomePeriodico=EUROPEAN JOURNAL OF PUBLIC HEALTH" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 1101-1262, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Readiness+of+Community+Health+Agents+for+COVID19%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/20733" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>13. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">SOUSA, MARIA DO SOCORRO DE</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2022</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1590/interface.210530" target="_blank" tabindex="190"></a>SOUSA, MARIA DO SOCORRO DE ; MACHADO, M. F. A. S. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="191">Vieira, A P G F</a></b> ; VIEIRA, Neiva Francenely Cunha ; DIAS, MARIA SOCORRO DE ARAÚJO . Planejamento estratégico de um programa de pós-graduação em rede: experiência de construção coletiva. Interface (Botucatu. Online)<sup><img id="18075762_13" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="18075762" original-title="Interface-Comunicacao Saude Educacao (1807-5762)<br />Fator de impacto (JCR 2022): 0.5"></sup>, v. 26, p. e210530, 2022. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1590/interface.210530&amp;issn=18075762&amp;volume=26&amp;issue=&amp;paginaInicial=e210530&amp;titulo=Planejamento estratégico de um programa de pós-graduação em rede: experiência de construção coletiva&amp;sequencial=13&amp;nomePeriodico=Interface (Botucatu. Online)" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85138931382&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">1</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A3, ISSN 1807-5762, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Planejamento+estrat%C3%A9gico+de+um+programa+de+p%C3%B3s-gradua%C3%A7%C3%A3o+em+rede%3A+experi%C3%AAncia+de+constru%C3%A7%C3%A3o+coletiva%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/12045" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>14. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2022</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1590/interface.220555" target="_blank" tabindex="192"></a><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="193">VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</a></b>; FREIRE, ROBERTO WAGNER JUNIOR ; DIAS, Maria Socorro de Araújo ; SILVA JÚNIOR, FERNANDO JOSÉ GUEDES DA ; PINTO, ANTÔNIO GERMANE ALVES ; FORTE, FRANKLIN DELANO SOARES . Estratégia de Saúde da Família: ressonâncias na atenção, gestão, educação e promoção da Saúde. Interface (Botucatu. Online)<sup><img id="18075762_14" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="18075762" original-title="Interface-Comunicacao Saude Educacao (1807-5762)<br />Fator de impacto (JCR 2022): 0.5"></sup>, v. 26, p. e220555, 2022. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1590/interface.220555&amp;issn=18075762&amp;volume=26&amp;issue=&amp;paginaInicial=e220555&amp;titulo=Estratégia de Saúde da Família: ressonâncias na atenção, gestão, educação e promoção da Saúde&amp;sequencial=14&amp;nomePeriodico=Interface (Botucatu. Online)" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A3, ISSN 1807-5762, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Estrat%C3%A9gia+de+Sa%C3%BAde+da+Fam%C3%ADlia%3A+resson%C3%A2ncias+na+aten%C3%A7%C3%A3o%2C+gest%C3%A3o%2C+educa%C3%A7%C3%A3o+e+promo%C3%A7%C3%A3o+da+Sa%C3%BAde%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/12045" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>15. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">BARREIRA FILHO, EDENILO BALTAZAR</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2022</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.4215/rm2022.e21021" target="_blank" tabindex="194"></a>BARREIRA FILHO, EDENILO BALTAZAR ; NUTO, SHARMÊNIA DE ARAÚJO SOARES ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="195">Saintrain, Maria Vieira de Lima</a> ; SILVA, PAULO GOBERLÂNIO DE BARROS ; <b>VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</b> . SPATIAL DISTRIBUTION OF RISK FACTORS AND FLUORIDE LEVELS IN PUBLIC WATER SUPPLIES ON A MUNICIPAL SCALE. MERCATOR (FORTALEZA. ONLINE)<sup></sup>, v. 21, p. 1-11, 2022. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.4215/rm2022.e21021&amp;issn=19842201&amp;volume=21&amp;issue=&amp;paginaInicial=1&amp;titulo=SPATIAL DISTRIBUTION OF RISK FACTORS AND FLUORIDE LEVELS IN PUBLIC WATER SUPPLIES ON A MUNICIPAL SCALE&amp;sequencial=15&amp;nomePeriodico=MERCATOR (FORTALEZA. ONLINE)" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 1984-2201, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22SPATIAL+DISTRIBUTION+OF+RISK+FACTORS+AND+FLUORIDE+LEVELS+IN+PUBLIC+WATER+SUPPLIES+ON+A+MUNICIPAL+SCALE%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>16. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">CORREIA, FRANCISCO</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2021</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1093/trstmh/traa182" target="_blank" tabindex="196"></a>CORREIA, FRANCISCO ; KERR, LIGIA ; FROTA, CRISTIANE ; BARRETO, IVANA ; ALMEIDA, ROSA ; PAMPLONA, LUCIANO ; ARAÚJO, LARISSA ; MYIAJIMA, FÁBIO ; LIMA, SHIRLENE ; ARAÚJO, FERNANDA ; SIMÕES, LEDA ; JÚNIOR, FRANCISCO ; MARTINS, ADRIANO ; DIAS, LIVIA ; ALCÂNTARA, ILKA ; SILVA, ANA ; SANHUEZA, CARLOS ; RIBEIRO, MARCO ; TEIXEIRA, MARIA ; <b>Meyer, Anya</b> . Factors associated with Chikungunya infection in a cohort of women aged 15-39 y in Fortaleza, Brazil. TRANSACTIONS OF THE ROYAL SOCIETY OF TROPICAL MEDICINE AND HYGIENE<sup><img id="00359203_16" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="00359203" original-title="Transactions Of The Royal Society Of Tropical Medicine And Hygiene (0035-9203)<br />Fator de impacto (JCR 2022): 2.2"></sup>, v. ahead, p. 1-10, 2021. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1093/trstmh/traa182&amp;issn=00359203&amp;volume=ahead&amp;issue=&amp;paginaInicial=1&amp;titulo=Factors associated with Chikungunya infection in a cohort of women aged 15-39 y in Fortaleza, Brazil&amp;sequencial=16&amp;nomePeriodico=TRANSACTIONS OF THE ROYAL SOCIETY OF TROPICAL MEDICINE AND HYGIENE" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000696252100001&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">2</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85116152071&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">1</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A3, ISSN 0035-9203, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Factors+associated+with+Chikungunya+infection+in+a+cohort+of+women+aged+15-39+y+in+Fortaleza%2C+Brazil%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/23046" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>17. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">VIEIRA, NEIVA FRANCENELY CUNHA</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2021</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1590/interface.200516" target="_blank" tabindex="197"></a>VIEIRA, NEIVA FRANCENELY CUNHA ; MACHADO, MARIA DE FÁTIMA ANTERO SOUSA ; NOGUEIRA, PAULA SACHA FROTA ; LOPES, KAROLINA DE SOUSA ; <b>VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</b> ; MORAIS, Ana Patrícia Pereira ; CAMPELO, ISABELLA LIMA BARBOSA ; GUIMARÃES, JOSÉ MARIA XIMENES ; NUTO, SHARMÊNIA DE ARAÚJO SOARES ; FREITAS, ROBERTO WAGNER JÚNIOR FREIRE DE . Fatores presentes na satisfação dos usuários na Atenção Básica. Interface (Botucatu. Online)<sup><img id="18075762_17" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="18075762" original-title="Interface-Comunicacao Saude Educacao (1807-5762)<br />Fator de impacto (JCR 2022): 0.5"></sup>, v. 25, p. 1-15, 2021. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1590/interface.200516&amp;issn=18075762&amp;volume=25&amp;issue=&amp;paginaInicial=1&amp;titulo=Fatores presentes na satisfação dos usuários na Atenção Básica&amp;sequencial=17&amp;nomePeriodico=Interface (Botucatu. Online)" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000609184400001&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">2</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A3, ISSN 1807-5762, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Fatores+presentes+na+satisfa%C3%A7%C3%A3o+dos+usu%C3%A1rios+na+Aten%C3%A7%C3%A3o+B%C3%A1sica%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/12045" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>18. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2021</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1590/1413-81232021262.29922020" target="_blank" tabindex="198"></a><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="199">VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</a></b>; MORAIS, Ana Patrícia Pereira ; CAMPELO, ISABELLA LIMA BARBOSA ; GUIMARÃES, JOSÉ MARIA XIMENES . Violência e vulnerabilidade no território do agente comunitário de saúde: implicações no enfrentamento da COVID-19. Ciência &amp; Saúde Coletiva<sup><img id="16784561_18" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="16784561" original-title="Ciencia &amp; Saude Coletiva (1678-4561)<br />Fator de impacto (JCR 2022): 1.7"></sup>, v. 26, p. 657-668, 2021. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1590/1413-81232021262.29922020&amp;issn=16784561&amp;volume=26&amp;issue=&amp;paginaInicial=657&amp;titulo=Violência e vulnerabilidade no território do agente comunitário de saúde: implicações no enfrentamento da COVID-19&amp;sequencial=18&amp;nomePeriodico=Ciência &amp; Saúde Coletiva" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000618757700028&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">5</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85101535625&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">5</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 1678-4561, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Viol%C3%AAncia+e+vulnerabilidade+no+territ%C3%B3rio+do+agente+comunit%C3%A1rio+de+sa%C3%BAde%3A+implica%C3%A7%C3%B5es+no+enfrentamento+da+COVID-19%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>19. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">SILVA, ADNA DE ARAÚJO</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2021</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1590/0034-7167-2019-0769" target="_blank" tabindex="200"></a>SILVA, ADNA DE ARAÚJO ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5035858383519296" target="_blank" tabindex="201">GUBERT, FABIANE DO AMARAL</a> ; BARBOSA FILHO, VALTER CORDEIRO ; FREITAS, ROBERTO WAGNER JÚNIOR FREIRE DE ; <b>VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</b> ; PINHEIRO, MARIA TALYTA MOTA ; REBOUÇAS, LIDIANE NOGUEIRA . Health promotion actions in the School Health Program in Ceará: nursing contributions. REVISTA BRASILEIRA DE ENFERMAGEM<sup><img id="19840446_19" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="19840446" original-title="Revista Brasileira De Enfermagem (1984-0446)<br />Fator de impacto (JCR 2022): 1.3"></sup>, v. 74, p. e20190769-8, 2021. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1590/0034-7167-2019-0769&amp;issn=19840446&amp;volume=74&amp;issue=&amp;paginaInicial=e20190769&amp;titulo=Health promotion actions in the School Health Program in Ceará: nursing contributions&amp;sequencial=19&amp;nomePeriodico=REVISTA BRASILEIRA DE ENFERMAGEM" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000744167300052&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">3</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85103745578&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">6</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A4, ISSN 1984-0446, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Health+promotion+actions+in+the+School+Health+Program+in+Cear%C3%A1%3A+nursing+contributions%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>20. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">SAINTRAIN, MARIA VIEIRA DE LIMA</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2021</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1371/journal.pone.0249275" target="_blank" tabindex="202"></a>SAINTRAIN, MARIA VIEIRA DE LIMA ; SAINTRAIN, SUZANNE VIEIRA ; DE OLIVEIRA BRANCO, JULY GRASSIELY ; PEIXOTO CALDAS, JOSÉ MANUEL ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9886955951397288" target="_blank" tabindex="203">LOURENÇO, CAROLINE BARBOSA</a> ; <b>VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</b> . Dependence in instrumental activities of daily living and its implications for older adults? oral health. PLoS One<sup><img id="19326203_20" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="19326203" original-title="Plos One (1932-6203)<br />Fator de impacto (JCR 2022): 3.7"></sup>, v. 16, p. e0249275, 2021. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1371/journal.pone.0249275&amp;issn=19326203&amp;volume=16&amp;issue=&amp;paginaInicial=e0249275&amp;titulo=Dependence in instrumental activities of daily living and its implications for older adults? oral health&amp;sequencial=20&amp;nomePeriodico=PLoS One" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000664603100017&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">8</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85104977444&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">6</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 1932-6203, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Dependence+in+instrumental+activities+of+daily+living+and+its+implications+for+older+adults%3F+oral+health%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/10600153309" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>21. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">GUBERT, FABIANE DO AMARAL</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2021</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1590/1413-81232021265.05352021" target="_blank" tabindex="204"></a><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5035858383519296" target="_blank" tabindex="205">GUBERT, FABIANE DO AMARAL</a> ; BARBOSA FILHO, VALTER CORDEIRO ; QUEIROZ, REJANE CHRISTINE DE SOUSA ; MARTINS, MARIANA CAVALCANTE ; ALVES, RENATA DE SOUSA ; ROLIM, ISAURA LETÍCIA TAVARES PALMEIRA ; LOPES, MARIA DO SOCORRO VIEIRA ; <b>VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</b> . Qualidade da Atenção Primária à Saúde infantil em estados da região Nordeste. Ciência &amp; Saúde Coletiva<sup><img id="16784561_21" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="16784561" original-title="Ciencia &amp; Saude Coletiva (1678-4561)<br />Fator de impacto (JCR 2022): 1.7"></sup>, v. 26, p. 1757-1766, 2021. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1590/1413-81232021265.05352021&amp;issn=16784561&amp;volume=26&amp;issue=&amp;paginaInicial=1757&amp;titulo=Qualidade da Atenção Primária à Saúde infantil em estados da região Nordeste&amp;sequencial=21&amp;nomePeriodico=Ciência &amp; Saúde Coletiva" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000657508200016&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">2</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85107401217&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">2</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 1678-4561, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Qualidade+da+Aten%C3%A7%C3%A3o+Prim%C3%A1ria+%C3%A0+Sa%C3%BAde+infantil+em+estados+da+regi%C3%A3o+Nordeste%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>22. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">SILVA, GLADSTON THALLES DA</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2021</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1590/1413-81232021265.03722021" target="_blank" tabindex="206"></a>SILVA, GLADSTON THALLES DA ; CARVALHO, FERDINANDO OLIVEIRA ; <b>VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</b> ; GOMES, GESSYKA MAYARA SOARES ; BEZERRA, LIS MARIA MACHADO RIBEIRO ; CAMELO, RAYSSA VERAS ; BARBOSA FILHO, VALTER CORDEIRO . Planejamento e apoio no processo de trabalho das equipes de atenção básica no Nordeste: análise do PMAQ-AB (3º ciclo). Ciência &amp; Saúde Coletiva<sup><img id="16784561_22" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="16784561" original-title="Ciencia &amp; Saude Coletiva (1678-4561)<br />Fator de impacto (JCR 2022): 1.7"></sup>, v. 26, p. 1749-1755, 2021. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1590/1413-81232021265.03722021&amp;issn=16784561&amp;volume=26&amp;issue=&amp;paginaInicial=1749&amp;titulo=Planejamento e apoio no processo de trabalho das equipes de atenção básica no Nordeste: análise do PMAQ-AB (3º ciclo)&amp;sequencial=22&amp;nomePeriodico=Ciência &amp; Saúde Coletiva" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85107412319&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">2</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 1678-4561, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Planejamento+e+apoio+no+processo+de+trabalho+das+equipes+de+aten%C3%A7%C3%A3o+b%C3%A1sica+no+Nordeste%3A+an%C3%A1lise+do+PMAQ-AB+(3%C2%BA+ciclo)%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>23. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">MAIA, L. A.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2021</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1590/0103-1104202112912" target="_blank" tabindex="207"></a>MAIA, L. A. ; <b>PIMENTEL G. F. VIEIRA-MEYER, ANYA</b> ; SAINTRAIN, M. V. L. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0636763091396917" target="_blank" tabindex="208">NUTO, S. A. S.</a> ; MORAIS, A. P. P. . Satisfação e acesso à saúde bucal das pessoas que vivem com HIV/Aids no nordeste brasileiro. Saúde em Debate<sup></sup>, v. 45, p. 406-419, 2021. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1590/0103-1104202112912&amp;issn=01031104&amp;volume=45&amp;issue=&amp;paginaInicial=406&amp;titulo=Satisfação e acesso à saúde bucal das pessoas que vivem com HIV/Aids no nordeste brasileiro&amp;sequencial=23&amp;nomePeriodico=Saúde em Debate" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A4, ISSN 0103-1104, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Satisfa%C3%A7%C3%A3o+e+acesso+%C3%A0+sa%C3%BAde+bucal+das+pessoas+que+vivem+com+HIV%2FAids+no+nordeste+brasileiro%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>24. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">CANDEIRO, GEORGE TM</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2021</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.5005/jp-journals-10024-3096" target="_blank" tabindex="209"></a>CANDEIRO, GEORGE TM ; NERI, JIOVANNE R ; DE CARVALHO, BRUNA MDF ; FEIJÃO, CAMILA P ; AVELAR, RAFAEL L ; LEMOS, JOSÉ VM ; <b>Vieira-Meyer, Anya PGF</b> . Repercussions of COVID-19 in Brazilian Dentists’ Personal and Professional Routines: An Online Survey. THE JOURNAL OF CONTEMPORARY DENTAL PRACTICE<sup></sup>, v. 22, p. 491-500, 2021. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.5005/jp-journals-10024-3096&amp;issn=15263711&amp;volume=22&amp;issue=&amp;paginaInicial=491&amp;titulo=Repercussions of COVID-19 in Brazilian Dentists’ Personal and Professional Routines: An Online Survey&amp;sequencial=24&amp;nomePeriodico=THE JOURNAL OF CONTEMPORARY DENTAL PRACTICE" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B2, ISSN 1526-3711, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Repercussions+of+COVID-19+in+Brazilian+Dentists%E2%80%99+Personal+and+Professional+Routines%3A+An+Online+Survey%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/25682" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>25. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">LIMA JÚNIOR, ANTONIO LEONEL DE</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2021</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1590/1413-81232021265.04432021" target="_blank" tabindex="210"></a>LIMA JÚNIOR, ANTONIO LEONEL DE ; BARRETO, IVANA CRISTINA DE HOLANDA CUNHA ; MARANHÃO, ROBERTO RIBEIRO ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0636763091396917" target="_blank" tabindex="211">NUTO, S. A. S.</a> ; BENEVIDES, BRUNO SOUZA ; SAINTRAIN, MARIA VIEIRA DE LIMA ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="212">Vieira, A P G F</a></b> . Formatos de contratação médica na Estratégia Saúde da Família e o desempenho de seus atributos essenciais. Ciência &amp; Saúde Coletiva<sup><img id="16784561_25" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="16784561" original-title="Ciencia &amp; Saude Coletiva (1678-4561)<br />Fator de impacto (JCR 2022): 1.7"></sup>, v. 26, p. 1657-1668, 2021. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1590/1413-81232021265.04432021&amp;issn=16784561&amp;volume=26&amp;issue=&amp;paginaInicial=1657&amp;titulo=Formatos de contratação médica na Estratégia Saúde da Família e o desempenho de seus atributos essenciais&amp;sequencial=25&amp;nomePeriodico=Ciência &amp; Saúde Coletiva" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000657508200007&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">1</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85107456795&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">1</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 1678-4561, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Formatos+de+contrata%C3%A7%C3%A3o+m%C3%A9dica+na+Estrat%C3%A9gia+Sa%C3%BAde+da+Fam%C3%ADlia+e+o+desempenho+de+seus+atributos+essenciais%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>26. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">NUTO, S. A. S.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2021</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1590/1413-81232021265.04352021" target="_blank" tabindex="213"></a><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0636763091396917" target="_blank" tabindex="214">NUTO, S. A. S.</a> ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="215">Vieira, A P G F</a></b> ; VIEIRA, NEIVA FRANCENELY CUNHA ; Freitas, Roberto Wagner Júnior Freire de ; AMORIM, KARLA PATRÍCIA CARDOSO ; DIAS, MARIA SOCORRO DE ARAÚJO ; VASCONCELOS, MARISTELA INÊS OSAWA ; MACHADO, MARIA DE FÁTIMA ANTERO SOUSA . Programa de Pós-Graduação em Saúde da Família no nordeste brasileiro: repercussões no exercício profissional dos egressos. Ciência &amp; Saúde Coletiva<sup><img id="16784561_26" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="16784561" original-title="Ciencia &amp; Saude Coletiva (1678-4561)<br />Fator de impacto (JCR 2022): 1.7"></sup>, v. 26, p. 1713-1725, 2021. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1590/1413-81232021265.04352021&amp;issn=16784561&amp;volume=26&amp;issue=&amp;paginaInicial=1713&amp;titulo=Programa de Pós-Graduação em Saúde da Família no nordeste brasileiro: repercussões no exercício profissional dos egressos&amp;sequencial=26&amp;nomePeriodico=Ciência &amp; Saúde Coletiva" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000657508200012&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">5</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85107420451&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">6</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 1678-4561, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Programa+de+P%C3%B3s-Gradua%C3%A7%C3%A3o+em+Sa%C3%BAde+da+Fam%C3%ADlia+no+nordeste+brasileiro%3A+repercuss%C3%B5es+no+exerc%C3%ADcio+profissional+dos+egressos%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>27. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">MARQUES, PAULO LEONARDO PONTE</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2020</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.22161/ijaers.74.42" target="_blank" tabindex="216"></a>MARQUES, PAULO LEONARDO PONTE ; SAINTRAIN, MARIA VIEIRA DE LIMA ; LIBÓRIO, ALEXANDRE BRAGA ; PEQUENO, LUCIANNA LEITE ; <b>VIEIRA-MEYER, ANYA PIMENTEL GOMES FERNANDES</b> . Oral Health conditions in hemodialysis Patients. INTERNATIONAL JOURNAL OF ADVANCED ENGINEERING RESEARCH AND SCIENCE<sup></sup>, v. 7, p. 334-343, 2020. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.22161/ijaers.74.42&amp;issn=23496495&amp;volume=7&amp;issue=&amp;paginaInicial=334&amp;titulo=Oral Health conditions in hemodialysis Patients&amp;sequencial=27&amp;nomePeriodico=INTERNATIONAL JOURNAL OF ADVANCED ENGINEERING RESEARCH AND SCIENCE" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> C, ISSN 2349-6495, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Oral+Health+conditions+in+hemodialysis+Patients%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>28. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">MAIA, L. A.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2020</span>MAIA, L. A. ; NUTO, SHARMÊNIA DE ARAÚJO SOARES ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="217">SAINTRAIN, MARIA V. DE L.</a> ; MORAIS, ANA PATRICIA P ; SOUSA, F. B. ; <b>Vieira-Meyer, Anya PGF</b> . Access to Oral Health Care of Persons with HIV/AIDS at Public Health Care Facilities in Brazil. Iranian Journal of Public Health<sup><img id="22516085_28" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="22516085" original-title="Iranian Journal Of Public Health (2251-6085)<br />Fator de impacto (JCR 2022): 1.4"></sup>, v. 49, p. 1184-1185, 2020. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=22516085&amp;volume=49&amp;issue=&amp;paginaInicial=1184&amp;titulo=Access to Oral Health Care of Persons with HIV/AIDS at Public Health Care Facilities in Brazil&amp;sequencial=28&amp;nomePeriodico=Iranian Journal of Public Health" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B3, ISSN 2251-6085, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Access+to+Oral+Health+Care+of+Persons+with+HIV%2FAIDS+at+Public+Health+Care+Facilities+in+Brazil%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/21484" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>29. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">VIEIRA-MEY, ANYA PIMENTEL GOMES FERNANDES</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2020</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.11606/s1518-8787.2020054001878" target="_blank" tabindex="218"></a><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="219">VIEIRA-MEY, ANYA PIMENTEL GOMES FERNANDES</a></b>; MORAIS, Ana Patrícia Pereira ; GUIMARÃES, JOSÉ MARIA XIMENES ; CAMPELO, ISABELLA LIMA BARBOSA ; VIEIRA, NEIVA FRANCENELY CUNHA ; MACHADO, MARIA DE FÁTIMA ANTERO SOUSA ; NOGUEIRA, PAULA SACHA FROTA ; NUTO, SHARMÊNIA DE ARAÚJO SOARES ; FREITAS, ROBERTO WAGNER JÚNIOR FREIRE DE . Infrastructure and work process in primary health care: PMAQ in Ceará. REVISTA DE SAÚDE PÚBLICA (ONLINE)<sup><img id="15188787_29" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="15188787" original-title="Revista De Saude Publica (1518-8787)<br />Fator de impacto (JCR 2022): 2.8"></sup>, v. 54, p. 62, 2020. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.11606/s1518-8787.2020054001878&amp;issn=15188787&amp;volume=54&amp;issue=&amp;paginaInicial=62&amp;titulo=Infrastructure and work process in primary health care: PMAQ in Ceará&amp;sequencial=29&amp;nomePeriodico=REVISTA DE SAÚDE PÚBLICA (ONLINE)" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000542214300001&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">9</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85086704079&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">6</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 1518-8787, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Infrastructure+and+work+process+in+primary+health+care%3A+PMAQ+in+Cear%C3%A1%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/22596" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>30. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">MARÇAL, FELIPE FRANCO</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2020</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1111/scd.12495" target="_blank" tabindex="220"></a>MARÇAL, FELIPE FRANCO ; COSTA, ANA CAROLINA DE FIGUEIREDO ; MAGALHÃES, IANA ARAGÃO ; GUERRA, LIS MONTEIRO DE CARVALHO ; LOPES, LUIZA LASSI DE ARAÚJO ; ALVES E SILVA, FRANCISCO CLÁUDIO FERNANDES ; <b>VIEIRA'MEYER, ANYA PIMENTEL GOMES FERNANDES</b> . Orthodontic treatment and aesthetic rehabilitation in a patient with Turner syndrome: A case report. SPECIAL CARE IN DENTISTRY<sup><img id="02751879_30" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="02751879" original-title="Special Care In Dentistry (0275-1879)<br />Fator de impacto (JCR 2022): 1.4"></sup>, v. 40, p. scd.12495, 2020. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1111/scd.12495&amp;issn=02751879&amp;volume=40&amp;issue=&amp;paginaInicial=scd.12495&amp;titulo=Orthodontic treatment and aesthetic rehabilitation in a patient with Turner syndrome: A case report&amp;sequencial=30&amp;nomePeriodico=SPECIAL CARE IN DENTISTRY" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000560040600001&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">1</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85089443496&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">2</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B2, ISSN 0275-1879, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Orthodontic+treatment+and+aesthetic+rehabilitation+in+a+patient+with+Turner+syndrome%3A+A+case+report%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/23565" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>31. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">VIEIRA-Meyer, Anya P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2020</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1017/dmp.2020.342" target="_blank" tabindex="221"></a><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="222">VIEIRA-Meyer, Anya P G F</a></b>; COUTINHO, MAÍRA B ; GUERRA, HELENA P ; SAINTRAIN, MARIA V ; CANDEIRO, GEORGE T DE M . Brazilian primary and secondary public oral health attention: are dentists ready to face COVID-19 Pandemic?. Disaster Medicine and Public Health Preparedness<sup><img id="19357893_31" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="19357893" original-title="Disaster Medicine And Public Health Preparedness (1935-7893)<br />Fator de impacto (JCR 2022): 2.7"></sup>, v. sep, p. 1-22, 2020. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1017/dmp.2020.342&amp;issn=19357893&amp;volume=sep&amp;issue=&amp;paginaInicial=1&amp;titulo=Brazilian primary and secondary public oral health attention: are dentists ready to face COVID-19 Pandemic?&amp;sequencial=31&amp;nomePeriodico=Disaster Medicine and Public Health Preparedness" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000799041500042&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">2</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85091762332&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">6</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B1, ISSN 1935-7893, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Brazilian+primary+and+secondary+public+oral+health+attention%3A+are+dentists+ready+to+face+COVID-19+Pandemic%3F%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/11600153403" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>32. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">MARANHÃO, ROBERTO RIBEIRO</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2020</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1590/interface.190640" target="_blank" tabindex="223"></a>MARANHÃO, ROBERTO RIBEIRO ; BARRETO, IVANA CRISTINA DE HOLANDA CUNHA ; ANDRADE, LUIZ ODORICO MONTEIRO DE ; <b>VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</b> ; LIMA JÚNIOR, ANTÔNIO LEONEL DE . Como se relacionam o escopo de práticas profissionais, a formação e a titulação de médicos de Família e Comunidade?. Interface (Botucatu. Online)<sup><img id="18075762_32" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="18075762" original-title="Interface-Comunicacao Saude Educacao (1807-5762)<br />Fator de impacto (JCR 2022): 0.5"></sup>, v. 24, p. e190640, 2020. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1590/interface.190640&amp;issn=18075762&amp;volume=24&amp;issue=&amp;paginaInicial=e190640&amp;titulo=Como se relacionam o escopo de práticas profissionais, a formação e a titulação de médicos de Família e Comunidade?&amp;sequencial=32&amp;nomePeriodico=Interface (Botucatu. Online)" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000609175900019&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">2</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A3, ISSN 1807-5762, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Como+se+relacionam+o+escopo+de+pr%C3%A1ticas+profissionais%2C+a+forma%C3%A7%C3%A3o+e+a+titula%C3%A7%C3%A3o+de+m%C3%A9dicos+de+Fam%C3%ADlia+e+Comunidade%3F%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/12045" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>33. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">COUTINHO, MAÍRA BARBOSA</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2020</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1089/tmj.2020.0138" target="_blank" tabindex="224"></a>COUTINHO, MAÍRA BARBOSA ; DAMASCENO, JULIANA XIMENES ; MESQUITA CALS DE OLIVEIRA, PEDRO CÉSAR ; ALVES MARINHO, ISABELLE MONTENEGRO ; MARÇAL, EDGAR ; <b>GOMES FERNANDES VIEIRA-MEYER, ANYA PIMENTEL</b> . A Novel Mobile App Intervention to Reduce Dental Anxiety in Infant Patients. Telemedicine and e-Health<sup><img id="15305627_33" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="15305627" original-title="Telemedicine And E-Health (1530-5627)<br />Fator de impacto (JCR 2022): 4.7"></sup>, v. sep, p. 32924876, 2020. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1089/tmj.2020.0138&amp;issn=15305627&amp;volume=sep&amp;issue=&amp;paginaInicial=32924876&amp;titulo=A Novel Mobile App Intervention to Reduce Dental Anxiety in Infant Patients&amp;sequencial=33&amp;nomePeriodico=Telemedicine and e-Health" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000570779800001&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">2</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85108024634&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">3</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A2, ISSN 1530-5627, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22A+Novel+Mobile+App+Intervention+to+Reduce+Dental+Anxiety+in+Infant+Patients%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/23728" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>34. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">CANDEIRO, GEORGE T DE M</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2020</span>CANDEIRO, GEORGE T DE M ; GAVINI, G. ; VIVAN, R. R. ; CARVALHO, B. M. D. F. ; DUARTE, M. A. H. ; FEIJAO, C. P. ; <b>VIEIRA-Meyer, Anya P G F</b> . Knowledge about Coronavirus disease 19 (COVID-19) and its professional repercussions among brazilian endodontists. BRAZILIAN ORAL RESEARCH<sup><img id="18073107_34" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="18073107" original-title="Brazilian Oral Research (1807-3107)<br />Fator de impacto (JCR 2022): 2.5"></sup>, v. 34, p. 11, 2020. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=18073107&amp;volume=34&amp;issue=&amp;paginaInicial=11&amp;titulo=Knowledge about Coronavirus disease 19 (COVID-19) and its professional repercussions among brazilian endodontists&amp;sequencial=34&amp;nomePeriodico=BRAZILIAN ORAL RESEARCH" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A2, ISSN 1807-3107, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Knowledge+about+Coronavirus+disease+19+(COVID-19)+and+its+professional+repercussions+among+brazilian+endodontists%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>35. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">SANTOS, GIDEON BORGES DOS</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2019</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1590/1413-81232018243.30922016" target="_blank" tabindex="225"></a>SANTOS, GIDEON BORGES DOS ; HORTALE, VIRGINIA ALONSO ; SOUZA, KATIA MENDES DE ; <b>VIEIRA-MEYER, ANYA PIMENTEL GOMES FERNANDES</b> . Similaridades e diferenças entre o Mestrado Acadêmico e o Mestrado Profissional enquanto política pública de formação no campo da Saúde Pública. Ciência &amp; Saúde Coletiva<sup><img id="16784561_35" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="16784561" original-title="Ciencia &amp; Saude Coletiva (1678-4561)<br />Fator de impacto (JCR 2022): 1.7"></sup>, v. 24, p. 941-952, 2019. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1590/1413-81232018243.30922016&amp;issn=16784561&amp;volume=24&amp;issue=&amp;paginaInicial=941&amp;titulo=Similaridades e diferenças entre o Mestrado Acadêmico e o Mestrado Profissional enquanto política pública de formação no campo da Saúde Pública&amp;sequencial=35&amp;nomePeriodico=Ciência &amp; Saúde Coletiva" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000461481100026&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">5</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85063290963&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">5</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 1678-4561, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Similaridades+e+diferen%C3%A7as+entre+o+Mestrado+Acad%C3%AAmico+e+o+Mestrado+Profissional+enquanto+pol%C3%ADtica+p%C3%BAblica+de+forma%C3%A7%C3%A3o+no+campo+da+Sa%C3%BAde+P%C3%BAblica%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>36. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">ROLIM, LEONARDO BARBOSA</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2019</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1590/0034-7167-2018-0033" target="_blank" tabindex="226"></a>ROLIM, LEONARDO BARBOSA ; MONTEIRO, JANÁSSIA GONDIM ; <b>MEYER, ANYA PIMENTEL GOMES FERNANDES VIEIRA</b> ; NUTO, SHARMÊNIA DE ARAÚJO SOARES ; ARAÚJO, MÁRCIO FLÁVIO MOURA DE ; FREITAS, ROBERTO WAGNER JUNIOR FREIRE . Evaluation of Primary Health Care attributes of Fortaleza city, Ceará State, Brazil. REVISTA BRASILEIRA DE ENFERMAGEM<sup><img id="19840446_36" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="19840446" original-title="Revista Brasileira De Enfermagem (1984-0446)<br />Fator de impacto (JCR 2022): 1.3"></sup>, v. 72, p. 19-26, 2019. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1590/0034-7167-2018-0033&amp;issn=19840446&amp;volume=72&amp;issue=&amp;paginaInicial=19&amp;titulo=Evaluation of Primary Health Care attributes of Fortaleza city, Ceará State, Brazil&amp;sequencial=36&amp;nomePeriodico=REVISTA BRASILEIRA DE ENFERMAGEM" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000462174800004&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">3</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85063929955&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">5</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A4, ISSN 1984-0446, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Evaluation+of+Primary+Health+Care+attributes+of+Fortaleza+city%2C+Cear%C3%A1+State%2C+Brazil%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>37. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">VIEIRA-MEYER, ANYA PIMENTEL GOMES FERNANDES</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2019</span>
# 		<img src="images/curriculo/ico_relevante.gif" style="icone-relevancia">
# 		<a class="icone-producao icone-doi" href="http://dx.doi.org/10.17269/s41997-019-00246-9" target="_blank" tabindex="227"></a><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="228">VIEIRA-MEYER, ANYA PIMENTEL GOMES FERNANDES</a></b>; DE ARAÚJO DIAS, MARIA SOCORRO ; VASCONCELOS, MARISTELA INES OSAWA ; ROUBERTE, EMILIA SOARES CHAVES ; DE ALMEIDA, ANA MATTOS BRITO ; DE ALBUQUERQUE PINHEIRO, THEMIS XAVIER ; DE LIMA SAINTRAIN, MARIA VIEIRA ; MACHADO, MARIA DE FATIMA ANTERO SOUSA ; DUFAULT, SUZANNE ; REYNOLDS, SARAH ANN ; FERNALD, LIA . What is the relative impact of primary health care quality and conditional cash transfer program in child mortality?. CANADIAN JOURNAL OF PUBLIC HEALTH-REVUE CANADIENNE DE SANTE PUBLIQUE<sup><img id="00084263_37" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="00084263" original-title="Canadian Journal Of Public Health-Revue Canadienne De Sante Publique (0008-4263)<br />Fator de impacto (JCR 2022): 4.3"></sup>, v. Julho, p. 29, 2019. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.17269/s41997-019-00246-9&amp;issn=00084263&amp;volume=Julho&amp;issue=&amp;paginaInicial=29&amp;titulo=What is the relative impact of primary health care quality and conditional cash transfer program in child mortality?&amp;sequencial=37&amp;nomePeriodico=CANADIAN JOURNAL OF PUBLIC HEALTH-REVUE CANADIENNE DE SANTE PUBLIQUE" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000501220900010&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">6</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85069918831&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">5</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A4, ISSN 0008-4263, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22What+is+the+relative+impact+of+primary+health+care+quality+and+conditional+cash+transfer+program+in+child+mortality%3F%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/20137" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>38. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">VIEIRA-MEYER, ANYA P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2019</span>
# 		<img src="images/curriculo/ico_relevante.gif" style="icone-relevancia">
# 		<a class="icone-producao icone-doi" href="http://dx.doi.org/10.1093/fampra/cmz040" target="_blank" tabindex="229"></a><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="230">VIEIRA-MEYER, ANYA P G F</a></b>; MACHADO, MARIA DE FATIMA A S ; GUBERT, FABIANE A ; MORAIS, ANA PATRICIA P ; PAULA SAMPAIO, YANA ; SAINTRAIN, MARIA VIEIRA L ; CAMERON, DREW ; REYNOLDS, SARAH ; RAM, TALA KATARINA ; FERNALD, LIA . Variation in primary health care services after implementation of quality improvement policy in Brazil. Family Practice<sup><img id="14602229_38" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="14602229" original-title="Family Practice (1460-2229)<br />Fator de impacto (JCR 2022): 2.2"></sup>, v. 19, p. 69-80, 2019. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1093/fampra/cmz040&amp;issn=14602229&amp;volume=19&amp;issue=&amp;paginaInicial=69&amp;titulo=Variation in primary health care services after implementation of quality improvement policy in Brazil&amp;sequencial=38&amp;nomePeriodico=Family Practice" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000558983900011&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">3</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85080843647&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">4</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A3, ISSN 1460-2229, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Variation+in+primary+health+care+services+after+implementation+of+quality+improvement+policy+in+Brazil%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/27901" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>39. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">SAINTRAIN, MARIA VIEIRA DE LIMA</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2018</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1017/s1041610218000145" target="_blank" tabindex="231"></a>SAINTRAIN, MARIA VIEIRA DE LIMA ; BEZERRA, THÂMARA MANOELA MARINHO ; SANTOS, FLAVIANO DA SILVA ; SAINTRAIN, SUZANNE VIEIRA ; PEQUENO, LUCIANNA LEITE ; DA SILVA, RAIMUNDA MAGALHÃES ; <b>VIEIRA-MEYER, ANYA PIMENTEL GOMES FERNANDES</b> . Subjective well-being and oral discomfort in older people. INTERNATIONAL PSYCHOGERIATRICS<sup><img id="10416102_39" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="10416102" original-title="International Psychogeriatrics (1041-6102)<br />Fator de impacto (JCR 2022): 7.0"></sup>, v. 1, p. 1-9, 2018. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1017/s1041610218000145&amp;issn=10416102&amp;volume=1&amp;issue=&amp;paginaInicial=1&amp;titulo=Subjective well-being and oral discomfort in older people&amp;sequencial=39&amp;nomePeriodico=INTERNATIONAL PSYCHOGERIATRICS" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000450197300012&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">3</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85049877213&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">4</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 1041-6102, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Subjective+well-being+and+oral+discomfort+in+older+people%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/16200" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>40. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">SAINTRAIN, MARIA VIEIRA DE LIMA</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2018</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1111/phn.12529" target="_blank" tabindex="232"></a>SAINTRAIN, MARIA VIEIRA DE LIMA ; SAINTRAIN, SUZANNE VIEIRA ; SAMPAIO, EDGAR GOMES MARQUES ; FERREIRA, BRUNO SOUZA P. ; NEPOMUCENO, TAFFAREL CANUTO ; FROTA, MIRNA ALBUQUERQUE ; <b>Vieira-Meyer, Anya PGF</b> . Older adults? dependence in activities of daily living: Implications for oral health. PUBLIC HEALTH NURSING<sup><img id="07371209_40" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="07371209" original-title="Public Health Nursing (0737-1209)<br />Fator de impacto (JCR 2022): 2.1"></sup>, v. 16, p. 1, 2018. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1111/phn.12529&amp;issn=07371209&amp;volume=16&amp;issue=&amp;paginaInicial=1&amp;titulo=Older adults? dependence in activities of daily living: Implications for oral health&amp;sequencial=40&amp;nomePeriodico=PUBLIC HEALTH NURSING" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000450016400003&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">18</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85050598865&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">19</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 0737-1209, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Older+adults%3F+dependence+in+activities+of+daily+living%3A+Implications+for+oral+health%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/29368" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>41. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">CALVASINA, PAOLA</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2018</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1186/s12889-018-6084-3" target="_blank" tabindex="233"></a>CALVASINA, PAOLA ; O?CAMPO, PATRICIA ; PONTES, MATEUS MOTA ; OLIVEIRA, JAMILLE BARRETO ; <b>VIEIRA-MEYER, ANYA P G F</b> . The association of the Bolsa Familia Program with children?s oral health in Brazil. BMC PUBLIC HEALTH<sup><img id="14712458_41" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="14712458" original-title="Bmc Public Health (1471-2458)<br />Fator de impacto (JCR 2022): 4.5"></sup>, v. 18, p. 1186, 2018. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1186/s12889-018-6084-3&amp;issn=14712458&amp;volume=18&amp;issue=&amp;paginaInicial=1186&amp;titulo=The association of the Bolsa Familia Program with children?s oral health in Brazil&amp;sequencial=41&amp;nomePeriodico=BMC PUBLIC HEALTH" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000447784600001&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">13</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85055079352&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">11</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 1471-2458, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22The+association+of+the+Bolsa+Familia+Program+with+children%3Fs+oral+health+in+Brazil%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/19621" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>42. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">BARRETO, I. C. H. C.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2018</span>BARRETO, I. C. H. C. ; Pessoa, VM ; Sousa, MFA ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0636763091396917" target="_blank" tabindex="234">NUTO, S. A. S.</a> ; FREITAS, R. W. J. F. ; Ribeiro, K G ; <b>Anya P G F Vieira-Meyer</b> ; Andrade, LOM . Complexidade e potencialidade do trabalho dos Agentes Comunitários de Saúde no Brasil contemporâneo. Saúde em Debate<sup></sup>, v. 42, p. 116-129, 2018. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=01031104&amp;volume=42&amp;issue=&amp;paginaInicial=116&amp;titulo=Complexidade e potencialidade do trabalho dos Agentes Comunitários de Saúde no Brasil contemporâneo&amp;sequencial=42&amp;nomePeriodico=Saúde em Debate" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A4, ISSN 0103-1104, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Complexidade+e+potencialidade+do+trabalho+dos+Agentes+Comunita%CC%81rios+de+Sau%CC%81de+no+Brasil+contempora%CC%82neo%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>43. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">NOGUEIRA, CARLA MANUELA RODRIGUES</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2017</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1590/1981-22562017020.160070" target="_blank" tabindex="235"></a>NOGUEIRA, CARLA MANUELA RODRIGUES ; FALCÃO, LUCILIA MARIA NUNES ; NUTO, SHARMÊNIA DE ARAÚJO SOARES ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="236">Saintrain, Maria Vieira de Lima</a> ; <b>VIEIRA-MEYER, ANYA PIMENTEL GOMES FERNANDES</b> . Self-perceived oral health among the elderly: a household-based study. REVISTA BRASILEIRA DE GERIATRIA E GERONTOLOGIA<sup></sup>, v. 20, p. 7-19, 2017. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1590/1981-22562017020.160070&amp;issn=19812256&amp;volume=20&amp;issue=&amp;paginaInicial=7&amp;titulo=Self-perceived oral health among the elderly: a household-based study&amp;sequencial=43&amp;nomePeriodico=REVISTA BRASILEIRA DE GERIATRIA E GERONTOLOGIA" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A3, ISSN 1981-2256, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Self-perceived+oral+health+among+the+elderly%3A+a+household-based+study%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>44. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">SILVA, R. L.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2017</span>SILVA, R. L. ; MARTINS, D. G. P. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="237">SAINTRAIN, M. V. L.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0003575267121159" target="_blank" tabindex="238">VIEIRA, K. M. F.</a> ; <b>Vieira, A P G F</b> . The relationship between children and earegivers dental anxiety in Brazilian children. current pediatric research<sup></sup>, v. 21, p. 354-362, 2017. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=09719032&amp;volume=21&amp;issue=&amp;paginaInicial=354&amp;titulo=The relationship between children and earegivers dental anxiety in Brazilian children&amp;sequencial=44&amp;nomePeriodico=current pediatric research" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B3, ISSN 0971-9032, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22The+relationship+between+children+and+earegivers+dental+anxiety+in+Brazilian+children%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/14984" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>45. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">HORTALE, VIRGINIA ALONSO</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2017</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1590/1981-7746-sol00072" target="_blank" tabindex="239"></a>HORTALE, VIRGINIA ALONSO ; SANTOS, GIDEON BORGES DOS ; SOUZA, KÁTIA MENDES DE ; <b>VIEIRA-MEYER, ANYA PIMENTEL GOMES FERNANDES</b> . RELAÇÃO TEORIA-PRÁTICA NOS CURSOS DE MESTRADO ACADÊMICO E PROFISSIONAL NA ÁREA DA SAÚDE COLETIVA. TRABALHO, EDUCAÇÃO E SAÚDE (ONLINE)<sup></sup>, v. 15, p. 857-878, 2017. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1590/1981-7746-sol00072&amp;issn=19817746&amp;volume=15&amp;issue=&amp;paginaInicial=857&amp;titulo=RELAÇÃO TEORIA-PRÁTICA NOS CURSOS DE MESTRADO ACADÊMICO E PROFISSIONAL NA ÁREA DA SAÚDE COLETIVA&amp;sequencial=45&amp;nomePeriodico=TRABALHO, EDUCAÇÃO E SAÚDE (ONLINE)" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B1, ISSN 1981-7746, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22RELA%C3%87%C3%83O+TEORIA-PR%C3%81TICA+NOS+CURSOS+DE+MESTRADO+ACAD%C3%8AMICO+E+PROFISSIONAL+NA+%C3%81REA+DA+SA%C3%9ADE+COLETIVA%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>46. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">CARVALHO, M. F.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2017</span>CARVALHO, M. F. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5537817532828798" target="_blank" tabindex="240">VASCONCELOS, M. I. O.</a> ; SILVA, A. R. V. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="241">Vieira, A P G F</a></b> . Utilização de monitoramento e análise de indicadores na atenção primária a saúde. SANARE (SOBRAL)<sup></sup>, v. 16, p. 67-73, 2017. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=16768019&amp;volume=16&amp;issue=&amp;paginaInicial=67-73&amp;titulo=Utilização de monitoramento e análise de indicadores na atenção primária a saúde&amp;sequencial=46&amp;nomePeriodico=SANARE (SOBRAL)" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B3, ISSN 1676-8019, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Utiliza%C3%A7%C3%A3o+de+monitoramento+e+an%C3%A1lise+de+indicadores+na+aten%C3%A7%C3%A3o+prim%C3%A1ria+a+sa%C3%BAde%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>47. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">CARVALHO, M. F.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2016</span>CARVALHO, M. F. ; SILVA, A. R. V. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5537817532828798" target="_blank" tabindex="242">VASCONCELOS, M. I. O.</a> ; <b>Vieira, A P G F</b> . Programa de Melhoria do Acesso e Qualidade da Atenção Básica: estratégicas utilizadas em dois estados nordestinos. REVISTA GESTÃO &amp; SAÚDE (BRASÍLIA)<sup></sup>, v. 07, p. 875-888, 2016. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=19824785&amp;volume=07&amp;issue=&amp;paginaInicial=875&amp;titulo=Programa de Melhoria do Acesso e Qualidade da Atenção Básica: estratégicas utilizadas em dois estados nordestinos&amp;sequencial=47&amp;nomePeriodico=REVISTA GESTÃO &amp; SAÚDE (BRASÍLIA)" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B1, ISSN 1982-4785, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Programa+de+Melhoria+do+Acesso+e+Qualidade+da+Aten%C3%A7%C3%A3o+B%C3%A1sica%3A+estrat%C3%A9gicas+utilizadas+em+dois+estados+nordestinos%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>48. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">PETROLA, KRISHNA ANDRÉIA FEITOSA</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2016</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1371/journal.pone.0161244" target="_blank" tabindex="243"></a>PETROLA, KRISHNA ANDRÉIA FEITOSA ; BEZERRA, ÍTALO BARROSO ; DE MENEZES, ÉRICO ALEXANDRO VASCONCELOS ; CALVASINA, PAOLA ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="244">Saintrain, Maria Vieira de Lima</a> ; <b>PIMENTEL G. F. VIEIRA-MEYER, ANYA</b> . Provision of Oral Health Care to Children under Seven Covered by Bolsa Família Program. Is This a Reality?. Plos One<sup><img id="19326203_48" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="19326203" original-title="Plos One (1932-6203)<br />Fator de impacto (JCR 2022): 3.7"></sup>, v. 11, p. e0161244-15, 2016. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1371/journal.pone.0161244&amp;issn=19326203&amp;volume=11&amp;issue=&amp;paginaInicial=e0161244&amp;titulo=Provision of Oral Health Care to Children under Seven Covered by Bolsa Família Program. Is This a Reality?&amp;sequencial=48&amp;nomePeriodico=Plos One" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000381577000075&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">7</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=84984853607&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">9</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 1932-6203, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Provision+of+Oral+Health+Care+to+Children+under+Seven+Covered+by+Bolsa+Fam%C3%ADlia+Program.+Is+This+a+Reality%3F%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/10600153309" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>49. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Vieira, A P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2016</span><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="245">Vieira, A P G F</a></b>; HORTALE, V. A. ; SANTOS, G. B. ; SOUZA, K. M. . Relação teoria-prática nos cursos de Mestrado Acadêmico e Profissional na área da Saúde Coletiva. Trabalho, Educação e Saúde (Online)<sup></sup>, v. 1, p. 1-5, 2016. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=19817746&amp;volume=1&amp;issue=&amp;paginaInicial=1&amp;titulo=Relação teoria-prática nos cursos de Mestrado Acadêmico e Profissional na área da Saúde Coletiva&amp;sequencial=49&amp;nomePeriodico=Trabalho, Educação e Saúde (Online)" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B1, ISSN 1981-7746, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Rela%C3%A7%C3%A3o+teoria-pr%C3%A1tica+nos+cursos+de+Mestrado+Acad%C3%AAmico+e+Profissional+na+%C3%A1rea+da+Sa%C3%BAde+Coletiva%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>50. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">DIAS, Maria Socorro de  Araújo</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2016</span>DIAS, Maria Socorro de  Araújo ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="246">Vieira, A P G F</a></b> . Rede de Formação em Saúde da Família: A experiência do nordeste brasileiro. Estudios Avanzados (En línea)<sup><img id="07185014_50" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="07185014" original-title="Estudios Avanzados (0718-5014)<br />Fator de impacto (JCR 2022): 0.1"></sup>, v. 24, p. 76-89, 2016. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=07185014&amp;volume=24&amp;issue=&amp;paginaInicial=76&amp;titulo=Rede de Formação em Saúde da Família: A experiência do nordeste brasileiro&amp;sequencial=50&amp;nomePeriodico=Estudios Avanzados (En línea)" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A2, ISSN 0718-5014, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Rede+de+Forma%C3%A7%C3%A3o+em+Sa%C3%BAde+da+Fam%C3%ADlia%3A+A+experi%C3%AAncia+do+nordeste+brasileiro%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>51. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">CARVALHO, M. F.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2016</span>CARVALHO, M. F. ; SILVA, A. R. V. ; CHAGAS, M. I. O. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="247">Vieira, A P G F</a></b> . Programa de Melhoria do Acesso e Qualidade da Atenção Básica: estratégicas utilizadas em dois estados nordestinos. Revista Eletrônica Gestão &amp; Saúde<sup></sup>, v. 07, p. 875-888, 2016. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=19824785&amp;volume=07&amp;issue=&amp;paginaInicial=875&amp;titulo=Programa de Melhoria do Acesso e Qualidade da Atenção Básica: estratégicas utilizadas em dois estados nordestinos&amp;sequencial=51&amp;nomePeriodico=Revista Eletrônica Gestão &amp; Saúde" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B1, ISSN 1982-4785, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Programa+de+Melhoria+do+Acesso+e+Qualidade+da+Aten%C3%A7%C3%A3o+B%C3%A1sica%3A+estrat%C3%A9gicas+utilizadas+em+dois+estados+nordestinos%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>52. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">LEAL ROCHA, LYANA</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2015</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1186/s12903-015-0022-x" target="_blank" tabindex="248"></a>LEAL ROCHA, LYANA ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="249">VIEIRA DE LIMA SAINTRAIN, MARIA</a> ; <b>PIMENTEL GOMES FERNANDES VIEIRA-MEY, ANYA</b> . Access to dental public services by disabled persons. BMC Oral Health (Online)<sup><img id="14726831_52" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="14726831" original-title="Bmc Oral Health (1472-6831)<br />Fator de impacto (JCR 2022): 2.9"></sup>, v. 15, p. 35-44, 2015. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1186/s12903-015-0022-x&amp;issn=14726831&amp;volume=15&amp;issue=&amp;paginaInicial=35&amp;titulo=Access to dental public services by disabled persons&amp;sequencial=52&amp;nomePeriodico=BMC Oral Health (Online)" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000351268600001&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">40</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=84928248325&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">45</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A3, ISSN 1472-6831, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Access+to+dental+public+services+by+disabled+persons%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/24341" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>53. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">CESAR, I. D.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2015</span><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0392878673146121" target="_blank" tabindex="250">CESAR, I. D.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="251">SAINTRAIN, M. V. L.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/6856975057726513" target="_blank" tabindex="252">CORREA, C. R. S.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/6769348795296207" target="_blank" tabindex="253">GALERA, S. C.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8494771655868048" target="_blank" tabindex="254">SAINTRAIN, S. V.</a> ; <b>Vieira, A P G F</b> . Functional Ability in Institutionalized Elders in Northeastern Brazil. Sci-Afric Journal of Scientific Issues<sup></sup>, v. 3, p. 639-644, 2015. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=23116188&amp;volume=3&amp;issue=&amp;paginaInicial=639&amp;titulo=Functional Ability in Institutionalized Elders in Northeastern Brazil&amp;sequencial=53&amp;nomePeriodico=Sci-Afric Journal of Scientific Issues" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> C, ISSN 2311-6188, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Functional+Ability+in+Institutionalized+Elders+in+Northeastern+Brazil%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>54. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">MOURAO, R. A. A.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2015</span>MOURAO, R. A. A. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="255">SAINTRAIN, M. V. L.</a> ; <b>Vieira, A P G F</b> . Quality of Life of Elderly People Suffering from Sequelae by Cerebro Vascular Accident. International Journal of Research Studies in Biosciences<sup></sup>, v. 3, p. 233-240, 2015. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=23490357&amp;volume=3&amp;issue=&amp;paginaInicial=233&amp;titulo=Quality of Life of Elderly People Suffering from Sequelae by Cerebro Vascular Accident&amp;sequencial=54&amp;nomePeriodico=International Journal of Research Studies in Biosciences" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> Não classificado, ISSN 2349-0357 <a href="https://scholar.google.com/scholar?q=intitle%3A%22Quality+of+Life+of+Elderly+People+Suffering+from+Sequelae+by+Cerebro+Vascular+Accident%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>55. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">BEZERRA, A. M. S.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2015</span>BEZERRA, A. M. S. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="256">SAINTRAIN, M. V. L.</a> ; <b>Anya Pimentel Gomes Fernandes Vieira-Meyer</b> . 'THROMBOLYSIS IN  PATIENTS \ilITH ISCHEMIC STROKE: Epidemiology, Deficits and  Disability in Hospital Admission and Discharge'. Internaüonal Journal of Research SÍudies in Biosciences (IJRSB)<sup></sup>, v. 3, p. 59-67, 2015. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=23490365&amp;volume=3&amp;issue=&amp;paginaInicial=59&amp;titulo='THROMBOLYSIS IN  PATIENTS \ilITH ISCHEMIC STROKE: Epidemiology, Deficits and  Disability in Hospital Admission and Discharge'&amp;sequencial=55&amp;nomePeriodico=Internaüonal Journal of Research SÍudies in Biosciences (IJRSB)" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> C, ISSN 2349-0365, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22'THROMBOLYSIS+IN++PATIENTS+%5CilITH+ISCHEMIC+STROKE%3A+Epidemiology%2C+Deficits+and++Disability+in+Hospital+Admission+and+Discharge'%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>56. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Saintrain, Maria Vieira de Lima</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2015</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1186/s13104-015-1538-5" target="_blank" tabindex="257"></a><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="258">Saintrain, Maria Vieira de Lima</a> ; CORREA, CARLOS ROBERTO SILVEIRA ; SAINTRAIN, SUZANNE VIEIRA ; NUTO, SHARMÊNIA DE ARAÚJO SOARES ; <b>VIEIRA-MEYER, ANYA PIMENTEL GOMES FERNANDES</b> . Brazilian adolescents? oral health trends since 1986: an epidemiological observational study. BMC Research Notes<sup><img id="17560500_56" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="17560500" original-title="Bmc Research Notes (1756-0500)<br />Fator de impacto (JCR 2022): 1.8"></sup>, v. 8, p. 1-7http://www.bio, 2015. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1186/s13104-015-1538-5&amp;issn=17560500&amp;volume=8&amp;issue=&amp;paginaInicial=1&amp;titulo=Brazilian adolescents? oral health trends since 1986: an epidemiological observational study&amp;sequencial=56&amp;nomePeriodico=BMC Research Notes" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=84944059745&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">4</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B1, ISSN 1756-0500, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Brazilian+adolescents%3F+oral+health+trends+since+1986%3A+an+epidemiological+observational+study%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/19600166028" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>57. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">MACHADO, MARIA DE FÁTIMA ANTERO SOUSA</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2015</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.7322/jhgd.96709" target="_blank" tabindex="259"></a>MACHADO, MARIA DE FÁTIMA ANTERO SOUSA ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5035858383519296" target="_blank" tabindex="260">GUBERT, FABIANE DO AMARAL</a> ; <b>MEYER, ANYA PIMENTEL G. F. VIEIRA</b> ; CORREIA SAMPAIO, YANA PAULA COÊLHO ; DIAS, MARIA DO SOCORRO DE ARAÚJO ; DE ALMEIDA, ANA MATTOS BRITO DE ALMEIDA BRITO ; MORAIS, ANA PATRICIA PEREIRA ; E SILVA, ANAMARIA CAVALCANTE ; CAMPOS, JOCILEIDE SALES ; CHAGAS, MARISTELA INÊS OSAWA ; CHAVES, EMÍLIA SOARES . THE HEALTH SCHOOL PROGRAMME: A HEALTH PROMOTION STRATEGY IN PRIMARY CARE IN BRAZIL. Journal of Human Growth and Development (Print)<sup></sup>, v. 25, p. 307-312, 2015. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.7322/jhgd.96709&amp;issn=01041282&amp;volume=25&amp;issue=&amp;paginaInicial=307&amp;titulo=THE HEALTH SCHOOL PROGRAMME: A HEALTH PROMOTION STRATEGY IN PRIMARY CARE IN BRAZIL&amp;sequencial=57&amp;nomePeriodico=Journal of Human Growth and Development (Print)" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=84949186653&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">14</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B2, ISSN 0104-1282, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22THE+HEALTH+SCHOOL+PROGRAMME%3A+A+HEALTH+PROMOTION+STRATEGY+IN+PRIMARY+CARE+IN+BRAZIL%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/5700160666" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>58. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">MAIA, L. A.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2015</span>MAIA, L. A. ; <b>Vieira, A P</b> ; MORAIS, A. P. P. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0636763091396917" target="_blank" tabindex="261">NUTO, S. A. S.</a> ; MENEZES, E. A. V. . Atenção à saúde bucal das Pessoas que  Vivem com HIV/Aids na perspectiva dos cirurgiões-dentistas. Saude em Debate<sup></sup>, v. 39, p. 730-747, 2015. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=01031104&amp;volume=39&amp;issue=&amp;paginaInicial=730&amp;titulo=Atenção à saúde bucal das Pessoas que  Vivem com HIV/Aids na perspectiva dos cirurgiões-dentistas&amp;sequencial=58&amp;nomePeriodico=Saude em Debate" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A4, ISSN 0103-1104, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Aten%C3%A7%C3%A3o+%C3%A0+sa%C3%BAde+bucal+das+Pessoas+que++Vivem+com+HIV%2FAids+na+perspectiva+dos+cirurgi%C3%B5es-dentistas%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>59. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">PEREIRA, A. K. A. M.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2015</span>PEREIRA, A. K. A. M. ; SILVA, G. S. N. ; RODRIGUES, M. P. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/6730902567211867" target="_blank" tabindex="262">ALVES, R. S.</a> ; <b>Vieira, A P G F</b> . CONCEPÇÕES E PRÁTICAS DE PROFISSIONAIS DE NÍVEL SUPERIOR EM EDUCAÇÃO EM  SAÚDE NA ESTRATÉGIA SAÚDE DA FAMÍLIA. Trabalho, Educação e Saúde (Online)<sup></sup>, v. 13, p. 131-152, 2015. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=19817746&amp;volume=13&amp;issue=&amp;paginaInicial=131&amp;titulo=CONCEPÇÕES E PRÁTICAS DE PROFISSIONAIS DE NÍVEL SUPERIOR EM EDUCAÇÃO EM  SAÚDE NA ESTRATÉGIA SAÚDE DA FAMÍLIA&amp;sequencial=59&amp;nomePeriodico=Trabalho, Educação e Saúde (Online)" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B1, ISSN 1981-7746, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22CONCEP%C3%87%C3%95ES+E+PR%C3%81TICAS+DE+PROFISSIONAIS+DE+N%C3%8DVEL+SUPERIOR+EM+EDUCA%C3%87%C3%83O+EM++SA%C3%9ADE+NA+ESTRAT%C3%89GIA+SA%C3%9ADE+DA+FAM%C3%8DLIA%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>60. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">MACHADO, M. F. A. S.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2015</span><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4194885068583094" target="_blank" tabindex="263">MACHADO, M. F. A. S.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/3162595976710799" target="_blank" tabindex="264">VIEIRA, N. F. C.</a> ; HORTALE, V. A. ; <b>Vieira, A P G F</b> ; MORAIS, A. P. P. ; AGUIAR, A. S. W. ; GOMES, A. M. A. ; SILVA, G. S. N. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9128715784318958" target="_blank" tabindex="265">COIMBRA, L. C.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9251743262592177" target="_blank" tabindex="266">DIAS, M. S. A.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5537817532828798" target="_blank" tabindex="267">VASCONCELOS, M. I. O.</a> ; MOREIRA, C. O. F. ; TOMAZ, J. B. C. ; FORTE, F. D. S. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4276567857553527" target="_blank" tabindex="268">MONTEIRO, C. F. S.</a> . O PROCESSO DE CONSTRUÇÃO DO CURRÍCULO NO MESTRADO PROFISSIONAL  EM SAÚDE DA FAMÍLIA. Trabalho, Educação e Saúde (Online)<sup></sup>, v. 13, p. 39-52, 2015. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=19817746&amp;volume=13&amp;issue=&amp;paginaInicial=39&amp;titulo=O PROCESSO DE CONSTRUÇÃO DO CURRÍCULO NO MESTRADO PROFISSIONAL  EM SAÚDE DA FAMÍLIA&amp;sequencial=60&amp;nomePeriodico=Trabalho, Educação e Saúde (Online)" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B1, ISSN 1981-7746, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22O+PROCESSO+DE+CONSTRU%C3%87%C3%83O+DO+CURR%C3%8DCULO+NO+MESTRADO+PROFISSIONAL++EM+SA%C3%9ADE+DA+FAM%C3%8DLIA%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>61. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">HORTALE, V. A.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2015</span>HORTALE, V. A. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9251743262592177" target="_blank" tabindex="269">DIAS, M. S. A.</a> ; <b>Vieira, A P G F</b> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4194885068583094" target="_blank" tabindex="270">MACHADO, M. F. A. S.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/3162595976710799" target="_blank" tabindex="271">VIEIRA, N. F. C.</a> . CONSTRUÇÃO TEÓRICO-METODOLÓGICA E APRENDIZADOS COM A EXPERIÊNCIA  NO MESTRADO PROFISSIONAL EM SAÚDE DA FAMÍLIA. Trabalho, Educação e Saúde (Online)<sup></sup>, v. 13, p. 11-23, 2015. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=19817746&amp;volume=13&amp;issue=&amp;paginaInicial=11&amp;titulo=CONSTRUÇÃO TEÓRICO-METODOLÓGICA E APRENDIZADOS COM A EXPERIÊNCIA  NO MESTRADO PROFISSIONAL EM SAÚDE DA FAMÍLIA&amp;sequencial=61&amp;nomePeriodico=Trabalho, Educação e Saúde (Online)" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B1, ISSN 1981-7746, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22CONSTRU%C3%87%C3%83O+TE%C3%93RICO-METODOL%C3%93GICA+E+APRENDIZADOS+COM+A+EXPERI%C3%8ANCIA++NO+MESTRADO+PROFISSIONAL+EM+SA%C3%9ADE+DA+FAM%C3%8DLIA%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>62. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Vieira, A P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2015</span><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="272">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9251743262592177" target="_blank" tabindex="273">DIAS, M. S. A.</a> . Rede de formação em saúde da família: a experiência  do nordeste brasileiro*. Estudios Avanzados (Impresa)<sup><img id="07185022_62" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="07185022" original-title="Estudios Avanzados (0718-5022)<br />Fator de impacto (JCR 2022): 0.1"></sup>, v. 01, p. 76-89, 2015. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=07185022&amp;volume=01&amp;issue=&amp;paginaInicial=76&amp;titulo=Rede de formação em saúde da família: a experiência  do nordeste brasileiro*&amp;sequencial=62&amp;nomePeriodico=Estudios Avanzados (Impresa)" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> Não classificado, ISSN 0718-5022 <a href="https://scholar.google.com/scholar?q=intitle%3A%22Rede+de+forma%C3%A7%C3%A3o+em+sa%C3%BAde+da+fam%C3%ADlia%3A+a+experi%C3%AAncia++do+nordeste+brasileiro*%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>63. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">SAINTRAIN, M. V. L.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2014</span><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="274">SAINTRAIN, M. V. L.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5516641709622899" target="_blank" tabindex="275">BIZERRIL, D. O.</a> ; <b>Vieira, A P G F</b> . Oral health of institutionalized and non-institutionalized Elders. Sci-afric Journal of Scientific Issues, Research and Essays<sup></sup>, v. 2, p. 160-165, 2014. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=23116188&amp;volume=2&amp;issue=&amp;paginaInicial=160&amp;titulo=Oral health of institutionalized and non-institutionalized Elders&amp;sequencial=63&amp;nomePeriodico=Sci-afric Journal of Scientific Issues, Research and Essays" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> C, ISSN 2311-6188, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Oral+health+of+institutionalized+and+non-institutionalized+Elders%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>64. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">MESQUITA, C. A. M.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2014</span><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/1849223683891811" target="_blank" tabindex="276">MESQUITA, C. A. M.</a> ; SOUSA, F. C. P. ; PETROLA, K. A. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4194885068583094" target="_blank" tabindex="277">MACHADO, M. F. A. S.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="278">Saintrain, Maria Vieira de Lima</a> ; VIEIRA-MEYER, A. P. G. F. ; <b>Vieira, A P G F</b> . Oral Care in the Brazilian Primary Health Care System. Family Medicine &amp; Medical Science Research<sup></sup>, v. 3, p. 1000120, 2014. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=23274972&amp;volume=3&amp;issue=&amp;paginaInicial=1000120&amp;titulo=Oral Care in the Brazilian Primary Health Care System&amp;sequencial=64&amp;nomePeriodico=Family Medicine &amp; Medical Science Research" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> Não classificado, ISSN 2327-4972 <a href="https://scholar.google.com/scholar?q=intitle%3A%22Oral+Care+in+the+Brazilian+Primary+Health+Care+System%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>65. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">SAINTRAIN, M. V. L.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2014</span><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="279">SAINTRAIN, M. V. L.</a> ; <b>VIEIRA, ANYA P. G. F.</b> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8494771655868048" target="_blank" tabindex="280">SAINTRAIN, S. V.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/1849223683891811" target="_blank" tabindex="281">MESQUITA, C. A. M.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9886955951397288" target="_blank" tabindex="282">LOURENÇO, CAROLINE BARBOSA</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7934153093330220" target="_blank" tabindex="283">MARQUES, P. L. P.</a> . The utilization of Primary Care Agents in the identification of  dental needs in the elderly population. SCI-AFRIC JOURNAL OF SCIENTIFIC ISSUES, RESEARCH AND ESSAYS<sup></sup>, v. 02, p. 327-331, 2014. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=23116188&amp;volume=02&amp;issue=&amp;paginaInicial=327&amp;titulo=The utilization of Primary Care Agents in the identification of  dental needs in the elderly population&amp;sequencial=65&amp;nomePeriodico=SCI-AFRIC JOURNAL OF SCIENTIFIC ISSUES, RESEARCH AND ESSAYS" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> C, ISSN 2311-6188, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22The+utilization+of+Primary+Care+Agents+in+the+identification+of++dental+needs+in+the+elderly+population%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>66. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">LAWRENCE, HERENIA</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2014</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.5020/18061230.2014.p190" target="_blank" tabindex="284"></a>LAWRENCE, HERENIA ; DE PAULA SOUSA, LUIZA ; DE LIMA GONÇALVES, FABÍOLA ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="285">VIEIRA DE LIMA SAINTRAIN, MARIA</a> ; <b>PIMENTEL GOMES FERNANDES VIEIRA, ANYA</b> . Acesso à saúde bucal pública pelo paciente especial: a ótica do cirurgião-dentista. Revista Brasileira em Promoção da Saúde (Print)<sup></sup>, v. 27, p. 190-197, 2014. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.5020/18061230.2014.p190&amp;issn=18061222&amp;volume=27&amp;issue=&amp;paginaInicial=190&amp;titulo=Acesso à saúde bucal pública pelo paciente especial: a ótica do cirurgião-dentista&amp;sequencial=66&amp;nomePeriodico=Revista Brasileira em Promoção da Saúde (Print)" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B2, ISSN 1806-1222, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Acesso+%C3%A0+sa%C3%BAde+bucal+p%C3%BAblica+pelo+paciente+especial%3A+a+%C3%B3tica+do+cirurgi%C3%A3o-dentista%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>67. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">SAINTRAIN, M. V. L.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2014</span><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="286">SAINTRAIN, M. V. L.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7934153093330220" target="_blank" tabindex="287">MARQUES, P. L. P.</a> ; LAMEIDA, L. H. P. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9886955951397288" target="_blank" tabindex="288">LOURENÇO, CAROLINE BARBOSA</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8577652222943813" target="_blank" tabindex="289">Silva, R M</a> ; <b>Vieira, A P G F</b> . The relation between gender and access to dental services and goods. Revista Brasileira em Promoção da Saúde (Online)<sup></sup>, v. 27, p. 1-5, 2014. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=18061230&amp;volume=27&amp;issue=&amp;paginaInicial=1&amp;titulo=The relation between gender and access to dental services and goods&amp;sequencial=67&amp;nomePeriodico=Revista Brasileira em Promoção da Saúde (Online)" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B2, ISSN 1806-1230, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22The+relation+between+gender+and+access+to+dental+services+and+goods%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>68. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">SAINTRAIN, M. V. L.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2014</span><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="290">SAINTRAIN, M. V. L.</a> ; MESQUITA, C. A. M. ; MARQUES, P. L. P. ; LOURENÇO, C. B. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="291">Vieira, A P G F</a></b> . The utilization of Primary Care Agents in the identification of dental needs in the elderly population. Sci-Afric Journal of Scientific Issues, Research and Essays<sup></sup>, v. 2, p. 327-331, 2014. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=23116188&amp;volume=2&amp;issue=&amp;paginaInicial=327&amp;titulo=The utilization of Primary Care Agents in the identification of dental needs in the elderly population&amp;sequencial=68&amp;nomePeriodico=Sci-Afric Journal of Scientific Issues, Research and Essays" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> C, ISSN 2311-6188, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22The+utilization+of+Primary+Care+Agents+in+the+identification+of+dental+needs+in+the+elderly+population%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>69. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">SAINTRAIN, M. V. L.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2014</span>SAINTRAIN, M. V. L. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5516641709622899" target="_blank" tabindex="292">BIZERRIL, D. O.</a> ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="293">Vieira, A P G F</a></b> . Oral health of institutionalized and non-institutionalized Elders. SCIENTIFIC RESEARCH AND ESSAYS<sup><img id="19922248_69" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="19922248" original-title="SCI RES ESSAYS (1992-2248)<br />Fator de impacto (JCR 2010): 0.445"></sup>, v. 2, p. 160-165, 2014. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=19922248&amp;volume=2&amp;issue=&amp;paginaInicial=160&amp;titulo=Oral health of institutionalized and non-institutionalized Elders&amp;sequencial=69&amp;nomePeriodico=SCIENTIFIC RESEARCH AND ESSAYS" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B3, ISSN 1992-2248, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Oral+health+of+institutionalized+and+non-institutionalized+Elders%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>70. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">SAINTRAIN, MARIA V. DE L.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2013</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1111/jgs.12181" target="_blank" tabindex="294"></a><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="295">SAINTRAIN, MARIA V. DE L.</a> ; GUIMARÃES, ANA VIRGÍNIA P. ; HONÓRIO, VIVIANE A. ; DE ALMEIDA, PAULO C. ; <b>VIEIRA, ANYA P. G. F.</b> . Depression Symptoms and Oral Discomfort in Elderly Adults. Journal of the American Geriatrics Society<sup><img id="00028614_70" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="00028614" original-title="Journal Of The American Geriatrics Society (0002-8614)<br />Fator de impacto (JCR 2022): 6.3"></sup>, v. 61, p. 651-652, 2013. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1111/jgs.12181&amp;issn=00028614&amp;volume=61&amp;issue=&amp;paginaInicial=651&amp;titulo=Depression Symptoms and Oral Discomfort in Elderly Adults&amp;sequencial=70&amp;nomePeriodico=Journal of the American Geriatrics Society" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000317907600026&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">8</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=84876220530&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">8</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 0002-8614, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Depression+Symptoms+and+Oral+Discomfort+in+Elderly+Adults%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/28514" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>71. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Vieira, A P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2013</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1111/jgs.12571" target="_blank" tabindex="296"></a><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="297">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="298">Saintrain, Maria Vieira de Lima</a> . Structural Violence on Elderly People's Oral Health. Journal of the American Geriatrics Society<sup><img id="00028614_71" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="00028614" original-title="Journal Of The American Geriatrics Society (0002-8614)<br />Fator de impacto (JCR 2022): 6.3"></sup>, v. 61, p. 2264-2265, 2013. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1111/jgs.12571&amp;issn=00028614&amp;volume=61&amp;issue=&amp;paginaInicial=2264&amp;titulo=Structural Violence on Elderly People's Oral Health&amp;sequencial=71&amp;nomePeriodico=Journal of the American Geriatrics Society" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 0002-8614, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Structural+Violence+on+Elderly+People's+Oral+Health%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/28514" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>72. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">LOURENÇO, CAROLINE BARBOSA</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2013</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1186/1471-2431-13-188" target="_blank" tabindex="299"></a><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9886955951397288" target="_blank" tabindex="300">LOURENÇO, CAROLINE BARBOSA</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="301">SAINTRAIN, MARIA VIEIRA</a> ; <b>VIEIRA, ANYA PIMENTEL</b> . Child, neglect and oral health. BMC Pediatrics (Online)<sup><img id="14712431_72" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="14712431" original-title="Bmc Pediatrics (1471-2431)<br />Fator de impacto (JCR 2022): 2.4"></sup>, v. 13, p. 188-195, 2013. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1186/1471-2431-13-188&amp;issn=14712431&amp;volume=13&amp;issue=&amp;paginaInicial=188&amp;titulo=Child, neglect and oral health&amp;sequencial=72&amp;nomePeriodico=BMC Pediatrics (Online)" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000327498500001&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">19</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=84887705161&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">32</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 1471-2431, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Child%2C+neglect+and+oral+health%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/14521" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>73. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">SAINTRAIN, M. V. L.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2013</span><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="302">SAINTRAIN, M. V. L.</a> ; BRAGA, J. O. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7934153093330220" target="_blank" tabindex="303">MARQUES, P. L. P.</a> ; <b>Vieira, A P G F</b> . Community Group Centers for health Education: Lessons from Northeastern Brazil. Education for Health (Abingdon. Print)<sup><img id="13576283_73" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="13576283" original-title="Education For Health (1357-6283)<br />Fator de impacto (JCR 2022): 0.7"></sup>, v. 26, p. 195-196, 2013. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=13576283&amp;volume=26&amp;issue=&amp;paginaInicial=195&amp;titulo=Community Group Centers for health Education: Lessons from Northeastern Brazil&amp;sequencial=73&amp;nomePeriodico=Education for Health (Abingdon. Print)" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B2, ISSN 1357-6283, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Community+Group+Centers+for+health+Education%3A+Lessons+from+Northeastern+Brazil%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/110201" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>74. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Holanda, A A</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2012</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1007/s10916-011-9801-3" target="_blank" tabindex="304"></a><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9854015115173972" target="_blank" tabindex="305">Holanda, A A</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/6275335471767492" target="_blank" tabindex="306">Sa, H L C</a> ; <b>Vieira, A P G F</b> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9948525011531885" target="_blank" tabindex="307">CATRIB, A. M. F.</a> . Use and Satisfaction with Electronic Health Record by Primary Care Physicians in a Health District in Brazil. Journal of Medical Systems<sup><img id="01485598_74" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="01485598" original-title="Journal Of Medical Systems (0148-5598)<br />Fator de impacto (JCR 2022): 5.3"></sup>, v. 36, p. 3141-3149, 2012. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1007/s10916-011-9801-3&amp;issn=01485598&amp;volume=36&amp;issue=&amp;paginaInicial=3141&amp;titulo=Use and Satisfaction with Electronic Health Record by Primary Care Physicians in a Health District in Brazil&amp;sequencial=74&amp;nomePeriodico=Journal of Medical Systems" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000307994400043&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">22</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=84867327914&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">27</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A2, ISSN 0148-5598, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Use+and+Satisfaction+with+Electronic+Health+Record+by+Primary+Care+Physicians+in+a+Health+District+in+Brazil%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/23710" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>75. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Saintrain, Maria Vieira de Lima</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2012</span><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="308">Saintrain, Maria Vieira de Lima</a> ; <b>Vieira, Anya Pimentel Gomes Fernandes</b> . Application of the Community Oral Health Indicator by Non-Dental Personnel and Its Contribution to Oral Healthcare. Plos One<sup><img id="19326203_75" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="19326203" original-title="Plos One (1932-6203)<br />Fator de impacto (JCR 2022): 3.7"></sup>, v. 7, p. e39733-7, 2012. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=19326203&amp;volume=7&amp;issue=&amp;paginaInicial=e39733&amp;titulo=Application of the Community Oral Health Indicator by Non-Dental Personnel and Its Contribution to Oral Healthcare&amp;sequencial=75&amp;nomePeriodico=Plos One" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000306950200012&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">8</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=84864415372&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">9</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 1932-6203, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Application+of+the+Community+Oral+Health+Indicator+by+Non-Dental+Personnel+and+Its+Contribution+to+Oral+Healthcare%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/10600153309" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>76. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">CORREA, C. R. S.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2012</span><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/6856975057726513" target="_blank" tabindex="309">CORREA, C. R. S.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="310">SAINTRAIN, M. V. L.</a> ; <b>Vieira, A P G F</b> . SAÚDE BUCAL DO ADOLESCENTE: NECESSIDADES ODONTOLÓGICAS LEVANTADAS PELOS AGENTES COMUNITÁRIOS DE SAÚDE. Revista Brasileira em Promoção da Saúde (Online)<sup></sup>, v. 25, p. 131-138, 2012. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=18061230&amp;volume=25&amp;issue=&amp;paginaInicial=131&amp;titulo=SAÚDE BUCAL DO ADOLESCENTE: NECESSIDADES ODONTOLÓGICAS LEVANTADAS PELOS AGENTES COMUNITÁRIOS DE SAÚDE&amp;sequencial=76&amp;nomePeriodico=Revista Brasileira em Promoção da Saúde (Online)" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B2, ISSN 1806-1230, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22SAU%CC%81DE+BUCAL+DO+ADOLESCENTE%3A+NECESSIDADES+ODONTOLO%CC%81GICAS+LEVANTADAS+PELOS+AGENTES+COMUNITA%CC%81RIOS+DE+SAU%CC%81DE%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>77. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Vieira, A P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2011</span><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="311">Vieira, A P G F</a></b>; Karbage, J.P. . Impacto da Atenção odontológica na Saúde Bucal de Crianças. Revista Brasileira em Promoção da Saúde (UNIFOR. Impresso)<sup></sup>, v. 24, p. 10-15, 2011. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=18061222&amp;volume=24&amp;issue=&amp;paginaInicial=10&amp;titulo=Impacto da Atenção odontológica na Saúde Bucal de Crianças&amp;sequencial=77&amp;nomePeriodico=Revista Brasileira em Promoção da Saúde (UNIFOR. Impresso)" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B2, ISSN 1806-1222, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Impacto+da+Aten%C3%A7%C3%A3o+odontol%C3%B3gica+na+Sa%C3%BAde+Bucal+de+Crian%C3%A7as%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>78. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Vieira, A P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2011</span><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="312">Vieira, A P G F</a></b>; Lourenço, C ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="313">SAINTRAIN, M. V. L.</a> . Is dental decay related to child neglect?. Caries Research (Online)<sup><img id="1421976X_78" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="1421976X" original-title="Caries Research (1421-976X)<br />Fator de impacto (JCR 2022): 4.2"></sup>, v. 45, p. 233-233, 2011. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=1421976X&amp;volume=45&amp;issue=&amp;paginaInicial=233&amp;titulo=Is dental decay related to child neglect?&amp;sequencial=78&amp;nomePeriodico=Caries Research (Online)" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 1421-976X, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Is+dental+decay+related+to+child+neglect%3F%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/24388" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>79. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">GRYNPAS, M. D.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2011</span>GRYNPAS, M. D. ; HANCOCK, R ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7549914789004407" target="_blank" tabindex="314">SAMPAIO, F. C.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7549914789004407" target="_blank" tabindex="315">SAMPAIO, FC</a> ; LIMEBACK, Hardy ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="316">Vieira, A P G F</a></b> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="317">SAINTRAIN, M. V.</a> . The relationship between dental fluorosis and tooth fluoride concentration - a study in an endemic area. Revista Brasileira em Promoção da Saúde (UNIFOR. Impresso)<sup></sup>, v. 24, p. 355-360, 2011. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=18061222&amp;volume=24&amp;issue=&amp;paginaInicial=355&amp;titulo=The relationship between dental fluorosis and tooth fluoride concentration - a study in an endemic area&amp;sequencial=79&amp;nomePeriodico=Revista Brasileira em Promoção da Saúde (UNIFOR. Impresso)" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B2, ISSN 1806-1222, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22The+relationship+between+dental+fluorosis+and+tooth+fluoride+concentration+-+a+study+in+an+endemic+area%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>80. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">COSTA, E. H. M.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2010</span>COSTA, E. H. M. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="318">SAINTRAIN, M. V. L.</a> ; <b>Anya Vieira</b> . Autopercepção da condição de saúde bucal em idosos institucionalizados e não institucionalizados. Ciência e Saúde Coletiva (Impresso)<sup><img id="14138123_80" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="14138123" original-title="Ciencia &amp; Saude Coletiva (1413-8123)<br />Fator de impacto (JCR 2022): 1.7"></sup>, v. 15, p. 2925-2930, 2010. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=14138123&amp;volume=15&amp;issue=&amp;paginaInicial=2925&amp;titulo=Autopercepção da condição de saúde bucal em idosos institucionalizados e não institucionalizados&amp;sequencial=80&amp;nomePeriodico=Ciência e Saúde Coletiva (Impresso)" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000282525400032&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">8</span></a><a href="http://www.scielo.org/applications/scielo-org/php/citedScielo.php?lang=pt&amp;pid=S1413-81232010000600030" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scielo.png"><span class="numero-citacao" data-tipo-ordenacao="2">26</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=78249267538&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">14</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 1413-8123, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Autopercep%C3%A7%C3%A3o+da+condi%C3%A7%C3%A3o+de+sa%C3%BAde+bucal+em+idosos+institucionalizados+e+n%C3%A3o+institucionalizados%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/4700151729" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>81. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">ALMEIDA, R. F. C.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2010</span><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/3992942738436467" target="_blank" tabindex="319">ALMEIDA, R. F. C.</a> ; <b>Vieira, A P G F</b> . Influência da  informação oral e escrita sobre antirretrovirais no conhecimento de usuários com HIV/AIDS. Revista Brasileira em Promoção da Saúde (UNIFOR. Impresso)<sup></sup>, v. 23, p. 251-258, 2010. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=18061222&amp;volume=23&amp;issue=&amp;paginaInicial=251&amp;titulo=Influência da  informação oral e escrita sobre antirretrovirais no conhecimento de usuários com HIV/AIDS&amp;sequencial=81&amp;nomePeriodico=Revista Brasileira em Promoção da Saúde (UNIFOR. Impresso)" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B2, ISSN 1806-1222, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Influ%C3%AAncia+da++informa%C3%A7%C3%A3o+oral+e+escrita+sobre+antirretrovirais+no+conhecimento+de+usu%C3%A1rios+com+HIV%2FAIDS%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>82. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">LIMA, D. L. F.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2009</span><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/1815947099924439" target="_blank" tabindex="320">LIMA, D. L. F.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7492450432942397" target="_blank" tabindex="321">Montenegro Jr, R.M.</a> ; <b>Vieira, A P G F</b> ; ALBANO, M.F. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5517143911175495" target="_blank" tabindex="322">REGO, D. M.</a> . Absence of periodontitis in acromegalic patients. Clinical Oral Investigations (Print)<sup><img id="14326981_82" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="14326981" original-title="Clinical Oral Investigations (1432-6981)<br />Fator de impacto (JCR 2022): 3.4"></sup>, v. 13, p. 165-169, 2009. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=14326981&amp;volume=13&amp;issue=&amp;paginaInicial=165&amp;titulo=Absence of periodontitis in acromegalic patients&amp;sequencial=82&amp;nomePeriodico=Clinical Oral Investigations (Print)" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000268207400005&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">15</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=67649389660&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">18</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 1432-6981, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Absence+of+periodontitis+in+acromegalic+patients%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/24409" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>83. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">AGUIAR, S. M.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2009</span>AGUIAR, S. M. ; <b>Vieira, A P G F</b> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0003575267121159" target="_blank" tabindex="323">VIEIRA, K. M. F.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/3220734838299055" target="_blank" tabindex="324">AGUIAR, S. M.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5885701115489299" target="_blank" tabindex="325">NOBREGA, J. O.</a> . Prevalência de sintomas de estresse nos estudantes de medicina. Jornal Brasileiro de Psiquiatria<sup></sup>, v. 58, p. 34-38, 2009. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=00472085&amp;volume=58&amp;issue=&amp;paginaInicial=34&amp;titulo=Prevalência de sintomas de estresse nos estudantes de medicina&amp;sequencial=83&amp;nomePeriodico=Jornal Brasileiro de Psiquiatria" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="http://www.scopus.com/scopus/inward/citedby.url?partnerID=bb5nvXTn&amp;rel=R6.0.0&amp;doi=null&amp;md5=f8a2d9f8af7e0fb17f9968fbbee28b49" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">6</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B3, ISSN 0047-2085, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Preval%C3%AAncia+de+sintomas+de+estresse+nos+estudantes+de+medicina%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/16240" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>84. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">ALMEIDA, R. F. C.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2009</span><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/3992942738436467" target="_blank" tabindex="326">ALMEIDA, R. F. C.</a> ; <b>Vieira, A P G F</b> . Evaluation of HIV/AIDS Patients? Knowledge on Antiretroviral Drugs. The Brazilian Journal of Infectious Diseases<sup><img id="14138670_84" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="14138670" original-title="Brazilian Journal Of Infectious Diseases (1413-8670)<br />Fator de impacto (JCR 2022): 3.4"></sup>, v. 13, p. 183-190, 2009. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=14138670&amp;volume=13&amp;issue=&amp;paginaInicial=183&amp;titulo=Evaluation of HIV/AIDS Patients? Knowledge on Antiretroviral Drugs&amp;sequencial=84&amp;nomePeriodico=The Brazilian Journal of Infectious Diseases" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000275105600006&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">12</span></a><a href="http://www.scielo.org/applications/scielo-org/php/citedScielo.php?lang=pt&amp;pid=S1413-86702009000300006" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scielo.png"><span class="numero-citacao" data-tipo-ordenacao="2">5</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=77749283195&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">11</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B1, ISSN 1413-8670, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Evaluation+of+HIV%2FAIDS+Patients%3F+Knowledge+on+Antiretroviral+Drugs%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/21807" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>85. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">AQUIAR, C. C. T.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2008</span>AQUIAR, C. C. T. ; <b>Vieira, A P G F</b> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4167422383882509" target="_blank" tabindex="327">Carvalho, A. F.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7492450432942397" target="_blank" tabindex="328">Montenegro Jr, R.M.</a> . Assessment instruments for Healt-Related Quality of Life in diabetes mellitus. Arquivos Brasileiros de Endocrinologia &amp; Metabologia<sup><img id="00042730_85" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="00042730" original-title="ARQUIVOS BRASILEIROS DE ENDOCRINOLOGIA E METABOLOGIA (0004-2730)<br />Fator de impacto (JCR 2017): 1.515"></sup>, v. 52, p. 931-939, 2008. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=00042730&amp;volume=52&amp;issue=&amp;paginaInicial=931&amp;titulo=Assessment instruments for Healt-Related Quality of Life in diabetes mellitus&amp;sequencial=85&amp;nomePeriodico=Arquivos Brasileiros de Endocrinologia &amp; Metabologia" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000259019700004&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">33</span></a><a href="http://www.scielo.org/applications/scielo-org/php/citedScielo.php?lang=pt&amp;pid=S0004-27302008000600004" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scielo.png"><span class="numero-citacao" data-tipo-ordenacao="2">30</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=53249108148&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">33</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B2, ISSN 0004-2730, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Assessment+instruments+for+Healt-Related+Quality+of+Life+in+diabetes+mellitus%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>86. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">CHACHRA, D.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2008</span>CHACHRA, D. ; <b>Vieira, A P G F</b> ; GRYNPAS, M. D. . Fluoride and mineralized tissues.. Crit Rev Biomed Eng<sup></sup>, v. 36, p. 183-221, 2008. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=0278940X&amp;volume=36&amp;issue=&amp;paginaInicial=183&amp;titulo=Fluoride and mineralized tissues.&amp;sequencial=86&amp;nomePeriodico=Crit Rev Biomed Eng" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=74349102755&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">27</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> B2, ISSN 0278-940X, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Fluoride+and+mineralized+tissues.%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/16311" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>87. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Vieira, A P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2007</span><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="329">Vieira, A P G F</a></b>; MENESES, J. M. ; MAIA, R. . Cementoblastoma related to a primary tooth: a case report. Journal of Oral Pathology and Medicine<sup><img id="09042512_87" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="09042512" original-title="Journal Of Oral Pathology &amp; Medicine (0904-2512)<br />Fator de impacto (JCR 2022): 3.3"></sup>, v. 36, p. 117-119, 2007. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=09042512&amp;volume=36&amp;issue=&amp;paginaInicial=117&amp;titulo=Cementoblastoma related to a primary tooth: a case report&amp;sequencial=87&amp;nomePeriodico=Journal of Oral Pathology and Medicine" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000243405000009&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">11</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=33846160140&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">18</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 0904-2512, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Cementoblastoma+related+to+a+primary+tooth%3A+a+case+report%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/26155" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>88. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Vieira, A P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2006</span><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="330">Vieira, A P G F</a></b>; HANCOCK, R. ; DIMITRIU, M. ; LIMEBACK, H. ; GRYNPAS, M. D. . Fluoride's effect on human dentin ultrasound velocity (elastic modulus) and tubule size. European Journal of Oral Sciences<sup><img id="09098836_88" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="09098836" original-title="European Journal Of Oral Sciences (0909-8836)<br />Fator de impacto (JCR 2022): 1.9"></sup>, v. 114, n.1, p. 83-88, 2006. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=09098836&amp;volume=114&amp;issue=&amp;paginaInicial=83&amp;titulo=Fluoride's effect on human dentin ultrasound velocity (elastic modulus) and tubule size&amp;sequencial=88&amp;nomePeriodico=European Journal of Oral Sciences" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000235169500013&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">19</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=33645061834&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">23</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A2, ISSN 0909-8836, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Fluoride's+effect+on+human+dentin+ultrasound+velocity+(elastic+modulus)+and+tubule+size%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/25025" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>89. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">LIMEBACK, H.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2006</span>LIMEBACK, H. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="331">Vieira, A P G F</a></b> ; LAWRENCE, H. . Improving esthetically-objectionable human enamel fluorosis with a simple microabrasion technique. European Journal of Oral Sciences<sup><img id="09098836_89" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="09098836" original-title="European Journal Of Oral Sciences (0909-8836)<br />Fator de impacto (JCR 2022): 1.9"></sup>, v. 114, n.in press, p. 123-126, 2006. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=09098836&amp;volume=114&amp;issue=&amp;paginaInicial=123&amp;titulo=Improving esthetically-objectionable human enamel fluorosis with a simple microabrasion technique&amp;sequencial=89&amp;nomePeriodico=European Journal of Oral Sciences" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A2, ISSN 0909-8836, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Improving+esthetically-objectionable+human+enamel+fluorosis+with+a+simple+microabrasion+technique%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/25025" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>90. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Vieira, A P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2005</span><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="332">Vieira, A P G F</a></b>; MAIA, R. . Auto-transplantation of central incisor with root dilaceration. International Journal of Oral and Maxillofacial Surgery<sup><img id="09015027_90" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="09015027" original-title="International Journal Of Oral And Maxillofacial Surgery (0901-5027)<br />Fator de impacto (JCR 2022): 2.4"></sup>, USA, v. 34, n.1, p. 89-91, 2005. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=09015027&amp;volume=34&amp;issue=&amp;paginaInicial=89&amp;titulo=Auto-transplantation of central incisor with root dilaceration&amp;sequencial=90&amp;nomePeriodico=International Journal of Oral and Maxillofacial Surgery" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000226402100015&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">17</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=11844258248&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">20</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 0901-5027, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Auto-transplantation+of+central+incisor+with+root+dilaceration%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/25629" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>91. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Vieira, A P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2005</span>
# 		<img src="images/curriculo/ico_relevante.gif" style="icone-relevancia">
# 		<b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="333">Vieira, A P G F</a></b>; HANCOCK, R. ; EGGERTSSON, H. ; EVERETT, E. T. ; GRYNPAS, M. D. . Tooth Quality in Fluorosis : Genetic and Environmental Factors. Calcified Tissue International<sup><img id="0171967X_91" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="0171967X" original-title="Calcified Tissue International (0171-967X)<br />Fator de impacto (JCR 2022): 4.2"></sup>, Estados Unidos, v. 76, n.1, p. 17-25, 2005. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=0171967X&amp;volume=76&amp;issue=&amp;paginaInicial=17&amp;titulo=Tooth Quality in Fluorosis : Genetic and Environmental Factors&amp;sequencial=91&amp;nomePeriodico=Calcified Tissue International" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000226557500004&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">51</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=13644270385&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">55</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 0171-967X, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Tooth+Quality+in+Fluorosis+%3A+Genetic+and+Environmental+Factors%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/25422" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>92. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Vieira, A P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2005</span><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="334">Vieira, A P G F</a></b>; LAWRENCE, H. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7549914789004407" target="_blank" tabindex="335">SAMPAIO, F.</a> ; LIMEBACK, H. ; GRYNPAS, M. D. . A Visual Analogue Scale for Dental Fluorosis severity. JOURNAL OF THE AMERICAN DENTAL ASSOCIATION<sup><img id="00028177_92" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="00028177" original-title="Journal Of The American Dental Association (0002-8177)<br />Fator de impacto (JCR 2022): 3.9"></sup>, Estados Unidos, v. 136, n.7, p. 895-901, 2005. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=00028177&amp;volume=136&amp;issue=&amp;paginaInicial=895&amp;titulo=A Visual Analogue Scale for Dental Fluorosis severity&amp;sequencial=92&amp;nomePeriodico=JOURNAL OF THE AMERICAN DENTAL ASSOCIATION" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000230528800017&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">20</span></a><a href="http://www.scopus.com/scopus/inward/citedby.url?partnerID=bb5nvXTn&amp;rel=R6.0.0&amp;doi=null&amp;md5=f8a2d9f8af7e0fb17f9968fbbee28b49" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">20</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 0002-8177, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22A+Visual+Analogue+Scale+for+Dental+Fluorosis+severity%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/23804" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>93. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Vieira, A P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2005</span><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="336">Vieira, A P G F</a></b>; HANCOCK, R. ; DIMITRIU, M. ; SCHWARTZ, M. ; LIMEBACK, H. ; GRYNPAS, M. D. . How does fluoride affect dentin microhardness and mineralization?. JOURNAL OF DENTAL RESEARCH<sup><img id="00220345_93" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="00220345" original-title="Journal Of Dental Research (0022-0345)<br />Fator de impacto (JCR 2022): 7.6"></sup>, Estados Unidos das Americas, v. 84, n.10, p. 951-957, 2005. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=00220345&amp;volume=84&amp;issue=&amp;paginaInicial=951&amp;titulo=How does fluoride affect dentin microhardness and mineralization?&amp;sequencial=93&amp;nomePeriodico=JOURNAL OF DENTAL RESEARCH" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000232059900015&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">28</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=32944456584&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">33</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 0022-0345, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22How+does+fluoride+affect+dentin+microhardness+and+mineralization%3F%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/25684" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>94. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Vieira, A P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2005</span><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="337">Vieira, A P G F</a></b>; MOUSNY, M. ; MAIA, R. ; HANCOCK, R. ; EVERETT, E. T. ; GRYNPAS, M. D. . Assessment of teeth as biomarkers for skeletal fluoride exposure. Osteoporosis International<sup><img id="0937941X_94" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="0937941X" original-title="Osteoporosis International (0937-941X)<br />Fator de impacto (JCR 2022): 4.0"></sup>, v. 16, p. 1576-1582, 2005. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=0937941X&amp;volume=16&amp;issue=&amp;paginaInicial=1576&amp;titulo=Assessment of teeth as biomarkers for skeletal fluoride exposure&amp;sequencial=94&amp;nomePeriodico=Osteoporosis International" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000233997600014&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">24</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=29044449460&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">29</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A2, ISSN 0937-941X, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Assessment+of+teeth+as+biomarkers+for+skeletal+fluoride+exposure%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/16453" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>95. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Vieira, A P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2004</span>
# 		<img src="images/curriculo/ico_relevante.gif" style="icone-relevancia">
# 		<b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="338">Vieira, A P G F</a></b>; HANCOCK, R. ; MAIA, R. ; LIMEBACK, H. ; GRYNPAS, M. D. . Is fluoride concentration in dentin and enamel a good indicator of dental fluorosis?. Journal of Dental Research<sup><img id="00220345_95" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="00220345" original-title="Journal Of Dental Research (0022-0345)<br />Fator de impacto (JCR 2022): 7.6"></sup>, Estados Unidos, v. 83, n.1, p. 76-80, 2004. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=00220345&amp;volume=83&amp;issue=&amp;paginaInicial=76&amp;titulo=Is fluoride concentration in dentin and enamel a good indicator of dental fluorosis?&amp;sequencial=95&amp;nomePeriodico=Journal of Dental Research" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000187792700014&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">52</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=1542757620&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">56</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 0022-0345, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22Is+fluoride+concentration+in+dentin+and+enamel+a+good+indicator+of+dental+fluorosis%3F%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/25684" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>96. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Vieira, A P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2003</span><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="339">Vieira, A P G F</a></b>; HANCOCK, R. ; SCHWARTZ, M. ; LIMEBACK, H. ; GRYNPAS, M. D. . How does fluoride concentration in tooth structure affect apatite crystal?. JOURNAL OF DENTAL RESEARCH<sup><img id="00220345_96" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="00220345" original-title="Journal Of Dental Research (0022-0345)<br />Fator de impacto (JCR 2022): 7.6"></sup>, Estados Unidos, v. 82, n.11, p. 909-913, 2003. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=00220345&amp;volume=82&amp;issue=&amp;paginaInicial=909&amp;titulo=How does fluoride concentration in tooth structure affect apatite crystal?&amp;sequencial=96&amp;nomePeriodico=JOURNAL OF DENTAL RESEARCH" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000186185400013&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">28</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=0642374092&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">35</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 0022-0345, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22How+does+fluoride+concentration+in+tooth+structure+affect+apatite+crystal%3F%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/25684" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>97. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">OMAR, J.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2002</span>OMAR, J. ; LUCAS, V. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="340">Vieira, A P G F</a></b> ; ROBERTS, G. J. . The relationship between odontogenic bacteraemia and orthodontic treatment procedures. European Journal of Orthodontics<sup><img id="01415387_97" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="01415387" original-title="European Journal Of Orthodontics (0141-5387)<br />Fator de impacto (JCR 2022): 2.6"></sup>, v. 24, p. 293-301, 2002. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=01415387&amp;volume=24&amp;issue=&amp;paginaInicial=293&amp;titulo=The relationship between odontogenic bacteraemia and orthodontic treatment procedures&amp;sequencial=97&amp;nomePeriodico=European Journal of Orthodontics" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000175693300007&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">50</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=0036616225&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">49</span></a></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> A1, ISSN 0141-5387, fonte <a href="https://sucupira.capes.gov.br/sucupira/public/consultas/coleta/veiculoPublicacaoQualis/listaConsultaGeralPeriodicos.jsf" target="_blank" title="">Qualis/CAPES</a> (quadriênio 2017-2020) <a href="https://scholar.google.com/scholar?q=intitle%3A%22The+relationship+between+odontogenic+bacteraemia+and+orthodontic+treatment+procedures%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a> <a href="https://www.scopus.com/sourceid/25026" target="_blank" title="Visualizar a página do periódico na Scopus">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scopus-icon-16.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>98. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">LOSSIO, J. J. A.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2000</span><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7884498527294072" target="_blank" tabindex="341">LOSSIO, J. J. A.</a> ; VALE, R. ; BENÍGNO, C. A. L. ; GINDIM, A. C. R. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="342">Vieira, A P G F</a></b> . Erros Na Fotopolimerização de Resinas Compostas. Causas e Consequências. REVISTA DO CURSO DE ODONTOLOGIA FFOE/UFC<sup></sup>, FORTALEZA-CE, v. 13, n.01, p. 25-30, 2000. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=01047361&amp;volume=13&amp;issue=&amp;paginaInicial=25&amp;titulo=Erros Na Fotopolimerização de Resinas Compostas. Causas e Consequências&amp;sequencial=98&amp;nomePeriodico=REVISTA DO CURSO DE ODONTOLOGIA FFOE/UFC" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> Não classificado, ISSN 0104-7361 <a href="https://scholar.google.com/scholar?q=intitle%3A%22Erros+Na+Fotopolimeriza%C3%A7%C3%A3o+de+Resinas+Compostas.+Causas+e+Consequ%C3%AAncias%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>99. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">SILVA, A. S. L. E.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">1999</span>SILVA, A. S. L. E. ; MOURA, W. V. B. ; <b>Vieira, A P G F</b> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9245670728518916" target="_blank" tabindex="343">BLEICHER, L.</a> ; BRAGA, M. V. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/6315856840792084" target="_blank" tabindex="344">CORDEIRO, C. M.</a> . Monitoramento do teor de flúor das águas de abastecimento público no município de Fortaleza - Ce. Revista Isegoria - Ação Coletiva em Revista<sup></sup>, v. 2, p. 37-40, 1999. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=22364870&amp;volume=2&amp;issue=&amp;paginaInicial=37&amp;titulo=Monitoramento do teor de flúor das águas de abastecimento público no município de Fortaleza - Ce&amp;sequencial=99&amp;nomePeriodico=Revista Isegoria - Ação Coletiva em Revista" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> Não classificado, ISSN 2236-4870 <a href="https://scholar.google.com/scholar?q=intitle%3A%22Monitoramento+do+teor+de+fl%C3%BAor+das+%C3%A1guas+de+abastecimento+p%C3%BAblico+no+munic%C3%ADpio+de+Fortaleza+-+Ce%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>100. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Vieira, A P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">1998</span><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="345">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7884498527294072" target="_blank" tabindex="346">LOSSIO, J. J. D.</a> . Erros na fotopolimerizacao de resinas compostas. Causas e consequencias. Revista odontologia da Universidade Federal do Ceará<sup></sup>, Ceara, v. 13, n.1, 1998. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=01047361&amp;volume=13&amp;issue=&amp;paginaInicial=&amp;titulo=Erros na fotopolimerizacao de resinas compostas. Causas e consequencias&amp;sequencial=100&amp;nomePeriodico=Revista odontologia da Universidade Federal do Ceará" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> Não classificado, ISSN 0104-7361 <a href="https://scholar.google.com/scholar?q=intitle%3A%22Erros+na+fotopolimerizacao+de+resinas+compostas.+Causas+e+consequencias%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>101. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">LOSSIO, J. J. A.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">1996</span><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7884498527294072" target="_blank" tabindex="347">LOSSIO, J. J. A.</a> ; GONDIM, A. C. R. ; EVANGELISTA, L. A. ; BENÍGNO, C. A. L. ; BRAGA, C. P. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="348">Vieira, A P G F</a></b> . Avaliação da Fotopolimerização Feita Pelo Dentista. XV Encontro Universitário da UFC- Resumo dos Trabalhos, Fortaleza-Ce, 1996. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=&amp;volume=&amp;issue=&amp;paginaInicial=&amp;titulo=Avaliação da Fotopolimerização Feita Pelo Dentista&amp;sequencial=101&amp;nomePeriodico=XV Encontro Universitário da UFC- Resumo dos Trabalhos" tooltip="Citações a partir de 1996"></span><p><img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/icon-16.png" style="margin-bottom:-4px"> Não classificado <a href="https://scholar.google.com/scholar?q=intitle%3A%22Avalia%C3%A7%C3%A3o+da+Fotopolimeriza%C3%A7%C3%A3o+Feita+Pelo+Dentista%22" target="_blank" title="Buscar o título da publicação no Google Acadêmico">
#             <img src="chrome-extension://cobekobjpobenpjdggbpkkklkcfoinen/images/scholar-hat-icon-18.png" style="margin-bottom:-4px">
#           </a></p><br></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# </div>
# <div class="cita-artigos">
# <b><a name="LivrosCapitulos" tabindex="349"></a>Capítulos de livros publicados</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-producao icone-doi" href="http://dx.doi.org/10.18310/9786554620512" target="_blank" tabindex="350"></a>GUILAM, M. C. R. ; <b>VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</b> ; TEIXEIRA, C. P. ; FACCHINI, L. A. . ENSINO EM REDE NA ÁREA DE SAÚDE COLETIVA: EXPERIÊNCIAS, DESAFIOS E TENDÊNCIAS. In: Carla Pacheco Teixeira; Kellen Cristina da Silva Gasque; Maria Cristina Rodrigues Guilam; Maria de Fátima Antero Sousa Machado; Néliton Gomes Azevedo; Rafael Fonseca de CAstro. (Org.). Educação na Saúde: Fundamentos e Perpectivas. 1ed.Porto Alegre: Redeunida, 2023, v. 1, p. 102-116. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-producao icone-doi" href="http://dx.doi.org/10.22533/at.ed.29121180111" target="_blank" tabindex="351"></a>SAINTRAIN, MARIA VIEIRA DE LIMA ; Almeida, Rosa Lívia Freitas de ; Braga, Débora Rosana Alves ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9886955951397288" target="_blank" tabindex="352">LOURENÇO, CAROLINE BARBOSA</a> ; Lima, Lia Vila Real ; Oliveira, Janayne de Sousa ; Vilar, Nathalie Barreto Saraiva ; Bezerra, Carina Bandeira ; <b>MEYER, ANYA PIMENTEL GOMES FERNANDES VIEIRA</b> . DÉFICE COGNITIVO EM PESSOAS IDOSAS: INTERFERÊNCIA DO EDENTULISMO. Pesquisa, Produção e Divulgação do Conhecimento na Odontologia 2. 1ed.: Atena Editora, 2021, v. , p. 114-121. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-producao icone-doi" href="http://dx.doi.org/10.22533/at.ed.1902127043" target="_blank" tabindex="353"></a>Almeida, Janaína Rocha de Sousa ; COUTINHO, MAÍRA BARBOSA ; Reis, Uhiana Braga ; Moreira, Manuela da Silva ; Saldanha, Kátia de Góis Holanda ; <b>VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</b> . A VIABILIDADE DO USO DA TÉCNICA RESTAURADORA ATRAUMÁTICA NA ATENÇÃO PRIMÁRIA À SAÚDE, NO CONTEXTO COVID-19. Odontologia: Da Dentística à Traumatologia. 1ed.: Atena Editora, 2021, v. , p. 17-30. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="354">PIMENTEL G. F. VIEIRA-MEYER, ANYA</a></b>. Prefácio II do livro Políticas, epidemiologia e experiências no Sistema Único de Saúde (SUS) ? possibilidades e desafi os do cenário brasileiro. In: I. Silva Junior, Fernando José Guedes da. org. II. Sales, Jaqueline Carvalho e Silva. org. III. Galiza, Francisca Tereza de. org. IV. Monteiro, Claudete Ferreira de Souza. (Org.). Políticas, epidemiologia e experiências no Sistema Único de Saúde (SUS) ? possibilidades e desafi os do cenário brasileiro. 1ed.Curitiba: Editora CRV, 2020, v. , p. 19-22. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="355">PIMENTEL G. F. VIEIRA-MEYER, ANYA</a></b>. Apresentação do livro Flores de Mandacarus. In: Maísa Paulino Rodrigues; Karla Patricia Cardoso Amorim; Rosana Lúcia Alves de Vilar; José Adailton da Silva; Ricardo Henrique Vieira de Melo. (Org.). Flores de Mandacarus: Produções Cultivadas no Mestrado Profissional em Saúde da Família da RENASF/UFRN. 1ed.Natal: EDUFRN, 2020, v. , p. 8-. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">PAIVA, S. M. ; PERAZZO, M. F. ; <b>PIMENTEL G. F. VIEIRA-MEYER, ANYA</b> ; FONTELES, C. S. R. ; MATTOS, F. F. ; PORDEUS, I. A. . A Epidemiologia abrindo caminho. In: Antônio Sergio Teixeira de Menezes; Fabrício Bitu Sousa. (Org.). Brasil/Guiné-Bissau: caminhos e conquistas entre a educação e a saúde bucal. 1ed.Fortaleza: EdUnichristus, 2020, v. , p. 74-81. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">PAIVA, S. M. ; MATTOS, F. F. ; FONTELES, C. S. R. ; <b>Anya, Pimentel Gomes Fernandes Vieira-Meyer</b> ; PERAZZO, M. F. ; PORDEUS, I. A. . Modelos de Levantamentos epidemiologicos de doenças orais na África. In: Antônio Sergio Teixeira de Menezes; Fabrício Bitu Sousa. (Org.). Brasil/Guiné-Bissau: caminhos e conquistas entre a educação e a saúde bucal. 1ed.Fortaleza: EdUnichristus, 2020, v. , p. 82-92. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="356">SAINTRAIN, MARIA V. DE L.</a> ; PINHEIRO, C. P. O. ; AMORIM, R. F. ; Santos, Z.M.S.A. ; SANTOS, FLAVIANO DA SILVA ; FROTA, M. A. ; <b>Anya Pimentel Gomes Fernandes Vieira-Meyer</b> . GENDER DIFFERENCES IN HEALTH SELF-PERCEPTION IN OLDER ADULTS IN PRIMARY HEALTH CARE IN BRAZIL. GENDER DIFFERENCES IN HEALTH SELF-PERCEPTION IN OLDER ADULTS IN PRIMARY HEALTH CARE IN BRAZIL. 1aed.Louven: EUSER, 2018, v. , p. 538-551. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="357">Anya Pimentel Gomes Fernandes Vieira</a></b>; DIAS, MARIA DO SOCORRO DE ARAÚJO . Redes educacionais: Tendência ou ousadia de uma estratégia de Gestão?. In: Percy Antonio Galimbertti; Izabelle MontÁlverne Napoleão Albuquerque. (Org.). Sociopolitica da Saúde: Desafios e oportunidades. 1aed.Sobral: UVA Editora, 2018, v. , p. 137-156. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>10. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">PETROLA, K. A. F. ; BEZERRA, I. B. ; MENEZES, E. A. V. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="358">SAINTRAIN, M. V. L.</a> ; <b>Vieira, A P G F</b> . AÇÕES DE PROMOÇÃO E PROTEÇÃO DA SAÚDE BUCAL DESENVOLVIDAS NA ESTRATÉGIA DE SAÚDE DA FAMÍLIA. In: Liberata Campos Coimbra, Geórgia Sibele Nogueira da Silva, Maísa Paulino Rodrigues, Annatália Meneses de Amorim Gomes. (Org.). ATENÇÃO INTEGRAL EM SAÚDE DA FAMÍLIA. 1ed.Natal/RN: edufrn, 2017, v. 1, p. 175-205. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>11. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">NUNE, R. C. ; SILVEIRA, P. V. ; MARCELO, M. E. M. F. ; FROTA, M. A. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="359">SAINTRAIN, M. V. L.</a> ; <b>Vieira, A P G F</b> . AVALIAÇÃO DE ATIVIDADES LÚDICAS NA PROMOÇÃO DE SAÚDE BUCAL. In: ZÉLIA MARIA DE SOUSA ARAÚJO SANTOS; MIRNA ALBUQUERQUE FROTA; ALINE BARBOSA TEIXEIRA MARTINS. (Org.). TECNOLOGIAS EM SAÚDE: da abordagem teórica a construção e aplicação no cenário do cuidado. 1ed.Fortaleza/ce: EDUECE, 2016, v. 01, p. 01-482. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>12. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">CARVALHO, M. F. ; <b>Vieira, A P G F</b> ; SILVA, A. R. V. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5537817532828798" target="_blank" tabindex="360">VASCONCELOS, M. I. O.</a> . As Dimensões Estratégicas do PMAQ-AB nos Estados do Piauí e Ceará. In: Altamira Pereira da Silva Reichert; Ana Patricia Pereira Morais; Anya Pimentel Gomes FErnandes Vieira Meyer; Renata de Sousa Alves. (Org.). Avaliação e Gestão do Cuidado na Saúde da Família. 01ed.Fortaleza/Ceará: Editora UFC, 2016, v. 01, p. 69-86. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>13. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="361">Vieira, A P G F</a></b>; DUTRA, E. M. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5537817532828798" target="_blank" tabindex="362">VASCONCELOS, M. I. O.</a> ; GUIMARAES, J. M. X. ; DIAS, M. S. A. . Operacionalização do PMAQ-AB na Perspectiva dos Gestores e Trabalhadores numa Região de Saúde do Ceará. In: Anya Pimentel Gomes Fernandes Vieira Meyer; Altamira Pereira da Silva Reichert; Ana Patricia Pereira Morais; Renata de Sousa Alves. (Org.). Avaliação e Gestão do Cuidado na Saúde da Família. 01ed.Fortaleza/CE: Edições UFC, 2016, v. 01, p. 375-388. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>14. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="363">Vieira, A P G F</a></b>; PRADO, J. P. ; OLIVEIRA, J. W. G. ; CASTRO FILHO, C. S. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5516641709622899" target="_blank" tabindex="364">BIZERRIL, D. O.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="365">SAINTRAIN, M. V. L.</a> . Satisfação do Idoso em Relação a Saúde Bucal na Atenção Básica. In: Maria Vieira de Lima Saintrain; Ana Paula Soares Gondim; Vanira Tereza Lopes da Silva. (Org.). O Sistema Único de Saúde Cuidando da Pessoa Idosa. 1ed.Fortaleza: Editora da UECE, 2014, v. 1, p. 1-200. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>15. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="366">Vieira, A P</a></b>; LANDIM, F. L. P. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9948525011531885" target="_blank" tabindex="367">CATRIB, A. M. F.</a> . Interfases da educação em saúde e da prevenção na prevalência da cárie dentária. Promoção da Saúde na Diversidade Humana e na Pluralidade de Itinerários terapêuticos. 1ed.Campinas: Saberes Editora, 2012, v. 1, p. 1-355. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>16. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5516641709622899" target="_blank" tabindex="368">BIZERRIL, D. O.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="369">SAINTRAIN, M. V.</a> ; <b>Vieira, A P G F</b> . Autoavaliação em Saúde Bucal Geriátrica. In: Maria Vieira de Lima Saintrain, Cleoneide Paulo Oliveira Pinheiro e Raimunda Magalhães da SIlva. (Org.). Saúde do Idoso: estudos e prática no processo do envelhecimento. 1ed.Fortaleza: Editora da UFC, 2012, v. 1, p. 289-310. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>17. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="370">SAINTRAIN, M. V. L.</a> ; PINHEIRO, C. P. O. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8577652222943813" target="_blank" tabindex="371">Silva, R M</a> ; <b>Vieira, A P</b> . Caracteristicas Epidemiologicas de Idosos Atendidos no Centro Integrado de Diabetes e Hipertensão do Ceará. In: Maria Vieira de Lima Saintrain; Cleoneide Paulo Oliveira Pinheiro; Raimundo Magalhães da Silva. (Org.). Saúde do Idoso: Estudos e Práticas no Processo do Envelhecimento. 1ed.Fortaleza: Edições UFC, 2012, v. 1, p. 1-329. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>18. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Gondim, APS ; Sousa,RA ; Vieira, L J E S ; Oliveira, A K A ; MARCELO, M. E. M. F. ; <b>Vieira, A P G F</b> . Saúde da Família como Estratégia de Articulação de Saberes e Práticas. In: Raimunda Magalhães da Silva e Maria Salete Bessa Jorge. (Org.). Cuidado em Saúde: Desafios e Práticas. Fortaleza: Ed UECE, 2011, v. , p. 271-289. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>19. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Lourenço, A N ; Amorim, R F ; Rodrigues, F J ; Nobre, MCQ ; Santos, Z.M.S.A. ; <b>Vieira, A P G F</b> . Modificações corporais em travestis: Abordagens conceituais e agravos à saúde. In: Rosendo Freitas de Amorim; Ana Maria Fontenelle Catrib; Maria Salete Bessa Jorge. (Org.). Olhares da saúde coletiva sobre o corpo. 1aed.Fortaleza: EdUECE, 2011, v. , p. 199-226. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>20. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="372">Vieira, A P</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9948525011531885" target="_blank" tabindex="373">CATRIB, A. M. F.</a> ; DIAS, M. S. A. ; FROTA, M. A. . Promoção da Saúde de crianças em alimentação complementar por meio da utilização dos alimentos regionais. In: Ana Maria Fontenelle Catrib; Maria Socorro de Araújo Dias; Mirna Albuquerque Frota. (Org.). Promoção da Saúde no Contexto da Estrategia Saúde da Família. 1ed.Campinas: Saberes, 2011, v. , p. 1-277. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>21. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="374">Vieira, A P G F</a></b>; SILVA, S. L. E. ; NUTO, S. . Vigilancia Sanitaria da Fluoretacao das Aguas de Abastecimento Publico no Municipio de Fortaleza-CE. In: Benedita Pessoa Forte; Maria Vilani Cavalcante Guedes. (Org.). Conhecimento e Poder em Saude: Cultura academica da pratica clinica e social aplicada. Fortaleza: Grafica da Universidade Federal do Ceara, 1997, v. , p. -. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b><a name="TrabalhosPublicadosAnaisCongresso" tabindex="375"></a>Resumos expandidos publicados em anais de congressos</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">NUNES, R. C. ; MARCELO, M. E. M. F. ; <b>Vieira, A P G F</b> . O aprendizado em Saúde Bucal por alunos participantes de atividades lúdico-educativas de promoção de saúde. In: IV Congresso Internacional em Avaliação Educacional, 2008, Fortaleza. IV Congresso Internacional em Avaliação Educacional - Avaliação: perspectivas para a escola contemporânea. Fortaleza, 2008. v. 01. p. 232-245. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">SILVEIRA, P. V. ; NUNES, R. C. ; <b>Vieira, A P G F</b> . Avaliação comparativa de aprendizagem no uso do lúdico e da aula expositiva no ensino fundamental. In: IV Congresso Internacional em Avaliação Educacional, 2008, Fortaleza. IV Congresso Internacional em Avaliação Educacional - Avaliação: perspectivas para a escola contemporânea. Fortaleza, 2008. v. 01. p. 119-130. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b><a name="TrabalhosPublicadosAnaisCongresso" tabindex="376"></a>Resumos publicados em anais de congressos</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="377">Vieira-MEYER, A. P. G. F.</a>; CALVASINA, P. G. ; <b>Anya P G F Vieira-Meyer</b> . Brazilian conditional cash transfer five year old children dental needs and access to dental care. In: 66th ORCA Congress, 2019, Cartagena. 66th ORCA Congress Abstract. London: Karger, 2019. v. 53. p. 357-410. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">PINHEIRO, T. X. A. ; <b>Vieira, A P G F</b> ; UCHOA, S. A. C. ; COELHO, A. A. ; SOUSA, C. S. M. . External evaluation of the Brazilian Primary Health Care Program. Are improvements being observed?. In: 9th European Public Health Conference, 2016, Austria. External evaluation of the Brazilian Primary Health Care Program. Are improvements being observed?. Estados Unidos: EUPHA, 2016. v. 26. p. 264-264. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">PINHEIRO, T. X. A. ; UCHOA, S. A. C. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4194885068583094" target="_blank" tabindex="378">MACHADO, M. F. A. S.</a> ; SAMPAIO, A. T. L. ; COELHO, A. A. ; SOUSA, C. S. M. ; ROCHA, P. M. ; <b>Vieira, A P G F</b> . The 3rd cycle of the Brazilian National Programme for Improving Access and Quality of Primary Care. In: EPH Conference reference 2016, 2016, Austria. The 3rd cycle of the Brazilian National Programme for Improving Access and Quality of Primary Care. Estados Unidos: EUPHA, 2016. v. 26. p. 452-452. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5516641709622899" target="_blank" tabindex="379">BIZERRIL, D. O.</a> ; <b>Vieira, A P G F</b> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="380">SAINTRAIN, M. V. L.</a> ; AGUIAR, A. S. W. . Resolutividade de Centros de Especialidades Odontológicas Regionais do Estado do Ceará. In: 10º Congresso Brasileiro de Saúde Coletiva, 2012, Porto Alegre - RS. Anais - 10º Congresso Brasileiro de Saúde Coletiva. Porto Alegre: Saúde Coletiva, 2012. v. 1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="381">Vieira, A P G F</a></b>. The Oral  Care the Brazilian Primary Health Care System. In: OMICS - Publishing Group, 2011, EUA. Anais - OMICS - Publishing Group. EUA: Editora da FMMSR, 2011. v. 1. p. 1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="382">SAINTRAIN, M. V. L.</a> ; MARCELO, M. E. M. F. ; SILVEIRA, P. V. ; <b>Vieira, A P G F</b> . Evaluation of an Oral Health Promotion Program in a poor area of Brazil. In: 20th IUHPE World Conference on Health Promotion, 2010, Genebra. 20th IUHPE World Conference on Health Promotion, 2010. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="383">SAINTRAIN, M. V. L.</a> ; SILVEIRA, P. V. ; MARCELO, M. E. M. F. ; <b>Vieira, A P G F</b> . Dental Care Program for Brazilian Industry Workers. In: 20th IUHPE World Conference on Health Promotion, 2010, Genebra. 20th IUHPE World Conference on Health Promotion, 2010. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">CASTRO FILHO, C. S. ; SILVEIRA, P. V. ; <b>Vieira, A P G F</b> . Influência da assistência odontológica na saúde bucal de trabalhadores. In: 27º CIOSP - Congresso Internacional de Odontologia de São Paulo, 2009, São Paulo. 27º CIOSP, 2009. v. 1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">CASTRO FILHO, C. S. ; SILVEIRA, P. V. ; <b>Vieira, A P G F</b> . Qualidade de vida em trabalhadores com e sem a assistência odontológica. In: 27º CIOSP - Congresso Internacional de Odontologia de São Paulo, 2009, São Paulo. 27º CIOSP, 2009. v. 1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>10. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="384">Vieira, A P G F</a></b>; LUNA, G. L. M. ; Vieira, L J E S ; MOREIRA, D. P. ; SILVA, J. G. . Notification of abuses against children and teenagers as an intrument for detection and prevention of violence in primary health care in Fortaleza, Ceará, Brasil. In: 12th World Congress on Public Health, 2009, Istanbul. 2009 World Congress on Public Health, 2009. v. 1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>11. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="385">SAINTRAIN, M. V. L.</a> ; <b>Vieira, A P G F</b> . Oral health epidemiological study realized dy health agents. In: 12th World Congress on Public Health, 2009, Istanbul. 2009 World Congress on Public Health, 2009. v. 1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>12. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="386">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="387">SAINTRAIN, M. V. L.</a> . Community oral health index for health agents. In: 12th World Congress on Public Health, 2009, Istanbul. 2009 World Congress on Public Health, 2009. v. 1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>13. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="388">Vieira, A P G F</a></b>; SILVA, J. G. ; Vieira, L J E S ; LUNA, G. L. M. ; SILVEIRA, E. A. R. ; MARTINS, E. M. P. . Instutionalized aged people: Reasons expressed in the collective subject discourse. In: 12th World Congress on Public Health, 2009, Istanbul. 2009 World Congress on Public Health, 2009. v. 1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>14. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="389">Vieira, A P G F</a></b>. Sedação e anestesia geral em odontopediatria. In: III Congresso Internacional de Odontologia, 2009, Fortaleza. III Congresso Internacional de Odontologia, 2009. v. 1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>15. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="390">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="391">SAINTRAIN, M. V. L.</a> . Levantamento de saúde bucal no município de Guaiuba-CE. In: III Congresso Internacional de Odontologia, 2009, Fortaleza. III Congresso Internacional de Odontologia, 2009. v. 1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>16. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="392">Vieira, A P G F</a></b>; GOES NETO, J. B. ; MARIANO, R. H. ; BRITO FILHO, F. G. . Prevalência de dor na coluna e alterações de postura em estudantes de odontologia do estado do Ceará. In: III Congresso Internacional de Odontologia, 2009, Fortaleza. III Congresso Internacional de Odontologia, 2009. v. 1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>17. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="393">Vieira, A P G F</a></b>; MARCELO, M. E. M. F. ; SILVEIRA, P. V. . Saúde bucal de mães e filhos com acompanhamento odontológico pré-natal e puerperal pela estratégia Saúde da Família. In: XI Congresso de Odontologia do Rio Grande do Norte, 2009, Natal. XI Congresso de Odontologia do RN, 2009. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>18. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="394">Vieira, A P G F</a></b>; MARCELO, M. E. M. F. ; SILVEIRA, P. V. . Serviço odontológico como benefício de empresas provadas no Ceará: satisfação do usuário.. In: XI Congresso de Odontologia do Rio Grande do Norte, 2009, Natal. XI Congresso de Odontologia do RN, 2009. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>19. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">OLIVEIRA, J. W. G. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5516641709622899" target="_blank" tabindex="395">BIZERRIL, D. O.</a> ; PRADO, J. P. ; CASTRO FILHO, C. S. ; <b>Vieira, A P G F</b> . Perspectivas e opiniões sobre a qualidade dos serviços da Equipe de Saúde Bucal no PSF. In: IX Congresso Brasileiro de Saúde Coletiva, 2009, Recife. Ciência e Saúde Coletiva (Impresso), 2009. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>20. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="396">Vieira, A P G F</a></b>; CASTRO FILHO, C. S. ; SILVEIRA, P. V. . Oral health and quality of life perception by adults with and without oral health  assistence. In: 12th World Congress on Public Health, 2009, Istambul. 12th World Congress on Public Health, 2009. v. 1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>21. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">MARCELO, M. E. M. F. ; SILVEIRA, P. V. ; <b>Vieira, A P G F</b> . Saúde bucal de mães e bebês companhados por Programa de Atenção Odontológica na estratégia Saúde da Família. In: IX Encontro de Pós-Graduação e Pesquisa da UNIFOR, 2009, Fortaleza. IX Encontro de Pós-Graduação e Pesquisa da UNIFOR, 2009. v. 1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>22. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">MARCELO, M. E. M. F. ; SILVEIRA, P. V. ; NUNES, R. C. ; <b>Vieira, A P G F</b> . Ações esporádicas de Promoção de Saúde Bucal utilizando atividades lúdicas e aula exportiva. In: IX Encontro de Pós-Graduação e Pesquisa da UNIFOR, 2009, Fortaleza. IX Encontro de Pós-Graduação e Pesquisa da UNIFOR, 2009. v. 1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>23. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">SILVEIRA, P. V. ; MARCELO, M. E. M. F. ; CASTRO FILHO, C. S. ; <b>Vieira, A P G F</b> . Satisfação com serviço odontológico na empresa. In: IX Encontro de Pós-Graduação e Pesquisa da UNIFOR, 2009, Fortaleza. IX Encontro de Pós-Graduação e Pesquisa da UNIFOR. v. 1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>24. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">ALENCAR, D. R. ; <b>Vieira, A P G F</b> ; CASTRO FILHO, C. S. ; OLIVEIRA, J. W. G. . Avaliação da satisfação dos usuários em relação as atividades da Equipe de Saúde Bucal na Regional 2 do município de Fortaleza. In: XV Encontro de Iniciação à Pesquisa da UNIFOR, 2009. XV Encontro de Iniciação à Pesquisa da UNIFOR, 2009. v. 1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>25. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">PRADO, J. P. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5516641709622899" target="_blank" tabindex="397">BIZERRIL, D. O.</a> ; <b>Vieira, A P G F</b> . Satisfação dos usuários do SUS em relação às atividades da equipe de Saúde Bucal na SER VI do Município. In: XV Encontro de Iniciação à Pesquisa da UNIFOR, 2009, Fortaleza. XV Encontro de Iniciação à Pesquisa da UNIFOR, 2009. v. 1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>26. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="398">SAINTRAIN, M. V. L.</a> ; <b>Vieira, A P G F</b> . Levantamento de necessidades de saúde bucal por agentes comunitários de saúde. In: 10º Congresso Brasileiro de Medicina de Família e Comunidade, 2009, Florianópolis. 10º Congresso Brasileiro de Medicina de Família e Comunidade, 2009. v. 1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>27. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">PRADO, J. P. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5516641709622899" target="_blank" tabindex="399">BIZERRIL, D. O.</a> ; MOTA, F. ; <b>Vieira, A P G F</b> . Avaliação dos usuários quanto às ESB na SER VI de Fortaleza-CE. In: 10º Congresso Brasileiro de Medicina de Família e Comunidade, 2009, Florianópolis. 10º Congresso Brasileiro de Medicina de Família e Comunidade, 2009. v. 1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>28. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="400">Vieira, A P G F</a></b>; ALENCAR, D. R. ; CASTRO FILHO, C. S. ; OLIVEIRA, J. W. G. . Como os usuários avaliam as equipes de saúde bucal de Fortaleza-CE?. In: 10º Congresso Brasileiro de Medicina de Família e Comunidade, 2009, Florianópolis. 10º Congresso Brasileiro de Medicina de Família e Comunidade, 2009. v. 1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>29. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">OLIVEIRA, J. W. G. ; ALMEIDA, V. H. ; REGADAS, P. ; <b>Vieira, A P G F</b> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="401">SAINTRAIN, M. V. L.</a> . A influência de um programa de prevenção da cárie dentária em escolares. In: 9º Congresso Brasileiro de Medicina de Família e Comunidade, 2º Congresso Cearense de Medicina de Família e Comunidade e 1º Simpósio Internacional de Ensino Aprendizagem em MFC e APS, 2008, Fortaleza. 9º Congresso Brasileiro de Medicina de Família e Comunidade, 2008. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>30. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">OLIVEIRA, J. W. G. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5516641709622899" target="_blank" tabindex="402">BIZERRIL, D. O.</a> ; PRADO, J. P. ; CASTRO FILHO, C. S. ; <b>Vieira, A P G F</b> . Como está a saúde bucal dos idosos em Fortaleza?. In: 9º Congresso Brasileiro de Medicina de Família e Comunidade, 2º Congresso Cearense de Medicina de Família e Comunidade e 1º Simpósio Internacional de Ensino Aprendizagem em MFC e APS, 2008, Fortaleza. 9º Congresso Brasileiro de Medicina de Família e Comunidade, 2008. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>31. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">OLIVEIRA, J. W. G. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5516641709622899" target="_blank" tabindex="403">BIZERRIL, D. O.</a> ; PRADO, J. P. ; CASTRO FILHO, C. S. ; <b>Vieira, A P G F</b> . A opinião pública sobre a equipe de saúde bucal do PSF. In: 9º Congresso Brasileiro de Medicina de Família e Comunidade, 2º Congresso Cearense de Medicina de Família e Comunidade e 1º Simpósio Internacional de Ensino Aprendizagem em MFC e APS, 2008, Fortaleza. 9º Congresso Brasileiro de Medicina de Família e Comunidade, 2008. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>32. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="404">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5516641709622899" target="_blank" tabindex="405">BIZERRIL, D. O.</a> ; PRADO, J. P. ; OLIVEIRA, J. W. G. ; CASTRO FILHO, C. S. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="406">SAINTRAIN, M. V. L.</a> . Unified health care system users satisfaction with dental treatment provided. In: 86th General Session &amp; Exhibition on the IADR, 2008, Toronto. IADR - General Session 2008, 2008. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>33. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5516641709622899" target="_blank" tabindex="407">BIZERRIL, D. O.</a> ; SAMPAIO, E. ; <b>Vieira, A P G F</b> . Oral health in the elder population. In: 55th ORCA Congress, 2008, Groninger. Caries Research, 2008. v. 42. p. 203-203. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>34. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="408">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="409">SAINTRAIN, M. V. L.</a> . Validation of the comunity oral health index. In: 55th ORCA Congress, 2008, Groninger. Caries Research, 2008. v. 42. p. 239-239. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>35. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="410">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5516641709622899" target="_blank" tabindex="411">BIZERRIL, D. O.</a> ; SAMPAIO, E. F. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="412">SAINTRAIN, M. V. L.</a> . Como está a saúde de nossos idosos (institucionalizados ou não). In: XVIII Congresso Mundial de Epidemiologia e VII Congresso Brasileiro de Epidemiologia, 2008, Porto Alegre. CD de Anais do EPI 2008, 2008. v. 01. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>36. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="413">SAINTRAIN, M. V. L.</a> ; <b>Vieira, A P G F</b> ; HONORIO, V. A. ; DIAS, M. B. R. L. . Interferência da educação em saúde e da prevenção na prevalência da cárie dentária. In: XVIII Congresso Mundial de Epidemiologia e VII Congresso Brasileiro de Epidemiologia, 2008, Porto Alegre. CD de Anais do EPI 2008, 2008. v. 01. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>37. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="414">Vieira, A P G F</a></b>; CASTRO FILHO, C. S. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5516641709622899" target="_blank" tabindex="415">BIZERRIL, D. O.</a> ; PRADO, J. P. ; OLIVEIRA, J. W. G. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="416">SAINTRAIN, M. V. L.</a> . Satisfação do usuário em relação aos dentistas do PSF em Fortaleza. In: VII Congresso Mundial de Epidemiologia e VII Congresso Brasileiro de Epidemiologia, 2008, Porto Alegre. CD de Anais do EPI 2008, 2008. v. 01. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>38. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="417">SAINTRAIN, M. V. L.</a> ; <b>Vieira, A P G F</b> ; SILVA, K. P. . Validação do Índice Comunitário em Saúde Bucal. In: VII Congresso Mundial de Epidemiologia e VII Congresso Brasileiro de Epidemiologia, 2008, Porto Alegre. CD de Anais do EPI 2008, 2008. v. 1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>39. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">NUNES, R. C. ; MARCELO, M. E. M. F. ; SILVEIRA, P. V. ; <b>Vieira, A P G F</b> ; FROTA, M. A. . Educação e motivação em Saúde Bucal através de estratégia lúdico-educativa.. In: II Mostra do Sistema Municipal de Saúde Escolar, 2008, Fortaleza. II Mostra do Sistema Municipal de Saúde Escolar, 2008. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>40. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">MARCELO, M. E. M. F. ; SILVEIRA, P. V. ; <b>Vieira, A P G F</b> . Estratégia de promoção de Saúde Bucal com mães e bebês na Atenção Básica. In: II Mostra do Sistema Municipal de Saúde Escolar, 2008, Fortaleza. II Mostra do Sistema Municipal de Saúde Escolar, 2008. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>41. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">PRADO, J. P. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5516641709622899" target="_blank" tabindex="418">BIZERRIL, D. O.</a> ; OLIVEIRA, J. W. G. ; CASTRO FILHO, C. S. ; <b>Vieira, A P G F</b> . Satisfação do usuário do SUS em relação à eficácia, eficiência e efetividade das atividades da equipe de Saúde Bucal. In: XIV Encontro de Iniciação à Pesquisa, 2008, Fortaleza. Encontros Científicos 2008 - CD-Rom. Fortaleza: Gráfica Unifor, 2008. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>42. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">SILVEIRA, P. V. ; CASTRO FILHO, C. S. ; <b>Vieira, A P G F</b> . Acesso à assistência de Odontologia no Município de Horizonte. In: II Congresso Brasileiro de Odontologia do Trabalho, 2008, Fortaleza. II Congresso Brasileiro de Odontologia do Trabalho, 2008. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>43. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5516641709622899" target="_blank" tabindex="419">BIZERRIL, D. O.</a> ; SAMPAIO, E. ; <b>Vieira, A P G F</b> . Saúde bucal em pessoas na terceira idade institucionalizadas. In: I Mostra do Sistema Municipal de Saúde Escola de Fortaleza, 2007, Fortaleza. Anais da I Mostra do Sistema Municipal de Saúde Escola de Fortaleza, 2007. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>44. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Vieira, L J E S ; Abreu, C A P ; Ferreira, R C ; <b>Vieira, A P G F</b> . School Violence: A Problem Compromising a Healthy Environment. In: 19th World Conference on Health Promotion and Health Education, 2007, Vancouver. Annais do 19th World Conference on Health Promotion and Health Education, 2007. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>45. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">NUNES, R. C. ; <b>Vieira, A P G F</b> . The Use of Educational Games on Oral Health Promotion. In: 19th World Conference on Health Promotion and Health Education, 2007, Vancouver. Annais da 19th World Conference on Health Promotion and Health Education, 2007. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>46. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">AQUIAR, C. C. T. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7492450432942397" target="_blank" tabindex="420">Montenegro Jr, R.M.</a> ; <b>Vieira, A P G F</b> . The Influence of Depression and Oral Health in Diabetics Patients Quality of Life. In: 19th World Conference on Health Promotion and Health Education, 2007, Vancouver. Annais da 19th World Conference on Health Promotion and Health Education, 2007. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>47. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/3992942738436467" target="_blank" tabindex="421">ALMEIDA, R. F. C.</a> ; Barros, E ; Pessoa, N ; <b>Vieira, A P G F</b> . Evaluating the Interest of the Community on Establishing a Service to Collect and Redistribute Unused Drugs. In: 19th World Conference on Health Promotion and Health Education, 2007, VAncouver. Annais do 19th World Conference on Health Promotion and Health Education, 2007. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>48. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="422">SAINTRAIN, M. V. L.</a> ; <b>Vieira, A P G F</b> ; COSTA, E. H. M. . Autopercepção da condição de saúde bucal em idosos institucionalizados e não institucionalizados. In: IV Congresso Brasileiro de Ciências Sociais e Humanas em Saúde, 2007, Salvador. Anais do IV Congresso Brasileiro de Ciências Sociais e Humanas em Saúde, 2007. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>49. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XAVIER, E. P. ; <b>Vieira, A P G F</b> ; ABREU, C. A. P. ; Abreu, C A P ; Vieira, L J E S ; Ferreira, R C . Violência na escola pública na percepção de alunos do ensino fundamental e médio, Fortaleza, Ceará. In: IV Congresso Brasileiro de Ciências Sociais e Humanas em Saúde, 2007, Salvador. Anais do IV Congresso Brasileiro de Ciências Sociais e Humanas em Saúde, 2007. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>50. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">VIEIRA, G. C. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/3162595976710799" target="_blank" tabindex="423">VIEIRA, N. F. C.</a> ; <b>Vieira, A P G F</b> . Pharmacists contribution on Healt promotion in the Brazilian Health System. In: 19th World Conference on Health Promotion and Health Education, 2007, Vancouver. Annais do 19th World Conference on Health Promotion and Health Education, 2007. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>51. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/3992942738436467" target="_blank" tabindex="424">ALMEIDA, R. F. C.</a> ; BARROS, E.P. ; PESSOA, N. T. ; <b>Vieira, A P G F</b> . Evaluating the interest of the community on establishing a service to collet and redistribute unused drugs. In: 19th World Conference on Health Promotions and Health Educacion, 2007, Vancouver. Annais 19th World Conference on Health Promotions and Health Educacion, 2007. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>52. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Vieira, L J E S ; ABREU, C. A. P. ; Ferreira, R C ; <b>Vieira, A P G F</b> . School violence: a problem compromising a healthy environment. In: Annais do 19th World Conference on Health Promotion and Health Education, 2007, Vancouver. Annais Annais do 19th World Conference on Health Promotion and Health Education, 2007. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>53. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">NUNES, R. C. ; <b>Vieira, A P G F</b> . The use of educational games on oral health promotion. In: 19ht World Conference on Health Promotion and Health Education, 2007, Vancouver. Annais 19ht World Conference on Health Promotion and Health Education, 2007. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>54. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">AQUIAR, C. C. T. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7492450432942397" target="_blank" tabindex="425">Montenegro Jr, R.M.</a> ; <b>Vieira, A P G F</b> . The influence of depression and oral health in diabetics patients quality of life. In: 19ht World Conference on Health Promotion and Health Education, 2007, Vancouver. Annais 19ht World Conference on Health Promotion and Health Education, 2007. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>55. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">VIEIRA, G. C. ; <b>Vieira, A P G F</b> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/3162595976710799" target="_blank" tabindex="426">VIEIRA, N. F. C.</a> . Percepção dos farmacêuticos sobre a inserção do farmacêutico no PSF. In: I Mostra do Sistema Municipal de Saúde Escola de Fortaleza, 2007, Fortaleza. Anais da I Mostra do Sistema Municipal de Saúde Escola de Fortaleza, 2007. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>56. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Ellery, AE ; <b>Vieira, A P G F</b> ; Andrade, LOM ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0636763091396917" target="_blank" tabindex="427">NUTO, S. A. S.</a> . O sistema de formação em saúde no Município de Fortaleza. In: I Mostra do Sistema Municipal de Saúde Escola de Fortaleza, 2007, Fortaleza. Anais da I Mostra do Sistema Municipal de Saúde Escola de Fortaleza, 2007. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>57. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">COSTA, E. H. M. ; <b>Vieira, A P G F</b> . Autopercepção da condição de saúde bucal em idosos institucionalizados e não institucionalizados. In: GERP 2007 - 5º Congresso Paulista de Geriatria e Gerontologia, 2007, Santos. GERP 2007, 2007. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>58. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="428">Vieira, A P G F</a></b>; COSTA, E. H. M. . Autopercepção da condição de saúde bucal em idosos institucionalizados e não institucionalizados. In: IV Congresso Brasileiro de Ciências Sociais e Humanas em Saúde - ABRASCO, 2007, Salvador. Anais do IV Congresso Brasileiro de Ciências Sociais e Humanas em Saúde, 2007. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>59. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="429">Vieira, A P G F</a></b>; GRYNPAS, M. D. . Vigilância sanitária das águas em área endêmica de fluorose dental. In: 11º Congresso Mundial de Saúde Pública, 2006, Rio de Janeiro. Annais do 11º Congresso Mundial de Saúde Pública, 2006. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>60. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="430">Vieira, A P G F</a></b>; GRYNPAS, M. D. . Quality of water in an Endemic Area of Dental Fluorosis. In: 84th General Session and Exhibition of the International Association for Dental Research, 2006, Brisbane. Annais do 84th General Session and Exhibition of the International Association for Dental Research. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>61. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Dalcastel, I. L. ; Monteiro, L.Z. ; Santos, Z.M.S.A. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="431">Vieira, A P G F</a></b> ; Montenegro Jr.,R.M. . Análise do conhecimento e da atuação de educadores físicos frente a portadores de hipertensão arterial. In: XXIX Simpósio internacional de ciências do esporte, 2006, São Paulo. Revista Brasielira de Ciência e Movimento, 2006. v. 14. p. 89. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>62. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Montenegro Jr.,R.M. ; Monteiro, L.Z. ; Dalcastel, I. L. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="432">Vieira, A P G F</a></b> ; Santos, Z.M.S.A. . Conhecimento e prática do educador físico frente a crianças portadoras de diabetes mellitus. In: XXIX Simpósio Internacional de Ciências do Esporte, 2006, São Paulo. Revista Brasileira de Ciência e Movimento, 2006. v. 14. p. 100. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>63. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Monteiro, L.Z. ; Dalcastel, I. L. ; Santos, Z.M.S.A. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="433">Vieira, A P G F</a></b> ; Silva, C.A.B. ; Montenegro Jr.,R.M. . Análise do conhecimento e da atuação de educadores físicos em academias de fortaleza frente a portadores de diabetes mellitus. In: 29th International Symposium on Sports Sciences, 2006, São Paulo. Revista Brasileira de Ciência e Movimento, 2006. v. 14. p. 73. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>64. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Monteiro, L.Z. ; Dalcastel, I. L. ; Santos, Z.M.S.A. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="434">Vieira, A P G F</a></b> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7492450432942397" target="_blank" tabindex="435">Montenegro Jr, R.M.</a> . Estudo da atuação de educadores físicos junto a pacientes com diabetes mellitus em academias de ginástica de Fortaleza (CE). In: 27º Congresso Brasileiro de Endocrinologia e Metabologia, 2006, Recife/Olinda. Endocrinologia &amp; Metabologia, 2006. v. 50. p. S527. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>65. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="436">Anya Vieira</a></b>; GRYNPAS, M. D. ; MENDES JR. I.R. . Quality of Water in an Endemic Area of Dental Fluorosis. In: IADR General Session &amp; Exhibition, 2006, Canadá. IADR General Session &amp; Exhibition, 2006. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>66. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="437">Vieira, A P G F</a></b>. Tooth F concentration, Dental fluorosis and Tooth Quality. In: Research Day, 2005, Toronto. Research Day - Abstracts, 2005. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>67. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="438">Vieira, A P G F</a></b>; HANCOCK, R. ; DIMITRIU, M. ; SCHWARTZ, M. ; LIMEBACK, H. ; GRYNPAS, M. D. . Tooth F concentration, Dental fluorosis and Tooth Quality in humans. In: 83rd General Session &amp; Exhibition of the IADR, 2005, Baltimore. J Dent Res, 2005. v. 84. p. 2149. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>68. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="439">Vieira, A P G F</a></b>; HANCOCK, R. ; EGGERTSSON, H. ; EVERETT, E. T. ; GRYNPAS, M. D. . Dental Fluorosis Genetic Susceptibility and teeth´s mechanical and material properties. In: 82nd General Session &amp; Exhibition of the IADR, 2004, Honolulu. 82nd General Session &amp; Exhibition of the IADR Anais, 2004. v. 83. p. 1420. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>69. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="440">Vieira, A P G F</a></b>; HANCOCK, R. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7549914789004407" target="_blank" tabindex="441">SAMPAIO, F.</a> ; LIMEBACK, H. ; GRYNPAS, M. D. . The relation between dentin and enamel fluoride concentration and the severity of human dental fluorosis- A study in an endemic dental fluorosis area. In: ORCA, 2004, Marburg. Caries Research, 2004. v. 38. p. 398. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>70. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="442">Vieira, A P G F</a></b>; HANCOCK, R. ; LIMEBACK, H. ; GRYNPAS, M. D. . Dentin vs Enamel as biomarkers for F induced changes in teeth. In: Research Day - University of Toronto, 2004, Toronto. Abstract Book, 2004. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>71. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="443">Vieira, A P G F</a></b>; LAWRENCE, H. ; LIMEBACK, H. ; GRYNPAS, M. D. . Visual Analogue Scale for Dental Fluorosis Journal of Dental Research. In: AADR/CADR Annual Meeting, 2003, San Antonio, TX. AADR/CADR Annual Meeting Anais, 2003. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>72. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="444">Vieira, A P G F</a></b>; HANCOCK, R. ; LIMEBACK, H. ; GRYNPAS, M. D. . Dentin vs Enamel as biomarkers for F induced changes in teeth.. In: IADR 81st General Session, 2003, Goteborg, Sweden. IADR 81st General Session Anais, 2003. v. 82. p. 2517. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>73. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="445">Vieira, A P G F</a></b>; LAWRENCE, H. ; LIMEBACK, H. ; GRYNPAS, M. D. . Visual Analogue Scale for Dental Fluorosis (DF VAS). In: Reserch DAy - University of Toronto, 2003, Toronto. Abstract Book, 2003. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>74. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="446">Vieira, A P G F</a></b>; GRYNPAS, M. D. ; LIMEBACK, H. ; MAIA, R. . Fluoride, Dental Fluorosis and microhardness in human dentin.. In: IADR Meeting, 2002, San Diego - CA USA. The IADR/AADR/CADR 80th General Session, 2002. p. 2183. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>75. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="447">Vieira, A P G F</a></b>. Myth or fact? Fluoride concentration in dentin and enamel correlates with severity of dental fluorosis - A study in areas of endemic dental fluorosis.. In: Gordon Research Conference, 2002, New London - NH. Gordon Research Conference, 2002. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>76. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="448">Vieira, A P G F</a></b>; GRYNPAS, M. D. ; LIMEBACK, H. ; NOGUEIRA, R. M. . Fluoride, Dental Fluorosis and Microhardness in human Dentin. In: Research Day - University of Toronto, 2002, Toronto. Abstract Book, 2002. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>77. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="449">Vieira, A P G F</a></b>. The relaitionship between odontogenic bacteremia and otrhodontic treatment procedures. In: British Society of Pediatric Dentistry Annual Meeting, 2000, Londres. Abstract Book, 2000. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>78. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="450">Vieira, A P G F</a></b>. Projeto de prevencao da carie em criancas na idade escolar. In: VII Encontro de Extensao - UFC, 1998, Fortaleza. VII Encontro de Extensao - UFC, 1998. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>79. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="451">Vieira, A P G F</a></b>. Vigilância sanitária de fluoretação das águas de abastecimento público do Município de Fortaleza. In: 18º Congresso Internacional de Odontologia de São Paulo, 1998, São Paulo. Livro de resumos, 1998. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>80. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="452">Vieira, A P G F</a></b>. Vigilância sanitária de fluoretação das águas de abastecimento público do Município de Fortaleza. In: XIV ENATESPO - Encontro Nacional dos Administradores e Técnicos do Serviço Público Odontológico, 1998, Fortaleza. XIV ENATESPO, 1998. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>81. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="453">Vieira, A P G F</a></b>. Infiltração gengival e oclusal em resinas compostas. In: 18º Congresso Internacional de Odontologia de São Paulo, 1998, São Paulo. Livro de Resumos, 1998. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>82. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="454">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7884498527294072" target="_blank" tabindex="455">LOSSIO, J. J. D.</a> . Influencia da intensidade de luz e do calor na dureza Knoop das resinas compostas. In: XXXIII Encontro Grupo Brasileiro Materiais Dentarios, 1997, Piracicaba. XXXIII Encontro Grupo Brasileiro Materiais Dentarios Anais, 1997. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>83. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="456">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7884498527294072" target="_blank" tabindex="457">LOSSIO, J. J. D.</a> . O dentista nao fotopolimeriza corretamente. In: XXXIII Encontro Grupo Brasileiro de Materiais Dentarios, 1997, Piracicaba. XXXIII Encontro Grupo Brasileiro de Materiais Dentarios, 1997. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>84. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="458">Vieira, A P G F</a></b>. Sistemática de Vigilância Sanitária no Sistema de Abastecimento d´àgua no Município de Fortaleza. In: XII Congresso Cearense de Odontologia, 1997, Fortaleza. XII Congresso Cearense de Odontologia, 1997. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>85. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="459">Vieira, A P G F</a></b>. Avaliação da Fotopolimerização de Resinas Compostas feita pelo Dentista. In: XII Congresso Cearense de Odontologia, 1997, Fortaleza. XII Congresso Cearense de Odontologia, 1997. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>86. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="460">Vieira, A P G F</a></b>; SILVA, A. M. N. ; SAMPAIO, C. F. . Anatomia topografica das hemorragias faciais. In: XVII Congresso Brasileiro de Anatomia, 1996, Fortaleza. XVII Congresso Brasileiro de Anatomia Anais, 1996. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>87. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="461">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7884498527294072" target="_blank" tabindex="462">LOSSIO, J. J. D.</a> . Avaliacao da fotopolimerizacao feita pelo dentista. In: XV Encontro Universitario de Iniciacao a Pesquisa, 1996, Fortaleza. XV Encontro Universitario de Iniciacao a Pesquisa Anais. Fortaleza: Grafica da Universidade Federal do Ceara. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>88. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="463">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7884498527294072" target="_blank" tabindex="464">LOSSIO, J. J. D.</a> . Importancia da intensidade de luz e do calor na dureza Knoop das resinas compostas. In: XV Encontro Universitario de Iniciacao a Pesquisa, 1996, Fortaleza. XV Encontro Universitario de Iniciacao a Pesquisa Anais. Fortaleza: Grafica da Universidade Federal do Ceara. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>89. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="465">Vieira, A P G F</a></b>. Sistema de Vigilancia na Fluoretacao das aguas do municipio de Fortaleza. In: XV Encontro Universitario de Iniciacao a Pesquisa, 1996, Fortaleza. XV Encontro Universitario de Iniciacao a Pesquisa Anais, 1996. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>90. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="466">Vieira, A P G F</a></b>. Problemas na Fotopolimerização. In: XXVIII Congresso Brasileiro de Odontologia / IX Congresso Bahiano de Odontologia, 1996, Salvador. Anais do XVIII Congresso Brasileiro de Odontologia / IX Congresso Bahiano de Odontologia, 1996. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>91. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="467">Vieira, A P G F</a></b>. Transplantes Dentários Autógenos. In: I Jornada Odontopet - UFC, 1996. I Jornada Odontopet - UFC. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>92. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="468">Vieira, A P G F</a></b>. Anatomia topografica das hemorragias faciais. In: XXVII Congresso Brasileiro de Odontologia, 1995, Natal. Anais do XXVII Congresso Brasileiro de Odontologia, 1995. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b>Resumos publicados em anais de congressos (artigos)</b>
# </div>
# <br class="clear">
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Vieira, A P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2005</span><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="469">Vieira, A P G F</a></b>; HANCOCK, R. ; DIMITRIU, M. ; SCHWARTZ, M. ; LIMEBACK, H. ; GRYNPAS, M. D. . Tooth F concentration, Dental Fluorosis and Tooth quality in Humans. Journal of Dental Research<sup></sup>, v. 84, n.Spec Iss A, p. 2149, 2005. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=0022345&amp;volume=84&amp;issue=&amp;paginaInicial=2149&amp;titulo=Tooth F concentration, Dental Fluorosis and Tooth quality in Humans&amp;sequencial=1&amp;nomePeriodico=Journal of Dental Research" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Vieira, A P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2004</span><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="470">Vieira, A P G F</a></b>; HANCOCK, R. ; EGGERTSSON, H. ; EVERETT, E. T. ; GRYNPAS, M. D. . Is Tooth Quality in Fluorosis more Affected by Genetic or Environmental Factors?. Journal of Dental Research<sup></sup>, v. 83, n.sp issue A, p. 1420, 2004. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=0022345&amp;volume=83&amp;issue=&amp;paginaInicial=1420&amp;titulo=Is Tooth Quality in Fluorosis more Affected by Genetic or Environmental Factors?&amp;sequencial=2&amp;nomePeriodico=Journal of Dental Research" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Vieira, A P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2004</span><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="471">Vieira, A P G F</a></b>; HANCOCK, R. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7549914789004407" target="_blank" tabindex="472">SAMPAIO, F.</a> ; LIMEBACK, H. ; GRYNPAS, M. D. . The relationship between dentin and enamel fluoride concentration and the severity of human dental fluorosis - A study in an endemic dental fluorosis area. Caries Research<sup></sup>, Inglaterra, v. 38, n.4, p. 398, 2004. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=0008568&amp;volume=38&amp;issue=&amp;paginaInicial=398&amp;titulo=The relationship between dentin and enamel fluoride concentration and the severity of human dental fluorosis - A study in an endemic dental fluorosis area&amp;sequencial=3&amp;nomePeriodico=Caries Research" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Vieira, A P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2003</span><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="473">Vieira, A P G F</a></b>; LAWRENCE, H. ; LIMEBACK, H. ; GRYNPAS, M. D. . Visual Analogue Scale for Dental Fluorosis. Journal of Dental Research<sup></sup>, Estados Unidos, v. 82, n.Spec issue, p. 1360, 2003. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=0022345&amp;volume=82&amp;issue=&amp;paginaInicial=1360&amp;titulo=Visual Analogue Scale for Dental Fluorosis&amp;sequencial=4&amp;nomePeriodico=Journal of Dental Research" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Vieira, A P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2003</span><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="474">Vieira, A P G F</a></b>; HANCOCK, R. ; LIMEBACK, H. ; GRYNPAS, M. D. . Dentin vs Enamel as biomarkers for F induced changes in teeth.. Journal of Dental Research<sup></sup>, Estados Unidos, v. 82, n.Spec issue, p. 2517, 2003. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=0022345&amp;volume=82&amp;issue=&amp;paginaInicial=2517&amp;titulo=Dentin vs Enamel as biomarkers for F induced changes in teeth.&amp;sequencial=5&amp;nomePeriodico=Journal of Dental Research" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Vieira, A P G F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2002</span><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="475">Vieira, A P G F</a></b>; GRYNPAS, M. D. ; LIMEBACK, H. ; NOGUEIRA, R. M. . Fluoride, Dental Fluorosis and microhardness in human dentin. Journal of Dental Research<sup></sup>, Estados Unidos, v. 81, n.spec issue, p. 2183, 2002. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=0022345&amp;volume=81&amp;issue=&amp;paginaInicial=2183&amp;titulo=Fluoride, Dental Fluorosis and microhardness in human dentin&amp;sequencial=6&amp;nomePeriodico=Journal of Dental Research" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="cita-artigos">
# <b><a name="ArtigosAceitos" tabindex="476"></a>Artigos  aceitos para publicação</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">VIEIRA-MEYER, A. P. G. F. ; MORAIS, A. P. P. ; Campelo, ILB ; GUIMARAES, J. M. X. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/3162595976710799" target="_blank" tabindex="477">VIEIRA, N. F. C.</a> ; MACHADO, MARIA DE FÁTIMA ANTERO SOUSA ; Nogueira, P S F ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0636763091396917" target="_blank" tabindex="478">NUTO, S. A. S.</a> ; FREITAS, R. W. J. F. ; <b>Vieira, A P</b> . Infraestrutura e processo de trabalho na atenção primária à saúde: PMAQ no Ceará. REVISTA DE SAUDE PUBLICA<sup><img id="00348910_1" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="00348910" original-title="Revista De Saude Publica (0034-8910)<br />Fator de impacto (JCR 2022): 2.8"></sup>, 2020. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">SAINTRAIN, S. V. L. ; MEYER, A. P. G. F. V. ; Vieira-MEYER, A. P. G. F. ; <b>VIEIRA-MEYER, ANYA PIMENTEL GOMES FERNANDES</b> . Older Adults? Dependence in Activities of Daily Living: implications for oral health. PUBLIC HEALTH NURSING<sup><img id="15251446_2" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="15251446" original-title="Public Health Nursing (1525-1446)<br />Fator de impacto (JCR 2022): 2.1"></sup>, 2018. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">ROLIM, L. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0636763091396917" target="_blank" tabindex="479">NUTO, S. A. S.</a> ; ARAUJO, M. F. M. ; <b>Anya Pimentel Gomes Fernandes Vieira-Meyer</b> ; FREITAS, R. W. J. F. . Avaliação dos atributos da atenção primária à saúde de Fortaleza-CE. REBEN - REVISTA BRASILEIRA DE ENFERMAGEM<sup><img id="00347167_3" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="00347167" original-title="Revista Brasileira De Enfermagem (0034-7167)<br />Fator de impacto (JCR 2022): 1.3"></sup>, 2018. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">LAWRENCE, H. ; SOUSA, L. P. ; GONCALVES, F. L. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="480">SAINTRAIN, M. V.</a> ; <b>Vieira, A P G F</b> . ACESSO DO PACIENTE ESPECIAL À SAÚDE BUCAL NO SISTEMA ÚNICO DE SAÚDE: A VISÃO DO CIRURGIÃO-DESNTISTA. Revista Brasileira em Promoção da Saúde (Online)<sup></sup>, 2015. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="481">Vieira, A P G F</a></b>; MAIA, L. A. . ATENÇÃO À SAÚDE BUCAL DAS PESSOAS QUE VIVEM COM HIV/AIDS NA PERSPECTIVA DOS CIRURGIÕES-DENTISTAS. Saude em Debate<sup></sup>, 2015. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">ROCHA, L. L. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="482">SAINTRAIN, M. V. L.</a> ; <b>Vieira, A P G F</b> . Access to Dental Public Services by Disabled Persons. BMC Oral Health (Online)<sup><img id="14726831_6" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="14726831" original-title="Bmc Oral Health (1472-6831)<br />Fator de impacto (JCR 2022): 2.9"></sup>, 2015. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="483">SAINTRAIN, M. V. L.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5516641709622899" target="_blank" tabindex="484">BIZERRIL, D. O.</a> ; <b>Vieira, A P G F</b> . Oral health of institutionalizad and non-institutionalized Elders. Sci-Afric Journal of Scientific Issues, Research and Essays<sup></sup>, 2014. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="485">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="486">SAINTRAIN, M. V. L.</a> . Reflexões sobre a saúde bucal no Brasil. Revista Brasileira em Promoção da Saúde (Online)<sup></sup>, 2014. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="487">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="488">Saintrain, Maria Vieira de Lima</a> . COMMUNITY GROUP CENTERS FOR HEALTH EDUCATION: Lessons from Northeastern Brazil. Education for Health (Abingdon. Print)<sup><img id="13576283_9" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="13576283" original-title="Education For Health (1357-6283)<br />Fator de impacto (JCR 2022): 0.7"></sup>, 2013. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b><a name="ApresentacoesTrabalho" tabindex="489"></a>Apresentações de Trabalho</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Fechine, I. ; SANTOS, H. P. G. ; <b>Anya P G F Vieira-Meyer</b> . Percepcoes das gestantes beneficiárias do programa bolsa família sobre as condicionalidades de saúde e suas repercussões na Regional II. 2019. (Apresentação de Trabalho/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Dantas, G.M.N. ; Faco, S.L. ; C, N. ; VIANA, T. S. A. ; <b>Anya P G F Vieira-Meyer</b> ; DAMASCENO, J. X. . A importancia do Cirurgiao-dentista na equipe multidisciplinar no acompanhamento de pacientes infantis com paralisia cerebral. 2019. (Apresentação de Trabalho/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Borges, M.M.F. ; Marinho, I.M.A. ; VIANA, T. S. A. ; DAMASCENO, J. X. ; Carvalho, I.F. ; <b>Anya P G F Vieira-Meyer</b> . Sindrome do Zika Vírus congênita e sua Repercussão na Odontologia. 2019. (Apresentação de Trabalho/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Rodrigues, D.F.S. ; C, N. ; Borges, M.M.F. ; VIANA, T. S. A. ; <b>Anya P G F Vieira-Meyer</b> ; DAMASCENO, J. X. . NOMA: Aspectos Clinicos e perspectiva odontológica sob uma revisão da literatura. 2019. (Apresentação de Trabalho/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Faco, S.L. ; Dantas, G.M.N. ; C, N. ; VIANA, T. S. A. ; <b>Anya P G F Vieira-Meyer</b> ; DAMASCENO, J. X. . Anemia Falciforme: Aspectos clinicos e tratamento odontológico - revisão de literatura. 2019. (Apresentação de Trabalho/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">SOOUSA, A. S. A. ; Matos Carlos, A.C. ; Marinho, I.M.A. ; DAMASCENO, J. X. ; VIANA, T. S. A. ; <b>Vieira-Meyer, Anya PGF</b> . Uso da anestesia geral como opção para o tratamento odontológico. 2019. (Apresentação de Trabalho/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Araújo, A.R.F. ; Marinho, I.M.A. ; <b>Anya P G F Vieira-Meyer</b> ; DAMASCENO, J. X. ; VIANA, T. S. A. . Manifestações Orais em crianças infectadas pelo virus da imunodeficiência Humana em Terapia Antirretroviral ativa. 2019. (Apresentação de Trabalho/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Marinho, I.M.A. ; SOOUSA, A. S. A. ; Araújo, A.R.F. ; <b>Anya P G F Vieira-Meyer</b> ; VIANA, T. S. A. ; DAMASCENO, J. X. . Cuidados no atendimento odontológico de pacientes com sindrome de down. 2019. (Apresentação de Trabalho/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">SOOUSA, A. S. A. ; Matos Carlos, A.C. ; Marinho, I.M.A. ; DAMASCENO, J. X. ; <b>Anya P G F Vieira-Meyer</b> ; VIANA, T. S. A. . Atendimento odontológico à pacientes com transplante renal: da hemodiálise ao pós-transplante. 2019. (Apresentação de Trabalho/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>10. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">C, N. ; Rodrigues, D.F.S. ; Dantas, G.M.N. ; VIANA, T. S. A. ; DAMASCENO, J. X. ; <b>Anya P. G. F. Vieira-Meyer</b> . Deficiência de G6PD e sua implicância na odontologia. 2019. (Apresentação de Trabalho/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>11. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Matos Carlos, A.C. ; <b>Anya P G F Vieira-Meyer</b> ; DAMASCENO, J. X. ; Borges, M.M.F. ; SOOUSA, A. S. A. ; VIANA, T. S. A. . Manejo odontológico do paciente valvulopata sob uso de anticoagulantes-revisão de literatura. 2019. (Apresentação de Trabalho/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>12. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="490">VIEIRA-MEYER, ANYA P G F</a></b>; PINHEIRO, T. X. A. ; UCHOA, S. A. C. ; MACHADO, M. F. A. S. ; SAMPAIO, A. T. L. ; ROCHA, P. M. . A National Programme for Improving Access and Quality of a multiprofessional Primary Care system. 2018. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>13. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="491">Vieira, A P G F</a></b>. The implications of the Brazilian National Programme for improving Access and Quality of Primary Care. 2017. (Apresentação de Trabalho/Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>14. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">MAGALHAES, I. A. ; MALTA, C. E. N. ; <b>Vieira, A P G F</b> ; DAMASCENO, J. X. ; SOUSA, F. B. ; VIANA, T. S. A. . SÍNDROME DE PRADER-WILLI - RELATO DE CASO. 2017. (Apresentação de Trabalho/Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>15. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="492">Vieira, A P G F</a></b>; ROCHA, L. V. ; DAMASCENO, J. X. ; MALTA, C. E. N. . O exercício do cirurgião-dentista no atendimento a pacientes com transtorno do espectro do autismo. 2017. (Apresentação de Trabalho/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>16. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">SOUSA, L. P. ; VIEIRA-MEYER, A. P. G. F. ; Lima, M C F ; Neponuceno, R C A ; Teixeira, D M ; <b>VIEIRA-MEYER, ANYA P G F</b> . Oral Health Care to Schizophrenic Patients by Primary Health Care. 2017. (Apresentação de Trabalho/Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>17. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="493">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="494">SAINTRAIN, M. V. L.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8494771655868048" target="_blank" tabindex="495">SAINTRAIN, S. V.</a> ; PEQUENO, L. L. ; PINHEIRO, C. P. O. . Atividade da vida diária e condição de saúde bucal de idosos. 2016. (Apresentação de Trabalho/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>18. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Lourenço, C ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="496">SAINTRAIN, M. V. L.</a> ; <b>Vieira, A P G F</b> . Child, Negligence and Oral Health. 2012. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>19. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="497">Vieira, A P G F</a></b>; Lourenço, C . Is dental decay related to child neglect?. 2011. (Apresentação de Trabalho/Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>20. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="498">Vieira, A P G F</a></b>. Puericultura integral. 2009. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>21. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="499">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="500">SAINTRAIN, M. V. L.</a> . The Oral Health Care System in Brazil. 2009. (Apresentação de Trabalho/Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>22. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="501">Vieira, A P G F</a></b>. Múltiplas abordagens no atendimento ao paciente pediátrico (Projeto Crescendo sem Cárie). 2008. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>23. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="502">Vieira, A P G F</a></b>. Sedação em Odontopediatria (Projeto Crescendo sem Cáries). 2008. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>24. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Ellery, AE ; Andrade, LOM ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="503">Vieira, A P G F</a></b> ; NUTO, S. . O sistema de formação em saúde no município de Fortaleza. 2007. (Apresentação de Trabalho/Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>25. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">VIEIRA, G. C. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="504">Vieira, A P G F</a></b> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/3162595976710799" target="_blank" tabindex="505">VIEIRA, N. F. C.</a> . Percepção dos farmacêuticos sobre a inserção do farmacêutico no PSF. 2007. (Apresentação de Trabalho/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>26. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="506">SAINTRAIN, M. V. L.</a> ; COSTA, E. H. M. ; <b>Vieira, A P G F</b> . Autopercepção da Condição de Saúde Bucal em Idosos Institucionalizados e Não nstitucionalizados. 2007. (Apresentação de Trabalho/Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>27. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="507">Vieira, A P G F</a></b>. Óxido Nitroso em Odontopediatria. 2007. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>28. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="508">Vieira, A P G F</a></b>. O uso do óxido nitroso na odontologia. 2007. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>29. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="509">Vieira, A P G F</a></b>. Projeto Sorriso Grisalho. 2007. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>30. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5516641709622899" target="_blank" tabindex="510">BIZERRIL, D. O.</a> ; SAMPAIO, E. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="511">Vieira, A P G F</a></b> . Saude bucal em pessoas na terceira idade institucionalizadas. 2007. (Apresentação de Trabalho/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>31. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="512">Vieira, A P G F</a></b>. Manejo do paciente com procedimento não farmacológico (técnicas de hipnotismo) e manejo do paciente com procedimento farmacológico (anestesia geral, óxido nitroso sedação oral). 2005. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>32. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="513">Vieira, A P G F</a></b>. Promoção de Saúde -  Vigilância sanitária da fluoretação das água de abastecimento público no Município de Fortaleza. 1997. (Apresentação de Trabalho/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>33. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="514">Vieira, A P G F</a></b>. Fórum Acadêmico: Inter-relação periodontia/dentística. 1997. (Apresentação de Trabalho/Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>34. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="515">Vieira, A P G F</a></b>. Fórum Acadêmico Anatomia Topográfica das Hemorragias Faciais. 1996. (Apresentação de Trabalho/Comunicação). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <a name="ProducaoTecnica" tabindex="516"></a><br class="clear"><div class="inst_back">
# <b>Produção técnica</b>
# </div>
# <div class="cita-artigos">
# <b><a name="AssessoriaConsultoria" tabindex="517"></a>Assessoria e consultoria</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="518">Vieira, A P G F</a></b>. Consultora Técnica. 2005. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b><a name="SoftwareSemPatente" tabindex="519"></a>Programas de computador sem registro</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="520">Vieira-Meyer, Anya PGF</a></b>. Olá Dentista!. 2019. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b><a name="ProdutosTecnologicos" tabindex="521"></a>Produtos tecnológicos</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="522">PIMENTEL G. F. VIEIRA-MEYER, ANYA</a></b>; Lima, R A ; Rolim, J P M L ; SOUSA, F. B. ; Aquino, P.B. . Revisão Protocolo de Atendimento para puericultua odontológica. 2020. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="523">Vieira-Meyer, Anya PGF</a></b>; Braga, J U ; Castro, D B . Curso EAD - Técnicas de Análise Espacial Aplicadas à Vigilância em Saúde na Atenção Básica - 1º Oferta. 2019. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b><a name="TrabalhosTecnicos" tabindex="524"></a>Trabalhos técnicos</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">SILVA, A. A. ; NORO, L. ; GOULART, B. ; CONDE, W. ; <b>Anya P G F Vieira-Meyer</b> . Membro da Comissão de Avaliação de Teses - Prêmio Capes de Teses na área de Saúde Coletiva. 2020. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="525">VIEIRA-MEYER, ANYA P G F</a></b>. Assessora Sociotécnica de projetos no ambito da chamada publica para Ações Emergenciais de Enfrentamento à COVID-19 junto a populações vulneráveis. 2020. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="526">PIMENTEL G. F. VIEIRA-MEYER, ANYA</a></b>; RAFAEL, R. M. R. ; PEIXOTO, S. W. V. . Coordenação do Fórum dos Coordenadores da Saúde Coletiva. 2019. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="527">Anya Pimentel Gomes Fernandes Vieira-Meyer</a></b>. Participou da Reunião para análise de APCN - Saúde Coletiva. 2018. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="528">Vieira, A P G F</a></b>. Comissão de Avaliação Quadrienal da Área de Saúde Coletiva. 2017. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="529">Anya Pimentel Gomes Fernandes Vieira-Meyer</a></b>. Participou da Reunião para análise de APCN - Saúde Coletiva. 2016. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="530">Vieira, A P</a></b>. Parecerista 6 projetos edital do programa de Pesquisa para o SUS: gestão comparilhada em saúde - PPSUS/UFPI. 2013. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="531">Anya Vieira</a></b>. Edital do Programa Pesquisa para o SUS: gestão compartilhada em saúde na UF do PIAUÍ - PI - 57 projetos avaliados. 2013. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="532">Anya Vieira</a></b>. Reunião da Comissão de Especialistas do Edital do Programa Pesquisa para o SUS -Gestão Compartilhada em Saúde na UF do: PIAUÍ - PI. 2013. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>10. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="533">Vieira, A P</a></b>. Parecerista 3 projetos edital do Programa de Pesquisa para o SUS: gestão compartilhada em saúde - PPSUS/UFMA. 2013. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>11. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Puricelli, E ; <b>Vieira, A P G F</b> . Coordenadora da Conferencia Nacional Tumores em Pacientes Pediátrico. 2009. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>12. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">DIAS, E. ; <b>Vieira, A P G F</b> . Coordenação de Conferência Nacional Traumatismo Maxilofacial Pediátrico. 2009. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>13. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="534">Vieira, A P G F</a></b>. Avaliadora Ad Hoc de trabalhos para Decit/ MS e Secretaria Estadual de Ciência e Tecnologia do Estado do Tocantins. 2009. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>14. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="535">Vieira, A P G F</a></b>. Avaliadora dos painéis do XV Encontro de Iniciação à Pesquisa da UNIFOR. 2009. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>15. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="536">Vieira, A P G F</a></b>. Parecerista ad hoc da 1ª fase de avaliação do Prêmio de Incentivo em Ciência e Tecnologia para o SUS 2009. 2009. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>16. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="537">Vieira, A P G F</a></b>. Parecerista Ad hoc da primeira fase de avaliação do Prêmio de Incentivo em Ciência e Tecnologia para o SUS 2009. 2009. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>17. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="538">Vieira, A P G F</a></b>. Comissão de avaliação de trabalhos do XV Encontro de Iniciação à Pesquisa da UNIFOR. 2009. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>18. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="539">Vieira, A P G F</a></b>. Comissão de avaliação de trabalhos do IX Encontro de Pós-Graduação e Pesquisa da UNIFOR. 2009. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>19. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="540">Vieira, A P G F</a></b>. Membro do Comitê Editorial Ad Hoc da Revista Brasileira em Promoção da Saúde. 2009. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>20. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="541">Vieira, A P G F</a></b>. Revisora de trabalhos do XVIII Congresso Mundial de Epidemiologia/ABRASCO. 2008. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>21. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="542">Vieira, A P G F</a></b>. Avaliadora da Sessão de Painéis no VIII Encontro de Pós-Graduação e Pesquisa da UNIFOR. 2008. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>22. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="543">Vieira, A P G F</a></b>. Professora do Curso Preparatório - Concurso Público PSF. 2006. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>23. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="544">Vieira, A P G F</a></b>. Coordenadora do Seminário Internacional de Promoção da Saúde. 2006. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>24. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="545">Vieira, A P G F</a></b>. Membro do Comitê Editorial Ad Hoc. 2006. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>25. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="546">Vieira, A P G F</a></b>. Professora do Curso de Aperfeiçoamento e Atualização em Saúde Família. 2006. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>26. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="547">Vieira, A P G F</a></b>. Professora da Jornada CRO/ABO-RO. 2005. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>27. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="548">Vieira, A P G F</a></b>. Professora no Curso de Especialização em Odontopediatria. 2005. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>28. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="549">Vieira, A P G F</a></b>. Professora do Curso Preparatório - Concurso Público PSF. 2005. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>29. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="550">Vieira, A P G F</a></b>. Coordenadora da Comissão de Recebimento. 1999. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>30. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="551">Vieira, A P G F</a></b>. Oradora da Turma. 1998. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>31. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="552">Vieira, A P G F</a></b>. Coordenadora da Comissão Acadêmica. 1997. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>32. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="553">Vieira, A P G F</a></b>. Membro da Comissão Acadêmica. 1995. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b><a name="EntrevistasMesasRedondas" tabindex="554"></a>Entrevistas, mesas redondas, programas e comentários na mídia</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="555">Vieira-MEYER, A. P. G. F.</a>; TASCHNER, N. P. ; STUENKEL, O. ; ARSENAULT, C. ; <b>Vieira-Meyer, Anya PGF</b> . Brazil has the most confirmed COVID-19 cases in Latin America. 'So what?' says President Bolsonaro. 2020. 
						
# 							(Programa de rádio ou TV/Entrevista).
# 						<img class="icone-video" src="http://estatico.cnpq.br/imagens/ico_pw_video.png" onclick="mostrarPreviaVideo('https://www.cbc.ca/news/world/brazil-has-the-most-confirmed-covid-19-cases-in-latin-america-so-what-says-president-bolsonaro-1.5550720')"></span>
# </div>
# <br class="clear">
# <br class="clear">
# <a name="DemaisProducaoTecnica" tabindex="556"></a><br class="clear"><div class="inst_back">
# <b>Demais tipos de produção técnica</b>
# </div>
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="557">PIMENTEL G. F. VIEIRA-MEYER, ANYA</a></b>. Abordagens Multidisciplinares em traumas dentofaciais. 2020. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="558">PIMENTEL G. F. VIEIRA-MEYER, ANYA</a></b>; Braga, J U ; Castro, D B . Curso EAD - Técnicas de Análise Espacial Aplicadas à Vigilância em Saúde na Atenção Básica - 1º Oferta. 2019. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="559">Vieira-Meyer, Anya PGF</a></b>; Braga, J U ; Castro, D B . Técnicas de Análise Espacial Aplicadas à Vigilância em Saúde na Atenção Básica.
# 							2019. (Desenvolvimento de material didático ou instrucional - MaterialDidático). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="560">PIMENTEL G. F. VIEIRA-MEYER, ANYA</a></b>; Braga, J U ; Castro, D B . Técnicas de Análise Espacial Aplicadas à Vigilância em Saúde na Atenção Básica.
# 							2019. (Desenvolvimento de material didático ou instrucional - MaterialDidático). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="561">Vieira, A P G F</a></b>. Coordenadora do Módulo Atenção Integral à Saúde da FAmília. 2013. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="562">Vieira, A P G F</a></b>; GOMES, A. M. A. ; MORAIS, A. P. P. ; AMORIM, A. K. A. ; FREITAS, C. A. S. ; AQUINO, D. M. C. ; GUBERT, F. A. ; SILVA, G. S. N. ; ALBUQUERQUE, I. M. N. ; RIBEIRO, M. T. A. M. . Caderno do Módulo: Atenção e Gestão do Cuidado à Saúde da Família - MPSF/RENASF.
# 							2013. (Desenvolvimento de material didático ou instrucional - MaterialDidático). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="563">Vieira, A P G F</a></b>; SALDANHA, A. C. ; MORAIS, A. P. P. ; AGUIAR, A. S. W. ; GOMES, A. M. A. ; MEDEIROS JUNIOR, A. ; SOUZA, E. C. F. ; BRANDAO, I. R. ; GUIMARAES, J. M. X. ; DIAS, M. S. A. ; MOTA, M. V. . Caderno do Módulo: Gestão do Processo de Trabalho na Atenção Primária - MPSF/RENASF.
# 							2013. (Desenvolvimento de material didático ou instrucional - Material Didático). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="564">Vieira, A P G F</a></b>; AMORIM, A. K. A. ; MORAIS, A. P. P. ; Oliveira, E N ; SILVA, G. S. N. ; CAMPOS, J. S. ; RIBEIRO, M. T. A. M. . Caderno do Módulo: Atenção Integral à Saúde da Família - MPSF/RENASF.
# 							2013. (Desenvolvimento de material didático ou instrucional - MaterialDidático). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="565">Vieira, A P</a></b>. Caderno do Módulo: Seminário de Acompanhamento I, II, III e IV.
# 							2013. (Desenvolvimento de material didático ou instrucional - Material Didático). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>10. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="566">Vieira, A P G F</a></b>. Urgências em odontopediatria. 2012. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>11. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="567">Vieira, A P G F</a></b>. Comissão avaliadora da XII Jornada Odontopet. 2012. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>12. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="568">Vieira, A P G F</a></b>. Módulo: Gestão do Processo de Trabalho na Atenção Primária - MPSF/RENASF. 2012. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>13. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="569">Vieira, A P G F</a></b>. Módulo: Atenção e Gestão do Cuidado à Saúde da Família - MPSF/RENASF. 2012. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>14. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="570">Vieira, A P G F</a></b>. Módulo: Atenção Integral à Saúde da Família - MPSF/RENASF. 2012. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>15. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="571">Vieira, A P G F</a></b>. Coordenadora do Módulo: Atenção Integral à Saúde da Família - MPSF/RENASF. 2012. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>16. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="572">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7934153093330220" target="_blank" tabindex="573">MARQUES, P. L. P.</a> ; PEQUENO, L. L. . Módulo II - Epidemiologia, planejamento e programação em saúde bucal - Curso de Aperfeiçoamento 'gestão em saúde bucal'.
# 							2012. (Desenvolvimento de material didático ou instrucional - Elaboração de Manual). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>17. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="574">Vieira, A P</a></b>. Guia do Participante.
# 							2012. (Desenvolvimento de material didático ou instrucional - ElaboraçãodeManual). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>18. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="575">Vieira, A P</a></b>; AMORIM, A. K. A. ; MORAIS, A. P. P. ; GOMES, A. M. A. ; FREITAS, C. A. S. ; AQUINO, D. M. C. ; GUBERT, F. A. ; SILVA, G. S. N. ; ALBUQUERQUE, I. M. N. ; RIBEIRO, M. T. A. M. ; VILAR, R. L. A. . Módulo: Atenção e Gestão do Cuidado.
# 							2012. (Desenvolvimento de material didático ou instrucional - Material Didático). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>19. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">AMORIM, A. K. A. ; <b>Vieira, A P</b> ; MORAIS, A. P. P. ; GOMES, A. M. A. ; OLIVEIRA, E. N. ; SILVA, G. S. N. ; CAMPOS, J. S. ; RIBEIRO, M. T. A. M. . Módulo: Atenção Integral a Saúde da Família.
# 							2012. (Desenvolvimento de material didático ou instrucional - Material Didático). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>20. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="576">Vieira, A P</a></b>. Guia do Participante.
# 							2012. (Desenvolvimento de material didático ou instrucional - Material Didático). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>21. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="577">Vieira, A P G F</a></b>. Universal2007FaixaB-Edital MCT/CNPq 15/2007 - Universal. 2010. (Relatório de pesquisa). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>22. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="578">Vieira, A P G F</a></b>. Aula no Módulo VI - Saúde da criança e Adolescente - Tema: Diagnóstico e tratamento de trauma na dentição decídua, mista e permanente - Curso de Especialização em Práticas Clínicas em Saúde da Família. 2009. (Curso de curta duração ministrado/Especialização). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>23. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="579">Vieira, A P G F</a></b>. Módulo: Sedação e anestesia geral em odontopediatria. 2009. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>24. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">NEVES FILHO, A. C. ; LOPES, M. S. V. ; <b>Vieira, A P G F</b> . Módulo VI - Saúde da Criança e Adolescente - Odontologia 2009.
# 							2009. (Desenvolvimento de material didático ou instrucional - Elaboração de Manual). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>25. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="580">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="581">SAINTRAIN, M. V. L.</a> . Apoio a projetos de pesquisa/Edital MCT/CNPq/MS-SCTIE-DECIT 23/2006 - Estudo da Gestão em Saúde. 2009. (Relatório de pesquisa). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>26. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="582">Vieira, A P G F</a></b>. Óxido Nitroso em Odontopediatria. 2007. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>27. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="583">Vieira, A P G F</a></b>. Programa de Formação de Facilitadores de Educação Permanente em Saúde / EAD. 2007. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>28. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="584">SAINTRAIN, M. V. L.</a> ; <b>Vieira, A P G F</b> . Oficina de validação do protocolo de treinamento para agentes comunitários de saúde na utilização do índice comunitário em saúde bucal. 2007. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>29. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="585">Vieira, A P G F</a></b>. Sedação em Odontologia. 2007. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>30. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="586">Vieira, A P G F</a></b>. II Congresso Internacional de Odontologia (Módulo de avanços em odontopediatria). 2006. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>31. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="587">Anya Vieira</a></b>. Curso de Aperfeiçoamento e Atualização em Saúde da Família. 2006. . </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>32. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="588">Anya Vieira</a></b>. Óxido nitroso uma opção viável na odontologia. 2006. . </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>33. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="589">Vieira, A P G F</a></b>. Professora II Curso preparatório - Concurso público PSF. 2006. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>34. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="590">Anya Vieira</a></b>. Editora Científica da Revista Brasileira em Promoção da Saúde - RBPS. 2006. (Editoração/Periódico). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>35. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="591">Vieira, A P G F</a></b>. Professora do I Curso Preparatório - Concurso Público PSF. 2005. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="Bancas" tabindex="592">
# <h1 tabindex="0">Bancas</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <a name="ParticipacaoBancasTrabalho" tabindex="593"></a><br class="clear"><div class="inst_back">
# <b>Participação em bancas de trabalhos de conclusão</b>
# </div>
# <div class="cita-artigos">
# <b>Mestrado</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">CALVASINA, P. G.; Oliveira, E N; <b>VIEIRA-Meyer, Anya P G F</b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0636763091396917" target="_blank" tabindex="594">NUTO, S. A. S.</a>.  Participação em banca de Aline Pinto Monteiro. Violência contra a mulher por parceiro íntimo: conhecimentos e condutas do cirurgião-dentista da Estratégia Saúde da Família. 2023. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>VIEIRA-Meyer, Anya P G F</b>; PERISSE, A. R. S.; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0636763091396917" target="_blank" tabindex="595">NUTO, S. A. S.</a>; BARRETO, I. C. H. C..  Participação em banca de Angélica Ribeiro Gurgel. Percepção dos profissionais da Estratégia Saúde Da Família sobre os possíveis impactos do Complexo Industrial e Portuário do Pecém na saúde da população. 2023. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>VIEIRA-Meyer, Anya P G F</b>; Lima Neto. A.S.; ALMEIDA, J. R. S.; SOUZA, M. L. P..  Participação em banca de ERLEMUS PONTE SOARES. REPERCUSSÃO DA COVID-19 NA ATENÇÃO À SAÚDE NA ESTRATÉGIA SAÚDE DA FAMÍLIA NO MUNICÍPIO DE FORTALEZA. 2023 - Fundacao Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>VIEIRA-Meyer, Anya P G F</b>; ICHIHARA, M. Y.; GUIMARAES, J. M. X.; ARAUJO, M. F. M..  Participação em banca de Gilmara Maria Batista Tavares Silva. Qualidade do pré-natal e seu impactos na redução da sífilis congênita no município de Fortaleza. 2023. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>VIEIRA-Meyer, Anya P G F</b>; GUERRA, LIS MONTEIRO DE CARVALHO; FONTELES, C. S. R..  Participação em banca de Nelsimar Bezerra Santos da Silva. Percepção e disponibilidade ao atendimento de pacientes especiais por graduandos em odontologia no município de Fortaleza-CE. 2023. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</b>; PINHEIRO JUNIOR, E. C.; Carvalho, I.F..  Participação em banca de PAULA GUERRA DA ROCHA. PREVALÊNCIA DE LESÕES CERVICAIS NÃO CARIOSAS E RECESSÃO GENGIVAL EM SUBPOPULAÇÕES DE GUINÉ-BISSAU. 2023. Dissertação (Mestrado em Ciências Odontológicas) - Centro Universitário Christus. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">FONTELES, C. S. R.; AGUIAR, A. S. W.; <b>VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</b>.  Participação em banca de RENATA TORREÃO VIANA DE MELO COSTA. Autopercepção da qualidade de vida e sua relação com discrepâncias maxilo-mandibulares, variáveis clínicas e terapia de reposição enzimática em pacientes com mucopolissacaridose na cidade de Fortaleza, Brasil. 2023. Dissertação (Mestrado em Odontologia) - Universidade Federal do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>VIEIRA-Meyer, Anya P G F</b>; SAINTRAIN, MARIA V; BARRETO, I. C. H. C.; NUTO, S..  Participação em banca de Elaine Lopes Pedrosa Bomfim. ?Atenção as pessoas acometidas por Hanseníase: ações realizadas por médicos e Cirurgiões-Dentistas da estratégia saúde da família. 2022. Dissertação (Mestrado em Mestrado Profissional em Saúde da Família - ProfSaúde) - FIOCRUZ/CEARÁ. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">MOREIRA, T. M. M.; GOMES, I. L. V.; FLORENCIO, R. S.; <b>VIEIRA-Meyer, Anya P G F</b>.  Participação em banca de JULIANA NEVES CESAR,. LESÕES PERIAPICAIS EM  PESSOAS COM DIABETES: PREVALÊNCIA E CARACTERÍSTICAS ASSOCIADAS. 2022. Dissertação (Mestrado em Saúde Pública) - Universidade Estadual do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>10. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Almeida, M E L; <b>VIEIRA-Meyer, Anya P G F</b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0636763091396917" target="_blank" tabindex="596">NUTO, S. A. S.</a>.  Participação em banca de Diego Ramos Aguiar. Avaliação da qualidade dos serviços prestados pelo Centro de Especialidades Odontológicas e Centro de Saúde da Família: percepção dos usuários. 2022. Dissertação (Mestrado em Mestrado Profissional em Saúde da Família - ProfSaúde) - FIOCRUZ/CEARÁ. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>11. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="597">SAINTRAIN, M. V. L.</a>; SILVA JUNIOR, G. B.; <b>Vieira-Meyer, Anya PGF</b>; Bezerra, Carina Bandeira.  Participação em banca de Débora Rosana Alves Braga. COVID-19: Aspectos econômicos e emocionais na população brasielira. 2021. Dissertação (Mestrado em Saúde Coletiva) - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>12. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">CBF, Valter; MMM, Thereza; <b>PIMENTEL G. F. VIEIRA-MEYER, ANYA</b>.  Participação em banca de Rayssa Veras Camelo. Ações das equipes de saúde da família na atenção a pessoa com hipertensão e diabetes no Brasil: uma análise secundária do PMAQ-AB 3o Ciclo. 2020. Dissertação (Mestrado em Saúde Pública) - Universidade Estadual do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>13. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>PIMENTEL G. F. VIEIRA-MEYER, ANYA</b>; Maçal, F.F.; Couto, J.L.P..  Participação em banca de Maira Barbosa Coutinho. A influencia de vídeos de antecipação de situação no controle dos níveis de ansiedade em pacientes odontopediatricos. 2020. Dissertação (Mestrado em Ciências Odontológicas) - Centro Universitário Christus. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>14. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">FREITAS, R. W. J. F.; GOMES, K. W. L.; ARAUJO, M. F. M.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="598">Vieira, A P G F</a></b>.  Participação em banca de Diego Magalhães Siqueira. Efetividade da auriculoterapia na analgesia das dores crônicas causadas pela Chikungunya: Ensaio clínico controlado e randomizado. 2019. Dissertação (Mestrado em Mestrado Profissional em Saúde da Família - ProfSaúde) - FIOCRUZ/CEARÁ. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>15. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="599">Vieira, A P G F</a></b>; Benevides, B. S.; BARRETO, I. C. H. C.; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0636763091396917" target="_blank" tabindex="600">NUTO, S. A. S.</a>.  Participação em banca de Antonio Leonel de Lima Junior. A PERCEPÇÃO DOS MÉDICOS EM RELAÇÃO AO IMPACTO DAS FORMAS DE CONTRATAÇÃO DESTES NO TRABALHO DA ESTRATÉGIA DE SAÚDE DA FAMÍLIA. 2019. Dissertação (Mestrado em Mestrado Profissional em Saúde da Família - ProfSaúde) - FIOCRUZ/CEARÁ. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>16. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">BARRETO, I. C. H. C.; Benevides, B. S.; <b>Anya P. G. F. Vieira - Meyer</b>; PAULA, J. B..  Participação em banca de Roberto Ribeiro Maranhão. Análise do escopo de práticas dos médicos na estratégia saúde da família em Fortaleza - Ceará. 2019. Dissertação (Mestrado em Mestrado Profissional em Saúde da Família - ProfSaúde) - FIOCRUZ/CEARÁ. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>17. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">VIEIRA-MEYER, A. P. G. F.; UCHIMURA, K. Y.; FREITAS, R. W. J. F.; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0636763091396917" target="_blank" tabindex="601">NUTO, S. A. S.</a>; <b>Vieira-Meyer, AP</b>.  Participação em banca de Helena Paula Guerra dos Santos. Como as gestantes beneficiárias do bolsa família percebem o programa e suas repercussões na saúde e na violência. 2019. Dissertação (Mestrado em Mestrado Profissional em Saúde da Família - MPSF/RENASF) - Fundação Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>18. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">SAINTRAIN, M. V. L.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="602">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5516641709622899" target="_blank" tabindex="603">BIZERRIL, D. O.</a>.  Participação em banca de Janayne de sousa Oliveira. Avaliação do fluxo salivar em idosos e aplicabilidade de intervenção nutricional. 2018. Dissertação (Mestrado em Saúde Coletiva) - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>19. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="604">Vieira, A P G F</a></b>; GUBERT, F. A.; Gondim, APS; LUNA, I. T..  Participação em banca de Germimma de Paiva Rêgo. Educação em saúde bucal como estratégia para promoção da autoimagem e prevenção de IST/HIV em adolescentes. 2016. Dissertação (Mestrado em SAÚDE DA FAMÍLIA) - Universidade Federal do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>20. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="605">Vieira, A P G F</a></b>; SILVA, A. S. R.; MARTINS, M. C.; CASTRO, R. C. M. B..  Participação em banca de Rosolinda Rodolfo de Sá Batista. Construção e validação de um Website educativo para promoção da amamentação e prevenção de hábitos bucais deletérios. 2016. Dissertação (Mestrado em SAÚDE DA FAMÍLIA) - Universidade Federal do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>21. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="606">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="607">SAINTRAIN, M. V. L.</a>; ARAUJO, M. A. M..  Participação em banca de Luiza de Paula Sousa. Atenção em Saúde Bucal oferecida aos pacientes com esquizofrenia na Estratégia Saúde da Família em Fortaleza/CE. 2016. Dissertação (Mestrado em Saúde da Família) - Fundação Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>22. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">OSTERNE, M. S. F.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="608">Vieira, A P G F</a></b>; FROTA, F. H. S..  Participação em banca de Jamile Pinheiro Karbage Salmito. Avaliação na atenção primária: experiências de aplicação da autoavaliação para melhoria do acesso e da qualidade - AMAQ no município de Fortaleza. 2015. Dissertação (Mestrado em Sociologia) - Universidade Estadual do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>23. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">PEDROSA, J. I. S.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="609">Vieira, A P G F</a></b>; NOGUEIRA, L. T.; CARDOSO, O. O..  Participação em banca de Lígia Nara Martins Santos. Qualidade na atenção básica: analise da satisfação do usuário no Estado do Piauí. 2014. Dissertação (Mestrado em Enfermagem) - Universidade Federal do Piauí. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>24. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="610">Vieira, A P G F</a></b>; ALVES, C. M. C.; THOMAZ, E. B. A. F.; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9128715784318958" target="_blank" tabindex="611">COIMBRA, L. C.</a>.  Participação em banca de Grazianne Medeiros Carvalho de Sousa. Centro de especialidades odontológicas: avaliação da atenção secundária em Saúde Bucal no Maranhão. 2013. Dissertação (Mestrado em Saúde Coletiva) - Universidade Federal do Maranhão. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>25. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="612">Vieira, A P G F</a></b>; Silva, C.A.B.; Araújo, M A L; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="613">SAINTRAIN, M. V. L.</a>.  Participação em banca de André Ricardo Bindá de Borba. Avaliação do serviço de atendimento domiciliar do município de Maracanaú-CE. 2013. Dissertação (Mestrado em Saúde Coletiva) - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>26. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="614">SAINTRAIN, M. V. L.</a>; LIBORIO, A. B.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="615">Vieira, A P G F</a></b>; Santos, Z.M.S.A..  Participação em banca de Paulo Leonardo Pontes Marques. Condições de Saúde Bucal de Pacientes em Hemodiálise. 2013. Dissertação (Mestrado em Mestrado em Saúde Coletiva) - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>27. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="616">SAINTRAIN, M. V. L.</a>; Cavalcante, T F; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="617">Vieira, A P G F</a></b>; Araújo, M A L.  Participação em banca de Aíla Maria da Silva Bezerra. Trombólise em Pacientes com Acidente Vascular Cerebral Isquêmico Agudo: aspectos epidemiológicos e funcionais na admissão e alta hospitalar. 2012. Dissertação (Mestrado em Mestrado em Saúde Coletiva) - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>28. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="618">Vieira, A P G F</a></b>; Gonçalves, M. L. C.; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="619">SAINTRAIN, M. V. L.</a>; Vieira, L J E S.  Participação em banca de Caroline Barbosa Lourenço. Crianças, negligência e saúde bucal. 2011. Dissertação (Mestrado em Mestrado em Saúde Coletiva) - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>29. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="620">Vieira, A P G F</a></b>; Almeida, M E L; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0636763091396917" target="_blank" tabindex="621">NUTO, S. A. S.</a>.  Participação em banca de Caroline Antero Machado. Atuação dos Cirurgiões-dentistas na Estratégia Saúde da Família. 2010. Dissertação (Mestrado em Mestrado em Saúde Coletiva) - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>30. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">OLIVEIRA, L.C.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="622">Vieira, A P G F</a></b>; VASCONCELOS FILHO, E. V.; ALMEIDA, M. I..  Participação em banca de Poliana Miranda Pinheiro. A prática do cirurgião dentista, no Programa Saúde da Família, no sentido da humanização da atenção à saúde bucal. 2009. Dissertação (Mestrado em Saúde Pública) - Universidade Estadual do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>31. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="623">Vieira, A P G F</a></b>; MONTALVERNE, D. G. B.; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7492450432942397" target="_blank" tabindex="624">Montenegro Jr, R.M.</a>; Silva, C.A.B..  Participação em banca de José Bezerra Goés Neto. Prevalência de dor na coluna e alterações de postura em estudantes de Odontologia do Estado do Ceará. 2009. Dissertação (Mestrado em Mestrado em Saúde Coletiva) - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>32. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">FONTELES, C. S. R.; CARVALHO, K. M.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="625">Vieira, A P G F</a></b>.  Participação em banca de Thyciana Rodrigues Ribeiro. Estudo do perfil de peptídios salivares de crianças com cárie da primeira infância. Uma visão da saliva como meio diagnóstico. 2009. Dissertação (Mestrado em Programa de Pós-Graduação em Odontologia) - Universidade Federal do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>33. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="626">Vieira, A P G F</a></b>.  Participação em banca de Maria Elisa Machado Ferreira Marcelo. Avaliação da Efetividade de um programa de atenção em saúde bucal com maes e bebes no serviço público. 2009. Dissertação (Mestrado em Mestrado em Saúde Coletiva) - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>34. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="627">Vieira, A P G F</a></b>.  Participação em banca de Paula Ventura da Silveira. Serviço odontológico em empresas privadas no Estado do Ceará. 2009. Dissertação (Mestrado em Mestrado em Saúde Coletiva) - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>35. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="628">Vieira, A P G F</a></b>; MENEZES, L. M. B.; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5517143911175495" target="_blank" tabindex="629">REGO, D. M.</a>; DIAS, A. A..  Participação em banca de Carlos Santos de Castro Filho. Saúde Bucal e qualidade de vida de trabalhadores com e sem assistência odontológica. 2008. Dissertação (Mestrado em Mestrado em Saúde Coletiva) - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>36. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="630">SAINTRAIN, M. V. L.</a>; Gonçalves, M. L. C.; DIAS, A. A.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="631">Vieira, A P G F</a></b>.  Participação em banca de Ana Karine Lima Moreira Mesquita. Condição epidemiológica de pacientes idosos portadores de Hipertensão e Diabetes. 2008. Dissertação (Mestrado em Mestrado em Saúde Coletiva) - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>37. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">MOREIRA NETO, J. J. S.; TELES, G. S.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="632">Vieira, A P G F</a></b>.  Participação em banca de Denise Lins de Sousa. Efeito antibacteriano do preparo biomecânico e de uma pasta à base de hidróxido de cálcio sobre bacterias presentes em canais radiculares de dentes decíduos necrosados após trauma. 2008. Dissertação (Mestrado em Programa de Pós-Graduação em Odontologia) - Universidade Federal do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>38. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">MOREIRA NETO, J. J. S.; TELES, G. S.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="633">Vieira, A P G F</a></b>.  Participação em banca de Rebeca Bastos Rocha Araújo. Conhecimento dos cirurgiões-dentistas e dos profissionais de emergências médicas sobre a avulsão dentária no município de Fortaleza, Ceará. 2008. Dissertação (Mestrado em Programa de Pós-Graduação em Odontologia) - Universidade Federal do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>39. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="634">Vieira, A P G F</a></b>; MIRANDA, A. S.; MONTENEGRO JUNIOR, R. M..  Participação em banca de Marco Túlio Aguiar Mourão Ribeiro. Avaliação da implantação de Programa de Residência em Medicina de Família e Comunidade de larga escala em capital do Nordeste.. 2008. Dissertação (Mestrado em Programa de Pós-Graduação em Saúde Coletiva) - Universidade Federal do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>40. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">MOREIRA NETO, J. J. S.; FONTELES, C. S. R.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="635">Vieira, A P G F</a></b>.  Participação em banca de Rosana Sales Dias. A frequência e o perfil epidemiológico do traumatismo dentário em escolares da Rede Pública Municipal de Fortaleza - Ceará. 2008. Dissertação (Mestrado em Programa de Pós-Graduação em Odontologia) - Universidade Federal do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>41. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="636">Vieira, A P G F</a></b>; Gonçalves, M. L. C.; VALE, O. C..  Participação em banca de Fabíola Monteiro de Castro. Avaliação das características baropodométricas de pacientes com diabetes mellitus tipo 2. 2007. Dissertação (Mestrado em Saúde coletiva) - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>42. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="637">Vieira, A P G F</a></b>.  Participação em banca de Regina Flávia de Castro Almeida. Avaliação da informação oral e escrita no conhecimento dos portadores de HIV/AIDS sobre medicamentos anti-retrovirais. 2007. Dissertação (Mestrado em Saúde coletiva) - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>43. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="638">Vieira, A P G F</a></b>.  Participação em banca de Carlos Clayton Tôrres Aguiar. Análise da qualidade de vida, depressão e saúde oral em pacientes diabéticos idosos. 2007. Dissertação (Mestrado em Saúde coletiva) - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>44. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="639">Vieira, A P G F</a></b>; Montenegro Jr.,R.M.; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4167422383882509" target="_blank" tabindex="640">Carvalho, A. F.</a>; Lima, M.C.O..  Participação em banca de Sâmia Mustafa Aguiar. Prevalência de sintomas de estresse e depressão em acadêmicos de Medicina e Odontologia. 2007. Dissertação (Mestrado em Saúde coletiva) - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>45. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">MENEZES, L. M. B.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="641">Vieira, A P G F</a></b>; Almeida, M E L.  Participação em banca de Cláudia Maria Frota Lima Botto. Análise do teor de flúor nas águas de abastecimento público do Ceará. 2007. Dissertação (Mestrado em Odontologia) - Universidade Federal do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>46. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="642">Vieira, A P G F</a></b>; Vieira, L J E S; Santos, Z.M.S.A.; FROTA, M. A..  Participação em banca de José Bernardo Rodrigues Ferreira. Programa educativo para prevención de afecciones por esfuerzos repetitivos em trabajadores del Frigorífico Filial 52-Ciudad de Onda Verde-SP/Brasil. 2006. Dissertação (Mestrado em Mestrado em Saúde Coletiva) - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>47. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Vieira, L J E S; FROTA, M. A.; Santos, Z.M.S.A.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="643">Vieira, A P G F</a></b>.  Participação em banca de Augustus Cézar Polimeno. Elaboración de uma guía de consulta rápida para el diagnóstico imagenológico de la neurofibromatose. 2006. Dissertação (Mestrado em Mestrado em Saúde Coletiva) - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b>Teses de doutorado</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">TOMASI, E.; DURO, S. M. S.; SANTOS, I. S.; SOARES, D. C.; <b>VIEIRA-Meyer, Anya P G F</b>.  Participação em banca de Maria del Pilar Flores Quispe. Qualidade da Atenção à Saúde da Criança na primeira semana de vida na Atenção Básica no Brasil: Programa Nacional de Melhoria do Acesso e da Qualidade da  Atenção Básica. 2022. Tese (Doutorado em Epidemiologia)  - Universidade Federal de Pelotas. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">RIBEIRO, T. R.; SILVA, P. G. B.; Couto, J.L.P.; <b>VIEIRA-MEYER, ANYA PIMENTEL GOMES FERNANDES</b>; DAMASCENO, J. X..  Participação em banca de Renata Asfor Rocha Carvalho Martins. Avaliação de parâmetros salivares, sorológicos, alimentares e socioeconômicos em indivíduos com e sem diabetes mellitus tipo 1. 2021. Tese (Doutorado em Odontologia)  - Universidade Federal do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">SAINTRAIN, MARIA V; Almeida, Rosa Lívia Freitas de; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8577652222943813" target="_blank" tabindex="644">Silva, R M</a>; <b>Anya, Pimentel Gomes Fernandes Vieira-Meyer</b>; BARRETO, I. C. H. C.; ARAGAO, J. A.; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5516641709622899" target="_blank" tabindex="645">BIZERRIL, D. O.</a>; SILVA JUNIOR, G. B..  Participação em banca de Caroline Barbosa Lourenço. Capacidade funcional em idosos: uma análise espacial em um município do Nordeste brasileiro. 2021. Tese (Doutorado em Programa de Pós-Graduação em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">DAMASCENO, M. M. C.; OLIVEIRA, A. P.; HISSA, M. N.; DANTAS, R. A. S.; VIEIRA-MEYER, A. P. G. F.; <b>Anya P G F Vieira-Meyer</b>.  Participação em banca de JOSÉ CLÁUDIO GARCIA LIRA NETO. ANÁLISE DA EFICÁCIA DA CANELA (CINNAMOMUM VERUM) NA REDUÇÃO DOS NÍVEIS GLICÊMICOS DE PESSOAS COM DIABETES MELLITUS TIPO 2: ENSAIO CLÍNICO ALEATORIZADO. 2020. Tese (Doutorado em Enfermagem)  - Universidade Federal do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="646">SAINTRAIN, MARIA V. DE L.</a>; FARIAS, M. R.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="647">Vieira, A P G F</a></b>; Silva, C.A.B.; Antunes, M F R.  Participação em banca de Lucianna Leite Pequeno. Saude bucal e estado nutricional em idosos. 2019. Tese (Doutorado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="648">SAINTRAIN, M. V. L.</a>; Coelho Filho, J M; BEZERRA, I. N.; <b>Vieira-Meyer, Anya PGF</b>; Silva, C.A.B..  Participação em banca de Carina Bandeira Bezerra. Avaliação clinicofuncional de idosos com diabetes mellitus: estudo comparativo entre brasileiros e Franceses. 2019. Tese (Doutorado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">DAMASCENO, M. M. C.; FIUZA, T. M.; VASCONCELOS, P. F.; VIEIRA-MEYER, A. P. G. F.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="649">Vieira, A P G F</a></b>.  Participação em banca de Eduarda Maria Duarte Rodrigues. Avaliação da atenção primária à saúde na perspectiva dos usuários adultos do município de Juazeiro do Norte-Ceará. 2018. Tese (Doutorado em Enfermagem)  - Universidade Federal do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="650">Vieira, A P G F</a></b>; ABREU, M. H. N. G.; WERNECK, M. A. F.; MARTINS, R. C.; ZINA, L. G.; CARVALHO, C. L.; PEIXOTO, S. W. V.; OLIVEIRA, A. C. B..  Participação em banca de Clarice Magalhães Rodrigues dos Reis. Analise do desempenho das Equipes de Saúde Bucal na Estratégia Saúde da Família: um estudo misto. 2016. Tese (Doutorado em Programa de Pós-Graduação em Odontologia)  - Universidade Federal de Minas Gerais. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">DAMASCENO, M. M. C.; SILVA, M. G. C.; ZANETTI, M. L.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="651">Vieira, A P G F</a></b>; GUBERT, F. A..  Participação em banca de Niciane Bandeira Pessoa Marinho. Satisfação dos usuários e dos profissionais do centro integrado de diabetes e hipertensão de Barbalha - CE. 2014. Tese (Doutorado em Enfermagem)  - Universidade Federal do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>10. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">MACHADO, C. V.; DIAS, A. A.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="652">Vieira, A P G F</a></b>; BAPTISTA, T. W. F.; COSTA, N. R..  Participação em banca de Liza Barreto Vieira. Política nacional e contexto local: a Atenção à Saúde Bucal em Municípios do estado do Ceará. 2013. Tese (Doutorado em Doutorado em Saúde Pública)  - Fundação Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b>Qualificações de Doutorado</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>VIEIRA-Meyer, Anya P G F</b>; MORAIS, C. L.; MACHADO, M. F. A. S.; ARAUJO, M. F. M.; PINTO, ANTÔNIO GERMANE ALVES; SOUZA, M. L. P..  Participação em banca de Yana Paula Coêlho Correia Sampaio. Ressonância da violência comunitária no processo de trabalho dos profissionais da ESF no cuidado integral à pessoa idosa. 2023 - Fundacao Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>Anya P G F Vieira-Meyer</b>; SILVEIRA, P. V.; Lima, R A; Gerage, LKAR.  Participação em banca de Emanuelle Albuquerque Carvalho Melo. Aprimoramento e satisfação com o uso do prontuário eletrônico do curso de odontologia da FFOE-UFC. 2019. Exame de qualificação (Doutorando em Odontologia)  - Universidade Federal do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="653">SAINTRAIN, M. V. L.</a>; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="654">Vieira, A P G F</a></b>; MOURA, W. V. B.; Abdon, A P V.  Participação em banca de Caroline Barbosa Lourenço. Dependencia funcional em idosos: uma análise espacieal no nordeste brasileiro. 2018. Exame de qualificação (Doutorando em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="655">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="656">SAINTRAIN, M. V. L.</a>; BEZERRA, I. N.; SILVA JUNIOR, G. B.; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8577652222943813" target="_blank" tabindex="657">Silva, R M</a>.  Participação em banca de Lucianna Leite Pequeno. Saúde bucal e estudo nutricional em idosos. 2017. Exame de qualificação (Doutorando em Programa de Pós-Graduação em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">DAMASCENO, M. M. C.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="658">Vieira, A P G F</a></b>; GUBERT, F. A.; ZANETTI, M. L..  Participação em banca de Niciane Bandeira Pessoa Marinho. Satisfação dos Usuários e dos Profissionais do Centro Integrado de Diabetes e Hipertensão de Barbalha-CE. 2013. Exame de qualificação (Doutorando em Enfermagem)  - Universidade Federal do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b>Qualificações de Mestrado</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">ARAUJO, C. E. L.; Andrade, LOM; <b>VIEIRA-Meyer, Anya P G F</b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0636763091396917" target="_blank" tabindex="659">NUTO, S. A. S.</a>.  Participação em banca de Hassã Pereira Lemos. Perfil de percepção dos Agentes Comunitários de Saúde sobre a resiliência da Atenção Primária à Saúde em Fortaleza-CE durante a pandemia de COVID-19. 2023 - Fundacao Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">DANTAS, T. S.; SILVA, P. G. B.; <b>VIEIRA-Meyer, Anya P G F</b>.  Participação em banca de Flávia Marília Azeredo de Oliveira. Comparação de um protocolo diário e em dias alternados de fotobiomodulação na prevenção de mucosite oral em pacientes submetidos à radioquimioterapia para câncer de boca. 2023. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>VIEIRA-Meyer, Anya P G F</b>; MOREIRA, D. M.; Gerage, LKAR.  Participação em banca de Paula Guerra da Rocha. Prevalência de Lesões cervicais não cariosas e recessão gengival em subpopulações de Guiné-Bissau. 2023. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">TEIXEIRA, A. K. M.; <b>Vieira-Meyer, Anya PGF</b>; FARIAS, M. R..  Participação em banca de Erika Basil Cavalcante Citó. Vicio em smartphones e o impacto na qualidade de vida e condições de saúde de estudantes da Universidade Federal do Ceará. 2021. Exame de qualificação (Mestrando em Odontologia)  - Universidade Federal do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0636763091396917" target="_blank" tabindex="660">NUTO, S. A. S.</a>; Almeida, M E L; <b>PIMENTEL G. F. VIEIRA-MEYER, ANYA</b>.  Participação em banca de Diego Ramos Aguiar. Avaliação da qualidade dos serviços prestados pelo Centro de Especialidades Odontológicas e Centro de Saúde da Família: percepção dos usuários. 2021. Exame de qualificação (Mestrando em Mestrado Profissional em Saúde da Família)  - Fundação Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>Anya Pimentel, Gomes Fernandes Vieira-Meyer</b>; ICHIHARA, M. Y.; GUIMARAES, J. M. X.; FREITAS, R. W. J. F..  Participação em banca de Gilmara Maria Batista Tavares da Silva. Qualificação - Qualidade do pré-natal e seu impacto na redução da sífilis congênita no município de Fortaleza. 2021. Exame de qualificação (Mestrando em Mestrado Profissional em Saúde da Família - MPSF/RENASF)  - Fundação Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>Anya, Pimentel Gomes Fernandes Vieira-Meyer</b>; NUTO, SHARMÊNIA DE ARAÚJO SOARES; PERISSE, A. R. S.; ARAUJO, M. F. M..  Participação em banca de Angelica Ribeiro Gurgel Barreira. Percepção dos profissionais da Estratégia Saúde da Família sobre os possiveis impactos do Complexo Industrial e Portuário do Pecém na Saúde da População. 2021. Exame de qualificação (Mestrando em Mestrado Profissional em Saúde da Família - ProfSaúde)  - FIOCRUZ/CEARÁ. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">SAINTRAIN, M. V. L.; BRASIL, C. C. P.; <b>VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</b>.  Participação em banca de Débora Rosana Alves Braga. CARACTERÍSTICAS SOCIOECONÔMICAS DE PESSOAS ACOMETIDAS PELA COVID-19 EM UM ESTADO DO NORDESTE BRASILEIRO. 2020. Exame de qualificação (Mestrando em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">CBF, Valter; <b>Anya P G F Vieira - Meyer</b>; MMM, Thereza; ACS, Helena.  Participação em banca de Rayssa Veras Camelo. Atuação das equipes de saude da familia na atencao a saude do hipertenso e diabetico no Brasil: uma análise secundária do PMAQ-AB 3o Ciclo. 2019. Exame de qualificação (Mestrando em Saúde Pública)  - Universidade Estadual do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>10. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>Anya P G F Vieira-Meyer</b>; Benevides, B. S.; Ramos Jr., A N.  Participação em banca de Elaine Lopes Pedrosa Bomfim. LINHAS DE CUIDADO AO PACIENTE COM HANSENÍASE: AÇÕES REALIZADAS POR MÉDICOS E CIRURGIÕES-DENTISTAS DA ESTRATÉGIA SAÚDE DA FAMÍLIA. 2019. Exame de qualificação (Mestrando em Mestrado Profissional em Saúde da Família - ProfSaúde)  - FIOCRUZ/CEARÁ. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>11. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>Vieira-Meyer, Anya PGF</b>; Rolim, J P M L; MOREIRA NETO, J. J. S..  Participação em banca de Maira Barbosa Coutinho. A influencia de vídeos de antecipação de situação no controle dos níveis de ansiedade em pacientes odontopediatricos. 2019. Exame de qualificação (Mestrando em Ciências Odontológicas)  - Centro Universitário Christus. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>12. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="661">Vieira, A P G F</a></b>.  Participação em banca de Carla Kerr Pontes. Prevalência de Sífilis em Homens que fazem sexo com homens em Fortaleza-Ceará-Brasil.. 2018. Exame de qualificação (Mestrando em Mestrado Profissional em Saúde da Família - ProfSaúde)  - FIOCRUZ/CEARÁ. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>13. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">UCHIMURA, K. Y.; FREITAS, R. W. J. F.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="662">Vieira, A P G F</a></b>.  Participação em banca de Helena Paula Guerra dos Santos. Como as gestantes beneficiárias do Bolsa Família percebem o programa e suas repercussões na saúde?. 2018. Exame de qualificação (Mestrando em Mestrado Profissional em Saúde da Família - MPSF/RENASF)  - Fundação Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>14. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="663">SAINTRAIN, M. V. L.</a>; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="664">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/1815947099924439" target="_blank" tabindex="665">LIMA, D. L. F.</a>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5516641709622899" target="_blank" tabindex="666">BIZERRIL, D. O.</a>.  Participação em banca de Janayne de Sousa Oliveira. Avaliação de fluxo salivas em idosos e a aplicabilidade de intervenção nutricional. 2017. Exame de qualificação (Mestrando em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>15. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="667">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="668">SAINTRAIN, M. V. L.</a>; DIAS, A. A.; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5516641709622899" target="_blank" tabindex="669">BIZERRIL, D. O.</a>.  Participação em banca de Sandy Kaena Soares de Freitas. Reabilitação oral protética e autopercepção da saúde bucal relacionada a qualidade de vida. 2017. Exame de qualificação (Mestrando em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>16. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="670">Vieira, A P G F</a></b>; BARRETO, I. C. H. C.; Andrade, LOM.  Participação em banca de Patrícia Symone Almeida Ivo. Desenvolvimento de cenários de utilização de tecnologias de informação e comunicação para o suporte a criança com transtorno do espectro autista. 2017. Exame de qualificação (Mestrando em Mestrado Profissional em Saúde da Família)  - Fundação Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>17. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="671">Vieira, A P G F</a></b>; FREITAS, R. W. J. F.; GOMES, K. W. L..  Participação em banca de Diego Magalhães Siqueira. Eficácia da auriculterapia na analgesia das dores crônicas causadas pela Chikungunya: ensaio clínico controlado e randomizado. 2017. Exame de qualificação (Mestrando em Mestrado Profissional em Saúde da Família)  - Fundação Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>18. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="672">Vieira, A P G F</a></b>; BARRETO, I. C. H. C.; CARVALHO, C. L..  Participação em banca de Roberto Ribeiro Maranhão. Análise do campo de práticas dos profissionais médicos na ESF em Fortaleza. 2017. Exame de qualificação (Mestrando em Mestrado Profissional em Saúde da Família)  - Fundação Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>19. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="673">Vieira, A P G F</a></b>; BARRETO, I. C. H. C.; BAPTISTA, T. W. F..  Participação em banca de Antonio Leonel de Lima Júnior. A percepção dos médicos em relação as formas de contratação desta categoria da Estratégia Saúde da Família. 2017. Exame de qualificação (Mestrando em Mestrado Profissional em Saúde da Família)  - Fundação Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>20. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="674">Vieira, A P G F</a></b>; VIEIRA, L. B.; ARAUJO, M. A. M..  Participação em banca de Luiza de Paula Sousa. Atenção em Saúde Bucal oferecida aos pacientes com esquizofrenia na Estratégia Saúde da Família em Fortaleza/CE. 2015. Exame de qualificação (Mestrando em Programa de Pós Graduação em Saúde da Familia)  - Fundação Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>21. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="675">Vieira, A P G F</a></b>; RODRIGUES, A.; CALVASINA, P. G.; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0636763091396917" target="_blank" tabindex="676">NUTO, S. A. S.</a>.  Participação em banca de Carla Manuela Rodrigues Nogueira. A Notificação Compulsória de Doenças ou Agravos na Atenção Primária: o caso dos atendimentos antirrábicos no município de Fortaleza/CE. 2015. Exame de qualificação (Mestrando em Mestrado Profissional em Saúde da Família)  - Fundação Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>22. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">FREITAS, R. W. J. F.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="677">Vieira, A P G F</a></b>; GOMES, K. W. L.; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5537817532828798" target="_blank" tabindex="678">VASCONCELOS, M. I. O.</a>.  Participação em banca de Leonardo Brabosa Rolim. Avaliação dos atributos da atenção primária à saúde de Fortaleza/CE: estudo através do PCATool-Brasil, versão profissional. 2015. Exame de qualificação (Mestrando em Saúde da Família)  - Fundação Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>23. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="679">SAINTRAIN, M. V. L.</a>; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="680">Vieira, A P G F</a></b>; AGUIAR, A. S. W..  Participação em banca de Marcus Antônio Melo Carvalho Filho. Prevalência de Trauma bucomaxilofacial em idosos hospitalizados em hospital terciário no Nordeste brasileiro. 2014. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>24. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">FROTA, F. H. S.; OSTERNE, M. S. F.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="681">Vieira, A P G F</a></b>.  Participação em banca de Jamile Pinheiro Karbage Salmito. Avaliação na atenção primária: experiência de aplicação da auto avaliação para melhoria do acesso e da qualidade - AMAQ no município de Fortaleza. 2014. Exame de qualificação (Mestrando em Sociologia)  - Universidade Estadual do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>25. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">SILVA, A. V. S. E.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="682">Vieira, A P G F</a></b>; GOMES, A. M. A.; MORAIS, A. P. P..  Participação em banca de Márcia Lúcia de Oliveira Gomes. Assistência  odontologica às pessoas com deficiência na 10ª região de saúde no Estado do Ceará. 2013. Exame de qualificação (Mestrando em Mestrado Profissional em Saúde da Família)  - Fundação Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>26. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">CHAGAS, M. I. O.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="683">Vieira, A P G F</a></b>; DIAS, M. S. A.; GUIMARAES, J. M. X..  Participação em banca de Ediléia Marcela Dutra. Programa nacional para melhoria do acesso e da qualidade da atenção básica - PMAQ: posicionamento de gestores e trabalhadores sobre o primeiro ciclo. 2013. Exame de qualificação (Mestrando em Mestrado Profissional em Saúde da Família)  - Fundação Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>27. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="684">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0636763091396917" target="_blank" tabindex="685">NUTO, S. A. S.</a>; SOUSA, F. B..  Participação em banca de Lizaldo Andrade Maia. Atenção à Saúde Bucal: arranjos e dispositivos operados no cuidado às pessoas. 2013. Exame de qualificação (Mestrando em Mestrado Profissional em Saúde da Família)  - Fundação Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>28. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="686">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="687">SAINTRAIN, M. V. L.</a>; AGUIAR, A. S. W..  Participação em banca de Krishna Andreia Feitosa Petrola. Investigação das ações de saúde bucal em crianças do bolsa família. 2013. Exame de qualificação (Mestrando em Mestrado Profissional em Saúde da Família)  - Fundação Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>29. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="688">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9128715784318958" target="_blank" tabindex="689">COIMBRA, L. C.</a>; CHAGAS, M. I. O..  Participação em banca de Mailson Fontes de Carvalho. Programa de melhoria do acesso e da qualidade da atenção básica em municípios do Piauí e Ceará. 2013. Exame de qualificação (Mestrando em Mestrado Profissional em Saúde da Família)  - Fundação Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>30. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="690">SAINTRAIN, M. V. L.</a>; LIBORIO, A. B.; AGUIAR, A. S. W.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="691">Vieira, A P G F</a></b>.  Participação em banca de Paulo Leonardo Ponte Marques. Alterações na saúde bucal de pacientes em hemodiálise. 2012. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>31. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">ROCHA, L. L.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="692">Vieira, A P G F</a></b>.  Participação em banca de Lyana Leal Rocha. Acessibilidade aos serviços públicos odontológico às pessoas com deficiencia em Fortaleza - CE. 2011. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>32. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">BRILHANTE, A. V. M.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="693">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9948525011531885" target="_blank" tabindex="694">CATRIB, A. M. F.</a>.  Participação em banca de aline Veras Morais Brilhante. A educação sexual na escola pública como mecanismo de promoção em saúde. 2011. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>33. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">BEZERRA, A. M. S.; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="695">SAINTRAIN, M. V. L.</a>; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="696">Vieira, A P G F</a></b>.  Participação em banca de Aila Maria da Silva Bezerra. Aspectos epidemiológicos e funcionais na admissão e alta hospitalar de pacientes com acidente vascular cerebral isquêmico. 2011. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>34. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">CARVALHO, I. L. N.; Gondim, APS; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="697">Vieira, A P G F</a></b>.  Participação em banca de Igho Leonardo do Nascimento Carvalho. Uso de psicofármacos em adolescentes atendidos pelo centro de atenção Psicossocial infato-juvenil. 2011. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>35. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Silva, C.A.B.; DINIZ, D. B.; HISSA, M. N.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="698">Vieira, A P G F</a></b>.  Participação em banca de Claudine Kênnia de Almeida Cezário. Avaliação do estado nutricional de adolescentes esportistas das vilas olímpicas de Fortaleza - Ceará. 2009. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>36. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="699">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0636763091396917" target="_blank" tabindex="700">NUTO, S. A. S.</a>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8577652222943813" target="_blank" tabindex="701">Silva, R M</a>.  Participação em banca de Caroline Antero Machado. Atuação dos cirurgiões-dentistas na ESB. 2009. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>37. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">SAMPAIO, J. J. C.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="702">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/0636763091396917" target="_blank" tabindex="703">NUTO, S. A. S.</a>.  Participação em banca de Mariana Ramalho de Farias. A (DES) integração do trabalho da equipe da saúde da família/saúde bucal no Programa Saúde da Família. 2008. Exame de qualificação (Mestrando em Saúde Pública)  - Universidade Estadual do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>38. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Silva, C.A.B.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="704">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8577652222943813" target="_blank" tabindex="705">Silva, R M</a>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9948525011531885" target="_blank" tabindex="706">CATRIB, A. M. F.</a>.  Participação em banca de Frederico Lemos Araújo. Detecção do risco cardiovascular em estudantes da cidade de Fortaleza. 2008. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>39. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="707">SAINTRAIN, M. V. L.</a>; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="708">Vieira, A P G F</a></b>; Gonçalves, M. L. C.; AMORIM, R. F..  Participação em banca de Ana Karine Lima Moreira. Fatores de risco associado a Hipertensão e a Diabetes Mellitus em pessoas idosas. 2008. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>40. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Valdés, M. T. M.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="709">Vieira, A P G F</a></b>; Gonçalves, M. L. C.; Santos, Z.M.S.A..  Participação em banca de Karla Simone dos Santos Oliveira Fróes. Qualidade de vida percebida de pacientes com sequelas de Acidente Vascular Cerebral. 2008. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>41. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">OLIVEIRA, L.C.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="710">Vieira, A P G F</a></b>; VASCONCELOS FILHO, E. V..  Participação em banca de Poliana Miranda Pinheiro. A prática do cirurgião dentista no Programa Saúde da Família - PSF de Fortaleza no sentido da humanização da atenção. 2008. Exame de qualificação (Mestrando em Saúde Pública)  - Universidade Estadual do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>42. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Gonçalves, M. L. C.; Silva, C.A.B.; Vieira, L J E S; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="711">Vieira, A P G F</a></b>.  Participação em banca de Helder Oliveira e Silva. prevalência da infecção pelo HIV em pacientes com tuberculose, em Fortaleza - CE. 2008. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>43. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="712">Vieira, A P G F</a></b>; MENEZES, L. M. B.; MACIEL, R. H. M. O.; Valdés, M. T. M..  Participação em banca de Carlos Santos de Castro Filho. Percepção de saúde bucal e qualidade de vida em trabalhadores com e sem assistência odontológica. 2008. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>44. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="713">Vieira, A P G F</a></b>; MONTALVERNE, D. G. B.; Silva, C.A.B.; Gonçalves, M. L. C..  Participação em banca de José Bezerra Góes Neto. Prevalância de dor na coluna e alterações de postura em estudantes de Odontologia do Estado do Ceará. 2008. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>45. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">MENEZES, L. M. B.; Gonçalves, M. L. C.; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="714">SAINTRAIN, M. V.</a>; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="715">Vieira, A P G F</a></b>.  Participação em banca de Paula Ventura da Silveira. Avaliação de serviço odontológico prestado por empresas privadas no estado do Ceará. 2008. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>46. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="716">Vieira, A P G F</a></b>; Vieira, L J E S; Pordeus, A M J; Oliveira Jr., J A.  Participação em banca de Samira Valentim Gama Lira. A conduta do mototaxista associada à prevalência e ocorrência de acidentes de trânsito. 2007. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>47. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="717">Vieira, A P G F</a></b>; Vieira, L J E S; FROTA, M. A.; OLIVEIRA, E. N..  Participação em banca de Geisy Lanne Muniz Luna. Notificação de maus tratos em crianças e adolescentes: subsídios para prática de educação e saúde. 2007. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>48. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="718">Vieira, A P G F</a></b>; Araújo, M A L; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8577652222943813" target="_blank" tabindex="719">Silva, R M</a>; Montagner, M. A..  Participação em banca de Ana Luíza de Araújo Campos. epidemiologia da sífilis em gestantes em Fortaleza - Ceará. 2007. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>49. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="720">Vieira, A P G F</a></b>; Matos, V C; Pereira, M L D; Gonçalves, M. L. C..  Participação em banca de Regina Flávia de Castro Almeida. Avaliação da informação oral e escrita no conhecimento dos usuários sobre medicamentos anti-retrovirais. 2007. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>50. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="721">Vieira, A P G F</a></b>; Gonçalves, M. L. C.; Vieira, L J E S; Araújo, M A L.  Participação em banca de Simone Paes de Melo. Avaliação do seguimento de crianças com diagnóstico de sífilis congênita em Fortaleza - Ceará. 2007. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>51. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="722">Vieira, A P G F</a></b>; Lima, M.C.O.; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4167422383882509" target="_blank" tabindex="723">Carvalho, A. F.</a>; MONTENEGRO JUNIOR, R. M..  Participação em banca de Sâmia Mustafa Aguiar. Prevalência de estresse e depressão em acadêmicos de medicina e odontologia da Universidade Federal do Ceará. 2007. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>52. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="724">Vieira, A P G F</a></b>; MENEZES, L. M. B.; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="725">SAINTRAIN, M. V. L.</a>.  Participação em banca de Ruth Cavalcanti Nunes. Utilização do lúdico como estratégia na promoção de saúde bucal. 2006. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>53. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="726">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4167422383882509" target="_blank" tabindex="727">Carvalho, A. F.</a>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7492450432942397" target="_blank" tabindex="728">Montenegro Jr, R.M.</a>.  Participação em banca de Carlos Calyton Torres Aguiar. Avaliação da qualidade de vida em idosos diabéticos e sua correlação com depresão, cognição e saúde oral. 2006. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>54. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Valdés, M. T. M.; Vieira, L J E S; ALENCAR, M. L.; FROTA, M. A.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="729">Vieira, A P G F</a></b>.  Participação em banca de Lia Machado Fiuza. Qualidade de vida na infância: visão de alunos da rede pública e privada de ensino. 2006. Exame de qualificação (Mestrando em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>55. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Silva, C.A.B.; CALDAS, C. P.; MONTALVERNE, D. G. B.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="730">Vieira, A P G F</a></b>.  Participação em banca de Silvana Mara Rocha Sydney Montenegro. Efeitos de um programa de fisioterapia como promotor de saúde na capacidade funcional de idosas institucionalizadas. 2006. Exame de qualificação (Mestrando em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b>Monografias de cursos de aperfeiçoamento/especialização</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="731">Vieira, A P G F</a></b>.  Participação em banca de Lyssa de Aragão Arrais. Disponibilidade ao atendimento de pacientes especiais pelos odontopediatras da cidade de Fortaleza. 2007. Monografia (Aperfeiçoamento/Especialização em Especialização em Odontologia)  - Associação Brasileira de Odontologia. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="732">Vieira, A P G F</a></b>.  Participação em banca de Suzana Rachel de Oliveira. Preparo e disponibilidade dos odontopediatras no município de Fortaleza em relação ao atendimento odontológico de crianças HIV positivas/AIDS. 2007. Monografia (Aperfeiçoamento/Especialização em Especialização em Odontopediatria)  - Associação Brasileira de Odontologia. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="733">Vieira, A P G F</a></b>.  Participação em banca de Carolina Barbosa Lourenço. Experiência multiprofissional no tratamento do bruxismo infantil. 2007. Monografia (Aperfeiçoamento/Especialização em Especialização em Odontopediatria)  - Associação Brasileira de Odontologia. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="734">Vieira, A P G F</a></b>.  Participação em banca de Jamile Pinheiro Karbage. Prevalência de cárie em crianças institucionalizadas, com e sem acesso à atenção odontológica. 2007. Monografia (Aperfeiçoamento/Especialização em Especialização em Odontopediatria)  - Associação Brasileira de Odontologia. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="735">Vieira, A P G F</a></b>.  Participação em banca de Mirela Andrade Campos. Autotrasplante: uma opção de tratamento. 2007. Monografia (Aperfeiçoamento/Especialização em Especialização em Odontopediatria)  - Associação Brasileira de Odontologia. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="736">Vieira, A P G F</a></b>.  Participação em banca de Maria Augusta Brito Neta. Avaliação do conhecimento das gestantes sobre a saúde bucal dos seus bebês no município de Acaraú - CE. 2007. Monografia (Aperfeiçoamento/Especialização em Especialização em Odontopediatria)  - Associação Brasileira de Odontologia. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="737">Vieira, A P G F</a></b>.  Participação em banca de Nara Gurgel Magalhães. Amamentação e hábito de sucção não-nutritivos. 2007. Monografia (Aperfeiçoamento/Especialização em Especialização em Odontopediatria)  - Associação Brasileira de Odontologia. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="738">Vieira, A P G F</a></b>.  Participação em banca de Silvana do Vale Cardoso. Avaliação do serviço odontológico no município de Aracati - CE, no período de 2001 a 2004. 2006. Monografia (Aperfeiçoamento/Especialização em Especialização em Odontologia em Saúde Coletiva)  - Universidade Estadual do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b>Trabalhos de conclusão de curso de graduação</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>Vieira-Meyer, Anya PGF</b>; DAMASCENO, J. X.; MAGALHAES, D. P..  Participação em banca de LIA DE VASCONCELOS ROCHA.Atendimento odontológico a paciente com hipoplasia dérmica focal (Síndrome de Goltz) com achados clínicos incomuns - relato de caso.
# 							2020. Trabalho de Conclusão de Curso (Graduação em odontologia)  - Centro Universitário Christus. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>Anya P G F Vieira-Meyer</b>; Carvalho, I.F.; Aquino, P.B..  Participação em banca de Marjorie Luiza Oliveira de Melo.Disponibilidade ao atendimento de Pacientes com necessidades especiais pelos odontopediatras cearenses.
# 							2019. Trabalho de Conclusão de Curso (Graduação em odontologia)  - Centro Universitário Christus. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="739">Vieira, A P G F</a></b>.  Participação em banca de Rodolpho Luiz Batista Couras.Percepção dos profissionais do PSF sobre a inclusão do farmacêutico no NASF.
# 							2009. Trabalho de Conclusão de Curso (Graduação em Graduação em Farmácia)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="740">Vieira, A P G F</a></b>.  Participação em banca de Rebeca B. Vasconcelos; Rachel B. Vasconcelos; Ana R. T. Car.Avaliação do tratamento odontologico ao paciente portador de necessidades especiais em ambiente não ambulatorial.
# 							2009. Trabalho de Conclusão de Curso (Graduação em Odontologia)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="741">Vieira, A P G F</a></b>.  Participação em banca de Denise Fontenelle; Krisna Lima; Mirella de Sousa Pereira.Grau de infiltração marginal de técnicas restauradoras com cimento ionomero de vidro e resina composta fotopolimerizável em molares decíduos.
# 							2009. Trabalho de Conclusão de Curso (Graduação em Odontologia)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">MAIA, T. M. C.; CARVALHO, P. G.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="742">Vieira, A P G F</a></b>.  Participação em banca de Andryne Vasconcelos Costa.Impacto da gravidez na adolescência.
# 							2009. Trabalho de Conclusão de Curso (Graduação em Graduação em Farmácia)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">ROMEU, G. A.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="743">Vieira, A P G F</a></b>; SILVEIRA, J. A. P..  Participação em banca de Mariana Macêdo Tavares.Avaliação da adesão ao tratamento anti-retroviral de pacientes portadores de HIV.
# 							2009. Trabalho de Conclusão de Curso (Graduação em Graduação em Farmácia)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">CARVALHO, P. G.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="744">Vieira, A P G F</a></b>; FIGUEIREDO, M. F..  Participação em banca de André Luiz Oliveira.Estudo de conhecimento e percepções do agente comunitário de saúde sobre a tuberculose pilmonar no município de Coreaú.
# 							2009. Trabalho de Conclusão de Curso (Graduação em Graduação em Farmácia)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">SPANOS, C. R.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="745">Vieira, A P G F</a></b>; MARTINS, M. G. A..  Participação em banca de Dayse M. P. Macedo, Marcella de O. Lima, Merlin E. Lopes.Frequência de amamentação, hábitos deletérios e maloclusões na dentição decídua - estudos realizado na escola pública professor Francisco Mattos Dourado.
# 							2009. Trabalho de Conclusão de Curso (Graduação em Odontologia)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>10. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="746">Vieira, A P G F</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="747">SAINTRAIN, M. V. L.</a>; ALBUQUERQUE, S. H. C..  Participação em banca de Davi Oliveira Bizerril.Saúde bucal em pessoas na terceira idade.
# 							2009. Trabalho de Conclusão de Curso (Graduação em Odontologia)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>11. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="748">Vieira, A P G F</a></b>; MAIA, M. C. G..  Participação em banca de Fernando Augusto Góes Mota.Satisfação dos usuários do SUS em relação às atividades de saúde na regional 6 do município de Fortaleza.
# 							2009. Trabalho de Conclusão de Curso (Graduação em Odontologia)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>12. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">SPANOS, C. R.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="749">Vieira, A P G F</a></b>; MARTINS, M. G. A..  Participação em banca de Aline M. Fontes, Bruno C. Martins e Érika M. Pinto Dinelly.A importância da recuperação de espaço após a perda precoce de dentes decíduos no desenvolvimento da oclusão normal - relato de caso.
# 							2008. Trabalho de Conclusão de Curso (Graduação em Odontologia)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>13. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="750">Vieira, A P G F</a></b>; FROTA, M. A..  Participação em banca de Fred Ribeiro Santiago.Fratura em idosos em decorrência de quedas com registro no Instituto Dr. José Frota (IJF) no ano de 2004.
# 							2006. Trabalho de Conclusão de Curso (Graduação em Fisioterapia)  - Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear"><br class="clear"><a name="ParticipacaoBancasComissoes" tabindex="751"></a><br class="clear"><div class="inst_back">
# <b>Participação em bancas de comissões julgadoras</b>
# </div>
# <div class="cita-artigos">
# <b>Professor titular</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="752">Vieira, A P G F</a></b>; MOREIRA, M. A. G.; VIEIRA, L. B.; NEVES, D. M. L.; NEVES, B. G.; FARIAS, M. R.. Comissão Julgadora Professor Efetivo - UFC. 2017. Universidade Federal do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b>Concurso público</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="753">Vieira, A P G F</a></b>. Comissão Julgadora - Concurso para Docente Superior da UFBA.
# 							2012. Universidade Federal da Bahia. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b>Avaliação de cursos</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="754">Vieira, A P G F</a></b>. Tema - Acadêmico III.
# 							2012. Universidade Federal do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b>Outras participações</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>Anya Pimentel, Gomes Fernandes Vieira-Meyer</b>; VIEIRA-MEYER, A. P. G. F.; ICHIHARA, M. Y.; GUIMARAES, J. M. X.; FREITAS, R. W. J. F.. Qualificação - Qualidade do pré-natal e seu impacto na redução da sífilis congênita no município de Fortaleza.
# 							2021.
# 							Fundação Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">SILVA, A. A.; NORO, L.; GOULART, B.; CONDE, W.; <b>Anya P G F Vieira-Meyer</b>. Comissão de Avaliação Prêmio Capes de Teses - Saúde Coletiva.
# 							2020.
# 							Coordenação de Aperfeiçoamento de Pessoal de Nível Superior. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>Vieira-Meyer, Anya PGF</b>; Gerage, LKAR; SILVEIRA, P. V.; Lima, R A. Qualificação - Elaboração de im Instrumento de Classificação de prioridades odontológicas para determinação de critérios de atendimento.
# 							2019.
# 							Universidade Federal do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="755">Vieira, A P G F</a></b>. Avaliador de trabalhos científicos II jornada Odontológica da Unichristus.
# 							2018.
# 							Centro Universitário Christus. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="756">Vieira, A P G F</a></b>. Avaliador de trabalhos do 1o Encontro Nordeste de Saúde da Família: em Defesa do SUS e da Estratégia Saúde da Família (ENESF 2018).
# 							2018.
# 							Governo do Estado do Ceará. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">VIEIRA-MEYER, A. P. G. F.; <b>Vieira-Meyer, Anya PGF</b>. Comissão avaliação PPSUS Alagoas - SISCT.
# 							2016.
# 							Fundação de Amparo à Pesquisa do Estado de Alagoas. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="757">Vieira, A P G F</a></b>. Reunião para análise de mérito de APCN´s Profissionais e Minter e Dinter.
# 							2014.
# 							Coordenação de Aperfeiçoamento de Pessoal de Nível Superior. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4640029618752231" target="_blank" tabindex="758">SAINTRAIN, M. V. L.</a>; LIBORIO, A. B.; AGUIAR, A. S. W.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="759">Vieira, A P G F</a></b>. Alterações na saúde bucal de pacientes em hemodiálise.
# 							2012. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Gonçalves, M. L. C.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="760">Vieira, A P G F</a></b>; Araújo, M A L; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8577652222943813" target="_blank" tabindex="761">Silva, R M</a>; Vieira, L J E S; Santos, Z.M.S.A.; LANDIM, F. L. P.; NATIOS, M. K.; MATSUE, R. Y.. Comissão Julgadora para Seleção da 5ª Turma do Mestrado em Saúde Coletiva da UNIFOR.
# 							2009.
# 							Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>10. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="762">Vieira, A P G F</a></b>. Comissão Julgadora de Trabalhos Científicos.
# 							2008.
# 							Sociedade Brasileira de Medicina de Familia e Comunidade. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>11. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="763">Vieira, A P G F</a></b>; Lima Neto. A.S.; Melo, M.E.. Comissão de seleção de bolsistas de extensão.
# 							2008.
# 							Secretaria Municipal de Saúde de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>12. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Gonçalves, M. L. C.; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9948525011531885" target="_blank" tabindex="764">CATRIB, A. M. F.</a>; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="765">Vieira, A P G F</a></b>; Silva, C.A.B.; Araújo, M A L; FROTA, M. A.. Comissão Julgadora para Seleção da IV turma do Mestrado em Saúde Coletiva.
# 							2008.
# 							Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>13. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="766">Vieira, A P G F</a></b>. Banca Examinadora da II Mostra do Sistema Municipal de Saúde Escola de Fortaleza, I Mostra do Agente Comunitário de Saúde de Fortaleza e Mostra Nacional Interativa Humaniza SUS.
# 							2008.
# 							Secretaria Municipal de Saúde. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>14. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8577652222943813" target="_blank" tabindex="767">Silva, R M</a>; Silva, C.A.B.; Santos, Z.M.S.A.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="768">Vieira, A P G F</a></b>; Leitinho, M C. Seleção da VIII Turma do Curso de Mestrado em Educação em Saúde.
# 							2007.
# 							Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>15. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="769">Vieira, A P G F</a></b>; FROTA, M. A.; Silva, C.A.B.; Valdés, M. T. M.; Gonçalves, M. L. C.; Montagner, M. A.. Seleção da III turma do Mestrado em Saúde Coletiva.
# 							2007.
# 							Universidade de Fortaleza. </span>
# </div>
# <br class="clear">
# <br class="clear">
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="Eventos" tabindex="770">
# <h1 tabindex="0">Eventos</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <div class="layout-cell layout-cell-12 data-cell">
# <a name="ParticipacaoEventos" tabindex="771"></a><br class="clear"><div class="inst_back">
# <b>Participação em eventos, congressos, exposições e feiras</b>
# </div>
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">15th European Public Health Conference. Readiness of Community Health Agents for COVID19. 2022. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Ágora Abrasco - Colóquio O papel da pós-graduação em Saúde Coletiva no enfrentamento à COVID-19.Colóquio O papel da pós-graduação em Saúde Coletiva no enfrentamento à COVID-19. 2020. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Feira Digital de Soluções para a Saúde Virtual 2020. Doutorados profissionais e a formação para o SUS: as experiências da Fiocruz. 2020. (Feira). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">II Congresso Integrado Unichristus. AVALIAÇÃO DA QUALIDADE DE VIDA DE CUIDADORES DE PACIENTES COM NECESSIDADES ESPECIAIS. 2020. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Seminário Internacional em Atenção Básica e as (re)existências na Pandemia.Diálogos (re)existenciais - Estratégia de Saúde na Família na pandemia. 2020. (Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">66th ORCA Congress. Brazilian conditional cash transfer five year old children dental needs and access to dental care. 2019. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Feira do Conhecimento. 2019. (Feira). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Fórum de Coordenadores dos Programas de Pós-Graduação da Abrasco.Ficha de Avaliação do Fórum e a visão dos representantes de área.
# 							2019. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Reunião Coordenadores dos Programas de Pós-Graduação.Preparando a avaliação quadrienal: planejamento, autoavaliação e produção técnica.
# 							2019. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>10. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Reunião Coordenadores dos Programas de Pós-Graduação.
# 							2019. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>11. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Seminário de Meio Termo da área de Saúde Coletiva. 2019. (Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>12. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">44o Congresso Brasileiro de Estomatologia e Patologia Oral. Atendimento odontológico a paciente com sindrome de goltz com achados clinicos raros - relato de caso. 2018. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>13. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">ENESF.Redes Educacionais.
# 							2018. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>14. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Fórum de Coordenadores dos Programas de Pós-Graduação.Fórum dos Coordenadores dos Programas de Pós-Graduação da ABRASCO. 2018. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>15. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">II Jornada Odontológica da Unichristus.Abordagem odontológica em Pacientes com Cardiopatia Infantil.
# 							2018. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>16. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">II Jornada Odontológica da Unichristus.Atendimento odontológico a paciente com achados incomuns associados a síndrome de Goltz - um relato de caso.
# 							2018. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>17. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">II Jornada Odontológica da Unichristus.Disponibilidade ao atendimentos de pacientes especiais pelos odontontopediatras cearenses.
# 							2018. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>18. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">International Conference of integrated Care.A National Programme for Improving Access and Quality of a multiprofessional Primary Care system. 2018. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>19. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">IV Mutirão de Síndrome da Zika Congênita.Avaliação de crianças com sindrome da Zika virus. 2018. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>20. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">The 18th International Conference on Integrated Care.A National Programme for Improving Access and Quality of a multiprofessional Primary Care system. 2018. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>21. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">VI Congresso Internacional de Odontologia. Sindrome de Goltz ou hipoplasia focal - revisão de literatura. 2018. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>22. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XV Encontro de Iniciação à Pesquisa e à Docencia e XIII Encontro de Pesquisadores da Unichristus.Experiência de atendimento de pacientes especiais infantis no Centro de Estudos em Pacientes com Necessidades Especiais Unichristus - CEPE.
# 							2018. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>23. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XV Jornada Odontopet.Cardiopatia infantil: cuidados no atendimento odontológico.
# 							2018. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>24. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XV Jornada Odontopet.Disponibilidade ao atendimentos de pacientes especiais pelos odontontopediatras cearenses.
# 							2018. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>25. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Congresso Internacional Saúde e Sociedade. 2017. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>26. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Fórum de Coordenadores. 2017. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>27. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">IADR. Oral Health Care to Schizophrenic Patients by Primary Health Care. 2017. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>28. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">The 21 st International World Congress of Epidemiology (WCE2017). the implications of the Brazilian National Programme for improving Access and Quality of Primary Care. 2017. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>29. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XVI JOIA.Cuidados no atendimento odontológico à pacientes com paralisia cerebral. 2017. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>30. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">22 nd IUHPE World Conference on Health Promotion.Satisfação e acesso à saúde bucal das pessoas que vivem com HIV/AIDS no Nordeste do Brasil na perspectiva dos usuários. 2016. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>31. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">22º IUHPE World Conference on Health Promotion.Viver com HIV/AIDS em Fortaleza/CE, Brasil: Sensibilizando para Ampliar o Acesso à Saúde Bucal na Atenção Primária e Secundária. 2016. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>32. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">4th Internacional Conference on Epidemiology &amp; Public Health.The Improvement of Brazilian Primary Health Care Program (Estrategia Sáude da Família)based on PMAQ extrenal avaliation. 2016. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>33. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">4th Internacional Conference on Epidemiology &amp; Public Health.The 3nd cycle of the Brazilian national programme for improving acess and quality of primary care (PMAQ). 2016. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>34. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">4th Internacional Conference on Epidemiology &amp; Public Health. 2016. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>35. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Fórum de Coordenadores de Pós-Graduação em Saúde Coletiva. 2016. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>36. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">The 9th European Public Health Conference. 2016. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>37. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">VI Jornada Odontológica de Sobral e VIII Jornada Sobralense de Estomatologia, Radiologia Odontológica e Patologia Oral.participou como avaliador de resumos de trabalhos científicos na VI Jornada Odontológica de Sobral e VIII Jornada Sobralense de Estomatologia, Radiologia Odontológica e Patologia Oral. 2016. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>38. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XVI ENCONTRO DE PÓS-GRADUAÇÃO E PESQUISA.ATIVIDADE DA VIDA DIÁRIA E CONDIÇÃO DE SAÚDE BUCAL DE IDOSOS.
# 							2016. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>39. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XV JOIA - III MOSTRA DE SAUDE BUCAL DO MUNICIPIO DE FORTALEZA.Sensibilizar os profissionais para ampliar o acesso à saude bucal no SUS das pessoas que vivem com HIV/AIDS em Fortaleza/CE. 2016. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>40. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">11º Congresso Brasileiro de Saúde Coletiva. Atenção à saúde bucal das pessoas que vivem com HIV/AIDS em Fortaleza-CE na perspectiva dos cirurgiões-dentistas. 2015. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>41. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">11º Congresso Brasileiro de Saúde Coletiva. A utilização das estratégias de desenvolvimento propostas pelo PMAQ-AB no Estado do Piauí, Brasil. 2015. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>42. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">1st Annual Internatiional Conference on Public Health. National Program of Health Services Evaluation: Methodological Innovations and Challenges in Brazil. 2015. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>43. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">1st Annual Internatiional Conference on Public Health. 2015. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>44. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">51º Congresso da Sociedade Brasileira de Medicina Tropical. 2015. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>45. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Curso"Avaliação de impactos utilizando desenhos observacionais v. 2015. (Oficina). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>46. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Curso Internacional sobre ?Avaliação de Impacto à Saúde de Grandes Empreendimentos?. 2015. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>47. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Fórum de Coordenadores de Pós-Graduação em Saúde Coletiva. 2015. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>48. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">II Mostra de saúde bucal do municipio de Fortaleza-CE.O que é preciso saber para atendimento odontológico às pessoas que vivem com HIV/AIDS? assistea o Vídeo., multiplique o conhecimento e faça a diferença. 2015. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>49. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">IV Congresso Internacional Ciências, Tecnologias y Culturas. 2015. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>50. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Modelos de Atenção em Saúde Perspectivas para o Estado do Ceará. 2015. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>51. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Programa de Pós-Graduação em Saúde Coletiva.Programa de Melhoria e Avaliação da Atenção Básica. 2015. (Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>52. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Reunião de Acompanhamento dos Programas de Pós-Graduação da área de Saúde Coletiva. 2015. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>53. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Seminário, Saúde e Sociedade do Futuro. 2015. (Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>54. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Curso Básico em Educação para as Profissões de Saúde. 2014. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>55. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Curso Básico em Educação para as Profissões de Saúde.Facilitadora. 2014. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>56. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Reunião do Fórum de Coordenadores dos Programas de Pós Graduação da Abrasco. 2014. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>57. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">V Seminário Internacional em Promoção da Saúde. 2014. (Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>58. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">V Seminário Internacional em Promoção da Saúde.Sintonia de Depressão e Desconforto Bucal em Pacientes Idosos. 2014. (Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>59. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">V  Seminário Internacional  em Promoção da Saúde.Sintonia de Depressão e desconforto bucal em pacientes idosos. 2014. (Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>60. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">V  Seminário Internacional  em Promoção da Saúde.Edentulismo na experiência vivida por idosos. 2014. (Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>61. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">V  Seminário Internacional  em Promoção da Saúde.Centro de convivência de idosos como espaço de promoção da saúde: uma experiência no Nordeste Brasileiro. 2014. (Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>62. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XIX Congresso Brasileiro de Geriatria e gerontologia. Violência contra a saúde bucal de pessoas idosas. 2014. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>63. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Convibra (online conference).Diagnóstico do acesso aos serviços de saúde bucal em um município cearense. 2013. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>64. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Encontro sobre Tendências Tecnológicas em Plataformas Vegetais. 2013. (Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>65. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Fórum de Coordenadores de Pós-Graduação em Saúde Coletiva. 2013. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>66. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Fórum de Coordenadores de Pós-Graduação em Saúde Coletiva. 2013. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>67. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Oficina para Seleção de Prioridades PPSUS 2012/2013.Oficina para Seleção de Prioridades PPSUS 2012/2013. 2013. (Oficina). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>68. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">RAIC - Reunião Anual de Iniciação Científica.Promovido pela VPPLR/Fundação Oswaldo Cruz/RJ. 2013. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>69. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Reunião do Colegiado do MPSF/RENASF.Reunião do Colegiado do MPSF/RENASF. 2013. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>70. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Seminário Temático I  - Saúde Coletiva e o SUS.Programa de Nacional de melhoria do acesso e da qualidade da atenção primária. 2013. (Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>71. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XIII Congresso das Secretarias Municipais de Saúde do Ceará. Desafios da Inovação em Saúde e o Papel do Complexo Produtor da Fiocruz Ceará. 2013. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>72. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">10º Congresso Brasileiro de Saúde Coletiva. Resolutividade de Centros de Especialidades Odontologicas Regionais do Estado do Ceará. 2012. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>73. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">2nd International Public Health Conference &amp; Nineteenth National Public Health Colloquium. ID: 2012/1290. 2012. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>74. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">2nd International Public Health Conference &amp; Nineteenth National Public Health Colloquium. Children, Neglect &amp; Oral Health. 2012. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>75. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">5ª Conferencia Latinoamericana y 4ª Conferencia Internamericana de Promoción de la Saludy Educación para la Salud           de l.Chlid, Negglicence and oral Health. 2012. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>76. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">ESP - CE - Como Profissionalizar os Editores Cieníificos.Editores Científicos da Saúde Coletiva.
# 							2012. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>77. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Fórum de Coordenadores de Pós-Graduação em Saúde Coletiva.Fórum de Coordenadores de Pós-Graduação em Saúde Coletiva. 2012. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>78. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">III Congresso Brasileiro de Odontologia do Trabalho. Estrtura de Serviços Odontológico em Empresas Privadas do Estado do Ceará. 2012. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>79. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">III Congresso Brasileiro de Odontologia do Trabalho. Serviços Odontológicos como Benefício de Empresas Privadas no Estado do Ceará: Satisfação do Trabalhador. 2012. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>80. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">IV Congresso Internacional de Odontologia do Ceará. O Lúdico como Prática Educativa em Saúde Bucal. 2012. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>81. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">IV Congresso Internacional de Odontologia do Ceará. 2012. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>82. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Iv Encontro Científico de Integração da Odontologia.Urgências em Odontopediatria.
# 							2012. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>83. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Oficina de Propriedades de Pesquisa da Secretaria de Gestão Estratégias  as.Oficina de Propriedades de Pesquisa da Secretaria de Gestão Estratégias  as. 2012. (Oficina). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>84. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Reunião da Rede de Pesquisa em Saúde Coletiva - ABRASCO. 2012. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>85. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">V Congresso de Hospital Infantil Albert Sabin. Sedação Consciente x Anestesia Geral. 2012. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>86. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">V Congresso de Hospital Infantil Albert Sabin. Urgências Odontológicas no Paciente Internado. 2012. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>87. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">VI Seminário Internacional de Atenção Básica. 2012. (Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>88. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">58th Annual ORCA Congress. Is dental decay related to child neglect?. 2011. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>89. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">IEA World Congress of Epidemiology. 2011. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>90. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">IV Encontro de Coordenadores de Mestrados Profissionais.
# 							2011. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>91. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Congresso Internacional "Os Tenores com o Malamed no Ceará". 2010. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>92. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">12th World Congress on Public Health. The Oral Health Care System in Brazil. 2009. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>93. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">III Congresso Internacional de Odontologia. 2009. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>94. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">IX Congresso Brasileiro de Saúde Coletiva. Perpectivas e opiniões sobre a qualidade dos serviços da Equipe de Saúde Bucal no PSF. 2009. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>95. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XX COBRAC - Congresso Brasileiro de Cirurgia e Tramatologia Buco-Maxilo-Facial. Coordenadora das Conferências Nacionais Tumores em pacientes pediátricos e Traumatismo Maxilo-Facial Pediátrico. 2009. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>96. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">55th Annual Congress of the European Organization for Caries Research (ORCA). 2008. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>97. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">9º Congresso Brasileiro de Medicina de Família e Comunidade, 2º Congresso Cearense de Medicina de Família e Comunidade e 1º Simpósio Internacional de Ensino Aprendizagem em MFC e APS. 2008. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>98. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">II Fórum de Saúde Coletiva.Mesa Redonda - Interdisciplinariedade na Saúde Coletiva: Desafios e Perspectivas. 2008. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>99. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">IV Congresso Internacional em Avaliação Educacional - Avaliação: perspectivas para a escola contemporânea. 2008. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>100. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XVIII Congresso Mundial de Epidemiologia. 2008. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>101. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">19th World Conference on Health Promotion and Health Education. School Violence: A Problem Compromising a Healthy Environment. 2007. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>102. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">21 Congresso Brasileiro de Odontopediatria. Debatedora no Simpósio sobre técnicas de tratamento pulpar na 1ª infância. 2007. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>103. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">21 Congresso Brasileiro de Odontopediatria. Óxido Nitroso em Odontopediatria. 2007. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>104. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">3º Curso de Odontogeriatria.Sedação em Odontologia.
# 							2007. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>105. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">I Mostra do Sistema Municipal de Saúde Escola de Fortaleza.Saude Bucal em pessoas na terceira idade institucionalizadas. 2007. (Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>106. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">11º Congresso Mundial de Saúde Pública. VIGILÂNCIA SANITÁRIA DAS ÁGUAS EM ÁREA ENDÊMICA DE FLUOROSE DENTAL. 2006. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>107. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">84th General Session and Exhibition of the International Association for Dental Research. Quality of water in an Endemic area of dental fluorosis. 2006. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>108. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">DST/AIDS. 2006. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>109. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">I Encontro Nacional de Educação de Saúde.
# 							2006. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>110. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">II Congresso Internacional de Odontologia. 2006. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>111. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">II Seminario Internacional de Atenção Primária: Saúde da Família.II Seminário Internacional de Atenção Primária: Saúde da Família. 2006. (Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>112. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Mundo Unifor - I Encontro Nacional de Educação de Saúde.
# 							2006. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>113. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Seminário de Informação para Docentes recém-ingressos na UNIFOR. 2006. (Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>114. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Seminário Internacional de Promoção da Saúde. 2006. (Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>115. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">VI Encontro Científico de Educação em Saúde.
# 							2006. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>116. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">X Fórum Brasileiro de Neuropsiquiatria Geriátrica/Curso: Repensar Gerontológico. X Fórum Brasileiro de Neuropsiquiatria Geriátrica/Curso: Repensar Gerontológico:Ressignificando Paradigmas. 2006. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>117. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XII Encontro de Iniciação à Pesquisa.Coordenador de Sessao.
# 							2006. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>118. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XII Encontro de Iniciação à Pesquisa.
# 							2006. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>119. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">83rd General Session &amp; Exhibition IADR. 83rd General Session &amp; Exhibition IADR. 2005. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>120. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Research Day - University of Toronto.Research Day - University of Toronto.
# 							2005. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>121. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">10th Annual Canadian Connective Tissue Conference. 10th Annual Canadian Connective Tissue Conference. 2004. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>122. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">51st ORCA meeting. 2004. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>123. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">82nd General Session &amp; Exhibition IADR. 2004. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>124. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Research Day - University of Toronto.
# 							2004. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>125. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">50th  ORCA meeting. 2003. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>126. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">81st General Session &amp; Exhibition IADR. 2003. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>127. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">AADR/CADR nnual Meeting. 2003. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>128. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">IADR Unilever Travel Award. 2003. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>129. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Research Day - University of Toronto.
# 							2003. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>130. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">20º Congresso Internacional de Odontologia de SP. 2002. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>131. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">AADR/CADR nnual Meeting. 2002. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>132. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Bone and Mineral Research Day.
# 							2002. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>133. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Gordon Research Conference. 2002. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>134. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Research Day - University of Toronto.
# 							2002. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>135. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">British Society of Paediatric Dentistry Annual Meeting.
# 							2000. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>136. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XIII Congresso Cearense de Odontologia. 1999. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>137. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">18o Congresso Internacional de Odontologia de SP. 1998. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>138. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">VII Encontro de Extensão - UFC.
# 							1998. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>139. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XII Congresso Cearense de Odontologia. 1997. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>140. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XXX Congresso Brasileiro de Odontologia. 1997. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>141. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XXXIII Encontro Grupo Brasieliro Materiais Dentários.
# 							1997. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>142. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">17o Congresso Internacional de Odontologia de SP. 1996. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>143. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XIII Congresso Pernambucano de Odontologia e VIII Congresso Norte-Nordeste de Odontologia. 1996. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>144. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XV Encontro Universitário de Iniciação a Pesquisa.
# 							1996. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>145. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XVII Congresso Brasileiro de Anatomia. 1996. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>146. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XXVIII Congresso Brasileiro de Odontologia. 1996. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>147. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XXXII Encontro Grupo Brasieliro Materiais Dentários.
# 							1996. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>148. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">I Jornada em Clínicas Integradas. 1995. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>149. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XI Congresso Cearense de Odontologia. 1995. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>150. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XXVII Congresso Brasileiro de Odontologia. 1995. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>151. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">I Jornada Científica do Instituto do Câncer do Ceará. 1994. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <a name="OrganizacaoEventos" tabindex="772"></a><br class="clear"><div class="inst_back">
# <b>Organização de eventos, congressos, exposições e feiras</b>
# </div>
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="773">VIEIRA-Meyer, Anya P G F</a></b>; MORAIS, A. P. P. ; MACHADO, MARIA DE FATIMA A S ; GUIMARAES, J. M. X. ; VIEIRA, NEIVA FRANCENELY CUNHA ; CHAGAS, M. I. O. ; NASCIMENTO, E. F. ; FORTE, FRANKLIN DELANO SOARES ; OLIVEIRA, A. L. S. ; FARIAS, S. F. ; DE ARAÚJO DIAS, MARIA SOCORRO ; CAMPELO, ISABELLA LIMA BARBOSA ; SOARES, E. P. ; FERREIRA, C. M. ; PEQUENO, A. M. C. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5537817532828798" target="_blank" tabindex="774">VASCONCELOS, M. I. O.</a> ; COSTA, M. S. ; MOREIRA, M. R. C. . Oficina ACS violência e Covid-19. 2022. (Outro). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="775">VIEIRA-MEYER, Anya Pimentel Gomes Fernandes</a></b>. 13º Congresso Brasileiro de Saúde Coletiva. 2022. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="776">Vieira, A P G F</a></b>. 23ª Reunião Anual de iniciação Cientifica - RAIC. 2015. (Outro). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="777">Vieira, A P G F</a></b>. 22ª Reunião Anual de iniciação Científica - RAIC. 2014. (Outro). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="778">Vieira, A P G F</a></b>. Coordenadora da 20ª Reunião Anual de Iniciação Científica. 2012. (Outro). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="779">Vieira, A P G F</a></b>. Coordenadora do Projeto Avaliação do Programa de Melhoria do Acesso e da Qualidade da Atenção Básica. 2012. (Outro). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="780">Vieira, A P G F</a></b>. Reunião da Rede de Pesquisa em Saúde Coletiva. 2012. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="781">Vieira, A P G F</a></b>. Coordenadora da Conferência Nacional: Traumatismo maxilofacial pediátrico no XX Congresso Brasileiro de Cirurgia e Traumatologia Buco-Maxilo-Faciais. 2009. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="782">Vieira, A P G F</a></b>. Coordenadora da Conferência Nacional: Tumores em pacientes pediátricos no XX Congresso Brasileiro de Cirurgia e Traumatologia Buco-Maxilo-Faciais. 2009. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>10. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="783">Vieira, A P G F</a></b>. Coordenadora do I Seminário Internacional de Promoção da Saúde. 2006. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>11. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="784">Vieira, A P G F</a></b>. Coordenador de sessão do XII Encontro de Iniciação à Pesquisa/UNIFOR. 2006. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="Orientacoes" tabindex="785">
# <h1 tabindex="0">Orientações</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <a name="Orientacaoemandamento" tabindex="786"></a><br class="clear"><div class="inst_back">
# <b>Orientações e supervisões em andamento</b>
# </div>
# <div class="cita-artigos">
# <b>Dissertação de mestrado</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Flávia Marília Azeredo de Oliveira. O uso de homeopatia para controle de ansiedade odontológica.
# 						Início: 2019.
# 						Dissertação (Mestrado em Ciências Odontológicas)  - Centro Universitário Christus. (Orientador). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b>Supervisão de pós-doutorado</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">José Maria Ximenes Guimarães.
# 						Início: 2020. FIOCRUZ/CEARÁ. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Ana Patricia Pereira Morais.
# 						Início: 2019. FIOCRUZ/CEARÁ. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b>Iniciação científica</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Thaís dos Santos Moreira. TRABALHO DOS AGENTES COMUNITÁRIOS DE SAÚDE NO CONTEXTO DA PANDEMIA POR COVID-19: REORGANIZAÇÃO DAS PRÁTICAS NOS TERRITÓRIOS DE VULNERABILIDADE SOCIAL. Início: 2021 - Fiocruz Ceara, Fundação Cearense de Apoio ao Desenvolvimento Científico e Tecnológico. (Orientador). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Ana Karoline do Nascimento Sales. VIOLÊNCIA NO TERRITÓRIO DO AGENTE COMUNITÁRIO DE SAÚDE: SENTIDOS E REPERCUSSÕES NO PROCESSO DE TRABALHO. Início: 2021 - Fiocruz Ceara, Fundação Cearense de Apoio ao Desenvolvimento Científico e Tecnológico. (Orientador). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">MAYARA AURELIANO RODRIGUES. COVID-19 E A VIOLÊNCIA URBANA: REPERCUSSÕES NO TRABALHO E NA SAÚDE MENTAL DE AGENTES COMUNITÁRIOS DE SAÚDE. Início: 2021 - Fiocruz Ceara, Iniciacao a Pesquisa. (Orientador). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <a name="Orientacoesconcluidas" tabindex="787"></a><br class="clear"><div class="inst_back">
# <b>Orientações e supervisões concluídas</b>
# </div>
# <div class="cita-artigos">
# <b>Dissertação de mestrado</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Nelsimar Bezerra Santos da Silva. Percepção e disponibilidade ao atendimento de pacientes especiais por graduandos em odontologia no município de Fortaleza-CE.
# 							2023.
						
# 					Dissertação  (Mestrado em Ciências Odontológicas)  - Centro Universitário Christus, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Angelica Ribeiro Gurgel Barreira. Percepção dos profissionais da Estratégia Saúde Da Família sobre os possíveis impactos do Complexo Industrial e Portuário do Pecém na saúde da população.
# 							2023.
						
# 					Dissertação  (Mestrado em Mestrado Profissional em Saúde da Família)  - Fundação Oswaldo Cruz, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">ERLEMUS PONTE SOARES. REPERCUSSÃO DA COVID-19 NA ATENÇÃO PRIMÁRIA À SAÚDE NA ESTRATÉGIA SAÚDE DA FAMÍLIA NO MUNICÍPIO DE FORTALEZA.
# 							2023.
						
# 					Dissertação  (Mestrado em Mestrado Profissional em Saúde da Família - MPSF/RENASF)  - Fundação Oswaldo Cruz, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Gilmara Maria Batista Tavares da Silva. Qualidade do pré-natal e seu impactos na redução da sífilis congênita no município de Fortaleza.
# 							2023.
						
# 					Dissertação  (Mestrado em Mestrado Profissional em Saúde da Família - MPSF/RENASF)  - Fundação Oswaldo Cruz, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">PAULA GUERRA DA ROCHA. PREVALÊNCIA DE LESÕES CERVICAIS NÃO CARIOSAS E RECESSÃO GENGIVAL EM SUBPOPULAÇÕES DE GUINÉ-BISSAU.
# 							2023.
						
# 					Dissertação  (Mestrado em Ciências Odontológicas)  - Centro Universitário Christus, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Elaine Lopes Pedrosa Bomfim. LINHAS DE CUIDADO AO PACIENTE COM HANSENÍASE: AÇÕES REALIZADAS POR MÉDICOS E CIRURGIÕES-DENTISTAS DA ESTRATÉGIA SAÚDE DA FAMÍLIA.
# 							2021.
						
# 					Dissertação  (Mestrado em Mestrado Profissional em Saúde da Família - ProfSaúde)  - FIOCRUZ/CEARÁ, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-lattes" target="_blank" href="http://lattes.cnpq.br/0040414833645815" tabindex="788"><img src="images/curriculo/logolattes.gif"></a>Maira Barbosa Coutinho. A influencia de vídeos de antecipação de situação no controle dos níveis de ansiedade em pacientes odontopediatricos.
# 							2020.
						
# 					Dissertação  (Mestrado em Ciências Odontológicas)  - Centro Universitário Christus, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">ANTONIO LEONEL DE LIMA JÚNIOR. A PERCEPÇÃO DOS MÉDICOS EM RELAÇÃO AS FORMAS DE CONTRATAÇÃO DESTA CATEGORIA NA ESTRATÉGIA DE SAÚDE DA FAMÍLIA.
# 							2019.
						
# 					Dissertação  (Mestrado em Mestrado Profissional em Saúde da Família - ProfSaúde)  - FIOCRUZ/CEARÁ, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Helena Paula Guerra dos Santos. COMO AS GESTANTES BENEFICIÁRIAS DO BOLSA FAMÍLIA PERCEBEM O PROGRAMA E SUAS REPERCUSSÕES NA SAÚDE E NA VIOLÊNCIA?.
# 							2019.
						
# 					Dissertação  (Mestrado em Mestrado Profissional em Saúde da Família)  - FIOCRUZ/CEARÁ, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>10. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">CARLA MANUELA RODRIGUES NOGUEIRA. A NOTIFICAÇÃO COMPULSÓRIA DE DOENÇAS OU AGRAVOS NO ÂMBITO DAS UNIDADES DE ATENÇÃO PRIMÁRIA À SAÚDE DO MUNICÍPIO DE FORTALEZA-CE: O CASO DOS ATENDIMENTOS ANTIRRÁBICOS.
# 							2016.
						
# 					Dissertação  (Mestrado em Mestrado Profissional em Saúde da Família)  - FIOCRUZ/CEARÁ, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>11. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-lattes" target="_blank" href="http://lattes.cnpq.br/5831281973848456" tabindex="789"><img src="images/curriculo/logolattes.gif"></a>LUIZA DE PAULA SOUSA. ATENÇÃO EM SAÚDE BUCAL OFERECIDA AOS PACIENTES COM ESQUIZOFRENIA NA ESTRATÉGIA SAÚDE DA FAMÍLIA.
# 							2016.
						
# 					Dissertação  (Mestrado em Mestrado Profissional em Saúde da Família)  - FIOCRUZ/CEARÁ, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>12. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">LEONARDO BARBOSA ROLIM. AVALIAÇÃO DOS ATRIBUTOS DA ATENÇÃO PRIMÁRIA À SAÚDE DE FORTALEZA-CE: estudo através do PCA Tool-Brasil, versão profissionais.
# 							2016.
						
# 					Dissertação  (Mestrado em Mestrado Profissional em Saúde da Família)  - FIOCRUZ/CEARÁ, . Coorientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>13. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-lattes" target="_blank" href="http://lattes.cnpq.br/6311072178482976" tabindex="790"><img src="images/curriculo/logolattes.gif"></a>Lizaldo Andrade Maia. Atenção à Saúde Bucal: arranjos e dispositivos operados no cuidado às pessoas que vivem com HIV/AIDS em Fortaleza-CE.
# 							2014.
						
# 					Dissertação  (Mestrado em Mestrado Profissional em Saúde da Família)  - Fundação Oswaldo Cruz, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>14. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Mailson Fontes de Carvalho. Programa da Melhoria do Acesso e da Qualidade da Atenção básica em Municípios do Piauí e Ceará.
# 							2014.
						
# 					Dissertação  (Mestrado em Mestrado Profissional em Saúde da Família)  - Fundação Oswaldo Cruz, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>15. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-lattes" target="_blank" href="http://lattes.cnpq.br/4652230653044815" tabindex="791"><img src="images/curriculo/logolattes.gif"></a>Krishna Andreia Feitosa Petrola. Ações de Saúde Bucal em Crianças do Programa Bolsa Família.
# 							2014.
						
# 					Dissertação  (Mestrado em Mestrado Profissional em Saúde da Família)  - Fundação Oswaldo Cruz, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>16. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-lattes" target="_blank" href="http://lattes.cnpq.br/4652230653044815" tabindex="792"><img src="images/curriculo/logolattes.gif"></a>Krishna Andréia Feitosa Petrola. Saúde Bucal na População Infantil.
# 							2012.
						
# 					Dissertação  (Mestrado em Mestrado Profissional em Saúde da Família)  - Fundação Oswaldo Cruz, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>17. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Mailson Fontes de Carvalho. Mestrado Profissional.
# 							2012.
						
# 					Dissertação  (Mestrado em Mestrado Profissional em Saúde da Família)  - Fundação Oswaldo Cruz, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>18. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Lyana Leal Rocha. Acessibilidade aos serviços públicos odontológico às pessoas com deficiencia em Fortaleza - CE.
# 							2012.
						
# 					Dissertação  (Mestrado em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza, . Coorientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>19. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-lattes" target="_blank" href="http://lattes.cnpq.br/5516641709622899" tabindex="793"><img src="images/curriculo/logolattes.gif"></a>Davi Oliveira Bizerril. Resolutividade dos Centros de Especialidades Odontologicas Regionais do Estado do Ceará.
# 							2011.
						
# 					Dissertação  (Mestrado em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza, Conselho Nacional de Desenvolvimento Científico e Tecnológico. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>20. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-lattes" target="_blank" href="http://lattes.cnpq.br/9886955951397288" tabindex="794"><img src="images/curriculo/logolattes.gif"></a>Caroline Barbosa Lourenço. Criança, negligência e saúde bucal.
# 							2011.
						
# 					Dissertação  (Mestrado em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>21. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Caroline Antero Machado. Atuação dos cirurgiões-dentistas na Equipe de Saúde Bucal.
# 							2010.
						
# 					Dissertação  (Mestrado em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>22. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-lattes" target="_blank" href="http://lattes.cnpq.br/4330252274453693" tabindex="795"><img src="images/curriculo/logolattes.gif"></a>Maria Elisa Machado Ferreira Marcelo. Avaliação da efetividade de um programa de atenção em saúde bucal com mães e bebês no serviço público.
# 							2009.
						
# 					Dissertação  (Mestrado em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>23. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-lattes" target="_blank" href="http://lattes.cnpq.br/7804203446863704" tabindex="796"><img src="images/curriculo/logolattes.gif"></a>José Bezerra Goés Neto. Prevalência de dor na coluna e alterações de postura em estudantes de Odontologia do Estado do Ceará.
# 							2009.
						
# 					Dissertação  (Mestrado em Saúde coletiva)  - Universidade de Fortaleza, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>24. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-lattes" target="_blank" href="http://lattes.cnpq.br/8226723633317599" tabindex="797"><img src="images/curriculo/logolattes.gif"></a>Paula Ventura da Silveira. Serviço odontológico em empresas privadas no Estado do Ceará.
# 							2009.
						
# 					Dissertação  (Mestrado em Saúde coletiva)  - Universidade de Fortaleza, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>25. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-lattes" target="_blank" href="http://lattes.cnpq.br/4573669416969289" tabindex="798"><img src="images/curriculo/logolattes.gif"></a>Carlos Santos de Castro Filho. Saúde Bucal e qualidade de vida de trabalhadores com e sem assistência odontológica.
# 							2008.
						
# 					Dissertação  (Mestrado em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza, Fundação Cearense de Apoio ao Desenvolvimento Científico e Tecnológico. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>26. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-lattes" target="_blank" href="http://lattes.cnpq.br/3992942738436467" tabindex="799"><img src="images/curriculo/logolattes.gif"></a>Regina Flávia Castro Almeida. Avaliação da informação oral e escrita no conhecimento dos usuários sobre medicamentos anti-retrovirais.
# 							2007.
						
# 					Dissertação  (Mestrado em Saúde Coletiva)  - Universidade de Fortaleza, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>27. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-lattes" target="_blank" href="http://lattes.cnpq.br/0937375336877981" tabindex="800"><img src="images/curriculo/logolattes.gif"></a>Carlos Clayton Tôrres Aguiar. Análise da qualidade de vida, depressão e saúde oral em pacientes diabéticos idosos.
# 							2007.
						
# 					Dissertação  (Mestrado em Mestrado em Saúde Coletiva)  - Universidade de Fortaleza, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>28. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-lattes" target="_blank" href="http://lattes.cnpq.br/1875665907908151" tabindex="801"><img src="images/curriculo/logolattes.gif"></a>Samia Mustafa Aguiar. Prevalência de estresse e depressão em estudantes de medicina e odontologia.
# 							2007.
						
# 					Dissertação  (Mestrado em Saúde Coletiva)  - Universidade de Fortaleza, . Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>29. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-lattes" target="_blank" href="http://lattes.cnpq.br/3131313477182583" tabindex="802"><img src="images/curriculo/logolattes.gif"></a>Ruth Cavalcanti Nunes. Utilização do Lúdico como estratégia na promoção da saúde bucal.
# 							2006.
						
# 					Dissertação  (Mestrado em Saúde Coletiva)  - Universidade de Fortaleza, Fundação Cearense de Apoio ao Desenvolvimento Científico e Tecnológico. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b>Supervisão de pós-doutorado</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Isabella Lima Campelo.
# 							2023. Fundacao Oswaldo Cruz, . Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Paola Gondim Calvasina.
# 							2015. FIOCRUZ/CEARÁ, Coordenação de Aperfeiçoamento de Pessoal de Nível Superior. Anya Pimentel Gomes Fernandes Vieira Meyer. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b>Monografia de conclusão de curso de aperfeiçoamento/especialização</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Luiza de Paula Sousa.
# 							Acesso do Paciente Especial  à clínica odontológica do SUS: A ótica do cirurgião dentista.
# 							2011.
# 							Monografia. (Aperfeiçoamento/Especialização em REsidência Multiprofissional em Saúde da FAmília e)  - Prefeitura Municipal de Fortaleza. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Fernando Côrte Pinheiro de Sousa.
# 							AVALIAÇÃO DA EQUIPE DE SAÚDE BUCAL NAS REGIONAIS I, II E V NO MUNICÍPIO DE FORTALEZA.
# 							2009.
# 							Monografia. (Aperfeiçoamento/Especialização em Curso de Especialização em Saúde da Família e Comu)  - Universidade Estadual do Ceará. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Krishna Andréia Feitosa Petrola.
# 							AVALIAÇÃO DA EQUIPE DE SAÚDE BUCAL NAS SECRETARIAS EXECUTIVAS REGIONAIS III, IV &amp; VI DO MUNICÍPIO DE FORTALEZA.
# 							2009.
# 							Monografia. (Aperfeiçoamento/Especialização em Curso de Especialização em Saúde da Família e Comu)  - Universidade Estadual do Ceará. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Suzana Rachel de Oliveira.
# 							Preparo e disponibilidade dos Odontopediatras do Município de Fortaleza em relação ao atendimento odontológico de crianças HIV positivas/AIDS.
# 							2007.
# 							Monografia. (Aperfeiçoamento/Especialização em Especialização em odontopediatria)  - Associacao Brasileira de Odontologia. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Jamile Pinheiro Karbage.
# 							Prevalência de cáries em crianças institucionalizadas e não institucionalizadas, com e sem acesso à atenção odontológica.
# 							2007.
# 							Monografia. (Aperfeiçoamento/Especialização em Odontopediatria)  - Universidade Estadual Vale do Acaraú. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Lyssa de Aragão Arrais.
# 							Disponibilidade ao atendimento de pacientes especiais pelos odontopediatras da cidade de Fortaleza-CE.
# 							2007.
# 							Monografia. (Aperfeiçoamento/Especialização em Especialização em Odontopediatria)  - Associação Brasileira de Odontologia. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b>Trabalho de conclusão de curso de graduação</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Tanandra de Fátima Rodrigues Duarte.
# 							Processo de Trabalho e saúde mental dos ACS da regional II em tempos de COVID-19.
# 							2021.
# 							Trabalho de Conclusão de Curso. (Graduação em odontologia)  - Centro Universitário Christus. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">LIA DE VASCONCELOS ROCHA.
# 							ATENDIMENTO ODONTOLÓGICO A PACIENTE COM SÍNDROME DE GOLTZ COM ACHADOS CLÍNICOS RAROS ? RELATO DE CASO.
# 							2020.
# 							Trabalho de Conclusão de Curso. (Graduação em odontologia)  - Centro Universitário Christus. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Marjorie Luiza Oliveira de Melo.
# 							DISPONIBILIDADE AO ATENDIMENTO DE PACIENTES ESPECIAIS PELOS ODONTOPEDIATRAS CEARENSES.
# 							2019.
# 							Trabalho de Conclusão de Curso. (Graduação em odontologia)  - Centro Universitário Christus. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">DANIELA ROCHA DE ALENCAR.
# 							Avaliação da equipe de saúde bucal no município de Fortaleza segundo a ótica dos gestores.
# 							2010.
# 							Trabalho de Conclusão de Curso. (Graduação em Odontologia)  - Universidade de Fortaleza. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">EMANUELLE CRAVEIRO DE MELO.
# 							Avaliação da equipe de saúde bucal no município de Fortaleza segundo a ótica dos gestores.
# 							2010.
# 							Trabalho de Conclusão de Curso. (Graduação em Odontologia)  - Universidade de Fortaleza. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">LIANA ROMERO MIRANDA.
# 							Avaliação da equipe de saúde bucal no município de Fortaleza segundo a ótica dos gestores.
# 							2010.
# 							Trabalho de Conclusão de Curso. (Graduação em Odontologia)  - Universidade de Fortaleza. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Fernando Augusto Góes Mota.
# 							Satisfação dos usuários do SUS em relação ás atividades de saúde na regional VI no município de Fortaleza.
# 							2009.
# 							Trabalho de Conclusão de Curso. (Graduação em Odontologia)  - Universidade de Fortaleza. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Davi de Oliveira Bizerril.
# 							Saúde bucal em pessoas da terceira idade.
# 							2009.
# 							Trabalho de Conclusão de Curso. (Graduação em Odontologia)  - Universidade de Fortaleza. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Rodolpho Luiz Batista Couras.
# 							Percepção dos profissionais do Programa Saúde da Família sobre a inclusão do farmacêutico no NASF.
# 							2009.
# 							Trabalho de Conclusão de Curso. (Graduação em Graduação em Farmácia)  - Universidade de Fortaleza. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>10. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Evanildo Henrique Macêdo da Costa.
# 							Autopercepção da condição de saúde bucal em idosos institucionalizados e não institucionalizados.
# 							2006.
# 							Trabalho de Conclusão de Curso. (Graduação em Odontologia)  - Universidade de Fortaleza. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b>Iniciação científica</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Tanandra de Fátima Rodrigues Duarte.
# 							VIOLÊNCIA URBANA: REPERCUSSÕES NO TRABALHO E NA SAÚDE MENTAL DE AGENTES COMUNITÁRIOS DE SAÚDE.
# 							2021.
# 							Iniciação Científica. (Graduando em odontologia)  - Centro Universitário Christus,
# 							Fundação Cearense de Apoio ao Desenvolvimento Científico e Tecnológico. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Antônio Diego Costa Bezerra.
# 							ANÁLISE DA SAÚDE DE MULHERES PRIVADAS DE LIBERDADE PORTADORAS DE DOENÇAS CRÔNICAS NÃO TRANSMISSÍVEIS E TUBERCULOSE NA PENITENCIÁRIA CEARENSE Ano II.
# 							2021.
# 							Iniciação Científica - Fundação Oswaldo Cruz,
# 							CNPq. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Antônio Diego Costa Bezerra.
# 							ANÁLISE DA SAÚDE DE MULHERES PRIVADAS DE LIBERDADE PORTADORAS DE DOENÇAS CRÔNICAS NÃO TRANSMISSÍVEIS E TUBERCULOSE NA PENITENCIÁRIA CEARENSE I.
# 							2020.
# 							Iniciação Científica - Fundação Oswaldo Cruz,
# 							CNPq. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Isabella Fechine de Oliveira.
# 							Percepções das gestantes beneficiarias do PBF sobre as condicionalidades de saúde e suas repercussoes na regional II.
# 							2019.
# 							Iniciação Científica. (Graduando em odontologia)  - Centro Universitário Christus. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Sara Viana Souza Fontenele.
# 							A influência do Programa Bolsa Família e da Estratégia da Saúde da Família  na violência doméstica e comunitária vivenciada na primeira infância.
# 							2018.
# 							Iniciação Científica. (Graduando em odontologia)  - Centro Universitário Christus,
# 							CNPq. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Erian Bezerra Belo de Codes.
# 							A influência do Programa Bolsa Família e da Estratégia da Saúde da Família  na violência doméstica e comunitária vivenciada na primeira infância.
# 							2017.
# 							Iniciação Científica. (Graduando em odontologia)  - Centro Universitário Christus,
# 							Coordenação de Aperfeiçoamento de Pessoal de Nível Superior. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Cassia Emanuella Nóbrega Malta.
# 							A influência do Programa Bolsa Família e da Estratégia da Saúde da Família  na violência doméstica e comunitária vivenciada na primeira infância.
# 							2017.
# 							Iniciação Científica. (Graduando em odontologia)  - Centro Universitário Christus,
# 							Coordenação de Aperfeiçoamento de Pessoal de Nível Superior. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Julia Queiroz.
# 							A influência do Programa Bolsa Família e da Estratégia da Saúde da Família  na violência doméstica e comunitária vivenciada na primeira infância.
# 							2017.
# 							Iniciação Científica. (Graduando em odontologia)  - Centro Universitário Christus,
# 							Coordenação de Aperfeiçoamento de Pessoal de Nível Superior. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Jamille Barreto Oliveira.
# 							O Programa Bolsa Família e Saúde da Família: violência doméstica.
# 							2016.
# 							Iniciação Científica. (Graduando em Odontologia)  - Universidade de Fortaleza,
# 							Coordenação de Aperfeiçoamento de Pessoal de Nível Superior. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>10. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Mateus Mota Campos.
# 							A influência do Programa Bolsa Família e Estratégia Saúde da Familia na  violência doméstica vivenciada na primeira infância.
# 							2016.
# 							Iniciação Científica. (Graduando em Odontologia)  - Universidade de Fortaleza,
# 							CNPq. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>11. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Danielle Costa Ferreira Bringel.
# 							A influência do Programa Bolsa Família e da Estratégia da Saúde da Família  na violência doméstica e comunitária vivenciada na primeira infância.
# 							2016.
# 							Iniciação Científica. (Graduando em Odontologia)  - Universidade de Fortaleza,
# 							Fundação Cearense de Apoio ao Desenvolvimento Científico e Tecnológico. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>12. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Anderson Maia Meneses.
# 							Avaliação da ESF no municipio de Fortaleza.
# 							2011.
# 							Iniciação Científica. (Graduando em Odontologia)  - Universidade de Fortaleza,
# 							Cnpq. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>13. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Camila Gonçalves Monteiro.
# 							estudo das propriedades antigênicas de peptídeos derivados de Mycobacterium leprae: ênfase em desenvolvimento de métodos para diagnósticos.
# 							2011.
# 							Iniciação Científica - Fundação Oswaldo Cruz. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>14. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Simone Mota de Holanda Lima.
# 							Avaliação da ESF no municipio de Fortaleza.
# 							2010.
# 							Iniciação Científica. (Graduando em Odontologia)  - Universidade de Fortaleza,
# 							Cnpq. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>15. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Daniella Rocha.
# 							Avalação do PSF no município de Fortaleza.
# 							2009.
# 							Iniciação Científica. (Graduando em Odontologia)  - Universidade de Fortaleza,
# 							Fundação Cearense de Apoio ao Desenvolvimento Científico e Tecnológico. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>16. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Davi Oliveira Bizerril.
# 							Utilização do Lúdico como estratégia na promoção da saúde.
# 							2008.
# 							Iniciação Científica. (Graduando em Odontologia)  - Universidade de Fortaleza,
# 							Cnpq. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>17. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Davi Oliveira Bizerril.
# 							Saúde bucal do idoso.
# 							2007.
# 							Iniciação Científica - Universidade de Fortaleza. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>18. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Juliana Paiva Prado.
# 							Satisfação do usuário em relação a Equipe de Saúde Bucal no município de Fortaleza.
# 							2007.
# 							Iniciação Científica. (Graduando em Odontologia)  - Universidade de Fortaleza. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>19. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">José Welley Gonçãlves de Oliveira.
# 							Percepção e satisfação do SUS em relação as atividades de equipe de saúde bucal.
# 							2007.
# 							Iniciação Científica. (Graduando em Odontologia)  - Universidade de Fortaleza. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>20. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Davi Oliveira Bizerril.
# 							Satisfação do usuário em relação a Equipe de Saúde Bucal no município de Fortaleza.
# 							2007.
# 							Iniciação Científica. (Graduando em Odontologia)  - Universidade de Fortaleza. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b>Orientações de outra natureza</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Maria Elisa Machado F. Marcelo e Paula Ventura da Silveira.
# 							Saúde bucal de mães e filhos com acompanhamento odontológico pré-natal e puerperal pela estratégia Saúde da Família.
# 							2009.
# 							Orientação de outra natureza. (Odontologia)  - Universidade de Fortaleza. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Maria Elisa Machado F. Marcelo e Paula Ventura da Silveira.
# 							Serviço odontológico como benefício de empresas privadas no Ceará: satisfação do usuário..
# 							2009.
# 							Orientação de outra natureza. (Odontologia)  - Universidade de Fortaleza. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Carlos Castro Filho.
# 							Percepção de saúde bucal e qualidade de vida.
# 							2008.
# 							Orientação de outra natureza. (Saúde Coletiva)  - Universidade de Fortaleza. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Juliana Paiva,Davi Bizerril,José W. Gonçalves,Carlos Santos.
# 							Satisfação do usuário do SUS em relação à eficácia, eficiência e efetividade das atividades da equipe de Saúde Bucal.
# 							2008.
# 							Orientação de outra natureza. (Odontologia)  - Universidade de Fortaleza. Orientador: Anya Pimentel Gomes Fernandes Vieira Meyer.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="PotencialInovacao" tabindex="803">
# <h1 tabindex="0">Inovação</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <a name="ProgramaComputadorSemRegistroPI" tabindex="804"></a><br class="clear"><div class="inst_back">
# <b>Programa de computador sem registro</b>
# </div>
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="805">Vieira-Meyer, Anya PGF</a></b>. Olá Dentista!. 2019. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <a name="OutrosProjetosPI" tabindex="806"></a><br class="clear"><div class="inst_back">
# <b>Outros projetos</b>
# </div>
# <div class="layout-cell layout-cell-12 data-cell"><a name="PP_Nos APS Brasil" tabindex="807"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2019 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Nos APS Brasil<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: Participação no Grupo de pesquisa Nós APS (Atenção Primária à Saúde) Brasil, grupo colaborativo que tem como interesse o entendimento do processo de trabalho dos atores envolvidos na Estratégia de Saúde da Família (ESF), principalmente dos Agentes Comunitários de Saúde (ACS), e a influência de estressores, como a Violência, a COVID-19, na organização do serviço, na labuta e saúde dos trabalhadores. O grupo foi constituído em 2020, a partir do encontro de pesquisadores de duas redes pré-existentes: a Rede Nordeste de Formação em Saúde da Família (RENASF) e a Rede de unidades da Fiocruz no Nordeste. As investigações iniciaram-se com a busca pela compreensão dos efeitos da violência urbana no cotidiano dos ACS, entendendo-se que, por serem linha de frente dos serviços de saúde da atenção primária no país, esses profissionais, que vivem e trabalham no território, tem o potencial de melhor compreender o impacto da violência urbana nos serviços de saúde e na atenção à saúde da população. O grupo tem também atuado em estudo com outros profissionais de saúde e investigando a resiliência dos Sistemas de Saúde na atenção primária. Ao compreender a realidade dos trabalhadores da atenção primária e dos territórios por eles assistidos, o grupo objetiva colaborar para o enfrentamento dos desafios encontrados. Como lócus de trabalho tem se priorizado municípios do Nordeste brasileiro (capitais e cidades do interior), coletando e analisando, desde 2019, dados quantitativos e qualitativos nas cidades de Fortaleza/CE, Recife/PE, Teresina/Pi, João Pessoa/PB Sobral/CE, Juazeiro do Norte/CE, Barbalha/CE e Crato/CE.. <br class="clear">Situação: Em andamento; Natureza: Outra. <br class="clear"><br class="clear">Integrantes: Anya Pimentel Gomes Fernandes Vieira Meyer - Coordenador / Neiva Francenely Cunha Vieira - Integrante / Andrea Silvia Walter de Aguiar - Integrante / Ana Patricia Pereira Morais - Integrante / Maristela Inês Osawa Chagas - Integrante / José Maria Ximenes Guimarães - Integrante / Alice Maria Correia Pequeno - Integrante / CORREIA SAMPAIO, YANA PAULA COÊLHO - Integrante / FORTE, FRANKLIN DELANO SOARES - Integrante / Ardigleusa Alves Coelho - Integrante / DE ARAÚJO DIAS, MARIA SOCORRO - Integrante / MACHADO, MARIA DE FATIMA A S - Integrante / Isabella Lima Barbosa Campelo - Integrante / André Luiz Sá Oliveira - Integrante / SILVA JÚNIOR, FERNANDO JOSÉ GUEDES DA - Integrante / FERREIRA, REGINA GLAUCIA LUCENA AGUIAR - Integrante / ALBUQUERQUE, GRAYCE ALENCAR - Integrante / FARIAS, SIDNEY FEITOZA - Integrante / NASCIMENTO, ELAINE FERREIRA DO - Integrante.<br class="clear">Financiador(es): Fundação Oswaldo Cruz - Auxílio
# 										financeiro / Fundação Cearense de Apoio ao Desenvolvimento Científico e Tecnológico - Auxílio
# 										financeiro / Instituto Lemann - Auxílio
# 										financeiro.</div>
# </div>
# </div><br class="clear"></div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="EducacaoPopularizacaoCTA" tabindex="808">
# <h1 tabindex="0">Educação e Popularização de C &amp; T</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <a name="ArtigosEPCTA" tabindex="809"></a><br class="clear"><div class="inst_back">
# <b>Artigos</b>
# </div>
# <div class="cita-artigos">
# <b><a name="ArtigosCompletos" tabindex="810"></a>Artigos completos publicados em periódicos</b>
# </div>
# <br class="clear">
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">MACHADO, MARIA DE FÁTIMA ANTERO SOUSA</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2015</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.7322/jhgd.96709" target="_blank" tabindex="811"></a>MACHADO, MARIA DE FÁTIMA ANTERO SOUSA ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5035858383519296" target="_blank" tabindex="812">GUBERT, FABIANE DO AMARAL</a> ; <b>MEYER, ANYA PIMENTEL G. F. VIEIRA</b> ; CORREIA SAMPAIO, YANA PAULA COÊLHO ; DIAS, MARIA DO SOCORRO DE ARAÚJO ; DE ALMEIDA, ANA MATTOS BRITO DE ALMEIDA BRITO ; MORAIS, ANA PATRICIA PEREIRA ; E SILVA, ANAMARIA CAVALCANTE ; CAMPOS, JOCILEIDE SALES ; CHAGAS, MARISTELA INÊS OSAWA ; CHAVES, EMÍLIA SOARES . THE HEALTH SCHOOL PROGRAMME: A HEALTH PROMOTION STRATEGY IN PRIMARY CARE IN BRAZIL. Journal of Human Growth and Development (Print)<sup></sup>, v. 25, p. 307-312, 2015. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.7322/jhgd.96709&amp;issn=01041282&amp;volume=25&amp;issue=&amp;paginaInicial=307&amp;titulo=THE HEALTH SCHOOL PROGRAMME: A HEALTH PROMOTION STRATEGY IN PRIMARY CARE IN BRAZIL&amp;sequencial=1&amp;nomePeriodico=Journal of Human Growth and Development (Print)" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=84949186653&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">14</span></a></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <a name="ProgramaComputadorEPCTA" tabindex="813"></a><br class="clear"><div class="inst_back">
# <b>Programa de Computador sem registro de patente</b>
# </div>
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="814">Vieira-Meyer, Anya PGF</a></b>. Olá Dentista!. 2019. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <a name="CursosCurtaEPCTA" tabindex="815"></a><br class="clear"><div class="inst_back">
# <b>Cursos de curta duração ministrados</b>
# </div>
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="816">Vieira, A P G F</a></b>. Módulo: Gestão do Processo de Trabalho na Atenção Primária - MPSF/RENASF. 2012. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="817">Vieira, A P G F</a></b>. Módulo: Atenção e Gestão do Cuidado à Saúde da Família - MPSF/RENASF. 2012. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="818">Vieira, A P G F</a></b>. Módulo: Atenção Integral à Saúde da Família - MPSF/RENASF. 2012. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="819">Vieira, A P G F</a></b>. Coordenadora do Módulo: Atenção Integral à Saúde da Família - MPSF/RENASF. 2012. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <a name="MaterialDidaticoEPCTA" tabindex="820"></a><br class="clear"><div class="inst_back">
# <b>Desenvolvimento de material didático ou instrucional</b>
# </div>
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="821">PIMENTEL G. F. VIEIRA-MEYER, ANYA</a></b>; Braga, J U ; Castro, D B . Técnicas de Análise Espacial Aplicadas à Vigilância em Saúde na Atenção Básica.
# 							2019. (Desenvolvimento de material didático ou instrucional - MaterialDidático). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <a name="EntrevistasEPCTA" tabindex="822"></a><br class="clear"><div class="inst_back">
# <b>Entrevistas, mesas redondas, programas e comentários na mídia </b>
# </div>
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7659758489387356" target="_blank" tabindex="823">Vieira-MEYER, A. P. G. F.</a>; TASCHNER, N. P. ; STUENKEL, O. ; ARSENAULT, C. ; <b>Vieira-Meyer, Anya PGF</b> . Brazil has the most confirmed COVID-19 cases in Latin America. 'So what?' says President Bolsonaro. 2020. 
						
# 							(Programa de rádio ou TV/Entrevista).
# 						<img class="icone-video" src="http://estatico.cnpq.br/imagens/ico_pw_video.png" onclick="mostrarPreviaVideo('https://www.cbc.ca/news/world/brazil-has-the-most-confirmed-covid-19-cases-in-latin-america-so-what-says-president-bolsonaro-1.5550720')"></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="OutrasInformacoesRelevantes" tabindex="824">
# <h1 tabindex="0">Outras informações relevantes</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <pre class="outras-informacoes">Outras Atividades:
# * Coordenadora da Comissão de Recepção do XIII Congresso Cearense de Odontologia, realizado no Centro de convenções Edson Queiroz, Maio 1999. Fortaleza - CE.
# * Oradora do corpo discente na cerimônia de graduação realizada na Concha Acústica (Auditório Reitor Antônio Martins Filho) UFC.  Agosto 1998.
# * Coordenadora da Comissão Acadêmica do XII Congresso Cearense de Odontologia, realizado no Centro de convenções Edson Queiroz, Maio 1997. Fortaleza - CE
# * Membro da Comissão Acadêmica do XI Congresso Cearense de Odontologia, realizado no Centro de Convenções Edson Queiroz, Maio 1995. Fortaleza - CE.</pre>
# </div><br class="clear"></div><br class="clear"><br class="clear"><div class="rodape-cv">Página gerada pelo Sistema Currículo Lattes em 30/03/2024 às 22:42:52</div><br class="clear"><div class="control-bar-wrapper"><a href="download.do?metodo=apresentar&amp;idcnpq=7659758489387356" target="_blank" class="button" tabindex="825"><span class="mini-ico mini-ico-down"></span>Baixar Currículo</a><a class="button" href="javascript:window.print()" tabindex="826"><span class="mini-ico mini-ico-list"></span>Imprimir Currículo</a>
# </div><br class="clear"></div>
# </div>
# </div>
# </div>
# </div>
# <div class="to-top-bar" id="toTop" style="display: none;">
# <a href="javascript:void(0)" class="topo-width center" tabindex="827">
# <div class="id-orgao">
# <span>CNPq</span> | Uma agência do Ministério da Ciência, Tecnologia e Inovação</div>
# </a>
# </div>


# <div id="menu-dados-gerais" class="megamenu" _megamenupos="0pos" style="z-index: 1001; display: none;">
		
# 		<div class="column"><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#Identificacao" tabindex="11">Identificação</a></li>
		
	
		
		
# 		<li><a href="#Endereco" tabindex="12">Endereço</a></li>
		
	
		
		
# 		<li><a href="#Idiomas" tabindex="13">Idiomas</a></li>
		
	
		
		
# 		<li><a href="#PremiosTitulos" tabindex="14">Prêmios e títulos</a></li>
		
	
		
		
# 		<li><a href="#OutrasInformacoesRelevantes" tabindex="15">Outras informações relevantes</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div><div id="menu-formacao" class="megamenu" _megamenupos="1pos" style="z-index: 1002; display: none;">
		
# 		<div class="column"><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#FormacaoAcademicaTitulacao" tabindex="17">Formação acadêmica/titulação</a></li>
		
	
		
		
# 		<li><a href="#FormacaoComplementar" tabindex="18">Formação complementar</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div><div id="menu-atuacao" class="megamenu" _megamenupos="2pos" style="z-index: 1013; left: 555.984px; top: 109px; display: none;">
		
# 		<div class="column"><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#AtuacaoProfissional" tabindex="20">Atuação profissional</a></li>
		
	
		
		
# 		<li><a href="#LinhaPesquisa" tabindex="21">Linhas de pesquisa</a></li>
		
	
		
		
# 		<li><a href="#MembroCorpoEditorial" tabindex="22">Membro de corpo editorial</a></li>
		
	
		
		
# 		<li><a href="#RevisorPeriodico" tabindex="23">Revisor de periódico</a></li>
		
	
		
		
# 		<li><a href="#AreasAtuacao" tabindex="24">Áreas de atuação</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div><div id="menu-projetos" class="megamenu" _megamenupos="3pos" style="z-index: 1004; display: none;">
		
# 		<div class="column"><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#ProjetosPesquisa" tabindex="26">Projetos de pesquisa</a></li>
		
	
		
		
# 		<li><a href="#ProjetosExtensao" tabindex="27">Projetos de extensão</a></li>
		
	
		
		
# 		<li><a href="#ProjetosDesenvolvimento" tabindex="28">Projetos de desenvolvimento</a></li>
		
	
		
		
# 		<li><a href="#OutrosProjetos" tabindex="29">Outros projetos</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div><div id="menu-producoes" class="megamenu" _megamenupos="4pos" style="z-index: 1005; display: none;">
		
# 		<div class="column"><div class="subtitMenu"><a href="#ProducaoBibliografica" tabindex="31">Produção Bibliográfica</a></div><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#ArtigosCompletos" tabindex="32">Artigos completos publicados em periódicos</a></li>
		
	
		
		
# 		<li><a href="#ArtigosAceitos" tabindex="33">Artigos aceitos para publicação</a></li>
		
	
		
		
# 		<li><a href="#LivrosCapitulos" tabindex="34">Livros e capítulos</a></li>
		
	
		
		
# 		<li><a href="#TextosJornaisRevistas" tabindex="35">Textos em jornais ou revistas (magazine)</a></li>
		
	
		
		
# 		<li><a href="#TrabalhosPublicadosAnaisCongresso" tabindex="36">Trabalhos publicados em anais de congressos</a></li>
		
	
		
		
# 		<li><a href="#ApresentacaoTrabalho" tabindex="37">Apresentações de trabalho</a></li>
		
	
		
		
# 		<li><a href="#OutrasProducoesBibliograficas" tabindex="38">Partitura musical</a></li>
		
	
		
		
# 		<li><a href="#OutrasProducoesBibliograficas" tabindex="39">Tradução</a></li>
		
	
		
		
# 		<li><a href="#OutrasProducoesBibliograficas" tabindex="40">Prefácio, pósfacio</a></li>
		
	
		
		
# 		<li><a href="#OutrasProducoesBibliograficas" tabindex="41">Outras produções bibliográficas</a></li>
		
# 	</ul></div><div class="column"><div class="subtitMenu"><a href="#ProducaoTecnica" tabindex="42">Produção Técnica</a></div><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#AssessoriaConsultoria" tabindex="43">Assessoria e consultoria</a></li>
		
	
		
		
# 		<li><a href="#SoftwareSemPatente" tabindex="44">Programas de computador sem registro</a></li>
		
	
		
		
# 		<li><a href="#ProdutosTecnologicos" tabindex="45">Produtos tecnológicos</a></li>
		
	
		
		
# 		<li><a href="#ProcessosTecnicas" tabindex="46">Processos e técnicas</a></li>
		
	
		
		
# 		<li><a href="#TrabalhosTecnicos" tabindex="47">Trabalhos técnicos</a></li>
		
	
		
		
# 		<li><a href="#DemaisProducaoTecnica" tabindex="48">Cartas, mapas ou similares</a></li>
		
	
		
		
# 		<li><a href="#DemaisProducaoTecnica" tabindex="49">Curso de curta duração ministrado</a></li>
		
	
		
		
# 		<li><a href="#DemaisProducaoTecnica" tabindex="50">Desenvolvimento de material didático ou instrucional</a></li>
		
	
		
		
# 		<li><a href="#DemaisProducaoTecnica" tabindex="51">Editoração</a></li>
		
	
		
		
# 		<li><a href="#ManutencaoObraArtistica" tabindex="52">Manutenção de obra artística</a></li>
		
	
		
		
# 		<li><a href="#DemaisProducaoTecnica" tabindex="53">Maquete</a></li>
		
	
		
		
# 		<li><a href="#EntrevistasMesasRedondas" tabindex="54">Entrevistas, mesas redondas, programas e comentários na mídia</a></li>
		
	
		
		
# 		<li><a href="#DemaisProducaoTecnica" tabindex="55">Relatório de pesquisa</a></li>
		
	
		
		
# 		<li><a href="#RedesSociais" tabindex="56">Redes sociais, websites e blogs</a></li>
		
	
		
		
# 		<li><a href="#DemaisProducoesTecnicas" tabindex="57">Outras produções técnicas</a></li>
		
# 	</ul></div><div class="column"><div class="subtitMenu"><a href="#ProducaoArtisticaCultural" tabindex="58">Produção Artística/Cultural</a></div><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#ArtesCenicas" tabindex="59">Artes cênicas</a></li>
		
	
		
		
# 		<li><a href="#Musica" tabindex="60">Música</a></li>
		
	
		
		
# 		<li><a href="#ArtesVisuais" tabindex="61">Artes visuais</a></li>
		
	
		
		
# 		<li><a href="#OutrasProducoesArtisticas" tabindex="62">Outras produções artísticas/culturais</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div><div id="menu-potencialInovacao" class="megamenu" _megamenupos="5pos" style="z-index: 1006; display: none;">
		
# 		<div class="column"><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#PatentePI" tabindex="64">Patente</a></li>
		
	
		
		
# 		<li><a href="#ProgramaComputadorRegistradoPI" tabindex="65">Programa de computador registrado</a></li>
		
	
		
		
# 		<li><a href="#CultivarProtegidaPI" tabindex="66">Cultivar protegida</a></li>
		
	
		
		
# 		<li><a href="#CultivarRegistradaPI" tabindex="67">Cultivar registrada</a></li>
		
	
		
		
# 		<li><a href="#DesenhoIndustrialRegistradoPI" tabindex="68">Desenho industrial registrado</a></li>
		
	
		
		
# 		<li><a href="#MarcaRegistradaPI" tabindex="69">Marca registrada</a></li>
		
	
		
		
# 		<li><a href="#TopografiaCircuitoIntegradoRegistradaPI" tabindex="70">Topografia de circuito integrado registrada</a></li>
		
	
		
		
# 		<li><a href="#ProgramaComputadorSemRegistroPI" tabindex="71">Programa de computador sem registro</a></li>
		
	
		
		
# 		<li><a href="#ProdutosPI" tabindex="72">Produtos</a></li>
		
	
		
		
# 		<li><a href="#ProcessosTecnicasPI" tabindex="73">Processos ou técnicas</a></li>
		
	
		
		
# 		<li><a href="#ProjetosPesquisaPI" tabindex="74">Projetos de pesquisa</a></li>
		
	
		
		
# 		<li><a href="#ProjetoDesenvolvimentoTecnologicoPI" tabindex="75">Projeto de desenvolvimento tecnológico</a></li>
		
	
		
		
# 		<li><a href="#ProjetoExtensaoPI" tabindex="76">Projeto de extensão</a></li>
		
	
		
		
# 		<li><a href="#OutrosProjetosPI" tabindex="77">Outros projetos</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div><div id="menu-educacaoPopularizacaoCTA" class="megamenu" _megamenupos="6pos" style="z-index: 1007; display: none;">
		
# 		<div class="column"><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#ArtigosEPCTA" tabindex="79">Artigos</a></li>
		
	
		
		
# 		<li><a href="#LivrosCapitulosEPCTA" tabindex="80">Livros e capítulos</a></li>
		
	
		
		
# 		<li><a href="#TextosJornaisEPCTA" tabindex="81">Textos em jornais de notícias/revistas</a></li>
		
	
		
		
# 		<li><a href="#ApresentacoesTrabalhoEPCTA" tabindex="82">Apresentações de trabalho</a></li>
		
	
		
		
# 		<li><a href="#ProgramaComputadorEPCTA" tabindex="83">Programa de computador sem registro de patente</a></li>
		
	
		
		
# 		<li><a href="#CursosCurtaEPCTA" tabindex="84">Cursos de curta duração ministrados</a></li>
		
	
		
		
# 		<li><a href="#MaterialDidaticoEPCTA" tabindex="85">Desenvolvimento de material didático ou instrucional</a></li>
		
	
		
		
# 		<li><a href="#EntrevistasEPCTA" tabindex="86">Entrevistas, mesas redondas, programas e comentários na mídia</a></li>
		
	
		
		
# 		<li><a href="#ProgramaComputadorRegistradoEPCTA" tabindex="87">Programa de Computador registrado</a></li>
		
	
		
		
# 		<li><a href="#OrganizacaoEventosEPCTA" tabindex="88">Organização de eventos, congressos, exposições e feiras</a></li>
		
	
		
		
# 		<li><a href="#RedesSociaisEPCTA" tabindex="89">Redes sociais, websites e blogs</a></li>
		
	
		
		
# 		<li><a href="#ArtesVisuaisEPCTA" tabindex="90">Artes Visuais</a></li>
		
	
		
		
# 		<li><a href="#ArtesCenicasEPCTA" tabindex="91">Artes Cênicas</a></li>
		
	
		
		
# 		<li><a href="#MusicaEPCTA" tabindex="92">Música</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div><div id="menu-eventos" class="megamenu" _megamenupos="7pos" style="z-index: 1008; display: none;">
		
# 		<div class="column"><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#ParticipacaoEventos" tabindex="94">Participação em eventos, congressos, exposições e feiras</a></li>
		
	
		
		
# 		<li><a href="#OrganizacaoEventos" tabindex="95">Organização de eventos, congressos, exposições e feiras</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div><div id="menu-orientacoes" class="megamenu" _megamenupos="8pos" style="z-index: 1009; display: none;">
		
# 		<div class="column"><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#Orientacoesconcluidas" tabindex="97">Orientações e supervisões concluídas</a></li>
		
	
		
		
# 		<li><a href="#Orientacaoemandamento" tabindex="98">Orientações e supervisões em andamento</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div><div id="menu-bancas" class="megamenu" _megamenupos="9pos" style="z-index: 1010; display: none;">
		
# 		<div class="column"><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#ParticipacaoBancasTrabalho" tabindex="100">Participação em bancas de trabalhos de conclusão</a></li>
		
	
		
		
# 		<li><a href="#ParticipacaoBancasComissoes" tabindex="101">Participação em bancas de comissões julgadoras</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div><div id="menu-citacoes" class="megamenu" _megamenupos="10pos" style="z-index: 1011; display: none;">
		
# 		<div class="column"><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#Citacoes" tabindex="103">Citações no ISI</a></li>
		
	
		
		
# 		<li><a href="#Citacoes" tabindex="104">Citações no SciELO</a></li>
		
	
		
		
# 		<li><a href="#Citacoes" tabindex="105">Citações no SCOPUS</a></li>
		
	
		
		
# 		<li><a href="#Citacoes" tabindex="106">Citações em outras bases bibliográficas</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div><div id="menu-informacoes-complementares" class="megamenu" _megamenupos="11pos" style="z-index: 1012; display: none;">
		
# 		<div class="column"><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="visualizacv.do?metodo=apresentar&amp;id=K4761906A8&amp;tipo=completo&amp;idiomaExibicao=1" tabindex="108">Mostrar informações complementares</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div></body></html>
# '''
```

```python
# html4 = r'''
# <html lang="pt-BR"><head>
# <meta http-equiv="Content-Type" content="text/html; charset=ISO-8859-1">
# <link href="images/v2/fav_ico_lattes.ico" rel="Shortcut Icon">
# <meta content="Sistema de Currículos Lattes 2.0 (CNPq)" http-equiv="generator">
# <meta content="currículo,curriculo,curriculum,cv,vitae,lattes,produção,producao,científica,cientifica,Brasil" http-equiv="keywords">
# <title>Currículo do Sistema de Currículos Lattes (Adriana Costa Bacelo)</title>
# <link type="text/css" rel="stylesheet" href="css/import.css">
# <link type="text/css" href="css/tablet-mobile.css" rel="stylesheet">
# <link type="text/css" rel="stylesheet" href="css/jquery.megamenu.css">
# <link type="text/css" rel="stylesheet" href="css/style_curiculo.css">
# <link type="text/css" rel="stylesheet" href="css/v2/espelho.css">
# <link type="text/css" rel="stylesheet" href="css/impressao.css">
# <script type="text/javascript" async="" src="https://ssl.google-analytics.com/ga.js"></script><script type="text/javascript" src="js/v2/jquery.min.js"></script><script src="js/jquery.cookie.js" type="text/javascript"></script><script src="js/acessibilidade.js" type="text/javascript"></script><script type="text/javascript" src="js/v2/tooltip-position.js"></script><script type="text/javascript" src="js/v2/jquery.tipsy.js"></script><script type="text/javascript" src="js/v2/jquery.megamenu.js"></script><script type="text/javascript" src="js/v2/byOrder.js"></script><script type="text/javascript" src="js/v2/font-size.js"></script><script type="text/javascript" src="js/v2/cv_citacoes.js"></script><script type="text/javascript" src="js/v2/hint_ajax.js"></script><script type="text/javascript" src="js/v2/menuFlowScroll.js"></script><script type="text/javascript" src="js/v2/videoUtils.js"></script><script type="text/javascript">
			
# 			var menu = {
# 				iniciar: function(id) {
# 					jkmegamenu.definemenu("ancora-menu-" + id, "menu-" + id, "focus");
# 				}
# 			};
			
# 			function abreLinkHistoricoPublicacoes(codigo,nome){
#             	var j = window.open("historicoAtualizacao.jsp?id="+codigo+"&nome="+nome+"&fromBusca=1", "historico","resizable=no,menubar=no,toolbar=no,status=yes,scrollbars=yes,width=550px,height=470px,left=0,top=0");
#             	if(j){
#             		j.focus();
#             	} else {
#             		alert('Por favor, desabilite o bloqueador de popups');
#             	}
# 			}
			
# 			var autores = {
# 				exibir: function() {
# 					$(this).hide().parent().find(".autores-et-al").show();
# 				},
# 				ocultar: function() {
# 					$(this).parent().hide().parent().find(".autores-et-al-plus").show();
# 				}
# 			};
				
# 			$(document).ready(function(){
			
# 				$(".ajaxJCR").ajaxJCR();
				
# 				$(".ajaxCAPES").ajaxCAPES();
				
# 				$(".ajaxINPI").ajaxINPI();
				
# 				$("span.citacoes").verificarCitacoes();
				
# 				$(".tooltip").tipsy({gravity: 's'});
				
# 				$(".autores-et-al-plus").click(autores.exibir);
				
# 				$(".autores-et-al-minus").click(autores.ocultar);
        			
#       		});	
      				  
# <!--       		$( function() { -->
# <!-- 			    $( document ).tooltip(); -->
# <!-- 			 } ); -->
      		                                
		
# 	</script><script type="text/javascript">
					
# 					  var _gaq = _gaq || [];
# 					  _gaq.push(['_setAccount', 'UA-35652740-1']);
# 					  _gaq.push(['_trackPageview']);
					
# 					  (function() {
# 					    var ga = document.createElement('script'); ga.type = 'text/javascript'; ga.async = true;
# 					    ga.src = ('https:' == document.location.protocol ? 'https://ssl' : 'http://www') + '.google-analytics.com/ga.js';
# 					    var s = document.getElementsByTagName('script')[0]; s.parentNode.insertBefore(ga, s);
# 					  })();
					
# 					</script>
# </head>
# <body>
# <div class="page min-width">
# <div class="header">
# <div class="header-content max-width">
# <div class="identity center">
# <div class="combo">
# <h1 class="escondido" tabindex="0"> Currículo Lattes - Busca Textual - Visualização do Currículo </h1>
# <a href="http://lattes.cnpq.br/" title="Currículo Lattes" target="_blank" tabindex="1"><img title="Currículo Lattes" alt="Currículo Lattes" src="images/titulo-sistema.png"></a>
# </div>
# <div class="menu-header" style="position: absolute; top: 15px;">
# <span class="tituloFlow"></span><span class="linksFlow"><a href="http://buscatextual.cnpq.br/buscatextual/visualizacv.do?id=K4778130Y3&amp;idiomaExibicao=2" class="btn-br" tabindex="2">English</a><a title="Imprimir currículo" class="bt-menu-header titBottom  icons-top icons-top-print" href="javascript:window.print()" tabindex="3"></a><a title="Aumentar a fonte" class="bt-menu-header titBottom fontMais icons-top icons-top-fontMais" href="javascript:void(0)" tabindex="4"></a><a title="Fonte padrão" class="bt-menu-header titBottom fontMenos icons-top icons-top-fontMenos" href="javascript:void(0)" tabindex="5"></a><a title="Alto contraste" class="bt-menu-header titBottom contraste icons-top icons-top-contraste-en" href="javascript:void(0)" tabindex="6"></a><a href="download.do?metodo=apresentar&amp;idcnpq=8861663346303670" target="_blank" class="bt-menu-header titBottom fontMenos icons-top icons-top-xml" title="Baixar Currículo em XML" tabindex="7"></a><a title="Ajuda" class="bt-menu-header titBottom icons-top icons-top-help" target="_blank" href="http://ajuda.cnpq.br/index.php" tabindex="8"></a></span>
# </div>
# <div class="logo">
# <a target="_blank" href="http://cnpq.br/" tabindex="9">
# <h2 tabindex="0">CNPq - Conselho Nacional de Desenvolvimento Científico e Tecnológico</h2>
# </a>
# </div>
# </div>
# </div>
# </div>
# <div class="navigation-wrapper">
# <div class="center navigation-wrapper-line">
# <div class="menuPrincipal">
# <div class="menuCtl menuAcesso">
# <div class="menucent"><script type="text/javascript">menu.iniciar("dados-gerais");</script>
		
		
# 		<a id="ancora-menu-dados-gerais" href="javascript:void(0)" class="acessibilidade" tabindex="10" _megamenupos="0pos"><span></span>Dados gerais</a>
		
	
		
		
		
		
# 	<script type="text/javascript">menu.iniciar("formacao");</script>
		
		
# 		<a id="ancora-menu-formacao" href="javascript:void(0)" class="acessibilidade separador" tabindex="16" _megamenupos="1pos"><span></span>Formação</a>
		
	
		
		
		
		
# 	<script type="text/javascript">menu.iniciar("atuacao");</script>
		
		
# 		<a id="ancora-menu-atuacao" href="javascript:void(0)" class="acessibilidade separador" tabindex="19" _megamenupos="2pos"><span></span>Atuação</a>
		
	
		
		
		
		
	
		
		
# 		<a id="ancora-menu-projetos" href="#ProjetosPesquisa" class="acessibilidade separador" tabindex="25" _megamenupos="3pos"><span></span>Projetos</a>
		
# 	<script type="text/javascript">menu.iniciar("projetos");</script>
		
		
		
		
# 	<script type="text/javascript">menu.iniciar("producoes");</script>
		
		
# 		<a id="ancora-menu-producoes" href="#ProducoesCientificas" class="acessibilidade separador" tabindex="28" _megamenupos="4pos"><span></span>Produções</a>
		
	
		
		
		
		
# 	<script type="text/javascript">menu.iniciar("potencialInovacao");</script>
		
		
# 		<a id="ancora-menu-potencialInovacao" href="#PotencialInovacao" class="acessibilidade separador" tabindex="61" _megamenupos="5pos"><span></span>Inovação</a>
		
	
		
		
		
		
# 	<script type="text/javascript">menu.iniciar("educacaoPopularizacaoCTA");</script>
		
		
# 		<a id="ancora-menu-educacaoPopularizacaoCTA" href="#EducacaoPopularizacaoCTA" class="acessibilidade separador" tabindex="76" _megamenupos="6pos"><span></span>Educação e Popularização de C &amp; T</a>
		
	
		
		
		
		
# 	<script type="text/javascript">menu.iniciar("eventos");</script>
		
		
# 		<a id="ancora-menu-eventos" href="#Eventos" class="acessibilidade separador" tabindex="91" _megamenupos="7pos"><span></span>Eventos</a>
		
	
		
		
		
		
# 	<script type="text/javascript">menu.iniciar("orientacoes");</script>
		
		
# 		<a id="ancora-menu-orientacoes" href="#Orientacoes" class="acessibilidade separador" tabindex="94" _megamenupos="8pos"><span></span>Orientações</a>
		
	
		
		
		
		
# 	<script type="text/javascript">menu.iniciar("bancas");</script>
		
		
# 		<a id="ancora-menu-bancas" href="#Bancas" class="acessibilidade separador" tabindex="97" _megamenupos="9pos"><span></span>Bancas</a>
		
	
		
		
		
		
# 	<script type="text/javascript">menu.iniciar("citacoes");</script>
		
		
# 		<a id="ancora-menu-citacoes" href="#Citacoes" class="acessibilidade separador" tabindex="100" _megamenupos="10pos"><span></span>Citações</a>
		
	
		
		
		
		
# 	<script type="text/javascript">menu.iniciar("informacoes-complementares");</script>
		
		
# 		<a id="ancora-menu-informacoes-complementares" href="javascript:void(0)" class="acessibilidade separador" tabindex="105" _megamenupos="11pos"><span></span>+</a>
		
	
		
		
		
		
# 	</div>
# </div>
# </div>
# </div>
# </div>
# <div class="content-wrapper">
# <div class="main-content max-width min-width center">
# <div class="layout-cell layout-cell-12">
# <div class="layout-cell-pad-main">
# <img class="foto" src="http://servicosweb.cnpq.br/wspessoa/servletrecuperafoto?tipo=1&amp;id=K4778130Y3"><div class="infpessoa">
# <h2 class="nome" tabindex="0">Adriana Costa Bacelo</h2>
# <ul class="informacoes-autor">
# <li>
# <span class="img_link icone-informacao-autor"></span>Endereço para acessar este CV: http://lattes.cnpq.br/8861663346303670</li>
# <li>
# <span class="img_identy icone-informacao-autor"></span>ID Lattes: <span style="font-weight: bold; color: #326C99;">8861663346303670</span>
# </li>
# <li>
# <span class="img_cert icone-informacao-autor"></span>Última atualização do currículo em 14/03/2024</li>
# </ul>
# </div><br class="clear"><div class="title-wrapper">
# <h1 class="ui-hidden" tabindex="0"></h1>
# <hr class="separator">
# <p class="resumo">Chefe de Gabinete da Fiocruz Ceará, nutricionista, tecnologista saúde pública da Fundação Oswaldo Cruz, atua em função de pesquisa como membro do Laboratório de Redes Integradas e Inteligentes em Saúde (LARIISA) pela Fiocruz Ceará, membro do grupo de pesquisa em Nutrição aplicada a doenças infecciosas do INI-Fiocruz e membro do grupo de pesquisa de micronutrientes do Instituto Josué de Castro  UFRJ. Além de atuar como articuladora de inovação, produção e empreendedorismo do Distrito de Inovação da Fiocruz Ceará. Atua ainda como docente do curso de Pós-Graduação Lato sensu de Nutrição aplicada a doenças infecciosas do INI- Fiocruz (desde fev/2017) e da residência multidisciplinar do INI- Fiocruz (desde fev/2020); como docente do mestrado acadêmico do Programa de Pós-graduação Stricto sensu em Nutrição e Saúde da Universidade Estadual do Ceará - EUCE, do mestrado profissional do programa de pós- graduação Stricto sensu em Saúde da Família  -PROFSAÚDE e do doutorado do Programa de pós-graduação Stricto sensu em Ciências/Biotecnologia e Saúde - Fiocruz Ceará. Formação: pós doutoranda em Saúde Digital na Fiocruz Ceará (em andamento), doutora em Ciências pelo instituto Nacional de Infectologia INI/FIOCRUZ (2015), mestre em Gerontologia Biomédica pela Pontifícia Universidade Católica do Rio Grande do Sul (2003), especialista em Nutrição Clínica (UGF- 1997, Santa Casa Misericórdia do RJ  1997);  em Nutrigenômica e nutrigenética na prática clínica para Unyleya (2022), em fitoterapia pela Pratiensino (2023) e em administração hospitalar (2001), dentre as formações complementares fez Capacitação em Pesquisa Clínica pelo Hospital Alemão Oswaldo Cruz (2018) é auditora interna de segurança alimentar FSSC 22.000 pela Max Consulting (2021). Tem experiência em gestão, docência, pesquisa e atuação clínica com ênfase nos seguintes temas: Saúde digital, genômica nutricional, saúde do idoso, doenças infecciosas e insegurança alimentar.<span style="font-weight: bold" class="texto"> (Texto informado pelo autor)</span>
# </p>
# </div><br class="clear"><div class="title-wrapper">
# <a name="Identificacao" tabindex="107">
# <h1 tabindex="0">Identificação</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Nome</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Adriana Costa Bacelo<img src="/buscatextual/images/icon-rfb.png" class="alinhamento-texto-img" title="Dados cadastrais validados junto a Receita Federal do Brasil"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Nome em citações bibliográficas</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">BACELO, A. C.;BACELO, ADRIANA COSTA;COSTA BACELO, ADRIANA;ADRIANA, BACELO</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Lattes iD</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><img src="./images/v2/icone_lattes.png" style="width: 15px; margin-right: 10px;">http://lattes.cnpq.br/8861663346303670</div>
# </div>
# </div>
# </div><br class="clear"><div class="title-wrapper">
# <a name="Endereco" tabindex="108">
# <h1 tabindex="0">Endereço</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Endereço Profissional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Fundação Oswaldo Cruz, Fundação Oswaldo Cruz - Ceará. <br class="clear">Rua São José, s/n<br class="clear">Precabura<br class="clear">61760000 - Eusébio, CE - Brasil - Caixa-postal: 61760000<br class="clear">Telefone: (85) 91584991<br class="clear">URL da Homepage: <a target="blank" href="https://ceara.fiocruz.br/portal/" tabindex="109">https://ceara.fiocruz.br/portal/</a></div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="FormacaoAcademicaTitulacao" tabindex="110">
# <h1 tabindex="0">Formação acadêmica/titulação</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2011 - 2015</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Doutorado em Pesquisa Clinica em Doenças Infecciosas. <br class="clear">Fundação Oswaldo Cruz, FIOCRUZ, Brasil.
		
# 	<br class="clear">Título: EVOLUÇÃO NUTRICIONAL DE PACIENTES COM TUBERCULOSE EM TRATAMENTO MEDICAMENTOSO, Ano de obtenção: 2015. <br class="clear">Orientador: <a class="icone-lattes" target="_blank" href="http://lattes.cnpq.br/0311400155430166" tabindex="111"><img src="images/curriculo/logolattes.gif"></a>valéria Cavalcanti Rolla. <br class="clear">Coorientador: Rejane Andréa Ramalho Nunes da Silva. <br class="clear">Palavras-chave: Desnutrição; AIDS; Tuberculose.<br class="clear">Grande área: Ciências da Saúde<br class="clear">Grande Área: Ciências da Saúde / Área: Nutrição. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2001 - 2003</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Mestrado em Gerontologia Biomedica. <br class="clear">Pontificia Univercidade Catolica do Rio Grande do Sul, PUCRS, Brasil.
		
# 	<br class="clear">Título: Impacto do Consumo de Sódio no Controle da Hipertensão Arterial do Idoso, Ano de Obtenção: 2003.<br class="clear">Orientador: <a class="icone-lattes" target="_blank" href="http://lattes.cnpq.br/8420962145466003" tabindex="112"><img src="images/curriculo/logolattes.gif"></a>Emilio Hideyuki Moriguchi.<br class="clear">Palavras-chave: Sódio; Hipertensão; Idoso.<br class="clear">Grande área: Ciências da Saúde</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2021 - 2023</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Especialização em Especialização em Fitoterapia.  (Carga Horária: 360h). <br class="clear">Pratiensino, PRATIENSINO, Brasil.
		
# 	<br class="clear">Título: .. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2021 - 2022</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Especialização em Nutrigenômica e Nutrigenética na Pratica Clinica.  (Carga Horária: 360h). <br class="clear">UNYLEYA EDITORA E CURSOS S/A, Unyleya, Brasil.
		
# 	<br class="clear">Título: .. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2001 - 2001</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Especialização em Administração Hopitalar. <br class="clear">Pontifícia Universidade Católica do Rio Grande do Sul, PUCRS, Brasil.
		
# 	<br class="clear">Título: Logistica de Implantação de uma Clinica Neurologica. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1997 - 1997</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Especialização em Nutrição Clinica. <br class="clear">Universidade Gama Filho, UGF, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1996 - 1997</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Especialização em Aperfeiçoamento Em Nutrição Clínica. <br class="clear">Santa Casa de Misericórdia do Rio de Janeiro, SCM/RJ, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1992 - 1996</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Graduação em Nutrição. <br class="clear">Universidade Gama Filho, UGF, Brasil.
		
# 	</div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="FormacaoAcademicaPosDoutorado" tabindex="113">
# <h1 tabindex="0">Pós-doutorado</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2021</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Pós-Doutorado. <br class="clear">Fundação Oswaldo Cruz, FIOCRUZ, Brasil.
		
# 	</div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="FormacaoComplementar" tabindex="114">
# <h1 tabindex="0">Formação Complementar</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2021 - 2022</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"> Extensão universitária em Curso de Nutrigenetica Pure. <br class="clear">Personalized Medicine Academy, PMA_NESTLE, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2021 - 2021</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"> Extensão universitária em Formação de Auditores Internos da Norma FSSC 22000 5.1.  (Carga horária: 20h). <br class="clear">Max Consulting, MAX CONSULTING, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2021 - 2021</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Certificação em Liderança, Capacidade de Aprender e Resiliência.  (Carga horária: 4h). <br class="clear">Pontifícia Universidade Católica do Rio Grande do Sul, PUCRS, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2021 - 2021</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Treinamento para operar o dispositivo e realizar exames com o Hilab.  (Carga horária: 2h). <br class="clear">HiLab, HILAB, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2020 - 2020</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">iGenesis: DNA, Diet e Wellness project.  (Carga horária: 8h). <br class="clear">DF MEDICA BRASIL, DF MEDICA, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2020 - 2020</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Longevity: Health project.  (Carga horária: 8h). <br class="clear">DF MEDICA BRASIL, DF MEDICA, Brasil.
		
# 	</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2017 - 2018</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Curso de Capacitação em Pesquisa Clinica.  (Carga horária: 360h). <br class="clear">Hospital Alemão Oswaldo Cruz, HAOC, Brasil.
		
# 	</div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="AtuacaoProfissional" tabindex="115">
# <h1 tabindex="0">Atuação Profissional</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell"><br class="clear"><div class="inst_back">
# <b>Fundação Oswaldo Cruz, FIOCRUZ, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2011 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Servidor Público, Enquadramento Funcional: Tecnologista Saúde Pública, Carga horária: 40</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Outras informações</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Atuação no IFF/FIOCRUZ de outubro/2011 à julho/2012Atuação no INI/FIOCRUZ de julho/2012 à julho de 2021Atuação na Fiocruz Ceará de agosto de 2021 até a presente data.Membro doLaboratório de Redes Integradas e Inteligentes em Saúde (LARIISA)Membro doGrupo de Pesquisa Clínica em Nutrição e Doenças Infecciosas INI / FiocruzMembro do Grupo de Pesquisa de Micronutrientes - Instituto Josué de Castro - UFRJDocente do Programa de Pós-Graduação em Nutrição e Saúde (PPGNS) da UniversidadeEstadual do Ceará (UECE)Docente do Mestrado Profissional PROFSAÚDE Pólo Fiocruz Ceará</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2004 - 2006</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Celetista, Enquadramento Funcional: nutricionista</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Atividades</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>02/2023 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Direção e administração, Fundação Oswaldo Cruz - Ceará. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Cargo ou função<br class="clear">Chefe de Gabinete.</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>07/2021 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Pesquisa e desenvolvimento, Fundação Oswaldo Cruz - Ceará. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Linhas de pesquisa <br class="clear"><a href="#LP_Saúde Digital" tabindex="116">Saúde Digital</a><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>02/2020 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Ensino, Pesquisa Clinica em Doenças Infecciosas, Nível: Pós-Graduação</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Disciplinas ministradas<br class="clear">Nutrição aplicada as doenças infecciosas<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>01/2017 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Ensino, Curso de Especialização em Nutrição Clínica Aplicada à Infectologia, Nível: Especialização</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Disciplinas ministradas<br class="clear">Nutrição Clinica aplicada a doenças infecciosas<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>01/2017 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Conselhos, Comissões e Consultoria, Instituto Nacional de Infectologia Evandro Chagas. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Cargo ou função<br class="clear">Sub-coordenação do curso de Curso de Especialização em Nutrição Clínica Aplicada à Infectologia.
				
# 				</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1/2004 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Pesquisa e desenvolvimento, Ipec, Serviço de Nutrição. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Linhas de pesquisa <br class="clear"><a href="#LP_Doenças infecciosas - Atenção integral em doenças infecciosas" tabindex="117">Doenças infecciosas - Atenção integral em doenças infecciosas</a><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>01/2006 - 1/2007</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Conselhos, Comissões e Consultoria, Instituto de Pesquisas Clinicas Evandro Chagas. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Cargo ou função<br class="clear">coordenador técnico da Equipe Multidisciplinar de Terapia Nutricional Enteral e Parenteral.
				
# 				</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>01/2004 - 1/2007</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Serviços técnicos especializados , Instituto de Pesquisas Clinicas Evandro Chagas. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Serviço realizado<br class="clear">Assistencia clinico nutricional a pacientes portadores de doenças infectocontagiosas a nivel ambulatorial e no centro de internação. </div>
# </div><br class="clear"><div class="inst_back">
# <b>Sociedade Unificada de Ensino Superior Augusto Motta, UNISUAM, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2005 - 2013</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Celetista, Enquadramento Funcional: Professor assistente, Carga horária: 12</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2007 - 2011</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Celetista formal, Enquadramento Funcional: Coordenador, Carga horária: 40</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Atividades</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>08/2012 - 12/2013</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Ensino, Nutrição, Nível: Graduação</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Disciplinas ministradas<br class="clear">Nutrição Clínica<br class="clear">Nutrição nas Fases da Vida<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>04/2010 - 12/2011</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Pesquisa e desenvolvimento, Departamento de Nutrição. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Linhas de pesquisa <br class="clear"><a href="#LP_Inovação e Desenvolvimento Tecnológico em Gastronomia e Nutrição Clínica." tabindex="118">Inovação e Desenvolvimento Tecnológico em Gastronomia e Nutrição Clínica.</a><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>02/2010 - 12/2011</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Ensino, Biologia, Nível: Graduação</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Disciplinas ministradas<br class="clear">Bioética<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>02/2010 - 12/2011</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Ensino, Biologia, Nível: Graduação</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Disciplinas ministradas<br class="clear">Bioética<br class="clear">Seminário de Pesquisa<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>02/2010 - 12/2011</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Ensino, Gastronomia, Nível: Graduação</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Disciplinas ministradas<br class="clear">Metodologia de Pesquisa<br class="clear">Projeto Profissional<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>02/2009 - 12/2011</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Ensino, Nutrição, Nível: Graduação</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Disciplinas ministradas<br class="clear">Aspectos Socioeconomicos e Psicologicos da Nutrição<br class="clear">Educação Nutricional<br class="clear">Exercicio da profissão<br class="clear">metodologia de Pesquisa<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>02/2010 - 07/2011</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Ensino, Nutrição, Nível: Graduação</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Disciplinas ministradas<br class="clear">Aspectos Socioculturais da Alimentação<br class="clear">Educação Nutricional<br class="clear">Exercício da Profissão<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2/2007 - 7/2011</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Direção e administração, Departamento de Nutrição. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Cargo ou função<br class="clear">coordenadora do curso de Nutrição.</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>02/2010 - 02/2011</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Ensino, Estetica e Cosmetologia, Nível: Graduação</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Disciplinas ministradas<br class="clear">Bioética e Legislação<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>02/2010 - 02/2011</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Ensino, Enfermagem, Nível: Graduação</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Disciplinas ministradas<br class="clear">Orientação de TCC<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>02/2007 - 02/2011</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Ensino, Enfermagem, Nível: Graduação</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Disciplinas ministradas<br class="clear">Introdução a Nutrição<br class="clear">Nutrição Aplicada a Enfermagem<br class="clear">Orientação de Monografia<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>02/2007 - 02/2010</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Ensino, Estetica e Cosmetologia, Nível: Graduação</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Disciplinas ministradas<br class="clear">Introdução a Nutrição<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>06/2008 - 06/2009</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Extensão universitária , Departamento de Nutrição. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Atividade de extensão realizada<br class="clear">`Módulod de Nutrição na Universidade Aberta a Terceira idade. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>06/2008 - 06/2009</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Extensão universitária , Departamento de Nutrição. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Atividade de extensão realizada<br class="clear">`Módulod de Nutrição na Universidade Aberta a Terceira idade. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>02/2005 - 02/2009</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Ensino, Nutrição, Nível: Graduação</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Disciplinas ministradas<br class="clear">Introdução a Nutrição<br class="clear">Nutrição em geriatria<br class="clear">Patologia da Nutrição e Dietoterapia I<br class="clear">Patologia da Nutrição e Dietoterapia II<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>8/2005 - 8/2006</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Extensão universitária , Departamento de Nutrição. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Atividade de extensão realizada<br class="clear">Atenção Nutricional a portadores de doenças Infectoparasitárias. </div>
# </div><br class="clear"><div class="inst_back">
# <b>Centro de Nutrição Vida e Saúde, CNVS, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2007 - 2011</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Gerente Técnico Administrativo, Enquadramento Funcional: Gerente Técnico-Administrativo, Carga horária: 10</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Outras informações</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Atividades desenvolvidas: gerenciamento administrativo; controle de custo; controle de pessoal; organização logística; negociação com fornecedores e elaboração de relatórios técnicos administrativos.</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Atividades</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>06/2007 - 12/2011</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Direção e administração, Ambulatórios. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Cargo ou função<br class="clear">Atendimento clinico nutricional individualizado.</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>06/2007 - 12/2011</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Serviços técnicos especializados , Ambulatórios. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Serviço realizado<br class="clear">Atendimento Clinico Nutricional Individualizado. </div>
# </div><br class="clear"><div class="inst_back">
# <b>Hospital São Lucas da Pucrs, HSLPUCRS, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2001 - 2003</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Celetista, Enquadramento Funcional: Nutricionista, Carga horária: 44</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Outras informações</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Atuação voltada para preceptoria de alunos de graduação em nutrição nas áreas de pediatria, adulto e idoso em unidades de clinica médica, terapia intensiva, emergência e ambulatório.</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Atividades</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>7/2001 - 11/2003</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Serviços técnicos especializados , Serviço de Nutrição e Dietética, Nutrição Assistencial. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Serviço realizado<br class="clear">Nutrição Clínica em UTI, Hemodialise, equipe de transplante Rim e pâncreas  e Centro Cirúrgico. </div>
# </div><br class="clear"><div class="inst_back">
# <b>Pontifícia Universidade Católica do Rio Grande do Sul, PUCRS, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1999 - 2000</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Celetista, Enquadramento Funcional: Nutricionista, Carga horária: 44</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Atividades</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1/1999 - 2/2000</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Serviços técnicos especializados , Grupo Memorial, Setor de Nutrição. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Serviço realizado<br class="clear">Chefia do Setor e atuação em Nutrição Clínica. </div>
# </div><br class="clear"><div class="inst_back">
# <b>Universidade Veiga de Almeida, UVA/RJ, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2004 - 2010</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Celetista, Enquadramento Funcional: Professor titular</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2004 - 2004</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Professor Visitante, Enquadramento Funcional: professor assistente, Carga horária: 6</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Atividades</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>10/2004 - 12/2010</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Ensino, Nutrição, Nível: Graduação</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Disciplinas ministradas<br class="clear">Nutrição na Fase adulta e Gerontologia<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1/2004 - 12/2010</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Pesquisa e desenvolvimento, Instituto de Ciências da Saúde, Faculdade de Nutrição. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Linhas de pesquisa <br class="clear"><a href="#LP_Gerontologia" tabindex="119">Gerontologia</a><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1/2004 - 12/2004</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Ensino, Nutrição, Nível: Especialização</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Disciplinas ministradas<br class="clear">Metabolismo<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1/2004 - 9/2004</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Ensino, Nutrição, Nível: Graduação</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Disciplinas ministradas<br class="clear">Nutriçaõ na fase adulta e gerontologia<br class="clear"></div>
# </div><br class="clear"><div class="inst_back">
# <b>Hospital Celina Gonçalves Veloso e Rocha Sa, HCG*, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1998 - 1999</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Celetista, Enquadramento Funcional: Gerente da Divisão de Economato, Carga horária: 44</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Outras informações</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Chefe do Serviço de Nutrição e Hotelaria; membro da equipe de infecção hospitalar e de nutrição enteral e parenteral, e nutricionista clínica atuando na unidade ambulatorial e de internação.</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Atividades</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>3/1998 - 11/1999</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Serviços técnicos especializados , Divisão de Economato, Hotelaria. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Serviço realizado<br class="clear">Administração Hospitalar e Nutrição Clínica. </div>
# </div><br class="clear"><div class="inst_back">
# <b>Hospital Memorial Fuad Chidid, MEMORIAL, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2000 - 2003</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Celetista formal, Enquadramento Funcional: Chefe do Serviço de NUtrição, Carga horária: 40</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Outras informações</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Licença sem vencimento de 2001 a 2003</div>
# </div><br class="clear"><br class="clear"><div class="inst_back">
# <b>Centro Cirúrgico Integrado da Policlínica de Botafogo, ICC-POLICLÍNICAB, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1996 - 1998</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Celetista, Enquadramento Funcional: Nutricionista, Carga horária: 30</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Outras informações</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Atuava como nutricionista do centro cirúrgico e da unidade de terapia intensiva.</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Atividades</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>4/1996 - 2/1998</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Serviços técnicos especializados , Centro Cirúrgico, Unidade de Internação. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Serviço realizado<br class="clear">Assistencia Nutricional pré e pós cirurgica. </div>
# </div><br class="clear"><div class="inst_back">
# <b>Hospital Geral da Santa Casa de Misericordia do Rio de Janeiro, HGSCMRJ, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1995 - 1995</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Colaborador, Enquadramento Funcional: Estagiária de Atendimento ambulatorial, Carga horária: 20</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1995 - 1995</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Colaborador, Enquadramento Funcional: Estágio Supervisionado, Carga horária: 20</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Atividades</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>8/1995 - 12/1995</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Estágios , Oitava Enfermaria, Serviço de Nutrição. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Estágio realizado<br class="clear">Nutrição em Hepatopatias e Endocrinologia. </div>
# </div><br class="clear"><div class="inst_back">
# <b>Hospital Maternidade da Ordem Terceira da Penitencia, HMOTP, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1995 - 1995</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Colaborador, Enquadramento Funcional: Estágio Extra Curricular de Suporte Nutricion, Carga horária: 25</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Atividades</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>3/1995 - 7/1995</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Estágios , Quality Medical Serviços Médicos e Cirúrgicos Ltda, Centro de Tratamento Intensivo Unidade de Terapia Nutricional. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b><br class="clear"></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Estágio realizado<br class="clear">Avaliação Nutricional e Nutrição Enteral e Parenteral. </div>
# </div><br class="clear"><div class="inst_back">
# <b>Universidade Gama Filho, UGF, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1994 - 1994</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Monitor, Enquadramento Funcional: Monitor da Disciplina de Dietoterapia, Carga horária: 12</div>
# </div><br class="clear"><div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1993 - 1993</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Monitor, Enquadramento Funcional: Monitor da Disciplina de Fisiologia, Carga horária: 12</div>
# </div><br class="clear"><br class="clear"><div class="inst_back">
# <b>AF Saúde Serviços Medicos LTDA, LE CHANTIER, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2020 - 2023</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: Colaborador, Enquadramento Funcional: Nutricionista, Carga horária: 8</div>
# </div><br class="clear"><br class="clear"><div class="inst_back">
# <b>AD &amp; CL Serviços Médicos LTDA, SAN GARDE, Brasil.
		
# 	</b>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right subtit-1">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Vínculo institucional</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5"><br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2022 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Vínculo: sócia (9% das quotas), Enquadramento Funcional: nutricionista, Carga horária: 10</div>
# </div><br class="clear"></div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="LinhaPesquisa" tabindex="120">
# <h1 tabindex="0">Linhas de pesquisa</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell"><a name="LP_Doenças infecciosas - Atenção integral em doenças infecciosas" tabindex="121"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Doenças infecciosas - Atenção integral em doenças infecciosas</div>
# </div><a name="LP_Saúde Digital" tabindex="122"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Saúde Digital</div>
# </div><a name="LP_Gerontologia" tabindex="123"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Gerontologia</div>
# </div><a name="LP_Inovação e Desenvolvimento Tecnológico em Gastronomia e Nutrição Clínica." tabindex="124"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Inovação e Desenvolvimento Tecnológico em Gastronomia e Nutrição Clínica.</div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="ProjetosPesquisa" tabindex="125">
# <h1 tabindex="0">Projetos de pesquisa</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell"><a name="PP_Análise comparativa de intervenções presenciais e virtuais para o acompanhamento nutricional de adultos: ensaio clínico de superioridade" tabindex="126"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2023 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Análise comparativa de intervenções presenciais e virtuais para o acompanhamento nutricional de adultos: ensaio clínico de superioridade<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: Introdução: O avanço tecnológico tem impulsionado transformações significativas na área da saúde, incluindo a gestão e o tratamento de doenças crônicas não transmissíveis, o que tem possibilitado a expansão e diversificação dos serviços de atendimento nutricional. Nesse cenário, as intervenções presenciais e virtuais têm se destacado, suscitando indagações acerca de sua eficácia, aplicabilidade e impacto nos resultados clínicos. Diante da crescente importância dessas abordagens para o contexto das doenças crônicas não transmissíveis, sabidamente crescente no mundo, torna-se imprescindível aprofundar a compreensão das diferenças entre ambas, com o propósito de embasar decisões e promover melhorias substanciais na área da nutrição.  Justificativa: A relevância deste estudo está pautada no potencial impacto que seus resultados poderão ter sobre a prática clínica em nutrição, contribuindo para a identificação de quais abordagens são mais adequadas a diferentes perfis de pacientes, considerando variáveis socioeconômicas e geográficas. Além disso, a análise comparativa permitirá uma avaliação mais precisa da efetividade e eficiência de cada modalidade de acompanhamento, fomentando a busca por soluções inovadoras e a utilização estratégica da tecnologia no âmbito da saúde nutricional. Objetivo: Comparar as intervenções presenciais e  virtuais no acompanhamento nutricional de adultos, com foco na adesão ao tratamento, evolução clínica e satisfação dos pacientes, a fim de fornecer evidências que auxiliem na tomada de decisões para aprimorar o atendimento nutricional de doenças crônicas não transmissíveis. Metodologia: Ensaio clínico de superioridade de comparação entre estratégias virtuais (teleatendimento) e convencionais (presenciais) de mapeamento e intervenção nutricional. Todos os 250 trabalhadores (servidores públicos, terceirizados, e bolsistas) e alunos, maiores de 19 anos, não sendo gestante, da unidade da Fiocruz Ceará serão convidados para participar do estudo. Os trabalhadores/alunos que assinarem o TCLE serão randomizados em dois grupos, a saber: a) grupo virtual (intervenção): estratégias de mapeamento e orientação/prescrição nutricional individuais e coletivas de forma virtual (Whatsapp ou Telegram e Zoom ou Teams); b) grupo presencial (comparação): estratégias de mapeamento e orientação/prescrição nutricional individuais e coletivas de forma presencial. Para ambos os grupos haverá um encontro presencial no Dzero (inclusão), D180 (6 meses de acompanhamento) e D360 (01 ano de acompanhamento) para avaliação de composição corporal. O grupo virtual passará por esta fase após a realização de uma autoavaliação virtual e ambos os resultados serão comparados. As consultas individuais serão trimestrais, e as e coletivas serão semanais. A avaliação nutricional incluirá exames clínicos (exame físico e história clínica), antropométricos (peso, altura, circunferências e perímetros corporais), bioimpedância e ultrassonografia de quadríceps femoral; além de aplicação da escala brasileira de insegurança alimentar (EBIA), marcador de estresse percebido (PSS), adesão auto-relatada, qualidade de sono autorreferida,  autopercepção de qualidade de vida e imagem corporal e um questionário de satisfação enquanto usuário (escala Likert) e ainda análise fenomenológica evolutiva dos discursos obtidos nas consultas coletivas. Uma sub amostra de 50 participantes obesos (IMC&gt;30 Kg/m2 e com idade entre 40 e 59 anos), 25 de cada grupo (virtual e presencial), será randomizada para investigação de saúde de precisão, para tal será realizada coleta de sangue e saliva nos tempos Dzero e D360. Resultados esperados: Espera-se que a disponibilização de estratégias digitais melhorem a adesão e por conseguinte os parâmetros clínicos, nutricionais, imunológicos, metabólicos e de qualidade de vida.. <br class="clear">Situação: Em andamento; Natureza: Pesquisa. <br class="clear">Alunos envolvidos: Graduação: (1) . <br class="clear"><br class="clear">Integrantes: Adriana Costa Bacelo - Coordenador / BRASIL, PEDRO EMMANUEL - Integrante / COPLE-RODRIGUES, CLAUDIA DOS SANTOS - Integrante / TORRES, PALOMA RIBEIRO - Integrante / Luiz Odorico Monteiro de Andrade - Integrante / Ivana Cristina de Holanda Cunha Barreto - Integrante / Kamila Maria Oliveira Sales - Integrante / Paulo Armando Cavalcante Aguilar - Integrante / José Xavier Neto - Integrante / Filipe Brito - Integrante / Marcela Helena Gambim Fonseca - Integrante / Júlio Cesar Martins Ximenes - Integrante / Ana Carolina Matias Dinelly Pinto - Integrante / Clarissa Perdigão Mello Ferraz - Integrante / William Cesar Bento Régis - Integrante / Carla Freire Celedonio Fernandes - Integrante / Márcia Andréa Gonçalves Leite - Integrante / Márcio Espíndola Freire Maia - Integrante / Francisco Leandro Fernandes Sampaio - Integrante / Elga Batista da Silva - Integrante / André Ribeiro Braga - Integrante.<br class="clear">Financiador(es): Conselho Nacional de Desenvolvimento Científico e Tecnológico - Auxílio
# 										financeiro.</div>
# </div><a name="PP_Desenvolvimento de uma solução digital para o controle de qualidade da indústria de alimentos baseados em requisitos internacionais" tabindex="127"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2021 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Desenvolvimento de uma solução digital para o controle de qualidade da indústria de alimentos baseados em requisitos internacionais<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: Introdução: A acreditação internacional das indústrias de alimentos brasileiros é uma necessidade de competitividade de mercado e para a segurança alimentar dos consumidores.  Justificativa: A FSSC 22000 foi reconhecida pela GFSI como um padrão de segurança alimentar internacional desde 2010, tornando-se um protocolo de certificação completo imprescindível, mas de aplicabilidade complexa. Mas impondo às indústrias de alimentos brasileiras uma necessidade de adequação para ingressão no mercado internacional, fator importante para o desenvolvimento econômico industrial brasileiro.  Objetivo: Propor um instrumento digital que apoie o controle de qualidade da indústria de alimentos brasileira, a fim de tornar mais prático, amigável e simplificado o processo de auditoria interna de segurança de alimentos demandado no mercado internacional. Metodologia: Estudo de desenvolvimento, validação e de confiabilidade de um aplicativo de auditoria de qualidade e segurança de alimentos a partir de critérios estabelecidos pela Norma FSSC 22000, em etapas de um aplicativo construído utilizando um software de programação multiplataformas. Numa primeira etapa será realizada a revisão sistemática sobre a aplicabilidade da FSSC 22000; posteriormente serão convidados especialistas auditores de segurança alimentar ISO 22 000, boas práticas de fabricação de alimentos (BPF), pontos críticos de controle (APPCC) e/ou FSSC 22000 para participar de um estudo etnográfico para a versão inicial do instrumento. Nesta etapa será seguida o instrumento será encaminhado ao mesmo painel de especialista conforme a técnica Delphi. Posteriormente, como base na verificação da concordância entre os especialistas será desenvolvido um aplicativo móvel de monitoramento da qualidade nos processos de produção de alimentos, utilizando um software de programação para multi plataformas, incluindo Android e posteriormente iOs, utilizando as tecnologias C# e JSON. Para o teste de confiabilidade o instrumento será aplicado sequencialmente 4 vezes, em semanas diferentes, numa indústria de alimentos da região metropolitana do Ceará por dois auditores que irão se revezar no uso das ferramentas tradicionais de auditoria e a solução digital proposta digital. Resultados esperados:  Espera-se apresentar uma solução digital, que traga agilidade e segurança no controle dos processos de segurança alimentar das indústrias alimentícias brasileiras, atendendo as normas do mercado de indústria de alimentos internacional.. <br class="clear">Situação: Em andamento; Natureza: Pesquisa. <br class="clear">Alunos envolvidos: Graduação: (1) . <br class="clear"><br class="clear">Integrantes: Adriana Costa Bacelo - Coordenador / Luiz Odorico Monteiro de Andrade - Integrante / Ivana Cristina de Holanda Cunha Barreto - Integrante / Margareth Catoia Varela - Integrante.<br class="clear"></div>
# </div><a name="PP_DESENVOLVIMENTO DE UM APLICATIVO MÓVEL DE REPRESENTAÇÃO DA INGESTÃO ALIMENTAR DE PACIENTES INTERNADOS: TESTE DE CONFIABILIDADE" tabindex="128"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2020 - 2023</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">DESENVOLVIMENTO DE UM APLICATIVO MÓVEL DE REPRESENTAÇÃO DA INGESTÃO ALIMENTAR DE PACIENTES INTERNADOS: TESTE DE CONFIABILIDADE<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: A desnutrição de pacientes hospitalizados é um problema significantemente crítico e geralmente é associado com um aumento na morbidade e mortalidade. Sendo de extrema importância que se realize a análise da capacidade de ingestão alimentar intra-hospitalar associada a avaliação do estado nutricional e clínico do paciente. Objetivo: testar a confiabilidade de um aplicativo móvel, como estratégia de virtualização de um instrumento de avaliação da aceitação e da qualidade das dietas hospitalares de pacientes internados. Metodologia: estudo de desenvolvimento e de confiabilidade Inter e intra observador de um aplicativo de monitoramento de ingestão alimentar intra-hospitalar, em etapas de um aplicativo construído utilizando um software de programação para multiplataformas a partir de um instrumento validado para pacientes com doenças infeciosas atendidos no Instituto Nacional de Infectologia Evandro Chagas/FIOCRUZ (INI). Para o teste de confiabilidade o instrumento será aplicado em 70 pacientes internados na Unidades hospitalar do Instituto Nacional de infectologia entre janeiro e julho de 2020, randomizados em dois grupos, onde um fator de confiabilidade a ser testado será a consistência do método na plataforma digital, verificando se o aplicativo promoverá dados equivalentes ao instrumento em papel, e por fim será realizado um plano de análise estatística. Espera-se que a implementação da tecnologia aplicada promova ainda mais a produtividade, tanto no controle das condições nutricionais do paciente, quanto no atendimento realizado pelos profissionais de saúde, esperando-se que haja um aumento significativo na efetividade do processo terapêutico, de reabilitação e nutricional no ambiente hospitalar em que foi testada.. <br class="clear">Situação: Concluído; Natureza: Pesquisa. <br class="clear">Alunos envolvidos: Mestrado acadêmico: (1) . <br class="clear"><br class="clear">Integrantes: Adriana Costa Bacelo - Coordenador / Claudio Fico Fonseca - Integrante / Alan Renier Jamal Occhioni Molter - Integrante.<br class="clear"><br class="clear">Número de produções C, T &amp; A: 2</div>
# </div><a name="PP_Desenvolvimento de um aplicativo móvel de representação da ingestão alimentar de pacientes internados: teste de confiabilidade" tabindex="129"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2020 - 2023</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Desenvolvimento de um aplicativo móvel de representação da ingestão alimentar de pacientes internados: teste de confiabilidade<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Situação: Concluído; Natureza: Pesquisa. <br class="clear">Alunos envolvidos: Mestrado acadêmico: (1) . <br class="clear"><br class="clear">Integrantes: Adriana Costa Bacelo - Coordenador / Marlete Pereira da Silva - Integrante / CARDOSO, CLÁUDIA SANTOS DE AGUIAR - Integrante / Brito, Patrícia Dias - Integrante / Claudio Fico Fonseca - Integrante / Alan Renier Jamal Occhioni Molter - Integrante.<br class="clear"></div>
# </div><a name="PP_DIAGNÓSTICO NUTRICIONAL E FUNÇÃO INTESTINAL DE PACIENTES SINTOMATICOS E ASSINTOMATICOS PORTADORES DE HTLV SUBMETIDOS A ORIENTAÇÃO DIETÉTICA COM E SEM A INCLUSÃO DA FARINHA DE OKARA" tabindex="130"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2016 - 2021</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">DIAGNÓSTICO NUTRICIONAL E FUNÇÃO INTESTINAL DE PACIENTES SINTOMATICOS E ASSINTOMATICOS PORTADORES DE HTLV SUBMETIDOS A ORIENTAÇÃO DIETÉTICA COM E SEM A INCLUSÃO DA FARINHA DE OKARA<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: A infecção pelo HTLV ainda é negligenciada pelos sistemas de atenção à saúde, possivelmente por aproximadamente 90% dos portadores serem assintomáticos e as complicações mais conhecidas ocorrem em apenas 5% dos infectados. Contudo, a morbidade das doenças associadas ao HTLV é elevada, a ocorrência de constipação intestinal nos pacientes com Paraparesia Espástica Tropical/Mielopatia é grande e a situação nutricional desta população é pouco estudada. Objetivo: Comparar estado nutricional de indivíduos adultos portadores de HTLV sintomáticos e assintomáticos e investigar o efeito da inclusão dietética da farinha de Okara sobre a função intestinal e o estado nutricional dos indivíduos sintomáticos. Metodologia: estudo prospectivo, longitudinal, analítico do acompanhamento nutricional, por 2 meses, de pacientes sintomáticos e assintomáticos com diagnóstico de HTLV, no período de setembro de 2015 à setembro de 2021. Serão realizadas as avaliações antropométrica, bioquímica e dietética (Recordatório de 24 horas e diversidade dietética) seguida de aconselhamento dietético e/ou suplementação de fibra às custas de farinha de Okara (grupo caso). A percepção de adesão dietética será avaliada por autorreferência nas diferentes consultas durante estudo. Será verificado o gasto energético por calorimetria indireta e prognóstico nutricional por ângulo de fase. Será considerada má nutrição quando pelo menos um dos resultados da avaliação nutricional estava fora dos valores de referência, sendo classificada de acordo com os parâmetros considerados, em: global (todos os parâmetros), calórica (Índice de Massa Corporal - IMC, dobra cutânea triciptal ? DCT, e circunferência do braço - CB), proteica (circunferência muscular do braço - CMB, albumina e transferrina), ou específica (marcadores séricos de ferro, selênio, zinco, retinol, D-25 hidroxicalciferol e ∂-tocoferol). Os dados serão analisados no programa R-project® versão 3.0.2. Os resultados divididos em tempo inicial e final. As diferenças serão consideradas significantes quando o valor de p foi ≤ 0.05. Resultados esperado: identificar as necessidades nutricionais do indivíduo com HTLV e contribuir para o conhecimento das repercussões dos sintomas no gasto energético, apontando a quantidade de fibra necessária para minimizar a constipação secundária a mielopatia espástica.. <br class="clear">Situação: Concluído; Natureza: Pesquisa. <br class="clear">Alunos envolvidos: Especialização: (2)  / Mestrado acadêmico: (1) . <br class="clear"><br class="clear">Integrantes: Adriana Costa Bacelo - Coordenador / Claudia dos santos cople - Integrante / eliane paiva - Integrante / Patricia Dias de Brito - Integrante / Cristiane Fonseca de Almeida - Integrante / Abelardo de Queiroz Campos Araújo - Integrante / Elizabethe Neves - Integrante / Ana Claudia Celestino - Integrante / Naise Oliveira da Rocha - Integrante / Paloma Ribeiro Torres - Integrante / Roseli dos Santos - Integrante.<br class="clear">Financiador(es): FAPERJ - Auxílio
# 										financeiro / FAPERJ - Bolsa.<br class="clear">Número de produções C, T &amp; A: 6</div>
# </div><a name="PP_AVALIAÇÃO DA RESPOSTA À INGESTÃO DE FIBRAS EM PACIENTES PORTADORES DE MIELOPATIA ASSOCIADA AO HTLV-1" tabindex="131"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2009 - 2010</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">AVALIAÇÃO DA RESPOSTA À INGESTÃO DE FIBRAS EM PACIENTES PORTADORES DE MIELOPATIA ASSOCIADA AO HTLV-1<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: O vírus HTLV é causador de uma infecção de extrema importância mundial, em algumas áreas, onde a infecção pelo HTLV-1 é endêmica, a taxa de prevalência gira em torno de 15% da população geral, elevando a soroprevalência com o aumento da idade. Outros grupos onde o HTLV-1também foi observado são os negros do sudeste dos EUA e imigrantes de áreas HTLV-I-endêmicas, que residem em Brooklyn, Nova Iorque, EUA. Trata-se de um vírus cujo contagio depende do contato com sangue (transfusão de sangue e seringas contaminadas) e com outros fluidos corpóreos (esperma, secreções vaginais e leite materno). Dentre os indivíduos infectados apenas 1 a 5% poderão ter alguma doença relacionada ao vírus. Os demais, 95 a 99% dos indivíduos não apresentam qualquer doença. A paraparesia espástica tropical, uma das mais importantes doenças associadas ao HTLV, é uma doença de progressão lenta e de baixa letalidade. Esta é caracterizada pela fraqueza progressiva dos membros inferiores, com espasticidade, hiperreflexia, perturbações sensoriais, distúrbios esfincterianos (incontinência urinária e constipação intestinal).Do ponto de vista nutricional tem se observado que o hábito alimentar da população mundial está cada vez mais restrito, no que tange a ingestão fibrínica e hídrica, tornando-se este mais um fator causal de constipação. A soja tem-se apresentado com grande potencial de aplicação para alimentação humana. Em destaque pelo seu valor nutricional e pelo alto teor em proteínas e fibras, a soja e seus derivados são fontes de isoflavonas, fitoestrógenos, apresentando alegações de propriedades funcionais. O extrato hidrossolúvel de soja (EHS) é um dos derivados não fermentado obtido a partir do grão desta leguminosa. Durante a fabricação do EHS, obtém-se o okara, um resíduo sólido, rico em proteínas, vitaminas, aminoácidos essenciais, fibras e isoflavonas. Atualmente, este resíduo industrial é utilizado como ração animal ou a maior parte dele tem sido descartado diretamente no lixo, contri. <br class="clear">Situação: Concluído; Natureza: Pesquisa. <br class="clear"><br class="clear">Integrantes: Adriana Costa Bacelo - Coordenador.<br class="clear"></div>
# </div><a name="PP_Evolução Nutricional de Pacientes com Tuberculose e AIDS em Tratamento com Análogo não Nucleosídeo Efavirenz Associado a Rifampicina" tabindex="132"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2008 - 2015</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Evolução Nutricional de Pacientes com Tuberculose e AIDS em Tratamento com Análogo não Nucleosídeo Efavirenz Associado a Rifampicina<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: Este estudo visa avaliar o efeito da terapia anti-tuberculose e anti-retroviral no estado nutricional de pacientes com TB e HIV.. <br class="clear">Situação: Concluído; Natureza: Pesquisa. <br class="clear"><br class="clear">Integrantes: Adriana Costa Bacelo - Coordenador / Valeria Rolla - Integrante / Rejane Andreia Ramalho - Integrante.<br class="clear"></div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="ProjetosDesenvolvimento" tabindex="133">
# <h1 tabindex="0">Projetos de desenvolvimento</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell"><a name="PP_Avaliação da Resposta a Ingestão de Fibras em Pacientes Portadores de Mielopatia Associada ao HTLV-1" tabindex="134"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2005 - 2007</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Avaliação da Resposta a Ingestão de Fibras em Pacientes Portadores de Mielopatia Associada ao HTLV-1<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Situação: Concluído; Natureza: Desenvolvimento. <br class="clear"><br class="clear">Integrantes: Adriana Costa Bacelo - Coordenador.<br class="clear"></div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="RevisorPeriodico" tabindex="135">
# <h1 tabindex="0">Revisor de periódico</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2015 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Periódico: Scientia Medica (PUCRS. Impresso)</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2015 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Periódico: Plos One</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2016 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Periódico: Cochrane Database of Systematic Reviews (Online)</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2017 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Periódico: Austin Journal of Nutrition &amp; Metabolism</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2019 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Periódico: CLINICAL NUTRITION ESPEN</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2019 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Periódico: Journal Research and Reports in Tropical Medicine</div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="RevisorProjetoFomento" tabindex="136">
# <h1 tabindex="0">Revisor de projeto de fomento</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2018 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Agência de fomento: Fundação de Amparo a Pesquisa do Estado do Amazonas</div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2016 - 2016</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Agência de fomento: Fundação Rondônia de Amparo ao Desenvolvimento das Ações Científicas e Tec</div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="AreasAtuacao" tabindex="137">
# <h1 tabindex="0">Áreas de atuação</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Grande área: Ciências da Saúde / Área: Medicina / Subárea: saúde digital. <br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Grande área: Ciências da Saúde / Área: Saúde Coletiva / Subárea: Saúde Coletiva. <br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Grande área: Ciências da Saúde / Área: Saúde Coletiva / Subárea: Nutrição Clinica aplicada a doenças infecciosas. <br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Grande área: Ciências da Saúde / Área: Saúde Coletiva / Subárea: Atendimento ambulatorial de nutrição clinica com foco em má nutrição. <br class="clear"></div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="Idiomas" tabindex="138">
# <h1 tabindex="0">Idiomas</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Inglês</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Compreende Bem, Fala Razoavelmente, Lê Bem, Escreve Razoavelmente. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>Espanhol</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Compreende Razoavelmente, Fala Pouco, Lê Razoavelmente, Escreve Pouco. </div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="PremiosTitulos" tabindex="139">
# <h1 tabindex="0">Prêmios e títulos</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2016</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Indicação para Premio Capes 2016 pelo Programa de Pesquisa Clínica em Doenças Infecciosas do INI, CAPES. </div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2011</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Honra ao Merito ao trabalho orientado de Conclusão de Curso do aluno Ivan Ramos Maia entitulado: AVALIAÇÃO DA OCORRÊNCIA DE ANEMIA EM PACIENTES COM TUBERCULOSE DURANTE O TRATAMENTO MEDICAMENTOSO., UNISUAM. </div>
# </div>
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="ProducoesCientificas" tabindex="140">
# <h1 tabindex="0">Produções</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <a name="ProducaoBibliografica" tabindex="141"></a><br class="clear"><div class="inst_back">
# <b>Produção bibliográfica</b>
# </div>
# <a name="Citacoes" tabindex="142"></a><div class="cita-artigos"> <b>Citações</b> </div><br class="clear"><div class="layout-cell layout-cell-12"><div class="layout-cell layout-cell-6"><div class="layout-cell-pad"> <div class="science_cont"><div class="web_s">Web of Science<a href="http://www.researcherid.com/rid/G-4336-2015" target="_blank" tabindex="143"><img src="/buscatextual/images/curriculo/researcherID.gif"></a></div><div class="trab">Total de trabalhos:2</div><div class="cita">Total de citações:19</div><div class="fator">Fator H:1</div><div class="detalhes"><a href="http://www.researcherid.com/rid/G-4336-2015" target="_blank" tabindex="144">Bacelo, Adriana C</a>&nbsp;&nbsp;Data:&nbsp;14/03/2024</div></div></div></div></div><br class="clear"><div class="layout-cell layout-cell-12"><div class="layout-cell layout-cell-6"><div class="layout-cell-pad"> <div class="science_cont"><div class="web_s">SCOPUS</div><div class="trab">Total de trabalhos:41</div><div class="cita">Total de citações:43</div><div class="detalhes">Bacelo, Adriana Costa&nbsp;&nbsp;Data:&nbsp;14/03/2024</div></div></div></div></div><br class="clear"><div id="artigos-completos">
# <div class="cita-artigos">
# <b><a name="ArtigosCompletos" tabindex="145"></a>Artigos completos publicados em periódicos</b>
# </div>
# <br class="clear"><br class="clear"><div class="layout-cell-6">
# <div class="sub_tit_form">
# <label for="combo-ordenacao-citacao">Ordenar por</label><select class="input-text input-login-select" id="combo-ordenacao-citacao"><option value="1"> Ordem Cronológica </option><option value="2"> Número de citações Web of science </option><option value="3"> Número de citações Scopus </option><option value="4"> Numero de citações Scielo </option><option value="5"> Primeiro autor </option><option value="6"> Impacto JCR </option><option value="7"> Ordem de Importância </option></select>
# </div>
# </div>
# <script type="text/javascript">
			
# 				var ordenacao = {
# 					engine: {
# 						textual: function(seletor) {
						
# 							$("#artigos-completos").find(".artigo-completo").sortElements(function(a, b){
# 								var v1 = $(a).find(seletor).text();
# 								var v2 = $(b).find(seletor).text();
# 								return (v1 > v2) ? 1 : -1;
# 							});
# 						},
# 						numerico: function(seletor) {
						
# 							$("#artigos-completos").find(".artigo-completo").sortElements(function(a, b){
# 								var v1 = parseInt($(a).find(seletor).text(), 10);
# 								var v2 = parseInt($(b).find(seletor).text(), 10);
# 								if (isNaN(v1)) v1 = 0;
# 								if (isNaN(v1)) v2 = 0;
# 								return (v1 < v2) ? 1 : -1;
# 							});
# 						},
# 						numerico_crescente: function(seletor) {
						
# 							$("#artigos-completos").find(".artigo-completo").sortElements(function(a, b){
# 								var v1 = parseInt($(a).find(seletor).text(), 10);
# 								var v2 = parseInt($(b).find(seletor).text(), 10);
# 								if (isNaN(v1)) v1 = 0;
# 								if (isNaN(v1)) v2 = 0;
# 								return (v1 > v2) ? 1 : -1;
# 							});
# 						}
# 					},
# 					change: function(e) {
					
# 						var combo = $(e.target),
# 							valor = combo.val();
							
# 						if (valor == 1) ordenacao.engine.numerico("span[data-tipo-ordenacao='ano']");
# 						if (valor == 2) ordenacao.engine.numerico("span[data-tipo-ordenacao='1']");
# 						if (valor == 3) ordenacao.engine.numerico("span[data-tipo-ordenacao='3']");						
# 						if (valor == 4) ordenacao.engine.numerico("span[data-tipo-ordenacao='2']");
# 						if (valor == 5) ordenacao.engine.textual("span[data-tipo-ordenacao='autor']");
# 						if (valor == 6) ordenacao.engine.numerico("span[data-tipo-ordenacao='jcr']");
# 						if (valor == 7) ordenacao.engine.numerico_crescente("span[data-tipo-ordenacao='importancia']");
							
# 					}
# 				};
			
# 				$("#combo-ordenacao-citacao").bind("change", ordenacao.change);
			
# 			</script>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">MOLTER, ALAN RENIER JAMAL OCCHIONI</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2024</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1186/s12911-023-02406-x" target="_blank" tabindex="146"></a>MOLTER, ALAN RENIER JAMAL OCCHIONI ; DA ROCHA CARVALHO, NAISE OLIVEIRA ; TORRES, PALOMA RIBEIRO ; DA SILVA, MARLETE PEREIRA ; DE BRITO, PATRÍCIA DIAS ; DO BRASIL, PEDRO EMMANUEL ALVARENGA AMERICANO ; FONSECA, CLAUDIO FICO ; <b>BACELO, ADRIANA COSTA</b> . Development of a mobile application to represent food intake in inpatients: dietary data systematization. BMC Medical Informatics and Decision Making<sup><img id="14726947_1" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="14726947" original-title="Bmc Medical Informatics And Decision Making (1472-6947)<br />Fator de impacto (JCR 2022): 3.5"></sup>, v. 24, p. 1-10, 2024. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1186/s12911-023-02406-x&amp;issn=14726947&amp;volume=24&amp;issue=&amp;paginaInicial=1&amp;titulo=Development of a mobile application to represent food intake in inpatients: dietary data systematization&amp;sequencial=1&amp;nomePeriodico=BMC Medical Informatics and Decision Making" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">PINTO, LUCIA DE MELLO COUTINHO</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2024</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1016/j.nutos.2024.02.008" target="_blank" tabindex="147"></a>PINTO, LUCIA DE MELLO COUTINHO ; MARTINS, REJANNE DE OLIVEIRA ; BRASIL, PEDRO EMMANUEL ALVARENGA AMERICANO DO ; <b>BACELO, ADRIANA COSTA</b> . AN EXPLORATORY STUDY ABOUT HEALTHCARE PROFESSIONALS AND STUDENTS FOOD CONSUMPTION DURING THE COVID-19 PANDEMIC. Clinical Nutrition Open Science<sup></sup>, v. 1, p. 1-27, 2024. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1016/j.nutos.2024.02.008&amp;issn=26672685&amp;volume=1&amp;issue=&amp;paginaInicial=1&amp;titulo=AN EXPLORATORY STUDY ABOUT HEALTHCARE PROFESSIONALS AND STUDENTS FOOD CONSUMPTION DURING THE COVID-19 PANDEMIC&amp;sequencial=2&amp;nomePeriodico=Clinical Nutrition Open Science" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">SILVA, M. P.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2022</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.12873/424pereira" target="_blank" tabindex="148"></a>SILVA, M. P. ; Santo, R.S. ; <b>BACELO, A. C.</b> ; ALMEIDA, C. F. ; Cardoso, C ; Simplicio, P. ; COLARES, L. G. T. ; PATRICIA, BRITO . Validación de contenido y aplicación de un instrumento para evaluar la aceptación de dietas hospitalarias.. NUTRICION CLINICA Y DIETETICA HOSPITALARIA<sup><img id="1989208X_3" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="1989208X" original-title="Nutricion Clinica Y Dietetica Hospitalaria (1989-208X)<br />Fator de impacto (JCR 2022): 0.5"></sup>, v. 42, p. 117-126, 2022. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.12873/424pereira&amp;issn=1989208X&amp;volume=42&amp;issue=&amp;paginaInicial=117&amp;titulo=Validación de contenido y aplicación de un instrumento para evaluar la aceptación de dietas hospitalarias.&amp;sequencial=3&amp;nomePeriodico=NUTRICION CLINICA Y DIETETICA HOSPITALARIA" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">PALOMA, TORRES</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2022</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1016/j.clnesp.2022.11.004" target="_blank" tabindex="149"></a>PALOMA, TORRES ; CLAUDIA, COPLE-RODRIGUES ; NAISE, ROCHA ; PATRICIA, BRITO ; CLEVIO, FONSECA ; MARCEL, QUINTANA ; PEDRO, BRASIL ; ABELARDO, ARAÚJO ; <b>ADRIANA, BACELO</b> . Evaluation of nutritional status and adherence to dietary monitoring among patients with Human T-Cell Leukemia Virus type 1 infection. CLINICAL NUTRITION ESPEN<sup><img id="24054577_4" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="24054577" original-title="Clinical Nutrition Espen (2405-4577)<br />Fator de impacto (JCR 2022): 3.0"></sup>, v. 1, p. 1-38, 2022. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1016/j.clnesp.2022.11.004&amp;issn=24054577&amp;volume=1&amp;issue=&amp;paginaInicial=1&amp;titulo=Evaluation of nutritional status and adherence to dietary monitoring among patients with Human T-Cell Leukemia Virus type 1 infection&amp;sequencial=4&amp;nomePeriodico=CLINICAL NUTRITION ESPEN" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">DA ROCHA, NAISE OLIVEIRA</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2021</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.12662/2317-3076jhbs.v9i1.3386.p1-5.2021" target="_blank" tabindex="150"></a>DA ROCHA, NAISE OLIVEIRA ; ARAÚJO, ABELARDO QUEIROZ CAMPOS ; SILVA, PAULA SIMPLICIO ; TORRES, PALOMA RIBEIRO ; Santos, Roseli dos ; PINTO, LUCIA MELLO COUTINHO ; Brito, Patrícia Dias ; DE ALMEIDA, CRISTIANE FONSECA ; <b>BACELO, ADRIANA COSTA</b> . Bioimpedância em doenças infecciosas: revisão crítica. JOURNAL OF HEALTH &amp; BIOLOGICAL SCIENCES<sup></sup>, v. 9, p. 1, 2021. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.12662/2317-3076jhbs.v9i1.3386.p1-5.2021&amp;issn=23173076&amp;volume=9&amp;issue=&amp;paginaInicial=1&amp;titulo=Bioimpedância em doenças infecciosas: revisão crítica&amp;sequencial=5&amp;nomePeriodico=JOURNAL OF HEALTH &amp; BIOLOGICAL SCIENCES" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">BACELO, ADRIANA COSTA</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2019</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.1016/j.clnesp.2019.09.007" target="_blank" tabindex="151"></a><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="152">BACELO, ADRIANA COSTA</a></b>; COPLE-RODRIGUES, CLAUDIA DOS SANTOS ; GONÇALVES, JULIANA LAUAR ; BORGES QUINTANA, MARCEL DE SOUZA ; TORRES, PALOMA RIBEIRO ; CARDOSO, CLÁUDIA SANTOS DE AGUIAR ; SILVA, MARLETE PEREIRA DA ; ESPIRITO SANTO, RAQUEL ; ALMEIDA, CRISTIANE FONSECA DE ; SILVA, PAULA SIMPLICIO DA ; BRITO, PATRICIA DIAS DE . Nutritional status of human T-lymphotropic virus 1 patients: A retrospective study. CLINICAL NUTRITION ESPEN<sup><img id="24054577_6" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="24054577" original-title="Clinical Nutrition Espen (2405-4577)<br />Fator de impacto (JCR 2022): 3.0"></sup>, v. 34, p. 32-36, 2019. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1016/j.clnesp.2019.09.007&amp;issn=24054577&amp;volume=34&amp;issue=&amp;paginaInicial=32&amp;titulo=Nutritional status of human T-lymphotropic virus 1 patients: A retrospective study&amp;sequencial=6&amp;nomePeriodico=CLINICAL NUTRITION ESPEN" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000493891000005&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">2</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85072823682&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">3</span></a></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">DELESDERRIER, EMÍLIA</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2019</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.3390/nu11092211" target="_blank" tabindex="153"></a>DELESDERRIER, EMÍLIA ; COPLE-RODRIGUES, CLÁUDIA S. ; OMENA, JULIANA ; KNEIP FLEURY, MARCOS ; BARBOSA BRITO, FLÁVIA ; <b>COSTA BACELO, ADRIANA</b> ; CORREA KOURY, JOSELY ; CITELLI, MARTA . Selenium Status and Hemolysis in Sickle Cell Disease Patients. Nutrients<sup><img id="20726643_7" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="20726643" original-title="Nutrients (2072-6643)<br />Fator de impacto (JCR 2022): 5.9"></sup>, v. 11, p. 2211, 2019. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.3390/nu11092211&amp;issn=20726643&amp;volume=11&amp;issue=&amp;paginaInicial=2211&amp;titulo=Selenium Status and Hemolysis in Sickle Cell Disease Patients&amp;sequencial=7&amp;nomePeriodico=Nutrients" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000487964600048&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">8</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85072516890&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">6</span></a></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">BACELO, ADRIANA COSTA</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2017</span>
# 		<img src="images/curriculo/ico_relevante.gif" style="icone-relevancia">
# 		<a class="icone-producao icone-doi" href="http://dx.doi.org/10.1016/j.clnesp.2016.11.001" target="_blank" tabindex="154"></a><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="155">BACELO, ADRIANA COSTA</a></b>; Brasil, P.E. ; COPLE-RODRIGUES, CLÁUDIA DOS SANTOS ; Georg, I. ; PAIVA, ELIANE ; RAMALHO, ANDREA ; ROLLA, V. . Dietary counseling adherence during tuberculosis treatment: A longitudinal study. Clinical Nutrition ESPEN<sup><img id="24054577_8" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="24054577" original-title="Clinical Nutrition Espen (2405-4577)<br />Fator de impacto (JCR 2022): 3.0"></sup>, v. 17, p. 44-53, 2017. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1016/j.clnesp.2016.11.001&amp;issn=24054577&amp;volume=17&amp;issue=&amp;paginaInicial=44&amp;titulo=Dietary counseling adherence during tuberculosis treatment: A longitudinal study&amp;sequencial=8&amp;nomePeriodico=Clinical Nutrition ESPEN" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000446653000008&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">5</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=85007488674&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">7</span></a></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">GONÇALVES, JULIANA LAUAR</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2017</span><a class="icone-producao icone-doi" href="http://dx.doi.org/10.12957/demetra.2016.22542" target="_blank" tabindex="156"></a>GONÇALVES, JULIANA LAUAR ; CARDOSO, CLÁUDIA SANTOS DE AGUIAR ; <b>BACELO, ADRIANA COSTA</b> ; SANTO, RAQUEL ESPÍRITO ; DE ALMEIDA, CRISTIANE FONSECA ; DA SILVA, PAULA SIMPLÍCIO ; DE BRITO, PATRÍCIA DIAS . COMPOSIÇÃO CORPORAL EM MULHERES TRANSGÊNERO VIVENDO COM HIV/AIDS: UMA DISCUSSÃO DAS TRANSFORMAÇÕES QUE INTERFEREM NA AVALIAÇÃO DO ESTADO NUTRICIONAL. DEMETRA: Alimentação, Nutrição &amp; Saúde<sup></sup>, v. 11, p. 1213-1223, 2017. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.12957/demetra.2016.22542&amp;issn=2238913X&amp;volume=11&amp;issue=&amp;paginaInicial=1213&amp;titulo=COMPOSIÇÃO CORPORAL EM MULHERES TRANSGÊNERO VIVENDO COM HIV/AIDS: UMA DISCUSSÃO DAS TRANSFORMAÇÕES QUE INTERFEREM NA AVALIAÇÃO DO ESTADO NUTRICIONAL&amp;sequencial=9&amp;nomePeriodico=DEMETRA: Alimentação, Nutrição &amp; Saúde" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>10. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Brito, P.D</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2017</span>Brito, P.D ; LAUAR, J. ; Simplicio, P. ; Cardoso, C ; SILVA, M. P. ; <b>BACELO, ADRIANA COSTA</b> ; Santo, R.S. ; ALMEIDA, C. F. . Educação alimentar e nutricional para o controle de comorbidades em pessoas com doenças infecciosas. REVISTA BRASILEIRA EM PROMOÇÃO DA SAÚDE (ONLINE)<sup></sup>, v. 30, p. 1-1, 2017. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=18061230&amp;volume=30&amp;issue=&amp;paginaInicial=1&amp;titulo=Educação alimentar e nutricional para o controle de comorbidades em pessoas com doenças infecciosas&amp;sequencial=10&amp;nomePeriodico=REVISTA BRASILEIRA EM PROMOÇÃO DA SAÚDE (ONLINE)" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="http://www.scopus.com/scopus/inward/citedby.url?partnerID=bb5nvXTn&amp;rel=R6.0.0&amp;doi=10.5020/18061230.2017.6842&amp;md5=574828f861a673f4c485b31927e7204b" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">1</span></a></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>11. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">BACELO, ADRIANA COSTA</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2015</span>
# 		<img src="images/curriculo/ico_relevante.gif" style="icone-relevancia">
# 		<a class="icone-producao icone-doi" href="http://dx.doi.org/10.1371/journal.pone.0134785" target="_blank" tabindex="157"></a><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="158">BACELO, ADRIANA COSTA</a></b>; RAMALHO, ANDREA ; BRASIL, PEDRO EMMANUEL ; COPLE-RODRIGUES, CLÁUDIA DOS SANTOS ; GEORG, INGEBOURG ; PAIVA, ELIANE ; ARGOLO, SHEILA VASQUES LEANDRO ; ROLLA, VALERIA CAVALCANTE . Nutritional Supplementation Is a Necessary Complement to Dietary Counseling among Tuberculosis and Tuberculosis-HIV Patients. PLoS One<sup><img id="19326203_11" src="/buscatextual/images/curriculo/jcr.gif" class="ajaxJCR jcrTip" data-issn="19326203" original-title="Plos One (1932-6203)<br />Fator de impacto (JCR 2022): 3.7"></sup>, v. 10, p. e0134785, 2015. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=10.1371/journal.pone.0134785&amp;issn=19326203&amp;volume=10&amp;issue=&amp;paginaInicial=e0134785&amp;titulo=Nutritional Supplementation Is a Necessary Complement to Dietary Counseling among Tuberculosis and Tuberculosis-HIV Patients&amp;sequencial=11&amp;nomePeriodico=PLoS One" tooltip="Citações a partir de 1996"><b class="label-citacoes">Citações:</b><a href="https://www.webofscience.com/api/gateway?GWVersion=2&amp;SrcApp=PARTNER_APP&amp;SrcAuth=LinksAMR&amp;KeyUT=WOS:000360144000010&amp;DestLinkType=CitingArticles&amp;DestApp=ALL_WOS&amp;UsrCustomerID=83e8e710a8e621f095e069bfe7c1e2bd" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><img src="images/v2/isi.gif"><span class="numero-citacao" data-tipo-ordenacao="1">19</span></a><a href="https://www.scopus.com/inward/citedby.uri?partnerID=HzOxMe3b&amp;scp=84943303286&amp;origin=inward" target="_blank" class="citacaoTip" original-title="CitaÃƒÂ§ÃƒÂµes a partir de 1996"><span class="separador-citacoes">|</span><img src="images/v2/scopus.png"><span class="numero-citacao" data-tipo-ordenacao="3">19</span></a></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>12. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">PAULA, C. G.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2014</span>PAULA, C. G. ; BOCCOLINI, C. S. ; SILVA, A. A. M. ; <b>BACELO, A. C.</b> ; TOSTE, F. ; CAPELLI, J. C. S. . Baixo peso ao nascer: fatores socioeconômicos, assistência pré-natal e nutricional ? uma revisão. REVISTA AUGUSTUS (UNISUAM. ONLINE)<sup></sup>, v. 14, p. 54-66, 2014. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=19811896&amp;volume=14&amp;issue=&amp;paginaInicial=54&amp;titulo=Baixo peso ao nascer: fatores socioeconômicos, assistência pré-natal e nutricional ? uma revisão&amp;sequencial=12&amp;nomePeriodico=REVISTA AUGUSTUS (UNISUAM. ONLINE)" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>13. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Souza,, Veronica Braga de</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2011</span>Souza,, Veronica Braga de ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5949190267189565" target="_blank" tabindex="159">GREGORIO, B. M.</a> ; FERNANDES-DOS-SANTOS, C. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="160">BACELO, A. C.</a></b> ; CATTA-PRETA, M. . Amamentação exclusiva : o cerne para uma melhor qualidade de vida. Nutrição em Pauta<sup></sup>, v. 19, p. 16, 2011. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=16762274&amp;volume=19&amp;issue=&amp;paginaInicial=16&amp;titulo=Amamentação exclusiva : o cerne para uma melhor qualidade de vida&amp;sequencial=13&amp;nomePeriodico=Nutrição em Pauta" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>14. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">VIEIRA T F</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2010</span>VIEIRA T F ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="161">BACELO, A. C.</a></b> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/1454150068332873" target="_blank" tabindex="162">CADIZ R L R M</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/3687045666859962" target="_blank" tabindex="163">CAPELLI, J.</a> . Hidration and Alimentary Habits in pré, during and post-training of a mens´s team Handball University, Rio de Janeiro. The FIEP Bulletin<sup></sup>, v. 80, p. 419-422, 2010. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=02566419&amp;volume=80&amp;issue=&amp;paginaInicial=419&amp;titulo=Hidration and Alimentary Habits in pré, during and post-training of a mens´s team Handball University, Rio de Janeiro&amp;sequencial=14&amp;nomePeriodico=The FIEP Bulletin" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>15. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">chagas</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2010</span>chagas ; VIEIRA T F ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="164">BACELO, A. C.</a></b> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/3687045666859962" target="_blank" tabindex="165">CAPELLI, J.</a> . Hidration and Alimentary Habits in pré, during and post-training an on University female futsal team, Rio de Janeiro. The FIEP Bulletin<sup></sup>, v. 80, p. 423-426, 2010. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=02566419&amp;volume=80&amp;issue=&amp;paginaInicial=423&amp;titulo=Hidration and Alimentary Habits in pré, during and post-training an on University female futsal team, Rio de Janeiro&amp;sequencial=15&amp;nomePeriodico=The FIEP Bulletin" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>16. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">Monteiro, Patrícia Guilherme</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2010</span>Monteiro, Patrícia Guilherme ; Souza-Mello, Vanessa ; GREGÓRIO, Bianca Martins ; Nascimento, FAM ; Reis, GSO ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="166">BACELO, A. C.</a></b> ; CATTA-PRETA, M. . Caracteres da síndrome metabólica e tratamento não farmacológico: uma revisão da literatura.. Revista augustus (UNISUAM. Online)<sup></sup>, v. 14, p. 11-23, 2010. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=19811896&amp;volume=14&amp;issue=&amp;paginaInicial=11&amp;titulo=Caracteres da síndrome metabólica e tratamento não farmacológico: uma revisão da literatura.&amp;sequencial=16&amp;nomePeriodico=Revista augustus (UNISUAM. Online)" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>17. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">CADIZ R L R M</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2009</span><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/1454150068332873" target="_blank" tabindex="167">CADIZ R L R M</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4704514110960279" target="_blank" tabindex="168">MELLO, D. B.</a> ; RECH, G. F. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="169">BACELO, A. C.</a></b> ; CAPELLI, J. C. S. . BODY IMAGE PERCEPTION OF PRACTICING ADOLESCENTS OF PHYSICAL ACTIVITY OF A SCHOOL OF PRIVATE NET OF TEACHING, RIO DE JANEIRO. The FIEP Bulletin<sup></sup>, v. 79, p. 21-24, 2009. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=02566419&amp;volume=79&amp;issue=&amp;paginaInicial=21&amp;titulo=BODY IMAGE PERCEPTION OF PRACTICING ADOLESCENTS OF PHYSICAL ACTIVITY OF A SCHOOL OF PRIVATE NET OF TEACHING, RIO DE JANEIRO&amp;sequencial=17&amp;nomePeriodico=The FIEP Bulletin" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>18. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">ALMEIDA, M. F. L.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2008</span><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/1794231232150051" target="_blank" tabindex="170">ALMEIDA, M. F. L.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9062811018583569" target="_blank" tabindex="171">PEREIRA, S. E.</a> ; ANASTASIO, A. ; TOSTE, F. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/3687045666859962" target="_blank" tabindex="172">CAPELLI, J.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5683437160866054" target="_blank" tabindex="173">MORAES, V. M.</a> ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="174">BACELO, A. C.</a></b> ; PARREIRA, M. T. ; GONCALVES, M. . Avaliação do Estado Nutricional de Funcionários participantes de um programa de qualidade de vida de uma Instituição de ensino superior da cidade do Rio de Janeiro. Revista Augustus (Rio de Janeiro)<sup></sup>, v. 1, p. 70, 2008. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=1415398X&amp;volume=1&amp;issue=&amp;paginaInicial=70&amp;titulo=Avaliação do Estado Nutricional de Funcionários participantes de um programa de qualidade de vida de uma Instituição de ensino superior da cidade do Rio de Janeiro&amp;sequencial=18&amp;nomePeriodico=Revista Augustus (Rio de Janeiro)" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>19. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">TOSTE, F.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2007</span>TOSTE, F. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="175">BACELO, A. C.</a></b> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/3687045666859962" target="_blank" tabindex="176">CAPELLI, J.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5683437160866054" target="_blank" tabindex="177">MORAES, V. M.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/2514251864249834" target="_blank" tabindex="178">PORTE, A.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9062811018583569" target="_blank" tabindex="179">PEREIRA, S. E.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4303234837741559" target="_blank" tabindex="180">CARDOSO, Claudia Santos de Aguiar</a> ; Seixas, P ; AMORIM, A. R. ; TOSTE, F. P. ; COPLE, Claudia dos Santos . A Contribuição da Nutrição nos Casos de Paraparesia Espástica Tropical por HTLV-1. Revista Augustus (Rio de Janeiro. Impresso)<sup></sup>, v. 12, p. 9-15, 2007. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=1415398X&amp;volume=12&amp;issue=&amp;paginaInicial=9&amp;titulo=A Contribuição da Nutrição nos Casos de Paraparesia Espástica Tropical por HTLV-1&amp;sequencial=19&amp;nomePeriodico=Revista Augustus (Rio de Janeiro. Impresso)" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>20. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">ALMEIDA, M. F. L.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2007</span><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/1794231232150051" target="_blank" tabindex="181">ALMEIDA, M. F. L.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/3687045666859962" target="_blank" tabindex="182">CAPELLI, J.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9062811018583569" target="_blank" tabindex="183">FREITAS, S. E. P.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5683437160866054" target="_blank" tabindex="184">MORAES, V. M.</a> ; TOSTE, F. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="185">BACELO, A. C.</a></b> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/2514251864249834" target="_blank" tabindex="186">PORTE, A.</a> . Estado Nutricional de Adolescentes Atendidos no Ambulatório de Nutrição do Centro de Educação Física Adalberto Nunes. Revista Augustus (Rio de Janeiro)<sup></sup>, v. 12, p. 16-21, 2007. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=1415398X&amp;volume=12&amp;issue=&amp;paginaInicial=16&amp;titulo=Estado Nutricional de Adolescentes Atendidos no Ambulatório de Nutrição do Centro de Educação Física Adalberto Nunes&amp;sequencial=20&amp;nomePeriodico=Revista Augustus (Rio de Janeiro)" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>21. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">CAPELLI, J.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2007</span><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/3687045666859962" target="_blank" tabindex="187">CAPELLI, J.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/1794231232150051" target="_blank" tabindex="188">ALMEIDA, M. F. L.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5683437160866054" target="_blank" tabindex="189">MORAES, V. M.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9062811018583569" target="_blank" tabindex="190">FREITAS, S. E. P.</a> ; TOSTE, F. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="191">BACELO, A. C.</a></b> ; REIS, R. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7872661608669228" target="_blank" tabindex="192">MATOS, P. E.</a> ; SIMOES, M. M. ; CAMPOS, E. E. . Projeto Saúde d=na Praça: Uma Ação Social Voltada para Avaliação Nutricional de Adultos Residentes em uma Comunidade Carente. Revista Augustus (Rio de Janeiro)<sup></sup>, v. 12, p. 31-35, 2007. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=1415398X&amp;volume=12&amp;issue=&amp;paginaInicial=31&amp;titulo=Projeto Saúde d=na Praça: Uma Ação Social Voltada para Avaliação Nutricional de Adultos Residentes em uma Comunidade Carente&amp;sequencial=21&amp;nomePeriodico=Revista Augustus (Rio de Janeiro)" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>22. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">FREITAS, S. E. P.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2007</span><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9062811018583569" target="_blank" tabindex="193">FREITAS, S. E. P.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/3687045666859962" target="_blank" tabindex="194">CAPELLI, J.</a> ; TOSTE, F. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/1794231232150051" target="_blank" tabindex="195">ALMEIDA, M. F. L.</a> ; ANASTASIO, A. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5683437160866054" target="_blank" tabindex="196">MORAES, V. M.</a> ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="197">BACELO, A. C.</a></b> ; PARREIRA, M. T. . Avaliação de Adesão ao Tratamento Nutricional em uma Clínica Escola de uma Instituição de Ensino Superior. Revista Augustus (Rio de Janeiro)<sup></sup>, v. 12, p. 53-59, 2007. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=1415398X&amp;volume=12&amp;issue=&amp;paginaInicial=53&amp;titulo=Avaliação de Adesão ao Tratamento Nutricional em uma Clínica Escola de uma Instituição de Ensino Superior&amp;sequencial=22&amp;nomePeriodico=Revista Augustus (Rio de Janeiro)" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>23. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">MORAES, V. M.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2007</span><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5683437160866054" target="_blank" tabindex="198">MORAES, V. M.</a> ; TOSTE, F. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/3687045666859962" target="_blank" tabindex="199">CAPELLI, J.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9062811018583569" target="_blank" tabindex="200">FREITAS, S. E. P.</a> ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="201">BACELO, A. C.</a></b> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/1794231232150051" target="_blank" tabindex="202">ALMEIDA, M. F. L.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/2514251864249834" target="_blank" tabindex="203">PORTE, A.</a> ; DIAS, E. ; SENNA, M. ; PIERRONI, L. . Impacto do Tratamento Nutricional na Redução de Peso Corporal de Pacientes Atendidos no Ambulatório de Nutrição do Centro Presbiteriano Beneficente - Rio de Janeiro. Revista Augustus (Rio de Janeiro)<sup></sup>, v. 12, p. 60-66, 2007. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=1415398X&amp;volume=12&amp;issue=&amp;paginaInicial=60&amp;titulo=Impacto do Tratamento Nutricional na Redução de Peso Corporal de Pacientes Atendidos no Ambulatório de Nutrição do Centro Presbiteriano Beneficente - Rio de Janeiro&amp;sequencial=23&amp;nomePeriodico=Revista Augustus (Rio de Janeiro)" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>24. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">BACELO, A. C.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2006</span><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="204">BACELO, A. C.</a></b>; Moriguchi ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7183837354797239" target="_blank" tabindex="205">Poli Figueredo</a> ; COPLE, Claudia dos Santos . Impacto da Redução da Ingestão de Sódio no Controle da Hipertensão Arterial Sistêmica em Idosos. Revista de Metabolismo e Nutrição<sup></sup>, v. 8, p. 114-119, 2006. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=01045989&amp;volume=8&amp;issue=&amp;paginaInicial=114&amp;titulo=Impacto da Redução da Ingestão de Sódio no Controle da Hipertensão Arterial Sistêmica em Idosos&amp;sequencial=24&amp;nomePeriodico=Revista de Metabolismo e Nutrição" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <div class="artigo-completo">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>25. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><span class="informacao-artigo" data-tipo-ordenacao="importancia"></span><span class="informacao-artigo" data-tipo-ordenacao="autor">BACELO, A. C.</span><span class="informacao-artigo" data-tipo-ordenacao="ano">2002</span><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="206">BACELO, A. C.</a></b>. efeitos prejudiciais do sal na alimentação, além da hipertensão. nutri-recall<sup></sup>, v. 1, p. 60-61, 2002. <span class="citado" cvuri="/buscatextual/servletcitacoes?doi=&amp;issn=16773268&amp;volume=1&amp;issue=&amp;paginaInicial=60&amp;titulo=efeitos prejudiciais do sal na alimentação, além da hipertensão&amp;sequencial=25&amp;nomePeriodico=nutri-recall" tooltip="Citações a partir de 1996"></span></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# </div>
# <div class="cita-artigos">
# <b><a name="LivrosCapitulos" tabindex="207"></a>Livros publicados/organizados ou edições</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-producao icone-doi" href="http://dx.doi.org/10.29327/522782" target="_blank" tabindex="208"></a>BRENNER, Augusto Madke ; GONZATTI, Carolina ; NESELLO, Alberto Chitolina ; FURTUNATO, Luis Fillipy ; ELVIR, Francisco Alfonso Rodriguez ; ROCHA, Naise Oliveira da ; Araújo, Abelardo de Queiroz Campos ; TORRES, PALOMA RIBEIRO ; Santos, Roseli dos ; Paiva, Eliane de Souza ; <b>BACELO, ADRIANA COSTA</b> ; SILVA, Carlos Eduardo da ; ARAÚJO, Isabelle Cássia Viana de ; Mota, Aline Carolina Castro ; SANTOS, Eliete Viana dos ; ALVES, João Victor Moura ; MORAES, Joel Campos de ; ALVES, Krisnna Mariana Aranda ; DANTAS, Alison Rener Araujo ; DIAS, Filipe Pereira da Silva . Saúde do Ensino à Pesquisa - Volume 1. 1. ed. Editora Pasteur, 2020. v. 1. 536p . </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b><a name="LivrosCapitulos" tabindex="209"></a>Capítulos de livros publicados</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="210">BACELO, A. C.</a></b>; PINTO, L. M. C. ; TOSTE, F. ; Fonseca, C.C. . Doenças Infecciosas e Parasitárias. In: Ilanna Marques Gomes da Rocha. (Org.). Modulação Intestinal ? Do Sequenciamento Genético à Prática Clínica. 1ed.Rio de Janeiro: Rubio, 2024, v. 1, p. 381-390. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-producao icone-doi" href="http://dx.doi.org/10.35170/ss.ed.9786586283068.14" target="_blank" tabindex="211"></a>TORRES, PALOMA RIBEIRO ; Araújo, Abelardo de Queiroz Campos ; Rocha, Naíse Oliveira da ; Santos, Roseli dos ; Paiva, Eliane de Souza ; Brito, Patrícia Dias ; <b>BACELO, ADRIANA COSTA</b> . OCORRÊNCIA DE ALTERAÇÕES GASTROINTESTINAIS E SUA CORRELAÇÃO COM IMC EM PACIENTES HTLV SINTOMÁTICOS ATENDIDOS NO AMBULATÓRIO DE NUTRIÇÃO DE UM CENTRO DE REFERÊNCIA DO RIO DE JANEIRO. Atualidades em Medicina Tropical no Brasil: Interdisciplinaridades. 1ed.: Stricto Sensu Editora, 2020, v. , p. 206-213. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Brito, P.D ; <b>BACELO, ADRIANA COSTA</b> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4303234837741559" target="_blank" tabindex="212">CARDOSO, Claudia Santos de Aguiar</a> ; Simplicio, P. ; Santo, R.S. ; LAUAR, J. . Uso de Alimentos Funcionais por Pessoas vivendo com HIV/AIDS. In: Neusa Maria Brunoro Costa; Carla de Oliveira Barbosa Rosa. (Org.). Alimentos funcionais - Componentes Bioativos e Efeitos fisiológicos. 2ed.Rio de Janeiro: Rubio, 2016, v. 1, p. 381-395. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="213">BACELO, A. C.</a></b>; Fonseca, C.C. ; COPLE, Claudia dos Santos . fome oculta e HIV. In: Andrea Ramalho. (Org.). Fome Oculta e doenças crônicas: um novo desafio. 1ed.Rio de Janeiro: Atheneu, 2014, v. 1, p. 135-144. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Durgante P. C. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="214">BACELO, A. C.</a></b> . . Terapia Nutricional no Paciente em Pos-operatorio de cirurgia cardiaca. In: Joao Carlos Vieira da Costa Guaragna. (Org.). Pos Operatorio em Cirurgia Cardiaca. 1ed.Rio de janeiro: Editora Guanabara Koogan S.A, 2005, v. 1, p. 329-334. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="215">BACELO, A. C.</a></b>. Avaliação do Estado Nutricional. In: Álvaro Réa Neto; Ciro Leite Mendes; Ederlon Alves de Carvalho Rezende; Fernando Suparregui Dias. (Org.). Monitorização em UTI. 1ed.Rio de Janeiro: REVINTER, 2004, v. 1, p. 255-259. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="216">BACELO, A. C.</a></b>. Interação Droga e Nutriente na Terceira Idade. In: Newton Luiz Terra; Beatriz Dornelles. (Org.). Envelhecimento Bem-Sucedido. 1ed.Porto Alegre: Edipucrs, 2003, v. 1, p. 401-407. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="217">BACELO, A. C.</a></b>; Milani . As Fibras e o seu Intestino. In: Beatriz Dornelles; Gilberto José Corrêa da Costa. (Org.). Investindo no Envelhecimento Saudável. 1ed.Porto Alegre: EDIPUCRS, 2003, v. 1, p. 309-312. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b><a name="TrabalhosPublicadosAnaisCongresso" tabindex="218"></a>Trabalhos completos publicados em anais de congressos</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">POZZA, R. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="219">BACELO, A. C.</a></b> ; POITEVIN, A. ; MORALES, J. V. ; TRAESEL, M. ; pimentel, N ; BRUNO, R. ; AVILA, D. O. D. . Nefroproteção: evidencias e recomendações. In: IX Jornada Gaúcha de Nefrologia e Enfermagem em Nefrologia, 2003, Porto Alegre. Revista da IX Jornada Gaúcha de Nefrologia e Enfermagem em Nefrologia, 2003. v. 1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b><a name="TrabalhosPublicadosAnaisCongresso" tabindex="220"></a>Resumos publicados em anais de congressos</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/7872661608669228" target="_blank" tabindex="221">MATOS, P. E.</a> ; SOUZA, G. ; COSTA, S. O. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="222">BACELO, A. C.</a></b> ; LEAL, A. ; TOSTE, F. . FOOD BANK - A TOOL IN PROMOTING FOOD SAFETY. In: XI Congresso Ibero Americano de Extensão Universitária, 2011, Santa Fé - Argentina. FOOD BANK - A TOOL IN PROMOTING FOOD SAFETY, 2011. p. 18-18. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">
# 		<img src="images/curriculo/ico_relevante.gif" style="icone-relevancia">
# 		<a class="icone-producao icone-doi" href="http://dx.doi.org/C:\Users\Adriana Bacelo\Downloads\Oasis, The Online Abstract Submission System4 (2).mht" target="_blank" tabindex="223"></a><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="224">BACELO, A. C.</a></b>; RAMALHO, R. A. ; COPLE, Claudia dos Santos ; ENNES, E. ; ARGOLO, S.V.L. ; FREIRE, M. ; PAIVA, E. ; INGEBOURG, G. ; ROLLA, V. . Interference of oxidative stress in iron protein depletion in Pacients Co-infected HIV/TB before and after drug intervention and nutritional guidance. In: AACC American Association for Clinical Chemistry, 2011, Atlanta - Georgia. 2011 AACC Annual Meeting. Atlanta - Georgia: OASIS, 2011. v. 11A. p. 1-1. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b><a name="ApresentacoesTrabalho" tabindex="225"></a>Apresentações de Trabalho</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="226">BACELO, ADRIANA COSTA</a></b>. Envelhecimento saudável: Nutrigenética e Nutrigenômica. 2022. (Apresentação de Trabalho/Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="227">BACELO, A. C.</a></b>. Nutrição de precisão: uma realidade ou uma perspectiva do futuro. 2022. (Apresentação de Trabalho/Simpósio). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">TORRES, PALOMA RIBEIRO ; Araújo, A.Q.C ; ROCHA, N. O. ; SANTOS, R. ; PAIVA, E. ; LEITE, L. C. C. B. ; LIMA, M. A. S. D. ; Cardoso, C ; BRITO, PATRICIA DIAS DE ; <b>BACELO, A. C.</b> . Ocorrência de alterações gastrointestinais e sua correlação com IMC em pacientes HTLV sintomáticos atendidos no ambulatório de nutrição de um centro de referencia do Rio de Janeiro. 2019. (Apresentação de Trabalho/Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">ROCHA, C. ; <b>BACELO, A. C.</b> ; TORRES, PALOMA RIBEIRO ; SANTOS, R. ; PAIVA, E. ; LEITE, L. C. C. B. ; SILVA, M. T. T. ; Santo, R.S. ; SILVA, MARLETE PEREIRA DA ; SILVA, PAULA SIMPLICIO DA ; ALMEIDA, CRISTIANE FONSECA DE ; Araújo, A.Q.C . Avaliação antropométrica de pacientes HTLV sintomáticos atendidos no ambulatório de nutrição de um centro de referencia do Rio de Janeiro. 2019. (Apresentação de Trabalho/Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">SILVA, M. P. ; Simplicio, P. ; <b>BACELO, A. C.</b> ; Santo, R.S. ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4303234837741559" target="_blank" tabindex="228">CARDOSO, Claudia Santos de Aguiar</a> . Segurança Alimentar e Estado Nutricional de pacientes com doença de chagas internados no centro de referencia do Rio de Janeiro. 2016. (Apresentação de Trabalho/Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="229">BACELO, ADRIANA COSTA</a></b>. Nutrição em Cirurgia bariátrica. 2015. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="230">BACELO, ADRIANA COSTA</a></b>. Nutrição em gestantes pós bariatrica. 2015. (Apresentação de Trabalho/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="231">BACELO, A. C.</a></b>; BRASIL, PEDRO EMMANUEL ; ARGOLO, SHEILA VASQUES LEANDRO ; COPLE-RODRIGUES, CLÁUDIA DOS SANTOS ; PAIVA, ELIANE ; ROLLA, V. ; RAMALHO, ANDREA ; Georg, I. ; LAUAR, J. ; ALMEIDA, C. F. . Body iron status and immunologic system in TB patients infected by Human Immunodeficiency Virus. 2015. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="232">BACELO, A. C.</a></b>; Almeida, C. ; Brito, P.D ; Santo, R.S. ; Dantas, J. ; Simplicio, P. ; Cardoso, C . Impacto do Estado nutricional na Evolução Clinica e Imunológica de Pacientes Graves com HIV. 2013. (Apresentação de Trabalho/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>10. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="233">BACELO, A. C.</a></b>; COPLE, Claudia dos Santos ; PAIVA, E. ; MAIA, I. R. ; RAMALHO, R. A. ; ROLLA, V. . ESTADO CORPORAL DE FERRO E VITAMINA A EM PESSOAS CO-INFECTADAS HIV/TB APÓS INTERVENÇÃO NUTRICIONAL NUM CENTRO DE REFERÊNCIA DE TUBERCULOSE NO RIO DE JANEIRO: ESTUDO PRELIMINAR. 2012. (Apresentação de Trabalho/Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>11. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="234">BACELO, A. C.</a></b>. Evaluación ellas nutricional pacientes con VIH en contra. 2012. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>12. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">
# 		<img src="images/curriculo/ico_relevante.gif" style="icone-relevancia">
# 		<b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="235">BACELO, A. C.</a></b>; RAMALHO, R. A. ; COPLE, Claudia dos Santos ; ENNES, E. ; ARGOLO, S.V.L. ; FREIRE, M. ; PAIVA, E. ; INGEBOURG, G. ; ROLLA, V. . Interference of oxidative stress in iron protein depletion in Pacients Co-infected HIV/TB before and after drug intervention and nutritional guidance. 2011. (Apresentação de Trabalho/Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>13. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">
# 		<img src="images/curriculo/ico_relevante.gif" style="icone-relevancia">
# 		<b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="236">BACELO, A. C.</a></b>; RAMALHO, R. A. ; COPLE, Claudia dos Santos ; MAIA, I. R. ; PAIVA, E. ; INGEBOURG, G. ; ARGOLO, S.V.L. ; FREIRE, M. ; ROLLA, V. . ASSESSMENT OF GLOBAL NUTRITION, VITAMIN E AND SELENIUM IN HIV/TUBERCULOSIS PATIENTS ON DIETARY COUNSELING. 2011. (Apresentação de Trabalho/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>14. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">MAIA, I. R. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="237">BACELO, A. C.</a></b> ; RAMALHO, R. A. ; COPLE, Claudia dos Santos ; PAIVA, E. ; INGEBOURG, G. ; ARGOLO, S.V.L. ; FREIRE, M. ; ROLLA, V. . AVALIAÇÃO DA OCORRÊNCIA DE ANEMIA EM PACIENTES COM TUBERCULOSE DURANTE O TRATAMENTO MEDICAMENTOSO. 2011. (Apresentação de Trabalho/Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>15. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">ROCHA, C. ; SOUZA, E. ; ABDALA, E. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="238">BACELO, A. C.</a></b> ; SOUZA, G. ; TOSTE, F. . Avaliação do Perfil Metabólico de Pacientes Submetidos à derivação Gastrica do Tipo Fobi-Capela. 2009. (Apresentação de Trabalho/Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>16. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">LEAL, A. ; CATTA-PRETA, M. ; REIS, G. S. O. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="239">BACELO, A. C.</a></b> . Análise Sensorial de Dois Suplementos Hipercalóricos e Hiperprotéicos. 2009. (Apresentação de Trabalho/Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>17. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">POITEVIN, A. ; TOSTE, F. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="240">BACELO, A. C.</a></b> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/1794231232150051" target="_blank" tabindex="241">ALMEIDA, M. F. L.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/9062811018583569" target="_blank" tabindex="242">PEREIRA, S. E.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/3687045666859962" target="_blank" tabindex="243">CAPELLI, J.</a> ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5683437160866054" target="_blank" tabindex="244">MORAES, V. M.</a> . Estudo das propriedades funcionais de sementes de CArica papaya L. (mamão) e de Cucurbita sp (abóboras).. 2007. (Apresentação de Trabalho/Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>18. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="245">BACELO, A. C.</a></b>; COPLE, Claudia dos Santos ; SOARES, Marcelo . Comparação dos Diferentes métodos de Avaliação estatural no adulto por Faixa etária. 2004. (Apresentação de Trabalho/Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>19. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="246">BACELO, A. C.</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4303234837741559" target="_blank" tabindex="247">CARDOSO, Claudia Santos de Aguiar</a> ; LESSA, Rosane Quaranta ; COPLE, Claudia dos Santos ; LOPES, Aline dos Santos . Perfil dos Pacientes com doença de Chagas atendidos no Ambulatório de Nutrição do IPEC - FIOCRUZ. 2004. (Apresentação de Trabalho/Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>20. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="248">BACELO, A. C.</a></b>. Nutrição Enteral em Cardiologia. 2003. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>21. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="249">BACELO, A. C.</a></b>. Avaliação Nutricional no Doente Crítico. 2003. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>22. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="250">BACELO, A. C.</a></b>. Avaliação e Monitoração da Terapia Nutricional. 2003. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>23. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="251">BACELO, A. C.</a></b>. Interação Fármado Nutriente no Idoso. 2003. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>24. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="252">BACELO, A. C.</a></b>. Semiologia Nutricional. 2003. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>25. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="253">BACELO, A. C.</a></b>. Relação Peso s Altura, Dobras Cutâneas, Bioimpedância e Exame Físico. 2003. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>26. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="254">BACELO, A. C.</a></b>. Avaliação Nutricional Subjetiva e Sinais Clínicos de Desnutrição. 2003. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>27. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="255">BACELO, A. C.</a></b>. Estudos Nutricionais com Idosos. 2003. (Apresentação de Trabalho/Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>28. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="256">BACELO, A. C.</a></b>. Alimentação e Qualidade de Vida no Trabalho. 2003. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>29. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">AVILA, D. O. D. ; POITEVIN, A. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="257">BACELO, A. C.</a></b> ; MORALES, J. V. ; TRAESEL, M. ; pimentel, N ; POZZA, R. ; BRUNO, R. . Nefrologia Baseada em Evidências. 2003. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>30. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="258">BACELO, A. C.</a></b>. Terapia nutricional conservadora para insuficiência renal crônica. 2002. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>31. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="259">BACELO, A. C.</a></b>. Atualização em Nutrição Geriátrica. 2002. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>32. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="260">BACELO, A. C.</a></b>. Interações entre drogas e nutrientes. 2002. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>33. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="261">BACELO, A. C.</a></b>. Perfil nutricional dos pacientes em programa de diálise no HSL da PUCRS - a análise retrospectiva. 2002. (Apresentação de Trabalho/Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>34. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="262">BACELO, A. C.</a></b>. Nutrição Enteral em Cardiopatia. 2002. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>35. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="263">BACELO, A. C.</a></b>. Perfil Nutricional dos Pacientes em Programa de Diálise no HSL da PUCRS. 2002. (Apresentação de Trabalho/Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>36. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="264">BACELO, A. C.</a></b>. Orientações Nutricionais para Pacientes em HD. 2002. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>37. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="265">BACELO, A. C.</a></b>. Atualização em Nutrição Geriatrica. 2002. (Apresentação de Trabalho/Conferência ou palestra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>38. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Milani ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="266">BACELO, A. C.</a></b> ; Durgante P. C. ; VINCIPROVA, A. P. ; BUENO, F. . Perfil dos Pacientes atendidos no Ambulatório de Nutrição em Geriatria do Hospital São Lucas da PUCRS. 2002. (Apresentação de Trabalho/Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>39. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">ANTUNES, M. T. ; Milani ; Durgante P. C. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="267">BACELO, A. C.</a></b> . Alimentação saudável na Terceira Idade. 2001. (Apresentação de Trabalho/Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>40. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="268">BACELO, A. C.</a></b>. Manipuladores de Alimentos. 1998. (Apresentação de Trabalho/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b><a name="OutrasProducoesBibliograficas" tabindex="269"></a>Outras produções bibliográficas</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="270">COSTA BACELO, ADRIANA</a></b>; Araújo, A.Q.C ; TORRES, P. R. ; Brito, P.D ; Almeida, C. ; Fonseca, C.C. ; QUINTANA, M. ; COPLE-RODRIGUES, CLAUDIA DOS SANTOS ; Brasil, P.E. ; ROCHA, N. O. . ASSOCIATION BETWEEN NUTRITIONAL STATE AND PHASE ANGLE IN SYMPTOMATIC AND ASYMPTOMATIC HTLV-1 INFECTED PATIENTS.
# 							Yale:
# 									medRxiv - Cold Spring Harbor Laboratory,  2022 (Preprint). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="271">BACELO, A. C.</a></b>; RAMALHO, ANDREA ; BRASIL, PEDRO EMMANUEL ; COPLE, Claudia dos Santos ; Georg, I. ; PAIVA, E. ; ARGOLO, SHEILA VASQUES LEANDRO ; ROLLA, V. . Nutritional Supplementation Is a Necessary Complement to Dietary Counseling among Tuberculosis and Tuberculosis-HIV Patients..
# 							Nova York:
# 									World Biomedical Frontiers,  2015 (Resumo expandido publicado em site de organização para divulgação das ciências médicas). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <a name="ProducaoTecnica" tabindex="272"></a><br class="clear"><div class="inst_back">
# <b>Produção técnica</b>
# </div>
# <div class="cita-artigos">
# <b><a name="ProdutosTecnologicos" tabindex="273"></a>Produtos tecnológicos</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">MOLTER, A. R. J. O. ; FONSECA, C. F. ; SILVA, M. P. ; PATRICIA, BRITO ; Brasil, P.E. ; <b>ADRIANA, BACELO</b> . ARIETI (Aplicativo móvel de Representação da Ingestão alimEntar de pacienteTes Internados). 2023. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">MOLTER, A. R. J. O. ; Brito, Patrícia Dias ; SILVA, M. P. ; Brasil, P.E. ; FONSECA, C. F. ; <b>COSTA BACELO, ADRIANA</b> . DESENVOLVIMENTO DE UM APLICATIVO MÓVEL DE REPRESENTAÇÃO DA INGESTÃO ALIMENTAR DE PACIENTES INTERNADOS: TESTE DE CONFIABILIDADE. 2022. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b><a name="EntrevistasMesasRedondas" tabindex="274"></a>Entrevistas, mesas redondas, programas e comentários na mídia</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="275">COSTA BACELO, ADRIANA</a></b>; GOMES, I. . Nutrição e Longevidade. 2022. 
						
# 							(Programa de rádio ou TV/Entrevista).
# 						<img class="icone-video" src="http://estatico.cnpq.br/imagens/ico_pw_video.png" onclick="mostrarPreviaVideo('https://youtu.be/MSJP_b5uOQw')"></span>
# </div>
# <br class="clear">
# <br class="clear">
# <a name="DemaisProducaoTecnica" tabindex="276"></a><br class="clear"><div class="inst_back">
# <b>Demais tipos de produção técnica</b>
# </div>
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="277">BACELO, A. C.</a></b>. MIcronutrientes e Processos Infecciosos. 2008. . </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="278">BACELO, A. C.</a></b>. Nutrição no Idoso. 2006. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="279">BACELO, A. C.</a></b>. Cuidados emNutrição em Doenças Infecto-Parasitárias. 2006. (Curso de curta duração ministrado/Extensão). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="280">BACELO, A. C.</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4303234837741559" target="_blank" tabindex="281">CARDOSO, Claudia Santos de Aguiar</a> ; Seixas, P ; Liporage, J ; Bozza, F . Manual de Rotinas em Terapia Nutricional.
# 							2006. (Desenvolvimento de material didático ou instrucional - Manual de Rotinas). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="282">BACELO, A. C.</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4303234837741559" target="_blank" tabindex="283">CARDOSO, Claudia Santos de Aguiar</a> ; LESSA, Rosane Quaranta . Protocolo de Avaliação nutricional.
# 							2005. (Desenvolvimento de material didático ou instrucional - Protocolo). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="284">BACELO, A. C.</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4303234837741559" target="_blank" tabindex="285">CARDOSO, Claudia Santos de Aguiar</a> ; LESSA, Rosane Quaranta . A Dieta faz parte do tratamento e da prevenção do diabetes.
# 							2004. (Desenvolvimento de material didático ou instrucional - Folder instrucional para Diabetes). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="286">BACELO, A. C.</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4303234837741559" target="_blank" tabindex="287">CARDOSO, Claudia Santos de Aguiar</a> ; LESSA, Rosane Quaranta . Orientação para o controle do consumo de potássio.
# 							2004. (Desenvolvimento de material didático ou instrucional - Folder de orientação nutricional). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="288">BACELO, A. C.</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4303234837741559" target="_blank" tabindex="289">CARDOSO, Claudia Santos de Aguiar</a> ; LESSA, Rosane Quaranta . Educação nutricional.
# 							2004. (Desenvolvimento de material didático ou instrucional - Folder de orientação nutricional). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="290">BACELO, A. C.</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4303234837741559" target="_blank" tabindex="291">CARDOSO, Claudia Santos de Aguiar</a> ; LESSA, Rosane Quaranta . Preciso aumentar meu peso?.
# 							2004. (Desenvolvimento de material didático ou instrucional - Folder de orientação nutricional). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>10. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="292">BACELO, A. C.</a></b>. Curso Avançado de Terapia Nutricional - Modulo Paciente Critico - Avaliação e Monitoramento da Terapia Nutricional. 2003. (Curso de curta duração ministrado/Extensão). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>11. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="293">BACELO, A. C.</a></b>; COLOSSI, Fernanda . Como Viever Bem. 2003. (Editoração/Periódico). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>12. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="294">BACELO, A. C.</a></b>. Resgatando o Cuidado. 2002. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>13. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="295">BACELO, A. C.</a></b>. Efeitos prejudiciais do sal na alimentação, além da hipertensão. 2002. (Editoração/Periódico). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>14. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="296">BACELO, A. C.</a></b>. Insuficiência Renal Crônica - Tratamento Conservador. 2001.
# 							(Palestra).
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>15. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="297">BACELO, A. C.</a></b>. Alimentação e Qualidade no Trabalho. 2001.
# 							(Palestra).
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>16. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="298">BACELO, A. C.</a></b>. Alimentação no Diabetes. 2000.
# 							(Palestra).
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>17. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="299">BACELO, A. C.</a></b>. Encontro de Adolescentes da CBCM - Oficina sobre o Tema Saúde e Alimentação. 1998. (Curso de curta duração ministrado/Outra). </span>
# </div>
# <br class="clear">
# <br class="clear">
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="Bancas" tabindex="300">
# <h1 tabindex="0">Bancas</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <a name="ParticipacaoBancasTrabalho" tabindex="301"></a><br class="clear"><div class="inst_back">
# <b>Participação em bancas de trabalhos de conclusão</b>
# </div>
# <div class="cita-artigos">
# <b>Mestrado</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>COSTA BACELO, ADRIANA</b>; SILVA, G. M. S.; OLIVEIRA, R. V. C.; OLIVEIRA, C. F..  Participação em banca de Monique da Silva Monção. Efeito do confinamento da COVID-19 sobre hábitos alimentares, segurança alimentar e nutricional e adesão às orientações dietéticas em pacientes com doenças infecciosas acompanhadas através de uma estratégia de atendimento não presencial. 2022. Dissertação (Mestrado em Pesquisa Clinica em Doenças Infecciosas) - Fundação Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>COSTA BACELO, ADRIANA</b>; GOMES, M.; CUNHA, A. L..  Participação em banca de Alessandra Araujo de Leonardo. Elaboração, Aprimoramento e Validação de Conteúdo de Instrumentos de Atendimento Nutricional para Identificação de Fatores de Risco Associados à Diminuição da Densidade Mineral Óssea em Crianças e Adolescentes com HIV/AIDS. 2021. Dissertação (Mestrado em Pesquisa Clinica em Doenças Infecciosas) - Fundação Oswaldo Cruz. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>BACELO, ADRIANA COSTA</b>; SOARES, A.; AVESANI, C..  Participação em banca de Estefania Costa Botelho. Avaliação da composição corporal e sua correlação com a função pulmonar em pacientes adultos com doença falciforme. 2017. Dissertação (Mestrado em Ciências Médicas) - Universidade do Estado do Rio de Janeiro. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>BACELO, ADRIANA COSTA</b>; SOARES, A.; AVESANI, C.; FLEURY, M..  Participação em banca de Karen Cordovil. composição corporal, gasto energético e indicadores de gravidade em adultos com doença falciforme. 2016. Dissertação (Mestrado em Ciências Médicas) - Universidade do Estado do Rio de Janeiro. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b>Teses de doutorado</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b>BACELO, A. C.</b>.  Participação em banca de Analucia Gomes Lopes Oliveira. Evolução do estado nutricional de pacientes com Leishmaniose Tegumentar Americana em tratamento. 2014. Tese (Doutorado em DOUTORADO EM CIÊNCIAS - PESQUISA CLINICA EM DOENÇAS INFECCIOSAS)  - Instituto Nacional de Infectologia Evandro Chagas. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b>Trabalhos de conclusão de curso de graduação</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="302">BACELO, A. C.</a></b>.  Participação em banca de Fabio Luis Francisco.O Papel do Enfermeiro nas Orientações a um Adulto diante do Diagnóstico de Tuberculosse.
# 							2009. Trabalho de Conclusão de Curso (Graduação em Nutrição)  - Centro Universitário Augusto Motta. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="303">BACELO, A. C.</a></b>.  Participação em banca de Márcia Paula de Lima.O Estresse em Cuidadores de Idosos com Pendência para as Atividades Básicas de Vida Diária.
# 							2009. Trabalho de Conclusão de Curso (Graduação em Nutrição)  - Centro Universitário Augusto Motta. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="304">BACELO, A. C.</a></b>.  Participação em banca de Lilian Torres Rodrigues Oliveira.Assistência de Enfermagem ao Paciente Portador de Pé Diabético.
# 							2009. Trabalho de Conclusão de Curso (Graduação em Nutrição)  - Centro Universitário Augusto Motta. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="305">BACELO, A. C.</a></b>.  Participação em banca de Jacy de Souza Vieira.Os Cuidados de Enfermagem no Programa de Aleitamento Materno.
# 							2009. Trabalho de Conclusão de Curso (Graduação em Nutrição)  - Centro Universitário Augusto Motta. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="306">BACELO, A. C.</a></b>.  Participação em banca de Maria de Fatima Fernandes.Atuação do Enfermeiro na Sexualidade da Terceira Idade.
# 							2009. Trabalho de Conclusão de Curso (Graduação em Nutrição)  - Centro Universitário Augusto Motta. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="307">BACELO, A. C.</a></b>.  Participação em banca de Lidia Faria Teixeira de Moura.Atuação do Enfermeiro na Prevenção de HIV / AIDS no Idoso.
# 							2009. Trabalho de Conclusão de Curso (Graduação em Nutrição)  - Centro Universitário Augusto Motta. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5683437160866054" target="_blank" tabindex="308">MORAES, V. M.</a>; TOSTE, F.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="309">BACELO, A. C.</a></b>.  Participação em banca de jaqueline Paulino da Silva2007.Avaliação do perfil Nutricional das Crianças Assistidas pela LBV (Legião da Boa Vontade) no período de 2005.
# 							2007. Trabalho de Conclusão de Curso (Graduação em Nutrição)  - Centro Universitário Augusto Motta. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5683437160866054" target="_blank" tabindex="310">MORAES, V. M.</a>; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="311">BACELO, A. C.</a></b>; TOSTE, F..  Participação em banca de Luana Bitencourt Emmerich.Risco de doença cardiovascular utilizando como preditor a relação cintura-quadril.
# 							2007. Trabalho de Conclusão de Curso (Graduação em Nutrição)  - Centro Universitário Augusto Motta. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">MORENO, M.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="312">BACELO, A. C.</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/2514251864249834" target="_blank" tabindex="313">PORTE, A.</a>.  Participação em banca de Isabel cristina Conceição da Silva.Algumas Causas de Contaminações de Alimentospor Staphylococcus Aureus.
# 							2007. Trabalho de Conclusão de Curso (Graduação em Nutrição)  - Centro Universitário Augusto Motta. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>10. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">MORENO, M.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="314">BACELO, A. C.</a></b>; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/2514251864249834" target="_blank" tabindex="315">PORTE, A.</a>.  Participação em banca de Luana Coelho dos Santos.A Importância do Aleitamento Materno para o Lactente.
# 							2007. Trabalho de Conclusão de Curso (Graduação em Nutrição)  - Centro Universitário Augusto Motta. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>11. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">LIMA, J. A.; COSTA, S. O.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="316">BACELO, A. C.</a></b>.  Participação em banca de Débora Lopes Souto.A Utilização da Contagem de Carboidratos na Adequação Dietética do Portador de Diabetes.
# 							2007. Trabalho de Conclusão de Curso (Graduação em Nutrição)  - Centro Universitário Augusto Motta. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>12. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/1794231232150051" target="_blank" tabindex="317">ALMEIDA, M. F. L.</a>; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="318">BACELO, A. C.</a></b>; MIRANDA, C..  Participação em banca de Sarita Leite Viana.Fibras Alimentares: propriedades Terapêuticas.
# 							2007. Trabalho de Conclusão de Curso (Graduação em Nutrição)  - Centro Universitário Augusto Motta. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>13. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/2514251864249834" target="_blank" tabindex="319">PORTE, A.</a>; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="320">BACELO, A. C.</a></b>; SOARES, L. L. S..  Participação em banca de Jacyra de Azevedo Paraense.Avaliações Higienico Sanitárias de Manipuladores de Alimetos.
# 							2007. Trabalho de Conclusão de Curso (Graduação em Nutrição)  - Centro Universitário Augusto Motta. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>14. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">CARDIM, C. V.; XAVIER, N.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="321">BACELO, A. C.</a></b>.  Participação em banca de medeiros.Fatores que Interferem à Nutrição do Idoso..
# 							2006. Trabalho de Conclusão de Curso (Graduação em Nutrição)  - Universidade Veiga de Almeida. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>15. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="322">BACELO, A. C.</a></b>; TOSTE, F.; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/5683437160866054" target="_blank" tabindex="323">MORAES, V. M.</a>.  Participação em banca de Marcia de Carvalho Pinto.Obesidade Infantil: Associação do Grau de Adiposidade com risco de Doença Cardiovascular.
# 							2006. Trabalho de Conclusão de Curso (Graduação em Nutrição)  - Centro Universitário Augusto Motta. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>16. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="324">BACELO, A. C.</a></b>; COSTA, S. O.; FURTADO, V..  Participação em banca de Débora de Jesus da Silva.Influência da Caseína e do Glúten na Dieta do Autista.
# 							2005. Trabalho de Conclusão de Curso (Graduação em Nutrição)  - Centro Universitário Augusto Motta. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>17. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">COSTA, S. O.; FURTADO, V.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="325">BACELO, A. C.</a></b>.  Participação em banca de Patrice Pereira Silva.Nutrição para Hipertensos em Tratamento Hemodialítico.
# 							2005. Trabalho de Conclusão de Curso (Graduação em Nutrição)  - Centro Universitário Augusto Motta. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>18. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">COSTA, S. O.; FURTADO, V.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="326">BACELO, A. C.</a></b>.  Participação em banca de Thaina Maia Paz.Internação Entre Fármacos e Nutrientes.
# 							2005. Trabalho de Conclusão de Curso (Graduação em Nutrição)  - Centro Universitário Augusto Motta. </span>
# </div>
# <br class="clear">
# <br class="clear"><br class="clear"><a name="ParticipacaoBancasComissoes" tabindex="327"></a><br class="clear"><div class="inst_back">
# <b>Participação em bancas de comissões julgadoras</b>
# </div>
# <div class="cita-artigos">
# <b>Outras participações</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="328">BACELO, A. C.</a></b>. I Seminário de pesquisa da Nutrição.
# 							2010.
# 							Centro Universitário Augusto Motta. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">SILVA, C.; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="329">BACELO, A. C.</a></b>. Seleção de docente para o curso de Nutrição da UNISUAM.
# 							2009.
# 							Centro Universitário Augusto Motta. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="330">BACELO, A. C.</a></b>. Comissão científica.
# 							2008.
# 							Associação Brasileira para o Estudo da Obesidade e Síndrome Metabólica. </span>
# </div>
# <br class="clear">
# <br class="clear">
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="Eventos" tabindex="331">
# <h1 tabindex="0">Eventos</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <div class="layout-cell layout-cell-12 data-cell">
# <a name="ParticipacaoEventos" tabindex="332"></a><br class="clear"><div class="inst_back">
# <b>Participação em eventos, congressos, exposições e feiras</b>
# </div>
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">I Simpósio do Curso de Especialização em Psicologia Hospitalar na Área de Infectologia do Instituto Nacional de Infectologia/Fiocruz.Reflexões sobre uma experiência multiprofissional no INI: possibilidades e limites.
# 							2017. (Simpósio). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">world Nutrition. ESTADO CORPORAL DE FERRO E VITAMINA A EM PESSOAS CO-INFECTADAS HIV/TB APÓS INTERVENÇÃO NUTRICIONAL NUM CENTRO DE REFERÊNCIA DE TUBERCULOSE NO RIO DE JANEIRO:ESTUDO PRELIMINAR. 2012. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Seminário de Alimentos do Curso de Nutrição.Variabilidade de Micronutrientes nos Alimentos. 2011. (Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">I Fórum da UNATI UNISUAM / UERJ.Análise Sensorial de Suplementos Nutricionais por Idosos (UNATI).
# 							2010. (Simpósio). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">10º Congresso Internacional de Nutrição, Longevidade &amp; Qualidade de Vida. Avaliação do Perfil Metabólico de Pacientes Submetidos à Derivação Gástrica do Tipo Fobi Capella. 2009. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Aspectos Sócio-culturais na Formação em Nutrição: resultados de uma pesquisa. 2009. (Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Inflamação Sistêmica x Inflamação Intestinal - Como devemos tratar?.I Meeting de Nutrição intensiva.
# 							2009. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Graduação em Nutrição:Qual a carga horária mínima necessária?.
# 							2008. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XII Congresso Brasileiro de Medicina Intensiva - Adulto - Pediatrico - Neonatal. 2008. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>10. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">I Seminário de Políticas pedagógicas para os Cursos da Área de Saúde. 2007. (Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>11. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Seminário dos Coordenadores de Cursos de Graduação das Áreas da saúde e agrárias. 2007. (Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>12. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">IV Semana Acadêmica das áreas de Ciências Biológicas e da Saúde.Terapia nutricional na insufuciência cardiaca.
# 							2006. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>13. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">IV Semana Acadêmica das Áreas de Ciências Biológicas e da Saúde.Distúrbios Alimentares e Padrões Estéticos - Uma abordagem Multidisciplinar.
# 							2006. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>14. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">VIII Fórum de Nutrição do Instituto Estadual de Doenças do Tórax Ary Parreiras.
# 							2006. (Simpósio). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>15. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XII Congresso Brasileiro de Medicina Intensiva - Adulto - Pediatrico - Neonatal. 2006. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>16. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Avaliação: instrumento de inclusão ou exclusão. 2005. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>17. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">10ª Encontro Farmamellitus de Educação em Diabetes.Dieta para Diabeticos.
# 							2003. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>18. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">24º Semana Acadêmica do Curso de Nutrição.WorkShop Avaliação Nutricional. 2003. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>19. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">IV Congresso Sulbrasileiro de Geriatria e Gerontologia e II Congresso de Geriatria e Gerontologia do Mercosul. Estudos Nutricionais com Idosos. 2003. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>20. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">IX Jornada de Terapia Intensiva e III Jornada de Enfermagem em Terapia Intensiva Adulto do HSL-PUCRS.Avaliação Nutricional do Paciente Crítico. 2003. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>21. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">IX Jornada de Terapia Intensiva e II Jornada de Enfermagem em Terapia Intensiva do HSL-PUCRS.Suporte Nutricional no Paciente Critico. 2003. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>22. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Semiologia Nutricional.Longevidade. 2003. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>23. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Semiologia Nutricional Imunológica. 2003. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>24. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">V Encontro de Nutrição do Hospital São Lucas da PUCRS.Interação Farmaco Nutriente no Idoso.
# 							2003. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>25. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">V Jornada de Nutrição em Cardiologia.Insuficiência Cardíaca - Nutrição Enteral em Cardiologia. 2003. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>26. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XI Congresso da Sociedade Brasileira de Hipertensão. 2003. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>27. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Interações Drogas e Nutrientes.Interações Drogas e Nutrientes. 2002. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>28. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">IV Encontro de Auxiliares e Técnicos de Enfermagem do Serviço de Nefrologia do Hospital São Lucas da PUCRS.Nutrição em Nefrologia.
# 							2002. (Encontro). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>29. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Seminário de Atualização em Geriatria e Gerontologia.Seminário de Atualização em Geriatria e Gerontologia. 2002. (Seminário). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>30. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">XX Curso de Inverno do IC-FUC.Nutrição Enteral em Cardiologia. 2002. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>31. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Terapia Nutricional Conservadora para Insuficiência Renal Crônica.Terapia Nutricional Conservadora para Insuficiência Renal Crônica. 2001. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>32. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Diabete Mellitus.Aliemtação do Diabético. 2000. (Outra). 
# 					</span>
# </div>
# <br class="clear">
# <br class="clear">
# </div>
# <a name="OrganizacaoEventos" tabindex="333"></a><br class="clear"><div class="inst_back">
# <b>Organização de eventos, congressos, exposições e feiras</b>
# </div>
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="334">BACELO, ADRIANA COSTA</a></b>; LEAL, A. ; Brito, P.D ; Simplicio, P. ; ALMEIDA, C. F. ; LAUAR, J. ; JAPIASSU, A. M. ; COPLE, Claudia dos Santos ; TOSTE, F. ; Durgante P. C. . VI Seminário de Nutrição em Infectologia do INI/Fiocruz. 2019. (Outro). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="335">BACELO, A. C.</a></b>; Simplicio, P. ; Brito, P.D ; LAUAR, J. ; JAPIASSU, A. . V Seminário de Nutrição em Infectologia. 2017. (Outro). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Almeida, C. ; Brito, P.D ; <a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/4303234837741559" target="_blank" tabindex="336">CARDOSO, Claudia Santos de Aguiar</a> ; Simplicio, P. ; Santo, R.S. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="337">BACELO, A. C.</a></b> . IV Seminário de Terapia Nutricional do INI. 2014. (Outro). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="338">BACELO, A. C.</a></b>. IV Simpósio de Docencia Superior UNISUAM: A Formação Docente na Sociedade do Conhecimento. 2011. (Outro). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">MANCINI, M. ; GELONESE-NETO, B. ; CASAGRANDE, D. ; PISCIOLARO, F. ; SALLES, J. E. N. ; LIMA, J. ; MANCINI, M. C. ; MELO, M. E. ; RODRIGUES, M. B. ; CARRA, M. K. ; BEYRUTI, M. ; <b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="339">BACELO, A. C.</a></b> . I Simposio de Nutrição Clínica da Abeso. 2008. (Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="340">BACELO, A. C.</a></b>. Encontro de Adolescentes da CBCM - Oficina sobre o Tema Saúde e Alimentação. 1998. (Outro). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="341">BACELO, A. C.</a></b>. VIII Jornada de Nutrição "I Seminário de Segurança Alimentar". 1994. (Outro). </span>
# </div>
# <br class="clear">
# <br class="clear">
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="Orientacoes" tabindex="342">
# <h1 tabindex="0">Orientações</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <a name="Orientacoesconcluidas" tabindex="343"></a><br class="clear"><div class="inst_back">
# <b>Orientações e supervisões concluídas</b>
# </div>
# <div class="cita-artigos">
# <b>Dissertação de mestrado</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-lattes" target="_blank" href="http://lattes.cnpq.br/0880732960289475" tabindex="344"><img src="images/curriculo/logolattes.gif"></a>LUCIA DE MELLO COUTINHO PINTO. FORNECIMENTO DE TERAPIA NUTRICIONAL COMO FATOR PREDITOR EM PESSOAS HOSPITALIZADAS COM COVID-19.
# 							2022.
						
# 					Dissertação  (Mestrado em Pesquisa Clinica em Doenças Infecciosas)  - Fundação Oswaldo Cruz, . Orientador: Adriana Costa Bacelo. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-lattes" target="_blank" href="http://lattes.cnpq.br/2653727116868514" tabindex="345"><img src="images/curriculo/logolattes.gif"></a>Alan Renier Jamal Occhioni Molter. Desenvolvimento de um aplicativo móvel de representação da ingestão alimentar de pacientes internados: sistematização dos dados dietéticos.
# 							2022.
						
# 					Dissertação  (Mestrado em Pesquisa Clinica em Doenças Infecciosas)  - Fundação Oswaldo Cruz, Coordenação de Aperfeiçoamento de Pessoal de Nível Superior. Orientador: Adriana Costa Bacelo. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-lattes" target="_blank" href="http://lattes.cnpq.br/0119681873564065" tabindex="346"><img src="images/curriculo/logolattes.gif"></a>PALOMA RIBEIRO TORRES. ESTADO NUTRICIONAL E PADRÃO DIETETICO DE PACIENTES HTLV SINTOMATICOS.
# 							2019.
						
# 					Dissertação  (Mestrado em Mestrado em Pesquisa Clínica em Doenças Infecciosas)  - Instituto Nacional de Infectologia Evandro Chagas, . Orientador: Adriana Costa Bacelo. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><a class="icone-lattes" target="_blank" href="http://lattes.cnpq.br/5393303702952755" tabindex="347"><img src="images/curriculo/logolattes.gif"></a>NAÍSE OLIVEIRA DA ROCHA. ASSOCIAÇÃO DO ÂNGULO DE FASE COM ESTADO NUTRICIONAL  DE PACIENTES HTLV SINTOMÁTICOS E ASSINTOMÁTICOS.
# 							2019.
						
# 					Dissertação  (Mestrado em Mestrado em Pesquisa Clínica em Doenças Infecciosas)  - Instituto Nacional de Infectologia Evandro Chagas, . Orientador: Adriana Costa Bacelo. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b>Monografia de conclusão de curso de aperfeiçoamento/especialização</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">PRISCILA RODRIGUES DE JESUS.
# 							AVALIAÇÃO DO USO DE PREBIÓTICOS, PROBIÓTICOS E SIMBIÓTICOS EM PACIENTES HIV COMO FATOR PROTETOR AO TRATO GASTROINTESTINAL.
# 							2018.
# 							Monografia. (Aperfeiçoamento/Especialização em Curso de Especialização em Nutrição Clínica Aplicada à Infectologia)  - Instituto Nacional de Infectologia Evandro Chagas. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">KATYELLEN MONIQUE MESQUITA DA SILVA.
# 							IMPORTÂNCIA DA INTERVENÇÃO NUTRICIONAL EM PACIENTES COM SÍNDROME DA LIPODISTROFIA DO HIV/AIDS SUBMETIDOS À TERAPIA  ANTIRRETROVIRAL.
# 							2018.
# 							Monografia. (Aperfeiçoamento/Especialização em Curso de Especialização em Nutrição Clínica Aplicada à Infectologia)  - Instituto Nacional de Infectologia Evandro Chagas. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Naise Oliveira da Rocha.
# 							Aplicação da Bioimpedância como indicador prognóstico em pacientes com  doenças infecciosas: uma revisão bibliográfica.
# 							2017.
# 							Monografia. (Aperfeiçoamento/Especialização em Curso de Especialização em Nutrição Clínica Aplicada à Infectologia)  - Instituto Nacional de Infectologia Evandro Chagas. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">PALOMA RIBEIRO TORRES.
# 							VITAMINA D NAS DOENÇAS INFECCIOSAS.
# 							2017.
# 							Monografia. (Aperfeiçoamento/Especialização em Curso de Especialização em Nutrição Clínica Aplicada à Infectologia)  - Instituto Nacional de Infectologia Evandro Chagas. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Rachel Brum.
# 							Estado Nutricional e Consumo alimentar de Crianças e adolescentes com doenças hematológicas.
# 							2016.
# 							Monografia. (Aperfeiçoamento/Especialização em terapia Nutricional)  - Universidade do Estado do Rio de Janeiro. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Anabel Godoy Alevado.
# 							A Obesidade como Fator Agravante nas Doenças Cardíacas associadas à Hiperten são Arterial.
# 							2005.
# 							Monografia. (Aperfeiçoamento/Especialização em Nutrição)  - Universidade Veiga de Almeida. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Luciana Barreto Carneiro.
# 							Relação entre Obesidade e Nutrição.
# 							2005.
# 							Monografia. (Aperfeiçoamento/Especialização em Nutrição)  - Universidade Veiga de Almeida. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Vânia Lúcia da Cruz Magalhães.
# 							Tratamento não farmacológico do Alzheimer.
# 							2005.
# 							Monografia. (Aperfeiçoamento/Especialização em Nutrição)  - Universidade Veiga de Almeida. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Vanda Marta de Lima Lichotti.
# 							A Importância do Trabalho Não Farmacológico na Hipertensão Arterial.
# 							2004.
# 							Monografia. (Aperfeiçoamento/Especialização em Nutrição)  - Universidade Veiga de Almeida. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b>Trabalho de conclusão de curso de graduação</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Luciane Santos nicolich.
# 							Riscos Nutricionais da Anorexia Nervosa.
# 							2007.
# 							Trabalho de Conclusão de Curso. (Graduação em Nutrição)  - Centro Universitário Augusto Motta. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>2. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Camila de Souza Figueiredo.
# 							Impacto da Suplementação no Estado Nutricional de Pré-escolares.
# 							2007.
# 							Trabalho de Conclusão de Curso. (Graduação em Nutrição)  - Centro Universitário Augusto Motta. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>3. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Cintia Gomes Cossich.
# 							Atuação do Aminoácido L-Triptofano e Ômega-3 no Tratamento e Prevenção do Transtorno de Humor.
# 							2006.
# 							Trabalho de Conclusão de Curso. (Graduação em Nutrição)  - Centro Universitário Augusto Motta. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>4. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Geysa Batista Serra.
# 							Anorexia Nervosa.
# 							2005. 50 f.
# 							Trabalho de Conclusão de Curso. (Graduação em Nutrição)  - Universidade Veiga de Almeida. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>5. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Adriana Mercês de Oliveira.
# 							Bulimia Nervosa do Indivíduo Adulto.
# 							2005. 29 f.
# 							Trabalho de Conclusão de Curso. (Graduação em Nutrição)  - Universidade Veiga de Almeida. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>6. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Marcelo Soares de Moraes.
# 							Conhecimento e Prática Materna sobre Alimentação Infantil para Crianças até 24 meses Atendidas em Unidade Básica de Saúde de Niterói.
# 							2005. 44 f.
# 							Trabalho de Conclusão de Curso. (Graduação em Nutrição)  - Universidade Veiga de Almeida. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>7. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Erica Leite Pereira.
# 							Condutas Nutricionais na Gestação: Um Caminho para Alimentação Saudável.
# 							2005. 48 f.
# 							Trabalho de Conclusão de Curso. (Graduação em Nutrição)  - Universidade Veiga de Almeida. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>8. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Thatiana de Freitas Nepomuceno.
# 							Segurança Alimentar no Aleitamento Materno.
# 							2005. 43 f.
# 							Trabalho de Conclusão de Curso. (Graduação em Nutrição)  - Universidade Veiga de Almeida. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>9. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Michele de Oliveira da Gama.
# 							Aspectos Nutricionais na Bulimia da Adolescente.
# 							2005. 46 f.
# 							Trabalho de Conclusão de Curso. (Graduação em Nutrição)  - Universidade Veiga de Almeida. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>10. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Daisi Carla Pizette Paranhos.
# 							Dislipidemia Infantil.
# 							2005. 33 f.
# 							Trabalho de Conclusão de Curso. (Graduação em Nutrição)  - Universidade Veiga de Almeida. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>11. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Joana Martins.
# 							Manual de Boas Práticas: Relato em uma Unidade de Alimentação e Nutrição.
# 							2005. 39 f.
# 							Trabalho de Conclusão de Curso. (Graduação em Nutrição)  - Universidade Veiga de Almeida. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>12. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Aline Batista.
# 							Associação entre Obesidade Infantil e Síndrome Metabólica.
# 							2005. 35 f.
# 							Trabalho de Conclusão de Curso. (Graduação em Nutrição)  - Universidade Veiga de Almeida. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>13. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Eliane Cabanez.
# 							Obesidade e Hipertensão Arterial.
# 							2005.
# 							Trabalho de Conclusão de Curso. (Graduação em Nutrição)  - Centro Universitário Augusto Motta. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>14. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Dayse Laísa Lima Ferreira.
# 							Controle Calêmico no Tratamento Conservador da Insuficiência Renal Crônica.
# 							2005.
# 							Trabalho de Conclusão de Curso. (Graduação em Nutrição)  - Centro Universitário Augusto Motta. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>15. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Rosana Mattos do Carmo.
# 							Intervenções Nutricionais na Caquexia no Idoso Portador de Insuficiência Cardíaca Congestiva.
# 							2005.
# 							Trabalho de Conclusão de Curso. (Graduação em Nutrição)  - Centro Universitário Augusto Motta. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>16. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Elaine Spinula Simões de Almeida.
# 							Dieta Hiperlipídica e Sedentarismo como Preditores de Obesidade na Vida Adulta.
# 							2005.
# 							Trabalho de Conclusão de Curso. (Graduação em Nutrição)  - Centro Universitário Augusto Motta. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>17. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Sofia Sesti.
# 							Perfil Nutricional dos Pacientes com Insuficiência Renal Crônica Terminal em Tratamento Dialitico - Análise de Pacientes atendidos no Serviço de Nefrologia do Hospital São Lucas da PUCRS.
# 							2002.
# 							Trabalho de Conclusão de Curso. (Graduação em Nutrição)  - Universidade Franciscana. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# <div class="cita-artigos">
# <b>Iniciação científica</b>
# </div>
# <br class="clear">
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">Marcelo Soares.
# 							Comparação dos Diferentes Métodos de Avaliação Estatural no Adulto por Faixa Etária.
# 							2004.
# 							Iniciação Científica. (Graduando em Nutrição)  - Universidade Veiga de Almeida. Orientador: Adriana Costa Bacelo.
# 						</span>
# </div>
# <br class="clear">
# <br class="clear">
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="PotencialInovacao" tabindex="348">
# <h1 tabindex="0">Inovação</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <a name="ProtudoTecnologicoPI" tabindex="349"></a><br class="clear"><div class="inst_back">
# <b>Produto tecnológico</b>
# </div>
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform">MOLTER, A. R. J. O. ; FONSECA, C. F. ; SILVA, M. P. ; PATRICIA, BRITO ; Brasil, P.E. ; <b>ADRIANA, BACELO</b> . ARIETI (Aplicativo móvel de Representação da Ingestão alimEntar de pacienteTes Internados). 2023. </span>
# </div>
# <br class="clear">
# <br class="clear">
# <a name="ProjetosPesquisaPI" tabindex="350"></a><br class="clear"><div class="inst_back">
# <b>Projetos de pesquisa</b>
# </div>
# <div class="layout-cell layout-cell-12 data-cell"><a name="PP_Análise comparativa de intervenções presenciais e virtuais para o acompanhamento nutricional de adultos: ensaio clínico de superioridade" tabindex="351"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2023 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Análise comparativa de intervenções presenciais e virtuais para o acompanhamento nutricional de adultos: ensaio clínico de superioridade<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: Introdução: O avanço tecnológico tem impulsionado transformações significativas na área da saúde, incluindo a gestão e o tratamento de doenças crônicas não transmissíveis, o que tem possibilitado a expansão e diversificação dos serviços de atendimento nutricional. Nesse cenário, as intervenções presenciais e virtuais têm se destacado, suscitando indagações acerca de sua eficácia, aplicabilidade e impacto nos resultados clínicos. Diante da crescente importância dessas abordagens para o contexto das doenças crônicas não transmissíveis, sabidamente crescente no mundo, torna-se imprescindível aprofundar a compreensão das diferenças entre ambas, com o propósito de embasar decisões e promover melhorias substanciais na área da nutrição.  Justificativa: A relevância deste estudo está pautada no potencial impacto que seus resultados poderão ter sobre a prática clínica em nutrição, contribuindo para a identificação de quais abordagens são mais adequadas a diferentes perfis de pacientes, considerando variáveis socioeconômicas e geográficas. Além disso, a análise comparativa permitirá uma avaliação mais precisa da efetividade e eficiência de cada modalidade de acompanhamento, fomentando a busca por soluções inovadoras e a utilização estratégica da tecnologia no âmbito da saúde nutricional. Objetivo: Comparar as intervenções presenciais e  virtuais no acompanhamento nutricional de adultos, com foco na adesão ao tratamento, evolução clínica e satisfação dos pacientes, a fim de fornecer evidências que auxiliem na tomada de decisões para aprimorar o atendimento nutricional de doenças crônicas não transmissíveis. Metodologia: Ensaio clínico de superioridade de comparação entre estratégias virtuais (teleatendimento) e convencionais (presenciais) de mapeamento e intervenção nutricional. Todos os 250 trabalhadores (servidores públicos, terceirizados, e bolsistas) e alunos, maiores de 19 anos, não sendo gestante, da unidade da Fiocruz Ceará serão convidados para participar do estudo. Os trabalhadores/alunos que assinarem o TCLE serão randomizados em dois grupos, a saber: a) grupo virtual (intervenção): estratégias de mapeamento e orientação/prescrição nutricional individuais e coletivas de forma virtual (Whatsapp ou Telegram e Zoom ou Teams); b) grupo presencial (comparação): estratégias de mapeamento e orientação/prescrição nutricional individuais e coletivas de forma presencial. Para ambos os grupos haverá um encontro presencial no Dzero (inclusão), D180 (6 meses de acompanhamento) e D360 (01 ano de acompanhamento) para avaliação de composição corporal. O grupo virtual passará por esta fase após a realização de uma autoavaliação virtual e ambos os resultados serão comparados. As consultas individuais serão trimestrais, e as e coletivas serão semanais. A avaliação nutricional incluirá exames clínicos (exame físico e história clínica), antropométricos (peso, altura, circunferências e perímetros corporais), bioimpedância e ultrassonografia de quadríceps femoral; além de aplicação da escala brasileira de insegurança alimentar (EBIA), marcador de estresse percebido (PSS), adesão auto-relatada, qualidade de sono autorreferida,  autopercepção de qualidade de vida e imagem corporal e um questionário de satisfação enquanto usuário (escala Likert) e ainda análise fenomenológica evolutiva dos discursos obtidos nas consultas coletivas. Uma sub amostra de 50 participantes obesos (IMC&gt;30 Kg/m2 e com idade entre 40 e 59 anos), 25 de cada grupo (virtual e presencial), será randomizada para investigação de saúde de precisão, para tal será realizada coleta de sangue e saliva nos tempos Dzero e D360. Resultados esperados: Espera-se que a disponibilização de estratégias digitais melhorem a adesão e por conseguinte os parâmetros clínicos, nutricionais, imunológicos, metabólicos e de qualidade de vida.. <br class="clear">Situação: Em andamento; Natureza: Pesquisa. <br class="clear">Alunos envolvidos: Graduação: (1) . <br class="clear"><br class="clear">Integrantes: Adriana Costa Bacelo - Coordenador / BRASIL, PEDRO EMMANUEL - Integrante / COPLE-RODRIGUES, CLAUDIA DOS SANTOS - Integrante / TORRES, PALOMA RIBEIRO - Integrante / Luiz Odorico Monteiro de Andrade - Integrante / Ivana Cristina de Holanda Cunha Barreto - Integrante / Kamila Maria Oliveira Sales - Integrante / Paulo Armando Cavalcante Aguilar - Integrante / José Xavier Neto - Integrante / Filipe Brito - Integrante / Marcela Helena Gambim Fonseca - Integrante / Júlio Cesar Martins Ximenes - Integrante / Ana Carolina Matias Dinelly Pinto - Integrante / Clarissa Perdigão Mello Ferraz - Integrante / William Cesar Bento Régis - Integrante / Carla Freire Celedonio Fernandes - Integrante / Márcia Andréa Gonçalves Leite - Integrante / Márcio Espíndola Freire Maia - Integrante / Francisco Leandro Fernandes Sampaio - Integrante / Elga Batista da Silva - Integrante / André Ribeiro Braga - Integrante.<br class="clear">Financiador(es): Conselho Nacional de Desenvolvimento Científico e Tecnológico - Auxílio
# 										financeiro.</div>
# </div><a name="PP_Desenvolvimento de uma solução digital para o controle de qualidade da indústria de alimentos baseados em requisitos internacionais" tabindex="352"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2021 - Atual</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Desenvolvimento de uma solução digital para o controle de qualidade da indústria de alimentos baseados em requisitos internacionais<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: Introdução: A acreditação internacional das indústrias de alimentos brasileiros é uma necessidade de competitividade de mercado e para a segurança alimentar dos consumidores.  Justificativa: A FSSC 22000 foi reconhecida pela GFSI como um padrão de segurança alimentar internacional desde 2010, tornando-se um protocolo de certificação completo imprescindível, mas de aplicabilidade complexa. Mas impondo às indústrias de alimentos brasileiras uma necessidade de adequação para ingressão no mercado internacional, fator importante para o desenvolvimento econômico industrial brasileiro.  Objetivo: Propor um instrumento digital que apoie o controle de qualidade da indústria de alimentos brasileira, a fim de tornar mais prático, amigável e simplificado o processo de auditoria interna de segurança de alimentos demandado no mercado internacional. Metodologia: Estudo de desenvolvimento, validação e de confiabilidade de um aplicativo de auditoria de qualidade e segurança de alimentos a partir de critérios estabelecidos pela Norma FSSC 22000, em etapas de um aplicativo construído utilizando um software de programação multiplataformas. Numa primeira etapa será realizada a revisão sistemática sobre a aplicabilidade da FSSC 22000; posteriormente serão convidados especialistas auditores de segurança alimentar ISO 22 000, boas práticas de fabricação de alimentos (BPF), pontos críticos de controle (APPCC) e/ou FSSC 22000 para participar de um estudo etnográfico para a versão inicial do instrumento. Nesta etapa será seguida o instrumento será encaminhado ao mesmo painel de especialista conforme a técnica Delphi. Posteriormente, como base na verificação da concordância entre os especialistas será desenvolvido um aplicativo móvel de monitoramento da qualidade nos processos de produção de alimentos, utilizando um software de programação para multi plataformas, incluindo Android e posteriormente iOs, utilizando as tecnologias C# e JSON. Para o teste de confiabilidade o instrumento será aplicado sequencialmente 4 vezes, em semanas diferentes, numa indústria de alimentos da região metropolitana do Ceará por dois auditores que irão se revezar no uso das ferramentas tradicionais de auditoria e a solução digital proposta digital. Resultados esperados:  Espera-se apresentar uma solução digital, que traga agilidade e segurança no controle dos processos de segurança alimentar das indústrias alimentícias brasileiras, atendendo as normas do mercado de indústria de alimentos internacional.. <br class="clear">Situação: Em andamento; Natureza: Pesquisa. <br class="clear">Alunos envolvidos: Graduação: (1) . <br class="clear"><br class="clear">Integrantes: Adriana Costa Bacelo - Coordenador / Luiz Odorico Monteiro de Andrade - Integrante / Ivana Cristina de Holanda Cunha Barreto - Integrante / Margareth Catoia Varela - Integrante.<br class="clear"></div>
# </div><a name="PP_Desenvolvimento de um aplicativo móvel de representação da ingestão alimentar de pacientes internados: teste de confiabilidade" tabindex="353"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2020 - 2023</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Desenvolvimento de um aplicativo móvel de representação da ingestão alimentar de pacientes internados: teste de confiabilidade<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Situação: Concluído; Natureza: Pesquisa. <br class="clear">Alunos envolvidos: Mestrado acadêmico: (1) . <br class="clear"><br class="clear">Integrantes: Adriana Costa Bacelo - Coordenador / Marlete Pereira da Silva - Integrante / CARDOSO, CLÁUDIA SANTOS DE AGUIAR - Integrante / Brito, Patrícia Dias - Integrante / Claudio Fico Fonseca - Integrante / Alan Renier Jamal Occhioni Molter - Integrante.<br class="clear"></div>
# </div><a name="PP_DIAGNÓSTICO NUTRICIONAL E FUNÇÃO INTESTINAL DE PACIENTES SINTOMATICOS E ASSINTOMATICOS PORTADORES DE HTLV SUBMETIDOS A ORIENTAÇÃO DIETÉTICA COM E SEM A INCLUSÃO DA FARINHA DE OKARA" tabindex="354"></a><div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b>2016 - 2021</b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">DIAGNÓSTICO NUTRICIONAL E FUNÇÃO INTESTINAL DE PACIENTES SINTOMATICOS E ASSINTOMATICOS PORTADORES DE HTLV SUBMETIDOS A ORIENTAÇÃO DIETÉTICA COM E SEM A INCLUSÃO DA FARINHA DE OKARA<br class="clear"></div>
# </div>
# <div class="layout-cell layout-cell-3 text-align-right">
# <div class="layout-cell-pad-5 text-align-right">
# <b></b>
# </div>
# </div>
# <div class="layout-cell layout-cell-9">
# <div class="layout-cell-pad-5">Descrição: A infecção pelo HTLV ainda é negligenciada pelos sistemas de atenção à saúde, possivelmente por aproximadamente 90% dos portadores serem assintomáticos e as complicações mais conhecidas ocorrem em apenas 5% dos infectados. Contudo, a morbidade das doenças associadas ao HTLV é elevada, a ocorrência de constipação intestinal nos pacientes com Paraparesia Espástica Tropical/Mielopatia é grande e a situação nutricional desta população é pouco estudada. Objetivo: Comparar estado nutricional de indivíduos adultos portadores de HTLV sintomáticos e assintomáticos e investigar o efeito da inclusão dietética da farinha de Okara sobre a função intestinal e o estado nutricional dos indivíduos sintomáticos. Metodologia: estudo prospectivo, longitudinal, analítico do acompanhamento nutricional, por 2 meses, de pacientes sintomáticos e assintomáticos com diagnóstico de HTLV, no período de setembro de 2015 à setembro de 2021. Serão realizadas as avaliações antropométrica, bioquímica e dietética (Recordatório de 24 horas e diversidade dietética) seguida de aconselhamento dietético e/ou suplementação de fibra às custas de farinha de Okara (grupo caso). A percepção de adesão dietética será avaliada por autorreferência nas diferentes consultas durante estudo. Será verificado o gasto energético por calorimetria indireta e prognóstico nutricional por ângulo de fase. Será considerada má nutrição quando pelo menos um dos resultados da avaliação nutricional estava fora dos valores de referência, sendo classificada de acordo com os parâmetros considerados, em: global (todos os parâmetros), calórica (Índice de Massa Corporal - IMC, dobra cutânea triciptal ? DCT, e circunferência do braço - CB), proteica (circunferência muscular do braço - CMB, albumina e transferrina), ou específica (marcadores séricos de ferro, selênio, zinco, retinol, D-25 hidroxicalciferol e ∂-tocoferol). Os dados serão analisados no programa R-project® versão 3.0.2. Os resultados divididos em tempo inicial e final. As diferenças serão consideradas significantes quando o valor de p foi ≤ 0.05. Resultados esperado: identificar as necessidades nutricionais do indivíduo com HTLV e contribuir para o conhecimento das repercussões dos sintomas no gasto energético, apontando a quantidade de fibra necessária para minimizar a constipação secundária a mielopatia espástica.. <br class="clear">Situação: Concluído; Natureza: Pesquisa. <br class="clear">Alunos envolvidos: Especialização: (2)  / Mestrado acadêmico: (1) . <br class="clear"><br class="clear">Integrantes: Adriana Costa Bacelo - Coordenador / Claudia dos santos cople - Integrante / eliane paiva - Integrante / Patricia Dias de Brito - Integrante / Cristiane Fonseca de Almeida - Integrante / Abelardo de Queiroz Campos Araújo - Integrante / Elizabethe Neves - Integrante / Ana Claudia Celestino - Integrante / Naise Oliveira da Rocha - Integrante / Paloma Ribeiro Torres - Integrante / Roseli dos Santos - Integrante.<br class="clear">Financiador(es): FAPERJ - Auxílio
# 										financeiro / FAPERJ - Bolsa.<br class="clear">Número de produções C, T &amp; A: 6</div>
# </div>
# </div><br class="clear"></div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="EducacaoPopularizacaoCTA" tabindex="355">
# <h1 tabindex="0">Educação e Popularização de C &amp; T</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <a name="ApresentacoesTrabalhoEPCTA" tabindex="356"></a><br class="clear"><div class="inst_back">
# <b>Apresentações de Trabalho</b>
# </div>
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="357">BACELO, ADRIANA COSTA</a></b>. Envelhecimento saudável: Nutrigenética e Nutrigenômica. 2022. (Apresentação de Trabalho/Congresso). </span>
# </div>
# <br class="clear">
# <br class="clear">
# <a name="EntrevistasEPCTA" tabindex="358"></a><br class="clear"><div class="inst_back">
# <b>Entrevistas, mesas redondas, programas e comentários na mídia </b>
# </div>
# <div class="layout-cell layout-cell-1 text-align-right">
# <div class="layout-cell-pad-6 text-align-right">
# <b>1. </b>
# </div>
# </div>
# <div class="layout-cell layout-cell-11">
# <span class="transform"><b><a class="tooltip" title="Clique para visualizar o currículo" href="http://lattes.cnpq.br/8861663346303670" target="_blank" tabindex="359">COSTA BACELO, ADRIANA</a></b>; GOMES, I. . Nutrição e Longevidade. 2022. 
						
# 							(Programa de rádio ou TV/Entrevista).
# 						<img class="icone-video" src="http://estatico.cnpq.br/imagens/ico_pw_video.png" onclick="mostrarPreviaVideo('https://youtu.be/MSJP_b5uOQw')"></span>
# </div>
# <br class="clear">
# <br class="clear">
# </div><br class="clear"></div><br class="clear"><div class="title-wrapper">
# <a name="OutrasInformacoesRelevantes" tabindex="360">
# <h1 tabindex="0">Outras informações relevantes</h1>
# </a>
# <hr class="separator">
# <div class="layout-cell layout-cell-12 data-cell">
# <pre class="outras-informacoes">Aprovação em  concurso publico para serviço técnico temporário 2011 do oficiais do corpo complementar de saúde, cargo Nutrição do Exercito do Brasil.- De 77 inscritos obteve na classificação final o 2°lugar, conforme links abaixo:
# 1° fase: http://www.1rm.eb.mil.br/images/stories/serv-mil/convocacao/pontuacao/oficiais/NUTRICAO.pdf
# 2° fase: http://www.1rm.eb.mil.br/images/stories/serv-mil/convocacao/repontuacao/oficiais/NUTRICAO_REPONTUADA.pdf</pre>
# </div><br class="clear"></div><br class="clear"><br class="clear"><div class="rodape-cv">Página gerada pelo Sistema Currículo Lattes em 03/04/2024 às 21:15:32</div><br class="clear"><div class="control-bar-wrapper"><a href="download.do?metodo=apresentar&amp;idcnpq=8861663346303670" target="_blank" class="button" tabindex="361"><span class="mini-ico mini-ico-down"></span>Baixar Currículo</a><a class="button" href="javascript:window.print()" tabindex="362"><span class="mini-ico mini-ico-list"></span>Imprimir Currículo</a>
# </div><br class="clear"></div>
# </div>
# </div>
# </div>
# </div>
# <div class="to-top-bar" id="toTop" style="display: none;">
# <a href="javascript:void(0)" class="topo-width center" tabindex="363">
# <div class="id-orgao">
# <span>CNPq</span> | Uma agência do Ministério da Ciência, Tecnologia e Inovação</div>
# </a>
# </div>


# <div id="menu-dados-gerais" class="megamenu" _megamenupos="0pos" style="z-index: 1001; display: none;">
		
# 		<div class="column"><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#Identificacao" tabindex="11">Identificação</a></li>
		
	
		
		
# 		<li><a href="#Endereco" tabindex="12">Endereço</a></li>
		
	
		
		
# 		<li><a href="#Idiomas" tabindex="13">Idiomas</a></li>
		
	
		
		
# 		<li><a href="#PremiosTitulos" tabindex="14">Prêmios e títulos</a></li>
		
	
		
		
# 		<li><a href="#OutrasInformacoesRelevantes" tabindex="15">Outras informações relevantes</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div><div id="menu-formacao" class="megamenu" _megamenupos="1pos" style="z-index: 1002; display: none;">
		
# 		<div class="column"><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#FormacaoAcademicaTitulacao" tabindex="17">Formação acadêmica/titulação</a></li>
		
	
		
		
# 		<li><a href="#FormacaoComplementar" tabindex="18">Formação complementar</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div><div id="menu-atuacao" class="megamenu" _megamenupos="2pos" style="z-index: 1003; display: none;">
		
# 		<div class="column"><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#AtuacaoProfissional" tabindex="20">Atuação profissional</a></li>
		
	
		
		
# 		<li><a href="#LinhaPesquisa" tabindex="21">Linhas de pesquisa</a></li>
		
	
		
		
# 		<li><a href="#RevisorPeriodico" tabindex="22">Revisor de periódico</a></li>
		
	
		
		
# 		<li><a href="#RevisorProjetoFomento" tabindex="23">Revisor de projeto de fomento</a></li>
		
	
		
		
# 		<li><a href="#AreasAtuacao" tabindex="24">Áreas de atuação</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div><div id="menu-projetos" class="megamenu" _megamenupos="3pos" style="z-index: 1004; display: none;">
		
# 		<div class="column"><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#ProjetosPesquisa" tabindex="26">Projetos de pesquisa</a></li>
		
	
		
		
# 		<li><a href="#ProjetosDesenvolvimento" tabindex="27">Projetos de desenvolvimento</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div><div id="menu-producoes" class="megamenu" _megamenupos="4pos" style="z-index: 1005; display: none;">
		
# 		<div class="column"><div class="subtitMenu"><a href="#ProducaoBibliografica" tabindex="29">Produção Bibliográfica</a></div><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#ArtigosCompletos" tabindex="30">Artigos completos publicados em periódicos</a></li>
		
	
		
		
# 		<li><a href="#ArtigosAceitos" tabindex="31">Artigos aceitos para publicação</a></li>
		
	
		
		
# 		<li><a href="#LivrosCapitulos" tabindex="32">Livros e capítulos</a></li>
		
	
		
		
# 		<li><a href="#TextosJornaisRevistas" tabindex="33">Textos em jornais ou revistas (magazine)</a></li>
		
	
		
		
# 		<li><a href="#TrabalhosPublicadosAnaisCongresso" tabindex="34">Trabalhos publicados em anais de congressos</a></li>
		
	
		
		
# 		<li><a href="#ApresentacaoTrabalho" tabindex="35">Apresentações de trabalho</a></li>
		
	
		
		
# 		<li><a href="#OutrasProducoesBibliograficas" tabindex="36">Partitura musical</a></li>
		
	
		
		
# 		<li><a href="#OutrasProducoesBibliograficas" tabindex="37">Tradução</a></li>
		
	
		
		
# 		<li><a href="#OutrasProducoesBibliograficas" tabindex="38">Prefácio, pósfacio</a></li>
		
	
		
		
# 		<li><a href="#OutrasProducoesBibliograficas" tabindex="39">Outras produções bibliográficas</a></li>
		
# 	</ul></div><div class="column"><div class="subtitMenu"><a href="#ProducaoTecnica" tabindex="40">Produção Técnica</a></div><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#AssessoriaConsultoria" tabindex="41">Assessoria e consultoria</a></li>
		
	
		
		
# 		<li><a href="#SoftwareSemPatente" tabindex="42">Programas de computador sem registro</a></li>
		
	
		
		
# 		<li><a href="#ProdutosTecnologicos" tabindex="43">Produtos tecnológicos</a></li>
		
	
		
		
# 		<li><a href="#ProcessosTecnicas" tabindex="44">Processos e técnicas</a></li>
		
	
		
		
# 		<li><a href="#TrabalhosTecnicos" tabindex="45">Trabalhos técnicos</a></li>
		
	
		
		
# 		<li><a href="#DemaisProducaoTecnica" tabindex="46">Cartas, mapas ou similares</a></li>
		
	
		
		
# 		<li><a href="#DemaisProducaoTecnica" tabindex="47">Curso de curta duração ministrado</a></li>
		
	
		
		
# 		<li><a href="#DemaisProducaoTecnica" tabindex="48">Desenvolvimento de material didático ou instrucional</a></li>
		
	
		
		
# 		<li><a href="#DemaisProducaoTecnica" tabindex="49">Editoração</a></li>
		
	
		
		
# 		<li><a href="#ManutencaoObraArtistica" tabindex="50">Manutenção de obra artística</a></li>
		
	
		
		
# 		<li><a href="#DemaisProducaoTecnica" tabindex="51">Maquete</a></li>
		
	
		
		
# 		<li><a href="#EntrevistasMesasRedondas" tabindex="52">Entrevistas, mesas redondas, programas e comentários na mídia</a></li>
		
	
		
		
# 		<li><a href="#DemaisProducaoTecnica" tabindex="53">Relatório de pesquisa</a></li>
		
	
		
		
# 		<li><a href="#RedesSociais" tabindex="54">Redes sociais, websites e blogs</a></li>
		
	
		
		
# 		<li><a href="#DemaisProducoesTecnicas" tabindex="55">Outras produções técnicas</a></li>
		
# 	</ul></div><div class="column"><div class="subtitMenu"><a href="#ProducaoArtisticaCultural" tabindex="56">Produção Artística/Cultural</a></div><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#ArtesCenicas" tabindex="57">Artes cênicas</a></li>
		
	
		
		
# 		<li><a href="#Musica" tabindex="58">Música</a></li>
		
	
		
		
# 		<li><a href="#ArtesVisuais" tabindex="59">Artes visuais</a></li>
		
	
		
		
# 		<li><a href="#OutrasProducoesArtisticas" tabindex="60">Outras produções artísticas/culturais</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div><div id="menu-potencialInovacao" class="megamenu" _megamenupos="5pos" style="z-index: 1006; display: none;">
		
# 		<div class="column"><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#PatentePI" tabindex="62">Patente</a></li>
		
	
		
		
# 		<li><a href="#ProgramaComputadorRegistradoPI" tabindex="63">Programa de computador registrado</a></li>
		
	
		
		
# 		<li><a href="#CultivarProtegidaPI" tabindex="64">Cultivar protegida</a></li>
		
	
		
		
# 		<li><a href="#CultivarRegistradaPI" tabindex="65">Cultivar registrada</a></li>
		
	
		
		
# 		<li><a href="#DesenhoIndustrialRegistradoPI" tabindex="66">Desenho industrial registrado</a></li>
		
	
		
		
# 		<li><a href="#MarcaRegistradaPI" tabindex="67">Marca registrada</a></li>
		
	
		
		
# 		<li><a href="#TopografiaCircuitoIntegradoRegistradaPI" tabindex="68">Topografia de circuito integrado registrada</a></li>
		
	
		
		
# 		<li><a href="#ProgramaComputadorSemRegistroPI" tabindex="69">Programa de computador sem registro</a></li>
		
	
		
		
# 		<li><a href="#ProdutosPI" tabindex="70">Produtos</a></li>
		
	
		
		
# 		<li><a href="#ProcessosTecnicasPI" tabindex="71">Processos ou técnicas</a></li>
		
	
		
		
# 		<li><a href="#ProjetosPesquisaPI" tabindex="72">Projetos de pesquisa</a></li>
		
	
		
		
# 		<li><a href="#ProjetoDesenvolvimentoTecnologicoPI" tabindex="73">Projeto de desenvolvimento tecnológico</a></li>
		
	
		
		
# 		<li><a href="#ProjetoExtensaoPI" tabindex="74">Projeto de extensão</a></li>
		
	
		
		
# 		<li><a href="#OutrosProjetosPI" tabindex="75">Outros projetos</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div><div id="menu-educacaoPopularizacaoCTA" class="megamenu" _megamenupos="6pos" style="z-index: 1013; left: 412.109px; top: 109px; display: none;">
		
# 		<div class="column"><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#ArtigosEPCTA" tabindex="77">Artigos</a></li>
		
	
		
		
# 		<li><a href="#LivrosCapitulosEPCTA" tabindex="78">Livros e capítulos</a></li>
		
	
		
		
# 		<li><a href="#TextosJornaisEPCTA" tabindex="79">Textos em jornais de notícias/revistas</a></li>
		
	
		
		
# 		<li><a href="#ApresentacoesTrabalhoEPCTA" tabindex="80">Apresentações de trabalho</a></li>
		
	
		
		
# 		<li><a href="#ProgramaComputadorEPCTA" tabindex="81">Programa de computador sem registro de patente</a></li>
		
	
		
		
# 		<li><a href="#CursosCurtaEPCTA" tabindex="82">Cursos de curta duração ministrados</a></li>
		
	
		
		
# 		<li><a href="#MaterialDidaticoEPCTA" tabindex="83">Desenvolvimento de material didático ou instrucional</a></li>
		
	
		
		
# 		<li><a href="#EntrevistasEPCTA" tabindex="84">Entrevistas, mesas redondas, programas e comentários na mídia</a></li>
		
	
		
		
# 		<li><a href="#ProgramaComputadorRegistradoEPCTA" tabindex="85">Programa de Computador registrado</a></li>
		
	
		
		
# 		<li><a href="#OrganizacaoEventosEPCTA" tabindex="86">Organização de eventos, congressos, exposições e feiras</a></li>
		
	
		
		
# 		<li><a href="#RedesSociaisEPCTA" tabindex="87">Redes sociais, websites e blogs</a></li>
		
	
		
		
# 		<li><a href="#ArtesVisuaisEPCTA" tabindex="88">Artes Visuais</a></li>
		
	
		
		
# 		<li><a href="#ArtesCenicasEPCTA" tabindex="89">Artes Cênicas</a></li>
		
	
		
		
# 		<li><a href="#MusicaEPCTA" tabindex="90">Música</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div><div id="menu-eventos" class="megamenu" _megamenupos="7pos" style="z-index: 1008; display: none;">
		
# 		<div class="column"><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#ParticipacaoEventos" tabindex="92">Participação em eventos, congressos, exposições e feiras</a></li>
		
	
		
		
# 		<li><a href="#OrganizacaoEventos" tabindex="93">Organização de eventos, congressos, exposições e feiras</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div><div id="menu-orientacoes" class="megamenu" _megamenupos="8pos" style="z-index: 1014; left: 658.094px; top: 109px; display: none;">
		
# 		<div class="column"><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#Orientacoesconcluidas" tabindex="95">Orientações e supervisões concluídas</a></li>
		
	
		
		
# 		<li><a href="#Orientacaoemandamento" tabindex="96">Orientações e supervisões em andamento</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div><div id="menu-bancas" class="megamenu" _megamenupos="9pos" style="z-index: 1010; display: none;">
		
# 		<div class="column"><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#ParticipacaoBancasTrabalho" tabindex="98">Participação em bancas de trabalhos de conclusão</a></li>
		
	
		
		
# 		<li><a href="#ParticipacaoBancasComissoes" tabindex="99">Participação em bancas de comissões julgadoras</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div><div id="menu-citacoes" class="megamenu" _megamenupos="10pos" style="z-index: 1011; display: none;">
		
# 		<div class="column"><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="#Citacoes" tabindex="101">Citações no ISI</a></li>
		
	
		
		
# 		<li><a href="#Citacoes" tabindex="102">Citações no SciELO</a></li>
		
	
		
		
# 		<li><a href="#Citacoes" tabindex="103">Citações no SCOPUS</a></li>
		
	
		
		
# 		<li><a href="#Citacoes" tabindex="104">Citações em outras bases bibliográficas</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div><div id="menu-informacoes-complementares" class="megamenu" _megamenupos="11pos" style="z-index: 1012; display: none;">
		
# 		<div class="column"><h3 tabindex="0"></h3><ul>
		
		
# 		<li><a href="visualizacv.do?metodo=apresentar&amp;id=K4778130Y3&amp;tipo=completo&amp;idiomaExibicao=1" tabindex="106">Mostrar informações complementares</a></li>
		
# 	</ul></div>
		
		
# 		<div class="clear"></div></div></body></html>
# '''
```
