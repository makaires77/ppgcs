```python
import requests
import pandas as pd

class DataRetriever:
    def __init__(self):
        self.base_url = "https://servicodados.ibge.gov.br/api/v3/agregados/"

    def get_economic_data_ceara(self, agregado, periodos, variaveis):
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
        }

        url = f"{self.base_url}{agregado}/periodos/{periodos}/variaveis/{variaveis}"
        params = {"localidades": "23"}  # 23 é o código do IBGE para o Ceará
        response = requests.get(url, headers=headers, params=params)
        if response.status_code == 200:
            data = response.json()
            df = pd.json_normalize(data)
            return df
        else:
            return None

# Exemplo de uso da classe
data_retriever = DataRetriever()
# Substitua 'agregado', 'periodos' e 'variaveis' pelos valores apropriados
economic_data = data_retriever.get_economic_data_ceara('agregado', 'periodos', 'variaveis')
```
