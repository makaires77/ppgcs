A classificação de nós é uma tarefa comum de aprendizado de máquina aplicada a gráficos: modelos de treinamento para classificar nós. Concretamente, os modelos de classificação de nós são usados para prever as classes de nós não rotulados como propriedades de nós com base em outras propriedades de nós. Durante o treinamento, a propriedade que representa a classe do nó é chamada de propriedade de destino. O GDS oferece suporte à classificação de nós binários e multiclasse.

No GDS, temos pipelines de classificação de nós que oferecem um fluxo de trabalho ponta a ponta, desde a extração de recursos até a classificação de nós. Os pipelines de treinamento residem no catálogo de pipelines . Quando um pipeline de treinamento é executado , um modelo de classificação é criado e armazenado no catálogo de modelos .

Um pipeline de treinamento é uma sequência de duas fases:
    O gráfico é aumentado com novas propriedades de nó em uma série de etapas.
    O gráfico aumentado é usado para treinar um modelo de classificação de nós.

Pode-se configurar quais etapas devem ser incluídas acima. As etapas executam algoritmos GDS que criam novas propriedades de nó. Após configurar as etapas de propriedade do nó, pode-se selecionar um subconjunto de propriedades do nó para ser usado como recursos. A fase de treinamento (II) treina vários candidatos a modelos usando validação cruzada, seleciona o melhor e relata métricas de desempenho relevantes.

Após treinar o pipeline , um modelo de classificação é criado. Este modelo inclui as etapas de propriedade do nó e a configuração de recursos do pipeline de treinamento e os utiliza para gerar os recursos relevantes para classificar nós não rotulados. O modelo de classificação pode ser aplicado para prever a classe de nós inéditos. Além da classe prevista para cada nó, a probabilidade prevista para cada classe também pode ser retida nos nós. A ordem das probabilidades corresponde à ordem das classes cadastradas no modelo.

A classificação só pode ser feita com um modelo de classificação (não com um pipeline de treinamento).

DAG algorithms

Directed Acyclic Graphs (DAGs) are directed graphs that do not contain cycles. These kind of graphs are commonly used to model dependencies between entities.

The canonical algorithm that goes hand in hand with DAGs is topological sort, for which GDS provides an efficient parallel implementation. Running topological sort is the best way to make sure the graph is a DAG.

Some of the problems that are computationally hard to solve in the general case can be solved efficiently when the scope is limited to DAGs. One of these is the longest path problem, for which GDS provides an efficient algorithm.

The Neo4j GDS library includes the following DAG algorithms:

Alpha

    Topological Sort
    Longest Path


### Modelar a escolha de tecnologias com o Algoritmo de Redes de Influência (ARI)
Para modelar a escolha de tecnologias com o ARI no contexto de um projeto de desenvolvimento, podemos seguir os seguintes passos:

#### 1. Definição dos Critérios:
Competências individuais da equipe:
Identificar as habilidades e conhecimentos relevantes para o projeto.
Exemplos: Expertise em linguagens de programação, frameworks, ferramentas, metodologias etc.
Quantificar as competências em diferentes níveis (iniciante, intermediário, avançado).

#### 2. Definição das Alternativas:
Tecnologias a serem consideradas:
Listar as tecnologias que podem ser utilizadas no projeto.
Exemplos: Linguagens de programação, frameworks, ferramentas, bibliotecas etc.
Considerar fatores como maturidade da tecnologia, curva de aprendizado, suporte da comunidade etc.

#### 3. Construção da Rede de Influência:
Relacionamentos entre critérios e alternativas:

    Para cada critério, determinar como ele influencia cada alternativa.

Exemplos:

    Expertise em linguagem X facilita o aprendizado de framework Y.
    Ferramenta Z aumenta a produtividade para o desenvolvimento de tecnologia W.
    Conhecimento em metodologia A melhora a qualidade do software B.
    Quantificar a força da influência com pesos (0-1).

#### 4. Cálculo da Influência:
Algoritmos como PageRank ou Katz:
Calcular a importância de cada tecnologia com base na influência que recebe dos critérios.
Considerar diferentes cenários de pesos e combinações de critérios.

#### 5. Seleção da Tecnologia:
Tecnologia com maior valor de influência:
Escolher a tecnologia com maior potencial para o sucesso do projeto.
Considerar outras variáveis além da influência, como custos, prazos e riscos.
Exemplo:

Critérios:

    C1: Expertise em linguagem X (peso 0.7)
    C2: Conhecimento em metodologia A (peso 0.6)
    C3: Experiência com ferramenta Z (peso 0.5)

Alternativas:

    A1: Tecnologia W (influenciada por C1 e C2)
    A2: Tecnologia Y (influenciada por C1 e C3)

Rede de Influência:

    C1 -> A1 (peso 0.8)
    C1 -> A2 (peso 0.5)
    C2 -> A1 (peso 0.7)
    C3 -> A2 (peso 0.6)

Cálculo da Influência:

    PageRank: A1 (0.65), A2 (0.35)
    Katz: A1 (0.72), A2 (0.28)

Seleção da Tecnologia:

    A1 (Tecnologia W) é a mais indicada, considerando a influência dos critérios.

#### Considerações:
O modelo pode ser adaptado para diferentes cenários e projetos.
É importante considerar os objetivos do projeto e as características das tecnologias.
O ARI fornece uma análise mais profunda e sensível das relações entre os critérios e alternativas.

```python
from neo4j import GraphDatabase

# Conexão com o banco de dados Neo4j
graph = GraphDatabase("bolt://localhost:7687", auth=("neo4j", "password"))

# Criação dos nós da rede
critica1 = graph.create_node("Criterio", nome="Critério 1")
critica2 = graph.create_node("Criterio", nome="Critério 2")
alternativa1 = graph.create_node("Alternativa", nome="Alternativa 1")
alternativa2 = graph.create_node("Alternativa", nome="Alternativa 2")

# Criação das arestas da rede
critica1.create_relationship("Influencia", critica2, peso=0.7)
critica2.create_relationship("Influencia", alternativa1, peso=0.5)
critica2.create_relationship("Influencia", alternativa2, peso=0.3)

# Cálculo da influência usando o PageRank
pagerank = graph.run("CALL algo.pageRank(nodes, {iterations: 10})")

# Impressão dos resultados
for result in pagerank:
    node = result["node"]
    rank = result["rank"]
    print(f"Nome: {node['nome']}, Rank: {rank}")
```
